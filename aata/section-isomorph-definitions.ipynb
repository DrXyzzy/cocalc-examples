<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 9.1 Definition and Examples"><span class="type">Section</span><span class="codenumber">9.1</span><span class="title">Definition and Examples</span></h2><a href="section-isomorph-definitions.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-1411">Two groups $(G, \cdot)$ and $(H, \circ)$ are <dfn class="terminology">isomorphic</dfn> if there exists a one-to-one and onto map $\phi : G \rightarrow H$ such that the group operation is preserved;  that is,</p><div class="displaymath">
\begin{equation*}
\phi( a \cdot b) = \phi( a) \circ \phi( b)
\end{equation*}
</div><p>for all $a$ and $b$ in $G\text{.}$ If $G$ is isomorphic to $H\text{,}$ we write $G \cong H\text{.}$  The map $\phi$ is called an <dfn class="terminology">isomorphism</dfn>.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-z4"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.1</span></h6><p id="p-1412">To show that ${\mathbb Z}_4 \cong \langle i \rangle\text{,}$ define a map $\phi: {\mathbb Z}_4 \rightarrow \langle i \rangle$ by $\phi(n) = i^n\text{.}$  We must show that $\phi$ is bijective and preserves the group operation.  The map $\phi$ is one-to-one and onto because</p><div class="displaymath">
\begin{align*}
\phi(0) & = 1\\
\phi(1) & = i\\
\phi(2) & = -1\\
\phi(3) & = -i.
\end{align*}
</div><p>Since</p><div class="displaymath">
\begin{equation*}
\phi(m + n) = i^{m+n} = i^m i^n = \phi(m) \phi( n),
\end{equation*}
</div><p>the group operation is preserved.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-real"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.2</span></h6><p id="p-1413">We can define an isomorphism $\phi$ from the additive group of real numbers $( {\mathbb R}, + )$ to the multiplicative group of positive real numbers  $( {\mathbb R^+}, \cdot )$  with the exponential map; that is,</p><div class="displaymath">
\begin{equation*}
\phi( x + y) = e^{x + y} = e^x e^y = \phi( x ) \phi( y).
\end{equation*}
</div><p>Of course, we must still show that $\phi$ is one-to-one and onto, but this can be determined using calculus.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-rational"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.3</span></h6><p id="p-1414">The integers are isomorphic to the subgroup of ${\mathbb Q}^\ast$ consisting of elements of the form $2^n\text{.}$  Define a map $\phi: {\mathbb Z} \rightarrow {\mathbb Q}^\ast$ by $\phi( n ) = 2^n\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
\phi( m + n ) = 2^{m + n} = 2^m 2^n = \phi( m ) \phi( n ).
\end{equation*}
</div><p>By definition the map $\phi$ is onto the subset $\{2^n :n \in {\mathbb Z} \}$ of  ${\mathbb Q}^\ast\text{.}$ To show that the map is injective, assume that $m \neq n\text{.}$  If we can show that $\phi(m) \neq \phi(n)\text{,}$ then we are done.  Suppose that $m \gt n$ and assume that $\phi(m) = \phi(n)\text{.}$  Then $2^m = 2^n$ or $2^{m-n} = 1\text{,}$ which is impossible since $m-n \gt 0\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-units"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.4</span></h6><p id="p-1415">The groups ${\mathbb Z}_8$ and ${\mathbb Z}_{12}$  cannot be isomorphic since they have different orders; however, it is true that $U(8) \cong U(12)\text{.}$  We know that</p><div class="displaymath">
\begin{align*}
U(8) & = \{1, 3, 5, 7 \}\\
U(12) & = \{1, 5, 7, 11 \}.
\end{align*}
</div><p>An isomorphism $\phi : U(8) \rightarrow U(12)$ is then given by</p><div class="displaymath">
\begin{align*}
1 & \mapsto  1\\
3 & \mapsto  5\\
5 & \mapsto  7\\
7 & \mapsto  11.
\end{align*}
</div><p>The map $\phi$ is not the only possible isomorphism between these two groups.  We could define another isomorphism $\psi$ by $\psi(1) = 1\text{,}$ $\psi(3) = 11\text{,}$ $\psi(5) = 5\text{,}$ $\psi(7) = 7\text{.}$ In fact, both of these groups are isomorphic to ${\mathbb Z}_2 \times {\mathbb Z}_2$ (see Example <a href="section-subgroups.ipynb#example-groups-z2xz2" class="xref" alt="Example 3.28 " title="Example 3.28 ">3.28</a> in Chapter <a href="groups.ipynb" class="xref" alt="Chapter 3 Groups" title="Chapter 3 Groups">3</a>).</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-not-isomorph"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.5</span></h6><p id="p-1416">Even though $S_3$ and ${\mathbb Z}_6$ possess the same number of elements, we would suspect that they are not isomorphic, because ${\mathbb Z}_6$ is abelian and $S_3$ is nonabelian.  To demonstrate that this is indeed the case, suppose that $\phi : {\mathbb Z}_6 \rightarrow  S_3$ is an isomorphism.  Let $a , b \in S_3$ be two elements such that $ab \neq ba\text{.}$  Since $\phi$ is an isomorphism, there exist elements $m$ and $n$ in ${\mathbb Z}_6$ such that</p><div class="displaymath">
\begin{equation*}
\phi( m )  = a \quad \text{and} \quad \phi( n )  = b.
\end{equation*}
</div><p>However,</p><div class="displaymath">
\begin{equation*}
ab = \phi(m ) \phi(n) = \phi(m + n) = \phi(n + m) = \phi(n ) \phi(m) = ba,
\end{equation*}
</div><p>which contradicts the fact that $a$ and $b$ do not commute.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-isomorph-1"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">9.6</span></h6><p id="p-1417">Let $\phi : G \rightarrow H$ be an isomorphism of two groups.  Then the following statements are true.</p><ol class="decimal"><li id="li-355"><p id="p-1418">$\phi^{-1} : H \rightarrow G$ is an isomorphism.</p></li><li id="li-356"><p id="p-1419">$|G| = |H|\text{.}$</p></li><li id="li-357"><p id="p-1420">If $G$ is abelian, then $H$ is abelian.</p></li><li id="li-358"><p id="p-1421">If $G$ is cyclic, then $H$ is cyclic.</p></li><li id="li-359"><p id="p-1422">If $G$ has a subgroup of order $n\text{,}$ then $H$ has a subgroup of order $n\text{.}$</p></li></ol></article><article class="proof" id="proof-57"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1423">Assertions (1) and (2) follow from the fact that $\phi$ is a bijection.  We will prove (3) here and leave the remainder of the theorem to be proved in the exercises.</p><p id="p-1424">(3) Suppose that $h_1$ and $h_2$ are elements of $H\text{.}$  Since $\phi$ is onto, there exist elements $g_1, g_2 \in G$ such that $\phi(g_1) = h_1$ and $\phi(g_2) = h_2\text{.}$  Therefore,</p><div class="displaymath">
\begin{equation*}
h_1 h_2 = \phi(g_1) \phi(g_2) =  \phi(g_1 g_2) = \phi(g_2 g_1) = \phi(g_2) \phi(g_1) = h_2 h_1.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><p id="p-1425">We are now in a position to characterize all cyclic groups.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-isomorph-2"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">9.7</span></h6><p id="p-1426">All cyclic groups of infinite order are isomorphic to ${\mathbb Z}\text{.}$</p></article><article class="proof" id="proof-58"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1427">Let $G$ be a cyclic group with infinite order and suppose that $a$ is a generator of $G\text{.}$  Define a map $\phi : {\mathbb Z} \rightarrow  G$ by $\phi : n \mapsto a^n\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
\phi( m+n ) = a^{m+n} = a^m a^n = \phi( m ) \phi( n ).
\end{equation*}
</div><p>To show that $\phi$ is injective, suppose that $m$ and $n$ are two elements in ${\mathbb Z}\text{,}$ where $m \neq n\text{.}$  We can assume that $m \gt n\text{.}$  We must show that $a^m \neq a^n\text{.}$ Let us suppose the contrary; that is, $a^m = a^n\text{.}$ In this case $a^{m - n} = e\text{,}$ where $m - n \gt 0\text{,}$ which contradicts the fact that $a$ has infinite order.  Our map is onto since any element in $G$ can be written as $a^n$ for some integer $n$ and $\phi(n) = a^n\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-isomorph-3"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">9.8</span></h6><p id="p-1428">If $G$ is a cyclic group of order $n\text{,}$ then $G$ is isomorphic to ${\mathbb Z}_n\text{.}$</p></article><article class="proof" id="proof-59"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1429">Let $G$ be a cyclic group of order $n$ generated by $a$ and define a map $\phi : {\mathbb Z}_n \rightarrow  G$ by $\phi : k \mapsto a^k\text{,}$ where $0 \leq k \lt n\text{.}$ The proof that $\phi$ is an isomorphism is one of the end-of-chapter exercises.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-isomorph-4"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">9.9</span></h6><p id="p-1430">If $G$ is a  group of order $p\text{,}$ where $p$ is a prime number, then $G$ is isomorphic to ${\mathbb Z}_p\text{.}$</p></article><article class="proof" id="proof-60"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1431">The proof is a direct result of Corollary <a href="section-lagranges-theorem.ipynb#corollary-cosets-theorem-7" class="xref" alt="Corollary 6.12 " title="Corollary 6.12 ">6.12</a>.</p></article></div>

<div class="mathbook-content"><p id="p-1432">The main goal in group theory is to classify all groups; however, it makes sense to consider two groups to be the same if they are isomorphic.  We state this result in the following theorem, whose proof is left as an exercise.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-isomorph-5"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">9.10</span></h6><p id="p-1433">The isomorphism of groups determines an equivalence relation on the class of all groups.</p></article></div>

<div class="mathbook-content"><p id="p-1434">Hence, we can modify our goal of classifying all groups to classifying all groups <dfn class="terminology">up to isomorphism</dfn>; that is, we will consider two groups to be the same if they are isomorphic.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Cayley's Theorem"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Cayley's Theorem</span></h3><a href="section-isomorph-definitions.ipynb#isomorph-subsection-cayleys-theorem" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-1435">Cayley proved that if $G$ is a group, it is isomorphic to a group of permutations on some set; hence, every group is a permutation group.  Cayley's Theorem is what  we call a representation theorem.  The aim of representation theory is to find an isomorphism of some group $G$ that we wish to study into a group that we know a great deal about, such as a group of permutations or matrices.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-cayley"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.11</span></h6><p id="p-1436">Consider the group ${\mathbb Z}_3\text{.}$  The Cayley table for ${\mathbb Z}_3$ is as follows.</p><div class="sidebyside" id="sidebyside-5"><div class="sbsrow" style="margin-left:0%;margin-right:0%;"><div class="sbspanel" style="width:100%;justify-content:flex-start;"><div class="displaymath" id="p-1437">
\begin{equation*}
\begin{array}{c|ccc}
+ & 0 & 1 & 2 \\ \hline
0       & 0 & 1 & 2 \\
1       & 1 & 2 & 0 \\
2       & 2 & 0 & 1 
\end{array}
\end{equation*}
</div></div></div></div><p id="p-1438">The addition table of ${\mathbb Z}_3$ suggests that it is the same as the permutation group $G = \{ (0), (0 1 2), (0 2 1) \}\text{.}$  The isomorphism here is</p><div class="displaymath">
\begin{align*}
0 & \mapsto
\begin{pmatrix}
0 & 1 & 2 \\ 0 & 1 & 2
\end{pmatrix}
= (0)\\
1 & \mapsto
\begin{pmatrix}
0 & 1 & 2 \\ 1 & 2 & 0
\end{pmatrix}
= (0 1 2)\\
2 & \mapsto
\begin{pmatrix}
0 & 1 & 2 \\ 2 & 0 & 1
\end{pmatrix}
= (0 2 1).
\end{align*}
</div></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-cayleys"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">9.12</span><span class="title">Cayley</span></h6><p id="p-1439">Every group is isomorphic to a group of permutations.</p></article><article class="proof" id="proof-61"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1440">Let $G$ be a group.  We must find a group of permutations $\overline{G}$ that is isomorphic to $G\text{.}$  For any $g \in G\text{,}$ define a  function $\lambda_g : G \rightarrow G$ by $\lambda_g(a) = ga\text{.}$  We claim that $\lambda_g$ is a permutation of $G\text{.}$  To show that $\lambda_g$ is one-to-one, suppose that $\lambda_g(a) = \lambda_g(b)\text{.}$  Then</p><div class="displaymath">
\begin{equation*}
ga =\lambda_g(a) = \lambda_g(b) = gb.
\end{equation*}
</div><p>Hence, $a = b\text{.}$  To show that $\lambda_g$ is onto, we must prove that for each $a \in G\text{,}$ there is a $b$ such that $\lambda_g (b) = a\text{.}$  Let $b = g^{-1} a\text{.}$</p><p id="p-1441">Now we are ready to define our group $\overline{G}\text{.}$ Let</p><div class="displaymath">
\begin{equation*}
\overline{G} = \{ \lambda_g : g \in G \}.
\end{equation*}
</div><p>We must show that $\overline{G}$ is a group under composition of functions and find an isomorphism between $G$ and $\overline{G}\text{.}$  We have closure under composition of functions since</p><div class="displaymath">
\begin{equation*}
(\lambda_g \circ \lambda_h )(a) = \lambda_g(ha) = gha = \lambda_{gh} (a).
\end{equation*}
</div><p>Also,</p><div class="displaymath">
\begin{equation*}
\lambda_e (a) = ea = a
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
(\lambda_{g^{-1}} \circ \lambda_g) (a) = \lambda_{g^{-1}} (ga) = g^{-1} g a = a = \lambda_e (a).
\end{equation*}
</div><p id="p-1442">We can define an isomorphism from $G$ to $\overline{G}$ by $\phi : g \mapsto \lambda_g\text{.}$ The group operation is preserved since</p><div class="displaymath">
\begin{equation*}
\phi(gh) = \lambda_{gh} = \lambda_g \lambda_h = \phi(g) \phi(h).
\end{equation*}
</div><p>It is also one-to-one, because if $\phi(g)(a) = \phi(h)(a)\text{,}$ then</p><div class="displaymath">
\begin{equation*}
ga = \lambda_g a = \lambda_h a=  ha.
\end{equation*}
</div><p>Hence, $g = h\text{.}$  That $\phi$ is onto follows from the fact that $\phi( g ) = \lambda_g$ for any $\lambda_g \in \overline{G}\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1443">The isomorphism $g \mapsto \lambda_g$ is known as the <dfn class="terminology">left regular representation</dfn> of $G\text{.}$</p></div>

<div class="mathbook-content"><p id="p-1444">Arthur Cayley was born in England in 1821, though he spent much of the first part of his life in Russia, where his father was a merchant.  Cayley was educated at Cambridge, where he took the first Smith's Prize in mathematics.  A lawyer for much of his adult life, he wrote several papers in his early twenties before entering the legal profession at the age of 25.  While practicing law he continued his mathematical research, writing more than 300 papers during this period of his life.  These included some of his best work.  In 1863 he left law to become a professor at Cambridge.  Cayley wrote more than 900 papers in fields such as group theory, geometry, and linear algebra. His legal knowledge was very valuable to Cambridge; he participated in the writing of many of the university's statutes.  Cayley was also one of the people responsible for the admission of women to Cambridge.</p></div>
