<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 16.4 Maximal and Prime Ideals"><span class="type">Section</span><span class="codenumber">16.4</span><span class="title">Maximal and Prime Ideals</span></h2><a href="section-maximal-prime-ideals.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-2384">In this particular section we are especially interested in certain ideals of commutative rings. These ideals give us special types of factor rings. More specifically, we would like to characterize those ideals $I$ of a commutative ring $R$ such that $R/I$ is an integral domain or a field.</p></div>

<div class="mathbook-content"><p id="p-2385">A proper ideal $M$ of a ring $R$ is a <dfn class="terminology">maximal ideal</dfn> of $R$ if the ideal $M$ is not a proper subset of any ideal of $R$ except $R$ itself. That is, $M$ is a  maximal ideal if for any ideal $I$ properly containing $M\text{,}$ $I = R\text{.}$ The following theorem completely characterizes maximal ideals for commutative rings with identity in terms of their corresponding factor rings.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-maximal-ideal-field"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">16.35</span></h6><p id="p-2386">Let $R$ be a commutative ring with identity and $M$ an ideal in $R\text{.}$ Then $M$ is a maximal ideal of $R$ if and only if $R/M$ is a field.</p></article><article class="proof" id="proof-112"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2387">Let $M$ be a maximal ideal in $R\text{.}$ If $R$ is a commutative ring, then $R/M$ must also be a commutative ring.  Clearly, $1 + M$ acts as an identity for $R/M\text{.}$ We must also show that every nonzero element in $R/M$ has an inverse.  If $a + M$ is a nonzero element in $R/M\text{,}$ then $a \notin M\text{.}$ Define $I$ to be the set $\{ ra + m : r \in R \text{ and } m \in M \}\text{.}$ We will show that $I$ is an ideal in $R\text{.}$ The set $I$ is nonempty since $0a+0=0$ is in $I\text{.}$ If $r_1 a + m_1$ and $r_2 a + m_2$ are two elements in $I\text{,}$ then</p><div class="displaymath">
\begin{equation*}
(r_1 a + m_1) - ( r_2 a + m_2) = (r_1 - r_2)a + (m_1 - m_2)
\end{equation*}
</div><p>is in $I\text{.}$ Also, for any $r \in R$ it is true that $rI \subset I\text{;}$ hence, $I$ is closed under multiplication and satisfies the necessary conditions to be an ideal. Therefore, by Proposition <a href="section-rings-definitions.ipynb#prop-subring" class="xref" alt="Proposition 16.10 " title="Proposition 16.10 ">16.10</a> and the definition of an ideal, $I$ is an ideal properly containing $M\text{.}$ Since $M$ is a maximal ideal, $I=R\text{;}$ consequently, by the definition of $I$ there must be an $m$ in $M$ and an element $b$ in $R$ such that $1=ab+m\text{.}$ Therefore,</p><div class="displaymath">
\begin{equation*}
1 + M = ab + M = ba + M = (a+M)(b+M).
\end{equation*}
</div><p id="p-2388">Conversely, suppose that $M$ is an ideal and $R/M$ is a field. Since $R/M$ is a field, it must contain at least two elements: $0 + M = M$ and $1 + M\text{.}$ Hence, $M$ is a proper ideal of $R\text{.}$  Let $I$ be any ideal properly containing $M\text{.}$ We need to show that $I = R\text{.}$ Choose $a$ in $I$ but not in $M\text{.}$ Since $a+ M$ is a nonzero element in a field, there exists an element $b +M$ in $R/M$ such that $(a+M)(b+M) = ab + M = 1+M\text{.}$  Consequently, there exists an element $m \in M$ such that $ab + m = 1$ and $1$ is in $I\text{.}$ Therefore, $r1 =r \in I$ for all $r \in R\text{.}$ Consequently, $I = R\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-prime-ideal"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.36</span></h6><p id="p-2389">Let $p{\mathbb Z}$ be an ideal in ${\mathbb Z}\text{,}$ where $p$ is prime. Then $p{\mathbb Z}$  is a maximal ideal since ${\mathbb Z}/ p {\mathbb Z} \cong {\mathbb Z}_p$ is a field.</p></article></div>

<div class="mathbook-content"><p id="p-2390">A proper ideal $P$ in a commutative ring $R$ is called a <dfn class="terminology">prime ideal</dfn> if whenever $ab \in P\text{,}$ then either $a \in P$ or $b \in P\text{.}$<span class="footnote"><a knowl="" class="id-ref" refid="hk-fn-5" id="fn-5"><sup> 5 </sup></a></span><span id="hk-fn-5" class="hidden-content tex2jax_ignore"><span class="footnote">It is possible to define prime ideals in a noncommutative ring. See [1] or [3].</span></span></p></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-z12-ideal"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.37</span></h6><p id="p-2391">It is easy to check that the set $P = \{ 0, 2, 4, 6, 8, 10  \}$ is an ideal in ${\mathbb Z}_{12}\text{.}$ This ideal is prime. In fact, it is a maximal ideal.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-36"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">16.38</span></h6><p id="p-2392">Let $R$ be a commutative ring with identity $1\text{,}$ where $1 \neq 0\text{.}$ Then $P$ is a prime ideal in $R$ if and only if $R/P$ is an integral domain.</p></article><article class="proof" id="proof-113"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2393">First let us assume that $P$ is an ideal in $R$ and $R/P$ is an integral domain.  Suppose that $ab \in P\text{.}$ If $a + P$ and $b + P$ are two  elements of $R/P$ such that $(a + P)(b + P) = 0 + P = P\text{,}$ then either $a + P = P$ or $b + P = P\text{.}$  This means that either $a$ is in $P$ or $b$ is in $P\text{,}$ which shows that $P$ must be prime.</p><p id="p-2394">Conversely, suppose that $P$ is prime and</p><div class="displaymath">
\begin{equation*}
(a + P)(b + P) = ab + P = 0 + P = P.
\end{equation*}
</div><p>Then $ab \in P\text{.}$ If $a \notin P\text{,}$ then $b$ must be in $P$ by the definition of a prime ideal; hence, $b + P = 0 + P$ and $R/P$ is an integral domain.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-nz-ideal"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.39</span></h6><p id="p-2395">Every ideal in ${\mathbb Z}$ is of the form $n {\mathbb Z}\text{.}$ The factor ring ${\mathbb Z} / n{\mathbb Z} \cong {\mathbb Z}_n$ is an integral domain only when $n$ is prime.  It is actually a field.  Hence, the nonzero prime ideals in ${\mathbb Z}$ are the ideals $p{\mathbb Z}\text{,}$ where $p$ is prime. This example really justifies the use of the word “prime” in our definition of prime ideals.</p></article></div>

<div class="mathbook-content"><p id="p-2396">Since every field is an integral domain, we have the following corollary.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-maximal-ideal"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">16.40</span></h6><p id="p-2397">Every maximal ideal in a commutative ring with identity is also a prime ideal.</p></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Historical Note"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Historical Note</span></h3><a href="section-maximal-prime-ideals.ipynb#rings-subsection-historical-note" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-2398">Amalie Emmy Noether, one of the  outstanding mathematicians of the twentieth century, was born in Erlangen, Germany in 1882. She was the daughter of Max Noether (1844–1921), a distinguished mathematician at the University of Erlangen. Together with Paul Gordon (1837–1912), Emmy Noether's father strongly influenced her early education. She entered the University of Erlangen at the age of 18. Although women had been admitted to universities in England, France, and Italy for decades, there was great resistance to their presence at universities in Germany. Noether was one of only two women among the university's 986 students. After completing her doctorate under Gordon in 1907, she continued to do research at Erlangen, occasionally lecturing when her father was ill.</p></div>

<div class="mathbook-content"><p id="p-2399">Noether went to Göttingen to study in 1916. David Hilbert and Felix Klein tried unsuccessfully to secure her an appointment at Göttingen. Some of the faculty objected to women lecturers, saying, “What will our soldiers think when they return to the university and are expected to learn at the feet of a woman?”  Hilbert, annoyed at the question, responded, “Meine Herren, I do not see that the sex of a candidate is an argument against her admission as a Privatdozent. After all, the Senate is not a bathhouse.”  At the end of World War I, attitudes changed and conditions greatly improved for women.  After Noether passed her habilitation examination in 1919, she was given a title and was paid a small sum for her lectures.</p></div>

<div class="mathbook-content"><p id="p-2400">In 1922, Noether became a Privatdozent at Göttingen. Over the next 11 years she used axiomatic methods to develop an abstract theory of rings and ideals. Though she was not good at lecturing, Noether was an inspiring teacher. One of her many students was B. L. van der Waerden, author of the first text treating abstract algebra from a modern point of view. Some of the other mathematicians  Noether influenced or closely worked with were Alexandroff, Artin, Brauer, Courant, Hasse, Hopf, Pontryagin, von Neumann, and Weyl. One of the high points of her career was an invitation to address the International Congress of Mathematicians in Zurich in 1932. In spite  of all the recognition she received from her colleagues, Noether's abilities were never recognized as they should have been during her lifetime. She was never promoted to full professor by the Prussian academic bureaucracy.</p></div>

<div class="mathbook-content"><p id="p-2401">In 1933, Noether, a Jew, was banned from participation in all academic activities in Germany. She emigrated to the United States, took a position at Bryn Mawr College, and became a member of the Institute for Advanced Study at Princeton. Noether died suddenly on April 14, 1935. After her death she was eulogized by such notable scientists as Albert Einstein.</p></div>
