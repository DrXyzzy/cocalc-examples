<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 5.1 Definitions and Notation"><span class="type">Section</span><span class="codenumber">5.1</span><span class="title">Definitions and Notation</span></h2><a href="section-permutation-definitions.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-784">In general, the permutations of a set $X$ form a group $S_X\text{.}$ If $X$ is a finite set, we can assume $X=\{ 1, 2, \ldots, n\}\text{.}$ In this case we write $S_n$ instead of $S_X\text{.}$  The following theorem says that $S_n$ is a group. We call this group the <dfn class="terminology">symmetric group</dfn> on $n$ letters.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-17"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">5.1</span></h6><p id="p-785">The symmetric group on $n$ letters, $S_n\text{,}$ is a group with $n!$ elements, where the binary operation is the composition of maps.</p></article><article class="proof" id="proof-27"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-786">The identity of $S_n$ is just the identity map that sends 1 to 1, 2 to 2, $\ldots\text{,}$ $n$ to $n\text{.}$ If $f : S_n \rightarrow S_n$ is a permutation, then $f^{-1}$ exists, since $f$ is one-to-one and onto; hence, every permutation has an inverse. Composition of maps is associative, which makes the group operation associative. We leave the proof that $|S_n|= n!$ as an exercise.</p></article></div>

<div class="mathbook-content"><p id="p-787">A subgroup of $S_n$ is called a <dfn class="terminology">permutation group</dfn>.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-permute-s5"><h6 class="heading"><span class="type">Example</span><span class="codenumber">5.2</span></h6><p id="p-788">Consider the subgroup $G$ of $S_5$ consisting of the identity permutation $\identity$ and the permutations</p><div class="displaymath">
\begin{align*}
\sigma
& =
\begin{pmatrix}
1 & 2 & 3 & 4 & 5 \\
1 & 2 & 3 & 5 & 4
\end{pmatrix}\\
\tau
& =
\begin{pmatrix}
1 & 2 & 3 & 4 & 5 \\
3 & 2 & 1 & 4 & 5
\end{pmatrix}\\
\mu
& =
\begin{pmatrix}
1 & 2 & 3 & 4 & 5 \\
3 & 2 & 1 & 5 & 4
\end{pmatrix}.
\end{align*}
</div><p>The following table tells us how to multiply elements in the permutation group $G\text{.}$</p><div class="sidebyside" id="sidebyside-2"><div class="sbsrow" style="margin-left:0%;margin-right:0%;"><div class="sbspanel" style="width:100%;justify-content:flex-start;"><div class="displaymath" id="p-789">
\begin{equation*}
\begin{array}{c|cccc}
\circ   & \identity     & \sigma & \tau   & \mu    \\
\hline
\identity     & \identity     & \sigma & \tau   & \mu    \\
\sigma & \sigma & \identity     & \mu    & \tau   \\
\tau   & \tau   & \mu    & \identity     & \sigma \\
\mu    & \mu    & \tau   & \sigma & \identity
\end{array}
\end{equation*}
</div></div></div></div></article></div>

<div class="mathbook-content"><article class="remark-like" id="remark-2"><h6 class="heading"><span class="type">Remark</span><span class="codenumber">5.3</span></h6><p id="p-790">Though it is natural to multiply elements in a group from left to right, functions are composed from right to left.  Let $\sigma$ and $\tau$ be permutations on a set $X\text{.}$ To compose $\sigma$ and $\tau$ as functions, we calculate $(\sigma \circ \tau)(x) = \sigma( \tau(x))\text{.}$ That is, we do $\tau$ first, then $\sigma\text{.}$ There are several ways to approach this inconsistency. <em class="emphasis">We will adopt the convention of multiplying permutations right to left. To compute $\sigma \tau\text{,}$ do $\tau$ first and then $\sigma\text{.}$</em> That is, by $\sigma \tau (x)$ we mean $\sigma( \tau( x))\text{.}$ (Another way of solving this problem would be to write functions on the right; that is, instead of writing $\sigma(x)\text{,}$ we could write $(x)\sigma\text{.}$ We could also multiply permutations left to right to agree with the usual way of multiplying elements in a group. Certainly all of these methods have been used.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-permute-s4-nonabelian"><h6 class="heading"><span class="type">Example</span><span class="codenumber">5.4</span></h6><p id="p-791">Permutation multiplication is not usually commutative. Let</p><div class="displaymath">
\begin{align*}
\sigma
& =
\begin{pmatrix}
1 & 2 & 3 & 4 \\
4 & 1 & 2 & 3
\end{pmatrix}\\
\tau
& =
\begin{pmatrix}
1 & 2 & 3 & 4 \\
2 & 1 & 4 & 3
\end{pmatrix}.
\end{align*}
</div><p>Then</p><div class="displaymath">
\begin{equation*}
\sigma \tau
=
\begin{pmatrix}
1 & 2 & 3 & 4 \\
1 & 4 & 3 & 2
\end{pmatrix},
\end{equation*}
</div><p>but</p><div class="displaymath">
\begin{equation*}
\tau \sigma
=
\begin{pmatrix}
1 & 2 & 3 & 4 \\
3 & 2 & 1 & 4
\end{pmatrix}.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Cycle Notation"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Cycle Notation</span></h3><a href="section-permutation-definitions.ipynb#permute-subsection-cycle-notation" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-792">The notation that we have used to represent permutations up to this point is cumbersome, to say the least. To work effectively with permutation groups, we need a more streamlined method of writing down and manipulating permutations.</p></div>

<div class="mathbook-content"><p id="p-793">A permutation $\sigma \in S_X$ is a <dfn class="terminology">cycle of length</dfn> $k$ if there exist elements $a_1, a_2, \ldots, a_k \in X$ such that</p><div class="displaymath">
\begin{align*}
\sigma( a_1 ) & = a_2\\
\sigma( a_2 ) & = a_3\\
& \vdots\\
\sigma( a_k ) & = a_1
\end{align*}
</div><p>and $\sigma( x) =x$ for all other elements $x \in X\text{.}$ We will write $(a_1, a_2, \ldots, a_k )$ to denote the cycle $\sigma\text{.}$  Cycles are the building blocks of all permutations.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-permute-cycle-notation"><h6 class="heading"><span class="type">Example</span><span class="codenumber">5.5</span></h6><p id="p-794">The permutation</p><div class="displaymath">
\begin{equation*}
\sigma =
\begin{pmatrix}
1 & 2 & 3 & 4 & 5 & 6 & 7\\
6 & 3 & 5 & 1 & 4 & 2 & 7
\end{pmatrix}
=
(1 6 2 3 5 4 )
\end{equation*}
</div><p>is a cycle of length 6, whereas</p><div class="displaymath">
\begin{equation*}
\tau =
\begin{pmatrix}
1 & 2 & 3 & 4 & 5 & 6 \\
1 & 4 & 2 & 3 & 5 & 6
\end{pmatrix}
=
(2 4 3)
\end{equation*}
</div><p>is a cycle of length 3.</p><p id="p-795">Not every permutation is a cycle. Consider the permutation</p><div class="displaymath">
\begin{equation*}
\begin{pmatrix}
1 & 2 & 3 & 4 & 5 & 6 \\
2 & 4 & 1 & 3 & 6 & 5
\end{pmatrix}
= (1 2 4 3)(5 6).
\end{equation*}
</div><p>This permutation actually contains a cycle of length 2 and a cycle of length 4.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-permute-cycle-mult"><h6 class="heading"><span class="type">Example</span><span class="codenumber">5.6</span></h6><p id="p-796">It is very easy to compute products of cycles. Suppose that</p><div class="displaymath">
\begin{equation*}
\sigma  = (1 3 5 2 ) \quad \text{and} \quad \tau = (2 5 6).
\end{equation*}
</div><p>If we think of $\sigma$ as</p><div class="displaymath">
\begin{equation*}
1  \mapsto  3, \qquad
3  \mapsto  5, \qquad
5  \mapsto  2, \qquad
2  \mapsto  1,
\end{equation*}
</div><p>and $\tau$ as</p><div class="displaymath">
\begin{equation*}
2  \mapsto  5, \qquad
5  \mapsto  6, \qquad
6  \mapsto  2,
\end{equation*}
</div><p>then for $\sigma \tau$ remembering that we apply $\tau$ first and then $\sigma\text{,}$ it must be the case that</p><div class="displaymath">
\begin{equation*}
1  \mapsto  3, \qquad
3  \mapsto  5, \qquad
5  \mapsto  6, \qquad
6  \mapsto  2 \mapsto 1,
\end{equation*}
</div><p>or $\sigma \tau =  (1 3 5 6 )\text{.}$  If $\mu = (1634)\text{,}$ then $\sigma \mu = (1 6 5 2)(3 4)\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-797">Two cycles in $S_X\text{,}$ $\sigma = (a_1, a_2, \ldots, a_k )$ and $\tau = (b_1, b_2, \ldots, b_l )\text{,}$ are <dfn class="terminology">disjoint</dfn> if $a_i \neq b_j$ for all $i$ and $j\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-permute-cycles-disjoint"><h6 class="heading"><span class="type">Example</span><span class="codenumber">5.7</span></h6><p id="p-798">The cycles $(1 3 5)$ and $(2 7 )$ are disjoint; however, the cycles $(1 3 5)$ and $(3 4 7 )$ are not.  Calculating their products, we find that</p><div class="displaymath">
\begin{align*}
(1 3 5)(2 7 ) & = (1 3 5)(2 7 )\\
(1 3 5)(3 4 7 ) & = (1 3 4 7 5).
\end{align*}
</div><p>The product of two cycles that are not disjoint may reduce to something less complicated; the product of disjoint cycles cannot be simplified.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-14"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">5.8</span></h6><p id="p-799">Let $\sigma$ and $\tau$ be two disjoint cycles in $S_X\text{.}$ Then $\sigma \tau = \tau \sigma\text{.}$</p></article><article class="proof" id="proof-28"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-800">Let $\sigma = (a_1, a_2, \ldots, a_k )$ and $\tau = (b_1, b_2, \ldots, b_l )\text{.}$ We must show that $\sigma \tau(x) = \tau \sigma(x)$ for all $x \in X\text{.}$ If $x$ is neither in $\{ a_1, a_2, \ldots, a_k \}$ nor $\{b_1, b_2, \ldots, b_l  \}\text{,}$ then both $\sigma$ and $\tau$ fix $x\text{.}$ That is, $\sigma(x)=x$ and $\tau(x)=x\text{.}$ Hence,</p><div class="displaymath">
\begin{equation*}
\sigma \tau(x) = \sigma( \tau(x)) = \sigma(x) = x = \tau(x) = \tau( \sigma(x)) =  \tau \sigma(x).
\end{equation*}
</div><p><em class="emphasis">Do not forget that we are multiplying permutations right to left, which is the opposite of the order in which we usually multiply group elements.</em>  Now suppose that $x \in \{ a_1, a_2, \ldots, a_k \}\text{.}$ Then $\sigma( a_i ) = a_{(i \bmod k) + 1}\text{;}$ that is,</p><div class="displaymath">
\begin{align*}
a_1 & \mapsto  a_2\\
a_2 & \mapsto  a_3\\
& \vdots\\
a_{k-1} & \mapsto  a_k\\
a_k & \mapsto  a_1. 
\end{align*}
</div><p>However, $\tau(a_i) = a_i$ since $\sigma$ and $\tau$ are disjoint. Therefore,</p><div class="displaymath">
\begin{align*}
\sigma \tau(a_i) & = \sigma( \tau(a_i))\\
& = \sigma(a_i)\\
& = a_{(i \bmod k)+1}\\
& = \tau( a_{(i \bmod k)+1} )\\
& = \tau( \sigma(a_i) )\\
& = \tau \sigma(a_i).
\end{align*}
</div><p>Similarly, if $x \in \{b_1, b_2, \ldots, b_l  \}\text{,}$ then $\sigma$ and $\tau$ also commute.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-18"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">5.9</span></h6><p id="p-801">Every permutation in $S_n$ can be written as the product of disjoint cycles.</p></article><article class="proof" id="proof-29"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-802">We can assume that $X = \{ 1, 2, \ldots, n \}\text{.}$ If $\sigma \in S_n$ and we define $X_1$ to be $\{ \sigma(1), \sigma^2(1), \ldots \}\text{,}$ then the set $X_1$ is finite since $X$ is finite. Now let $i$ be the first integer in $X$ that is not in $X_1$ and define $X_2$ by $\{ \sigma(i), \sigma^2(i), \ldots \}\text{.}$ Again, $X_2$ is a finite set.  Continuing in this manner, we can define finite disjoint sets $X_3, X_4, \ldots\text{.}$ Since $X$ is a finite set, we are guaranteed that this process will end and there will be only a finite number of these sets, say $r\text{.}$ If $\sigma_i$ is the cycle defined by</p><div class="displaymath">
\begin{equation*}
\sigma_i( x ) = 
\left\{\begin{array}{ll}
\sigma( x ) & x \in X_i \\
x   & x \notin X_i, 
\end{array}\right.
\end{equation*}
</div><p>then $\sigma = \sigma_1 \sigma_2 \cdots \sigma_r\text{.}$ Since the sets $X_1, X_2, \ldots, X_r$ are disjoint, the cycles $\sigma_1, \sigma_2, \ldots, \sigma_r$ must also be disjoint.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-permute-cycle-products"><h6 class="heading"><span class="type">Example</span><span class="codenumber">5.10</span></h6><p id="p-803">Let</p><div class="displaymath">
\begin{align*}
\sigma & =
\begin{pmatrix}
1 & 2 & 3 & 4 & 5 & 6 \\
6 & 4 & 3 & 1 & 5 & 2
\end{pmatrix}\\
\tau & =
\begin{pmatrix}
1 & 2 & 3 & 4 & 5 & 6 \\
3 & 2 & 1 & 5 & 6 & 4
\end{pmatrix}.
\end{align*}
</div><p>Using cycle notation, we can write</p><div class="displaymath">
\begin{align*}
\sigma      & = (1624)\\
\tau        & = (13)(456)\\
\sigma \tau & = (1 3 6) ( 2 4 5)\\
\tau \sigma & = (1 4 3 )(2 5 6).
\end{align*}
</div></article></div>

<div class="mathbook-content"><article class="remark-like" id="remark-3"><h6 class="heading"><span class="type">Remark</span><span class="codenumber">5.11</span></h6><p id="p-804">From this point forward we will find it convenient to use cycle notation to represent permutations. When using cycle notation, we often denote the identity permutation by $(1)\text{.}$</p></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Transpositions"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Transpositions</span></h3><a href="section-permutation-definitions.ipynb#permute-subsection-transpositions" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-805">The simplest permutation is a cycle of length 2. Such cycles are called  <dfn class="terminology">transpositions</dfn>. Since</p><div class="displaymath">
\begin{equation*}
(a_1, a_2, \ldots, a_n ) = (a_1 a_n ) (a_1 a_{n-1} ) \cdots ( a_1 a_3 ) (a_1 a_2 ), 
\end{equation*}
</div><p>any cycle can be written as the product of transpositions, leading to the following proposition.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-15"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">5.12</span></h6><p id="p-806">Any permutation of a finite set containing at least two elements can be written as the product of transpositions.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-permute-transpositions"><h6 class="heading"><span class="type">Example</span><span class="codenumber">5.13</span></h6><p id="p-807">Consider the permutation</p><div class="displaymath">
\begin{equation*}
( 1 6 ) (2 5 3) = (1 6 )( 2 3 )( 2 5 ) = (1 6 )( 4 5 )(2 3 )( 4 5 )(2 5 ).
\end{equation*}
</div><p>As we can see, there is no unique way to represent permutation as the product of transpositions. For instance, we can write the identity permutation as $(1 2 )(1 2 )\text{,}$ as $(1 3 )(2 4 )(1 3 )( 2 4 )\text{,}$ and in many other ways. However, as it turns out, no permutation can be written as the product of both an even number of transpositions and an odd number of transpositions. For instance, we could represent the permutation $(1 6)$ by</p><div class="displaymath">
\begin{equation*}
(2 3 )(1 6)( 2 3)
\end{equation*}
</div><p>or by</p><div class="displaymath">
\begin{equation*}
(3 5) (1 6) (1 3) (1 6) (1 3) (3 5) (5 6),
\end{equation*}
</div><p>but $(1 6)$ will always be the product of an odd number of transpositions.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-identity-even-trans"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">5.14</span></h6><p id="p-808">If the identity is written as the product of $r$ transpositions,</p><div class="displaymath">
\begin{equation*}
\identity = \tau_1 \tau_2 \cdots \tau_r, 
\end{equation*}
</div><p>then $r$ is an even number.</p></article><article class="proof" id="proof-30"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-809">We will employ  induction  on $r\text{.}$  A transposition cannot be the identity; hence,   $r \gt 1\text{.}$ If $r=2\text{,}$ then we are done. Suppose that $r \gt 2\text{.}$ In this case the product of the last two transpositions, $\tau_{r-1} \tau_r\text{,}$ must be one of the following cases:</p><div class="displaymath">
\begin{align*}
(a b)(a b) & = \identity\\
(b c)(a b) & = (a c)(b c)\\
(c d)(a b) & = (a b)(c d)\\
(a c)(a b) & = (a b)(b c),
\end{align*}
</div><p>where $a\text{,}$ $b\text{,}$ $c\text{,}$ and $d$ are distinct.</p><p id="p-810">The first equation simply says that a transposition is its own inverse. If this case occurs, delete $\tau_{r-1} \tau_r$ from the product to obtain</p><div class="displaymath">
\begin{equation*}
\identity = \tau_1 \tau_2 \cdots \tau_{r - 3} \tau_{r - 2}.
\end{equation*}
</div><p>By induction $r - 2$ is even; hence, $r$ must be even.</p><p id="p-811">In each of the other three cases, we can replace $\tau_{r - 1} \tau_r$ with the right-hand side of the corresponding equation to obtain a new product of $r$ transpositions for the identity. In this new product the last occurrence of $a$ will be in the next-to-the-last transposition. We can continue this process with $\tau_{r - 2} \tau_{r - 1}$ to obtain either a product of $r - 2$ transpositions or a new product of $r$ transpositions where the last occurrence of $a$ is in $\tau_{r - 2}\text{.}$ If the identity is the product of $r - 2$ transpositions, then again we are done, by our induction hypothesis; otherwise, we will repeat the procedure with $\tau_{r - 3} \tau_{r - 2}\text{.}$</p><p id="p-812">At some point either we will have two adjacent, identical transpositions canceling each other out or $a$ will be shuffled so that it will appear only in the first transposition. However, the latter case cannot occur, because the identity would not fix $a$ in this instance. Therefore, the identity permutation must be the product of $r-2$ transpositions and, again by our induction hypothesis, we are done.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-even-and-odd"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">5.15</span></h6><p id="p-813">If a permutation $\sigma$ can be expressed as the product of an even number of transpositions, then any other product of transpositions equaling $\sigma$ must also contain an even number of transpositions. Similarly, if $\sigma$ can be expressed as the product of an odd number of transpositions, then any other product of transpositions equaling $\sigma$ must also contain an odd number of transpositions.</p></article><article class="proof" id="proof-31"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-814">Suppose that</p><div class="displaymath">
\begin{equation*}
\sigma = \sigma_1 \sigma_2 \cdots \sigma_m = \tau_1 \tau_2 \cdots \tau_n, 
\end{equation*}
</div><p>where $m$ is even. We must show that $n$ is also an even number.  The inverse of $\sigma$ is $\sigma_m \cdots \sigma_1\text{.}$ Since</p><div class="displaymath">
\begin{equation*}
\identity = \sigma \sigma_m \cdots \sigma_1 = \tau_1  \cdots \tau_n \sigma_m \cdots \sigma_1, 
\end{equation*}
</div><p>$n$ must be even by Lemma <a href="section-permutation-definitions.ipynb#lemma-identity-even-trans" class="xref" alt="Lemma 5.14 " title="Lemma 5.14 ">5.14</a>. The proof for the case in which $\sigma$ can be expressed as an odd number of transpositions is left as an exercise.</p></article></div>

<div class="mathbook-content"><p id="p-815">In light of Theorem <a href="section-permutation-definitions.ipynb#theorem-even-and-odd" class="xref" alt="Theorem 5.15 " title="Theorem 5.15 ">5.15</a>, we define a permutation to be <dfn class="terminology">even</dfn> if it can be expressed as an even number of transpositions and <dfn class="terminology">odd</dfn> if it can be expressed as an odd number of transpositions.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  The Alternating Groups"><span class="type">Subsection</span><span class="codenumber" /><span class="title">The Alternating Groups</span></h3><a href="section-permutation-definitions.ipynb#permute-subsection-alternating-groups" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-816">One of the most important subgroups of $S_n$ is the set of all even permutations, $A_n\text{.}$   The group $A_n$ is called the <dfn class="terminology">alternating group on $n$ letters</dfn>.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-20"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">5.16</span></h6><p id="p-817">The set $A_n$ is a subgroup of $S_n\text{.}$</p></article><article class="proof" id="proof-32"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-818">Since the product of two even permutations must also be an even permutation, $A_n$ is closed.  The identity is an even permutation and therefore is in $A_n\text{.}$ If $\sigma$ is an even permutation, then</p><div class="displaymath">
\begin{equation*}
\sigma = \sigma_1 \sigma_2 \cdots \sigma_r,
\end{equation*}
</div><p>where $\sigma_i$ is a transposition and $r$ is even. Since the inverse of any transposition is itself,</p><div class="displaymath">
\begin{equation*}
\sigma^{-1} = \sigma_r \sigma_{r-1} \cdots \sigma_1
\end{equation*}
</div><p>is also in $A_n\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-16"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">5.17</span></h6><p id="p-819">The number of even permutations in $S_n\text{,}$ $n \geq 2\text{,}$ is equal to the number of odd permutations; hence, the order of $A_n$ is $n!/2\text{.}$</p></article><article class="proof" id="proof-33"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-820">Let $A_n$ be the set of even permutations in $S_n$ and $B_n$ be the set of odd permutations.  If we can show that there is a bijection between these sets, they must contain the same number of elements. Fix a transposition $\sigma$ in $S_n\text{.}$  Since $n \geq 2\text{,}$ such a $\sigma$ exists.  Define</p><div class="displaymath">
\begin{equation*}
\lambda_{\sigma} : A_n \rightarrow B_n
\end{equation*}
</div><p>by</p><div class="displaymath">
\begin{equation*}
\lambda_{\sigma} ( \tau ) = \sigma  \tau .
\end{equation*}
</div><p>Suppose that $\lambda_{\sigma} ( \tau ) = \lambda_{\sigma} ( \mu )\text{.}$ Then $\sigma  \tau = \sigma  \mu$ and so</p><div class="displaymath">
\begin{equation*}
\tau = \sigma^{-1} \sigma \tau = \sigma^{-1} \sigma \mu = \mu. 
\end{equation*}
</div><p>Therefore, $\lambda_{\sigma}$ is one-to-one.  We will leave the proof that $\lambda_{\sigma}$ is surjective to the reader.</p></article></div>

<div class="mathbook-content"><p id="p-822">Lagrange first thought of permutations as functions from a set to itself, but it was Cauchy who developed the basic theorems and notation for permutations. He was the first to use cycle notation. Augustin-Louis Cauchy (1789–1857) was born in Paris at the height of the French Revolution. His family soon left Paris for the village of Arcueil to escape the Reign of Terror. One of the family's neighbors there was Pierre-Simon Laplace (1749–1827), who encouraged him to seek a career in mathematics. Cauchy began his career as a mathematician by solving a problem in geometry given to him by Lagrange. Cauchy wrote over 800 papers on such diverse topics as differential equations, finite groups, applied mathematics, and complex analysis. He was one of the mathematicians  responsible for making calculus rigorous. Perhaps more theorems and concepts in mathematics have the name Cauchy attached to them than that of any other mathematician.</p></div>
