<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 11.1 Group Homomorphisms"><span class="type">Section</span><span class="codenumber">11.1</span><span class="title">Group Homomorphisms</span></h2><a href="section-group-homomorphisms.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-1706">A <dfn class="terminology">homomorphism</dfn> between groups $(G, \cdot)$ and $(H, \circ)$ is a map $\phi :G \rightarrow H$ such that</p><div class="displaymath">
\begin{equation*}
\phi( g_1 \cdot g_2 ) = \phi( g_1 ) \circ \phi( g_2 )
\end{equation*}
</div><p>for $g_1, g_2 \in G\text{.}$ The range of $\phi$ in $H$ is called the <dfn class="terminology">homomorphic image</dfn> of $\phi\text{.}$</p></div>

<div class="mathbook-content"><p id="p-1707">Two groups are related in the strongest possible way if they are isomorphic; however, a weaker relationship may exist between two groups.  For example, the symmetric group $S_n$ and the group ${\mathbb Z}_2$ are related by the fact that $S_n$ can be divided into even and odd permutations that exhibit a group structure like that ${\mathbb Z}_2\text{,}$ as shown in the following multiplication table.</p></div>

<div class="mathbook-content"><div class="sidebyside" id="sidebyside-8"><div class="sbsrow" style="margin-left:0%;margin-right:0%;"><div class="sbspanel" style="width:100%;justify-content:flex-start;"><div class="displaymath" id="p-1708">
\begin{equation*}
\begin{array}{c|cc}
& \text{even} & \text{odd} \\ \hline
\text{even} & \text{even} & \text{odd} \\
\text{odd}  & \text{odd}  & \text{even}
\end{array}
\end{equation*}
</div></div></div></div></div>

<div class="mathbook-content"><p id="p-1709">We use homomorphisms to study relationships such as the one we have just described.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-homomorph-zn"><h6 class="heading"><span class="type">Example</span><span class="codenumber">11.1</span></h6><p id="p-1710">Let $G$ be a group and $g \in G\text{.}$ Define a map $\phi : {\mathbb Z} \rightarrow G$ by $\phi( n ) = g^n\text{.}$ Then $\phi$ is a group homomorphism, since</p><div class="displaymath">
\begin{equation*}
\phi( m + n ) = g^{ m + n} = g^m g^n = \phi( m ) \phi( n ).
\end{equation*}
</div><p>This homomorphism maps ${\mathbb Z}$ onto the cyclic subgroup of $G$ generated by $g\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-homomorph-gl2"><h6 class="heading"><span class="type">Example</span><span class="codenumber">11.2</span></h6><p id="p-1711">Let $G = GL_2( {\mathbb R })\text{.}$ If</p><div class="displaymath">
\begin{equation*}
A =
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
\end{equation*}
</div><p>is in $G\text{,}$ then the determinant is  nonzero; that is, $\det(A) = ad - bc \neq 0\text{.}$  Also, for any two elements $A$ and $B$ in $G\text{,}$ $\det(AB) = \det(A) \det(B)\text{.}$ Using the determinant, we can define a homomorphism $\phi : GL_2( {\mathbb R }) \rightarrow {\mathbb R}^\ast$ by $A \mapsto \det(A)\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-homomorph-circle-group"><h6 class="heading"><span class="type">Example</span><span class="codenumber">11.3</span></h6><p id="p-1712">Recall that the circle group ${ \mathbb T}$ consists of all complex numbers $z$ such that $|z|=1\text{.}$ We can define a homomorphism $\phi$ from the additive group of real numbers ${\mathbb R}$ to ${\mathbb T}$ by $\phi : \theta \mapsto \cos \theta + i \sin \theta\text{.}$ Indeed,</p><div class="displaymath">
\begin{align*}
\phi( \alpha + \beta ) & = \cos( \alpha + \beta ) + i \sin( \alpha + \beta )\\
& = (\cos \alpha \cos \beta - \sin \alpha \sin \beta)  + i( \sin \alpha  \cos \beta + \cos \alpha \sin \beta )\\
& = (\cos \alpha + i \sin \alpha )(\cos \beta + i \sin \beta)\\
& = \phi( \alpha ) \phi( \beta ).
\end{align*}
</div><p>Geometrically, we are simply wrapping the real line around the circle in a group-theoretic fashion.</p></article></div>

<div class="mathbook-content"><p id="p-1713">The following proposition lists some basic properties of group homomorphisms.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-homorphism-subgroup"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">11.4</span></h6><p id="p-1714">Let $\phi : G_1 \rightarrow G_2$ be a homomorphism of groups. Then </p><ol class="decimal"><li id="li-405"><p id="p-1715">If $e$ is the identity of $G_1\text{,}$ then $\phi( e)$ is the identity of $G_2\text{;}$</p></li><li id="li-406"><p id="p-1716">For any element $g \in G_1\text{,}$ $\phi( g^{-1}) = [\phi( g )]^{- 1}\text{;}$</p></li><li id="li-407"><p id="p-1717">If $H_1$ is a subgroup of $G_1\text{,}$ then $\phi( H_1 )$ is a subgroup of $G_2\text{;}$</p></li><li id="li-408"><p id="p-1718">If $H_2$ is a  subgroup of $G_2\text{,}$ then $\phi^{-1}(H_2) = \{ g \in G _1: \phi(g) \in H_2 \}$ is a subgroup of $G_1\text{.}$ Furthermore, if $H_2$ is normal in $G_2\text{,}$ then $\phi^{-1}(H_2)$ is normal in $G_1\text{.}$</p></li></ol></article><article class="proof" id="proof-73"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1719">(1) Suppose that $e$ and $e'$ are the identities of $G_1$ and $G_2\text{,}$ respectively; then</p><div class="displaymath">
\begin{equation*}
e' \phi(e) = \phi(e) = \phi(e e) = \phi(e) \phi(e).
\end{equation*}
</div><p>By cancellation, $\phi(e) = e'\text{.}$</p><p id="p-1720">(2) This statement follows from the fact that</p><div class="displaymath">
\begin{equation*}
\phi( g^{-1}) \phi(g) = \phi(g^{-1} g) = \phi(e) = e'.
\end{equation*}
</div><p id="p-1721">(3) The set $\phi(H_1)$ is nonempty since the identity of $G_2$ is in $\phi(H_1)\text{.}$ Suppose that $H_1$ is a subgroup of $G_1$ and let $x$ and $y$ be in $\phi(H_1)\text{.}$ There exist elements $a, b \in H_1$ such that $\phi(a) = x$ and $\phi(b)=y\text{.}$ Since</p><div class="displaymath">
\begin{equation*}
xy^{-1} = \phi(a)[ \phi(b)]^{-1} = \phi(a b^{-1} ) \in \phi(H_1),
\end{equation*}
</div><p>$\phi(H_1)$ is a subgroup of $G_2$ by Proposition <a href="section-subgroups.ipynb#proposition-subgroup" class="xref" alt="Proposition 3.31 " title="Proposition 3.31 ">3.31</a>.</p><p id="p-1722">(4) Let $H_2$ be a subgroup of $G_2$ and define $H_1$ to be $\phi^{-1}(H_2)\text{;}$ that is, $H_1$ is the set of all $g \in G_1$ such that $\phi(g) \in H_2\text{.}$  The identity is in $H_1$ since $\phi(e) = e'\text{.}$ If $a$ and $b$ are in $H_1\text{,}$ then $\phi(ab^{-1}) = \phi(a)[ \phi(b) ]^{-1}$ is in $H_2$ since $H_2$ is a subgroup of $G_2\text{.}$  Therefore, $ab^{-1} \in H_1$ and $H_1$ is a subgroup of $G_1\text{.}$ If $H_2$ is normal in $G_2\text{,}$ we must show that $g^{-1} h g \in H_1$ for $h \in H_1$ and $g \in G_1\text{.}$ But</p><div class="displaymath">
\begin{equation*}
\phi( g^{-1} h g) = [ \phi(g) ]^{-1} \phi( h ) \phi( g ) \in H_2,
\end{equation*}
</div><p>since $H_2$ is a normal subgroup of $G_2\text{.}$  Therefore, $g^{-1}hg \in H_1\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1723">Let $\phi : G \rightarrow H$ be a group homomorphism and suppose that $e$ is the identity of $H\text{.}$ By Proposition <a href="section-group-homomorphisms.ipynb#proposition-homorphism-subgroup" class="xref" alt="Proposition 11.4 " title="Proposition 11.4 ">11.4</a>, $\phi^{-1} ( \{ e \} )$ is a subgroup of $G\text{.}$ This subgroup is called the <dfn class="terminology">kernel</dfn> of $\phi$ and will be denoted by $\ker \phi\text{.}$  In fact, this subgroup is a normal subgroup of $G$ since the trivial subgroup is normal in $H\text{.}$  We state this result in the following theorem, which says that with every homomorphism of groups we can naturally associate a normal subgroup.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-53"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">11.5</span></h6><p id="p-1724">Let $\phi : G \rightarrow H$ be a group homomorphism. Then the kernel of $\phi$ is a normal subgroup of $G\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-homomorph-g2-to-r"><h6 class="heading"><span class="type">Example</span><span class="codenumber">11.6</span></h6><p id="p-1725">Let us examine the homomorphism $\phi : GL_2( {\mathbb R }) \rightarrow {\mathbb R}^\ast$ defined by $A \mapsto \det( A )\text{.}$ Since 1 is the identity of ${\mathbb R}^\ast\text{,}$ the kernel of this homomorphism is all $2 \times 2$ matrices having determinant one. That is, $\ker \phi = SL_2( {\mathbb R })\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-homomorph-kernel"><h6 class="heading"><span class="type">Example</span><span class="codenumber">11.7</span></h6><p id="p-1726">The kernel of the group homomorphism $\phi : {\mathbb R} \rightarrow {\mathbb C}^\ast$ defined by $\phi( \theta ) = \cos \theta + i \sin \theta$ is $\{ 2 \pi n : n \in {\mathbb Z} \}\text{.}$ Notice that $\ker \phi \cong {\mathbb Z}\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-homomorph-z7"><h6 class="heading"><span class="type">Example</span><span class="codenumber">11.8</span></h6><p id="p-1727">Suppose that we wish to determine all possible homomorphisms $\phi$ from ${\mathbb Z}_7$ to  ${\mathbb Z}_{12}\text{.}$ Since the kernel of $\phi$ must be a subgroup of  ${\mathbb Z}_7\text{,}$ there are only two possible kernels, $\{ 0 \}$ and all of ${\mathbb Z}_7\text{.}$  The image of a subgroup of ${\mathbb Z}_7$ must be a subgroup of ${\mathbb Z}_{12}\text{.}$ Hence, there is no injective homomorphism; otherwise, ${\mathbb Z}_{12}$ would have a subgroup of order 7, which is impossible. Consequently, the only possible homomorphism from ${\mathbb Z}_7$ to  ${\mathbb Z}_{12}$ is the one mapping all elements to zero.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-homomorph-gn"><h6 class="heading"><span class="type">Example</span><span class="codenumber">11.9</span></h6><p id="p-1728">Let $G$ be a group. Suppose that  $g \in G$ and $\phi$ is the homomorphism from ${\mathbb Z}$ to $G$ given by $\phi( n ) = g^n\text{.}$ If the order of $g$ is infinite, then the kernel of this homomorphism is $\{ 0 \}$ since $\phi$ maps ${\mathbb Z}$ onto the cyclic subgroup of $G$ generated by $g\text{.}$ However, if the order of $g$ is finite, say $n\text{,}$ then the kernel of $\phi$ is $n {\mathbb Z}\text{.}$</p></article></div>
