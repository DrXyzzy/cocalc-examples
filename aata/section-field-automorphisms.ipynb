<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-3583">Our first task is to establish a link between group theory and field theory by examining automorphisms of fields.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-49"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">23.1</span></h6><p id="p-3584">The set of all automorphisms of a field $F$ is a group under composition of functions.</p></article><article class="proof" id="proof-190"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3585">If $\sigma$ and $\tau$ are automorphisms of $F\text{,}$ then so are $\sigma \tau$ and $\sigma^{-1}\text{.}$  The identity is certainly an automorphism; hence, the set of all automorphisms of a field $F$ is indeed a group.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-50"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">23.2</span></h6><p id="p-3586">Let $E$ be a field extension of $F\text{.}$  Then the set of all automorphisms of $E$ that fix $F$ elementwise is a group; that is, the set of all automorphisms $\sigma : E \rightarrow E$ such that $\sigma( \alpha ) = \alpha$ for all $\alpha \in F$ is a group.</p></article><article class="proof" id="proof-191"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3587">We need only show that the set of automorphisms of $E$ that fix $F$ elementwise is a subgroup of the group of all automorphisms of $E\text{.}$  Let $\sigma$ and $\tau$ be two automorphisms of $E$ such that $\sigma( \alpha ) = \alpha$ and $\tau( \alpha ) = \alpha$ for all $\alpha \in F\text{.}$  Then $\sigma \tau( \alpha ) = \sigma( \alpha) = \alpha$ and  $\sigma^{-1}( \alpha ) = \alpha\text{.}$  Since the identity fixes every  element of $E\text{,}$ the set of automorphisms of $E$ that leave elements of  $F$ fixed is a subgroup of the entire group of automorphisms of $E\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3588">Let $E$ be a field extension of $F\text{.}$  We will denote the full group of automorphisms of $E$ by $\aut(E)\text{.}$  We define the <dfn class="terminology">Galois group</dfn> of $E$ over $F$ to be the group of automorphisms of $E$ that fix $F$ elementwise; that is,</p><div class="displaymath">
\begin{equation*}
G(E/F) = \{ \sigma \in \aut(E) : \sigma(\alpha) = \alpha \text{ for all } \alpha \in F  \}.
\end{equation*}
</div><p>If $f(x)$ is a polynomial in $F[x]$ and $E$ is the splitting field of $f(x)$ over $F\text{,}$ then we define the Galois group of $f(x)$ to be $G(E/F)\text{.}$ </p></div>

<div class="mathbook-content"><article class="example-like" id="example-galois-complex-conj"><h6 class="heading"><span class="type">Example</span><span class="codenumber">23.3</span></h6><p id="p-3589">Complex conjugation, defined by $\sigma : a + bi \mapsto a - bi\text{,}$ is an automorphism of the complex numbers.  Since</p><div class="displaymath">
\begin{equation*}
\sigma(a) = \sigma(a + 0i) = a - 0i = a,
\end{equation*}
</div><p>the automorphism defined by complex conjugation must be in $G( {\mathbb C} / {\mathbb R} )\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-galois-q-sqrt5"><h6 class="heading"><span class="type">Example</span><span class="codenumber">23.4</span></h6><p id="p-3590">Consider the fields ${\mathbb Q} \subset {\mathbb Q}(\sqrt{5}\, ) \subset {\mathbb Q}( \sqrt{3}, \sqrt{5}\, )\text{.}$  Then for $a, b \in {\mathbb Q}( \sqrt{5}\, )\text{,}$</p><div class="displaymath">
\begin{equation*}
\sigma( a + b \sqrt{3}\, ) = a - b \sqrt{3}
\end{equation*}
</div><p>is an automorphism of ${\mathbb Q}(\sqrt{3}, \sqrt{5}\, )$ leaving ${\mathbb Q}( \sqrt{5}\, )$ fixed.  Similarly,</p><div class="displaymath">
\begin{equation*}
\tau( a + b \sqrt{5}\, ) = a - b \sqrt{5}
\end{equation*}
</div><p>is an automorphism of ${\mathbb Q}(\sqrt{3}, \sqrt{5}\, )$ leaving ${\mathbb Q}( \sqrt{3}\, )$ fixed. The automorphism $\mu = \sigma \tau$ moves both $\sqrt{3}$ and $\sqrt{5}\text{.}$  It will soon be clear that $\{ \identity, \sigma, \tau, \mu  \}$ is the Galois group of ${\mathbb Q}(\sqrt{3}, \sqrt{5}\, )$ over ${\mathbb Q}\text{.}$ The following table shows that this group is isomorphic to ${\mathbb Z}_2 \times {\mathbb Z}_2\text{.}$</p><div class="displaymath">
\begin{equation*}
\begin{array}{c|cccc}
& \identity & \sigma & \tau & \mu \\
\hline
\identity  & \identity & \sigma & \tau & \mu \\
\sigma & \sigma & \identity & \mu & \tau \\
\tau & \tau & \mu & \identity & \sigma \\
\mu & \mu & \tau & \sigma & \identity
\end{array}
\end{equation*}
</div><p>We may also regard the field ${\mathbb Q}( \sqrt{3}, \sqrt{5}\, )$ as a vector space over ${\mathbb Q}$ that has basis $\{ 1, \sqrt{3}, \sqrt{5}, \sqrt{15}\, \}\text{.}$  It is no coincidence that $|G( {\mathbb Q}( \sqrt{3}, \sqrt{5}\, ) /{\mathbb Q})| = [{\mathbb Q}(\sqrt{3}, \sqrt{5}\, ):{\mathbb Q})] = 4\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-roots-permute"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">23.5</span></h6><p id="p-3591">Let $E$ be a field extension of $F$ and $f(x)$ be a polynomial in $F[x]\text{.}$ Then any automorphism in $G(E/F)$ defines a permutation of the roots of $f(x)$ that lie in $E\text{.}$</p></article><article class="proof" id="proof-192"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3592">Let</p><div class="displaymath">
\begin{equation*}
f(x) = a_0 + a_1 x + a_2 x^2 + \cdots + a_n x^n
\end{equation*}
</div><p>and suppose that $\alpha \in E$ is a zero of $f(x)\text{.}$ Then for $\sigma \in G(E/F)\text{,}$</p><div class="displaymath">
\begin{align*}
0 & = \sigma( 0 )\\
& = \sigma( f( \alpha ))\\
& = \sigma(a_0 + a_1\alpha + a_2 \alpha^2 + \cdots + a_n \alpha^n)\\
& = a_0 + a_1 \sigma(\alpha) + a_2 [\sigma(\alpha)]^2 + \cdots + a_n [\sigma(\alpha)]^n;
\end{align*}
</div><p>therefore, $\sigma( \alpha )$ is also a zero of $f(x)\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3593">Let $E$ be an algebraic extension of a field $F\text{.}$  Two elements $\alpha, \beta \in E$ are <dfn class="terminology">conjugate</dfn> over $F$ if they have the same minimal polynomial. For example, in the field ${\mathbb Q}( \sqrt{2}\, )$ the elements $\sqrt{2}$ and $-\sqrt{2}$ are conjugate over ${\mathbb Q}$ since they are both roots of the irreducible polynomial $x^2 - 2\text{.}$</p></div>

<div class="mathbook-content"><p id="p-3594">A converse of the last proposition exists. The proof follows directly from LemmaÂ <a href="section-splitting-fields.ipynb#lemma-field-isomorphism" class="xref" alt="Lemma 21.32 " title="Lemma 21.32 ">21.32</a>.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-52"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">23.6</span></h6><p id="p-3595">If $\alpha$ and $\beta$ are conjugate over $F\text{,}$ there exists an isomorphism $\sigma : F( \alpha ) \rightarrow F( \beta )$ such that $\sigma$ is the identity when restricted to $F\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-extension-order"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">23.7</span></h6><p id="p-3596">Let $f(x)$ be a polynomial in $F[x]$ and suppose that $E$ is the splitting field for $f(x)$ over $F\text{.}$  If $f(x)$ has no repeated roots, then</p><div class="displaymath">
\begin{equation*}
|G(E/F)| = [E:F].
\end{equation*}
</div></article><article class="proof" id="proof-193"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3597">We will use mathematical induction on the degree of $f(x)\text{.}$ If the degree of $f(x)$ is 0 or 1, then $E = F$ and there is nothing to show. Assume that the result holds for all polynomials of degree $k$ with $0 \leq k \lt n\text{.}$  Suppose that the degree of $f(x)$ is $n\text{.}$  Let $p(x)$ be an irreducible factor of $f(x)$ of degree $r\text{.}$ Since all of the roots of $p(x)$ are in $E\text{,}$ we can choose one of these roots, say $\alpha\text{,}$ so that $F \subset F( \alpha ) \subset E\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
[E: F(\alpha)] = n/r \quad \text{and} \quad [F(\alpha): F] = r.
\end{equation*}
</div><p>If $\beta$ is any other root of $p(x)\text{,}$ then $F \subset F( \beta ) \subset E\text{.}$ By LemmaÂ <a href="section-splitting-fields.ipynb#lemma-field-isomorphism" class="xref" alt="Lemma 21.32 " title="Lemma 21.32 ">21.32</a>, there exists a unique isomorphism $\sigma: F( \alpha ) \rightarrow F( \beta )$ for each such $\beta$ that fixes $F$ elementwise.  Since $E$ is a splitting field of $p(x)\text{,}$ there are exactly $r$ such isomorphisms. For each of these automorphisms, we can use our induction hypothesis on $[E: F(\alpha)] = n/r \lt n$ to conclude that</p><div class="displaymath">
\begin{equation*}
|G(E/F(\alpha))| = [E:F(\alpha)].
\end{equation*}
</div><p>Consequently, there are</p><div class="displaymath">
\begin{equation*}
[E:F] = [E:F(\alpha)] [F( \alpha):F] = n
\end{equation*}
</div><p>possible automorphisms of $E$ that fix $F\text{,}$ or $|G(E/F)| = [E:F]\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-37"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">23.8</span></h6><p id="p-3598">Let $F$ be a finite field with a finite extension $E$ such that $[E:F]=k\text{.}$ Then $G(E/F)$ is cyclic of order $k\text{.}$</p></article><article class="proof" id="proof-194"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3599">Let $p$ be the characteristic of $E$ and $F$ and assume that the orders of $E$ and $F$ are $p^m$ and $p^n\text{,}$ respectively.  Then $nk = m\text{.}$  We can also assume that $E$ is the splitting field of $x^{p^m} - x$ over a subfield of order $p\text{.}$  Therefore, $E$ must also be the splitting field of $x^{p^m} - x$ over $F\text{.}$  Applying TheoremÂ <a href="section-field-automorphisms.ipynb#theorem-extension-order" class="xref" alt="Theorem 23.7 " title="Theorem 23.7 ">23.7</a>, we find that $|G(E/F)| = k\text{.}$</p><p id="p-3600">To prove that $G(E/F)$ is cyclic, we must find a generator for $G(E/F)\text{.}$  Let $\sigma : E \rightarrow E$ be defined by $\sigma(\alpha) = \alpha^{p^n}\text{.}$  We claim that $\sigma$ is the element in $G(E/F)$ that we are seeking.  We first need to show that $\sigma$ is in $\aut(E)\text{.}$  If $\alpha$ and $\beta$ are in $E\text{,}$</p><div class="displaymath">
\begin{equation*}
\sigma(\alpha + \beta) = (\alpha + \beta)^{p^n} = \alpha^{p^n} + \beta^{p^n} = \sigma(\alpha) + \sigma(\beta)
\end{equation*}
</div><p>by LemmaÂ <a href="section-finite-field.ipynb#lemma-freshmans-dream" class="xref" alt="Lemma 22.3 Freshman's Dream" title="Lemma 22.3 Freshman's Dream">22.3</a>  Also, it is easy to show that $\sigma(\alpha \beta) = \sigma( \alpha ) \sigma( \beta )\text{.}$ Since $\sigma$ is a nonzero homomorphism of fields, it must be injective.  It must also be onto, since $E$ is a finite field.  We know that $\sigma$ must be in $G(E/F)\text{,}$ since $F$ is the splitting field of $x^{p^n} - x$ over the base field of order $p\text{.}$ This means that $\sigma$ leaves every element in $F$ fixed.  Finally, we must show that the order of $\sigma$ is $k\text{.}$ By TheoremÂ <a href="section-field-automorphisms.ipynb#theorem-extension-order" class="xref" alt="Theorem 23.7 " title="Theorem 23.7 ">23.7</a>, we know that</p><div class="displaymath">
\begin{equation*}
\sigma^k( \alpha ) = \alpha^{p^{nk}} = \alpha^{p^m} = \alpha
\end{equation*}
</div><p>is the identity of $G( E/F)\text{.}$  However, $\sigma^r$ cannot be the identity for $1 \leq r \lt k\text{;}$ otherwise, $x^{p^{nr}} - x$ would have $p^m$ roots, which is impossible.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-galois-group-sqrt3-5"><h6 class="heading"><span class="type">Example</span><span class="codenumber">23.9</span></h6><p id="p-3601">We can now confirm that the Galois group of ${\mathbb Q}( \sqrt{3}, \sqrt{5}\, )$ over ${\mathbb Q}$ in ExampleÂ <a href="section-field-automorphisms.ipynb#example-galois-q-sqrt5" class="xref" alt="Example 23.4 " title="Example 23.4 ">23.4</a> is indeed isomorphic to ${\mathbb Z}_2 \times {\mathbb Z}_2\text{.}$  Certainly the group $H = \{ \identity, \sigma, \tau, \mu \}$ is a subgroup of $G({\mathbb Q}( \sqrt{3}, \sqrt{5}\, )/{\mathbb Q})\text{;}$ however,  $H$ must be all of $G({\mathbb Q}( \sqrt{3}, \sqrt{5}\, )/{\mathbb Q})\text{,}$ since</p><div class="displaymath">
\begin{equation*}
|H| = [{\mathbb Q}( \sqrt{3}, \sqrt{5}\, ):{\mathbb Q}] = |G({\mathbb Q}( \sqrt{3}, \sqrt{5}\, )/{\mathbb Q})| = 4.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="example-like" id="example-galois-group-x4"><h6 class="heading"><span class="type">Example</span><span class="codenumber">23.10</span></h6><p id="p-3602">Let us compute the Galois group of</p><div class="displaymath">
\begin{equation*}
f(x) = x^4 + x^3 + x^2 + x + 1
\end{equation*}
</div><p>over ${\mathbb Q}\text{.}$ We know that $f(x)$ is irreducible by ExerciseÂ <a href="exercises-poly.ipynb#exercise-poly-cyclotomic-polynomials" class="xref" alt="Exercise 17.4.20 Cyclotomic Polynomials" title="Exercise 17.4.20 Cyclotomic Polynomials">17.4.20</a> in ChapterÂ <a href="poly.ipynb" class="xref" alt="Chapter 17 Polynomials" title="Chapter 17 Polynomials">17</a>.  Furthermore, since $(x -1)f(x) = x^5 -1\text{,}$ we can use DeMoivre's Theorem to determine that the roots of $f(x)$ are $\omega^i\text{,}$  where  $i = 1, \ldots, 4$ and</p><div class="displaymath">
\begin{equation*}
\omega = \cos(2 \pi / 5 ) + i \sin(2 \pi / 5 ).
\end{equation*}
</div><p>Hence, the splitting field of $f(x)$ must be ${\mathbb Q}(\omega)\text{.}$  We can define automorphisms $\sigma_i$ of ${\mathbb Q}(\omega )$ by $\sigma_i( \omega ) = \omega^i$ for $i = 1, \ldots, 4\text{.}$  It is easy to check that these are indeed distinct automorphisms in $G( {\mathbb Q}( \omega) / {\mathbb Q} )\text{.}$  Since</p><div class="displaymath">
\begin{equation*}
[{\mathbb Q}( \omega) : {\mathbb Q}] = | G( {\mathbb Q}( \omega) / {\mathbb Q})| = 4,
\end{equation*}
</div><p>the $\sigma_i$'s must be all of $G( {\mathbb Q}( \omega) / {\mathbb Q} )\text{.}$ Therefore, $G({\mathbb Q}( \omega) / {\mathbb Q})\cong {\mathbb Z}_4$ since $\omega$ is a generator for the Galois group.</p></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Separable Extensions"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Separable Extensions</span></h3><a href="section-field-automorphisms.ipynb#galois-subsection-separable-extensions" class="permalink">Â¶</a></div>

<div class="mathbook-content"><p id="p-3603">Many of the results that we have just proven depend on the fact that a polynomial $f(x)$ in $F[x]$ has no repeated roots in its splitting field. It is evident that we need to know exactly when a polynomial factors into distinct linear factors in its splitting field. Let $E$ be the splitting field of a polynomial $f(x)$ in $F[x]\text{.}$ Suppose that $f(x)$ factors over $E$ as</p><div class="displaymath">
\begin{equation*}
f(x) = (x - \alpha_1)^{n_1} (x - \alpha_2)^{n_2} \cdots (x - \alpha_r)^{n_r} = \prod_{i = 1}^{r} (x - \alpha_i)^{n_i}.
\end{equation*}
</div><p>We define the <dfn class="terminology">multiplicity</dfn> of a root $\alpha_i$ of $f(x)$ to be $n_i\text{.}$  A root with multiplicity 1 is called a <dfn class="terminology">simple root</dfn>. Recall that a polynomial $f(x) \in F[x]$ of degree $n$ is <dfn class="terminology">separable</dfn> if it has $n$ distinct roots in its splitting field $E\text{.}$ Equivalently, $f(x)$ is separable if it factors into distinct linear factors over $E[x]\text{.}$ An extension $E$ of $F$ is a <dfn class="terminology">separable extension</dfn> of $F$ if every element in $E$ is the root of a separable polynomial in $F[x]\text{.}$ Also recall that $f(x)$ is separable if and only if $\gcd( f(x), f'(x)) = 1$ (LemmaÂ <a href="section-finite-field.ipynb#lemma-separable-derivative" class="xref" alt="Lemma 22.5 " title="Lemma 22.5 ">22.5</a>).</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-53"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">23.11</span></h6><p id="p-3604">Let $f(x)$ be an irreducible polynomial over $F\text{.}$ If the characteristic of $F$ is $0\text{,}$ then $f(x)$ is separable.  If the characteristic of $F$ is $p$ and $f(x) \neq g(x^p)$ for some $g(x)$ in $F[x]\text{,}$ then $f(x)$ is also separable.</p></article><article class="proof" id="proof-195"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3605">First assume that $\chr F = 0\text{.}$ Since $\deg f'(x) \lt \deg f(x)$ and $f(x)$ is irreducible, the only way $\gcd( f(x), f'(x)) \neq 1$ is if $f'(x)$ is the zero polynomial; however, this is impossible in a field of characteristic zero. If $\chr F = p\text{,}$ then $f'(x)$ can be the zero polynomial if every coefficient of $f'(x)$ is a multiple of $p\text{.}$  This can happen only if we have a polynomial of the form $f(x) = a_0 + a_1 x^p + a_2 x^{2p} + \cdots + a_n x^{np}\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3606">Certainly extensions of a field $F$ of the form $F(\alpha)$ are some of the easiest to study and understand.  Given a field extension $E$ of $F\text{,}$ the obvious question to ask is when it is possible to find an element $\alpha \in E$ such that $E = F( \alpha )\text{.}$ In this case, $\alpha$ is called a <dfn class="terminology">primitive element</dfn>. We already know that primitive elements exist for certain extensions. For example,</p><div class="displaymath">
\begin{equation*}
{\mathbb Q}( \sqrt{3}, \sqrt{5}\, ) = {\mathbb Q}( \sqrt{3} + \sqrt{5}\, )
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
{\mathbb Q}( \sqrt[3]{5}, \sqrt{5}\, i ) = {\mathbb Q}( \sqrt[6]{5}\, i ).
\end{equation*}
</div><p>CorollaryÂ <a href="section-finite-field.ipynb#corollary-finite-extension-finite-field" class="xref" alt="Corollary 22.12 " title="Corollary 22.12 ">22.12</a> tells us that there exists a primitive element for any finite extension of a finite field. The next theorem tells us that we can often find a primitive element.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-primitive-element"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">23.12</span><span class="title">Primitive Element Theorem</span></h6><p id="p-3607">Let $E$ be a finite separable extension of a field $F\text{.}$ Then there exists an $\alpha \in E$ such that $E=F( \alpha )\text{.}$</p></article><article class="proof" id="proof-196"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3608">We already know that there is no problem if $F$ is a finite field.  Suppose that $E$ is a finite extension of an infinite field. We will prove the result for $F(\alpha, \beta)\text{.}$  The general case easily follows when we use mathematical induction. Let $f(x)$ and $g(x)$ be  the minimal polynomials of $\alpha$ and $\beta\text{,}$ respectively. Let $K$ be the field in which both $f(x)$ and $g(x)$ split. Suppose that $f(x)$ has zeros $\alpha = \alpha_1, \ldots, \alpha_n$ in $K$ and $g(x)$ has zeros $\beta = \beta_1, \ldots, \beta_m$ in $K\text{.}$ All of these zeros have multiplicity 1, since $E$ is separable over $F\text{.}$ Since $F$ is infinite, we can find an $a$ in $F$ such that</p><div class="displaymath">
\begin{equation*}
a \neq \frac{\alpha_i - \alpha}{\beta - \beta_j}
\end{equation*}
</div><p>for all $i$ and $j$ with $j \neq 1\text{.}$ Therefore, $a( \beta - \beta_j ) \neq \alpha_i - \alpha\text{.}$ Let $\gamma = \alpha + a \beta\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
\gamma = \alpha + a \beta \neq \alpha_i + a \beta_j;
\end{equation*}
</div><p>hence, $\gamma - a \beta_j \neq \alpha_i$ for all $i, j$ with $j \neq 1\text{.}$  Define $h(x) \in F( \gamma )[x]$ by $h(x) = f( \gamma - ax)\text{.}$ Then $h( \beta ) = f( \alpha ) = 0\text{.}$ However, $h( \beta_j ) \neq 0$ for $j \neq 1\text{.}$ Hence, $h(x)$ and $g(x)$ have a single common factor in $F( \gamma )[x]\text{;}$ that is, the minimal polynomial of $\beta$ over $F( \gamma )$ must be linear, since $\beta$ is the only zero common to both $g(x)$ and $h(x)\text{.}$ So $\beta \in F( \gamma )$ and $\alpha = \gamma - a \beta$ is in $F( \gamma )\text{.}$ Hence, $F( \alpha, \beta ) = F( \gamma )\text{.}$</p></article></div>
