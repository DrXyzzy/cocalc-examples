<div class="mathbook-content"><p id="p-2402">The Chinese Remainder Theorem is a result from elementary number theory about the solution of systems of simultaneous congruences. The Chinese mathematician Sun-tsï wrote about the theorem in the first century A.D. This theorem has some interesting consequences in the design of software for parallel processors.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-chinese-remainder"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">16.41</span></h6><p id="p-2403">Let $m$ and $n$ be positive integers such that $\gcd( m, n) = 1\text{.}$ Then for $a, b \in {\mathbb Z}$ the system</p><div class="displaymath">
\begin{align*}
x & \equiv  a \pmod{m}\\
x & \equiv  b \pmod{n} 
\end{align*}
</div><p>has a solution.  If $x_1$ and $x_2$ are two solutions of the system, then  $x_1 \equiv x_2 \pmod{mn}\text{.}$</p></article><article class="proof" id="proof-114"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2404">The equation $x \equiv a \pmod{m}$ has a solution since $a + km$ satisfies the equation for all $k \in {\mathbb Z}\text{.}$  We must show that there exists an integer $k_1$ such that</p><div class="displaymath">
\begin{equation*}
a + k_1 m \equiv b \pmod{n}.
\end{equation*}
</div><p>This is equivalent to showing that</p><div class="displaymath">
\begin{equation*}
k_1 m \equiv (b-a) \pmod{n}
\end{equation*}
</div><p>has a solution for $k_1\text{.}$  Since $m$ and $n$ are relatively prime, there exist integers $s$ and $t$ such that $ms + nt = 1\text{.}$ Consequently,</p><div class="displaymath">
\begin{equation*}
(b-a) ms = (b-a) -(b-a) nt,
\end{equation*}
</div><p>or</p><div class="displaymath">
\begin{equation*}
[(b-a)s]m \equiv (b-a) \pmod{n}.
\end{equation*}
</div><p>Now let $k_1 = (b-a)s\text{.}$</p><p id="p-2405">To show that any two solutions are congruent modulo $mn\text{,}$ let $c_1$ and $c_2$ be two solutions of  the system. That is,</p><div class="displaymath">
\begin{align*}
c_i & \equiv  a \pmod{m}\\
c_i & \equiv  b \pmod{n}
\end{align*}
</div><p>for $i = 1, 2\text{.}$ Then</p><div class="displaymath">
\begin{align*}
c_2 & \equiv  c_1 \pmod{m}\\
c_2 & \equiv  c_1 \pmod{n}.
\end{align*}
</div><p>Therefore, both $m$ and $n$ divide $c_1 - c_2\text{.}$ Consequently, $c_2 \equiv c_1 \pmod{mn}\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-solve-system"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.42</span></h6><p id="p-2406">Let us solve the system</p><div class="displaymath">
\begin{align*}
x & \equiv  3 \pmod{4}\\
x & \equiv  4 \pmod{5}.
\end{align*}
</div><p>Using the Euclidean algorithm, we can find integers $s$ and $t$ such that $4s + 5t = 1\text{.}$ Two such integers are $s = 4$ and $t = -3\text{.}$ Consequently,</p><div class="displaymath">
\begin{equation*}
x = a + k_1 m = 3 + 4k_1 = 3 + 4[(5 - 4)4] = 19.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-87"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">16.43</span><span class="title">Chinese Remainder Theorem</span></h6><p id="p-2407">Let $n_1, n_2, \ldots, n_k$ be positive integers such that $\gcd(n_i, n_j) = 1$ for $i \neq j\text{.}$ Then for any integers $a_1, \ldots, a_k\text{,}$ the system</p><div class="displaymath">
\begin{align*}
x & \equiv  a_1 \pmod{n_1}\\
x & \equiv  a_2 \pmod{n_2}\\
&  \vdots \\
x & \equiv  a_k \pmod{n_k}
\end{align*}
</div><p>has a solution.  Furthermore, any two solutions of the system are congruent modulo $n_1 n_2 \cdots n_k\text{.}$</p></article><article class="proof" id="proof-115"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2408">We will use mathematical induction on the number of equations in the system. If there are $k= 2$ equations, then the theorem is true by  Lemma <a href="section-software-application.ipynb#lemma-chinese-remainder" class="xref" alt="Lemma 16.41 " title="Lemma 16.41 ">16.41</a>. Now suppose that the result is true for a system of $k$  equations or less and that we wish to find a solution of</p><div class="displaymath">
\begin{align*}
x & \equiv  a_1 \pmod{n_1}\\
x & \equiv  a_2 \pmod{n_2}\\
& \vdots \\
x & \equiv  a_{k+1} \pmod{n_{k+1}}.
\end{align*}
</div><p>Considering the first $k$ equations, there exists a solution that is unique modulo $n_1 \cdots n_k\text{,}$ say $a\text{.}$ Since $n_1 \cdots n_k$ and $n_{k+1}$ are relatively prime, the system</p><div class="displaymath">
\begin{align*}
x & \equiv  a \pmod{n_1 \cdots n_k }\\
x & \equiv  a_{k+1} \pmod{n_{k+1}}
\end{align*}
</div><p>has a solution that is unique modulo $n_1 \ldots n_{k+1}$ by the lemma.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-system-mod-4-5-9-7"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.44</span></h6><p id="p-2409">Let us solve the system</p><div class="displaymath">
\begin{align*}
x & \equiv  3 \pmod{4}\\
x & \equiv  4 \pmod{5}\\
x & \equiv  1 \pmod{9}\\
x & \equiv  5 \pmod{7}.
\end{align*}
</div><p>From Example <a href="section-software-application.ipynb#example-rings-solve-system" class="xref" alt="Example 16.42 " title="Example 16.42 ">16.42</a> we know that 19 is a solution of the first two congruences and any other solution of the system is congruent to $19 \pmod{20}\text{.}$ Hence, we can reduce the system to a system of three congruences:</p><div class="displaymath">
\begin{align*}
x & \equiv  19 \pmod{20}\\
x & \equiv  1 \pmod{9}\\
x & \equiv  5 \pmod{7}.
\end{align*}
</div><p>Solving the next two equations, we can reduce the system to</p><div class="displaymath">
\begin{align*}
x & \equiv  19 \pmod{180}\\
x & \equiv  5 \pmod{7}.
\end{align*}
</div><p>Solving this last system, we find that 19 is a solution for the system that is unique up to modulo 1260.</p></article></div>

<div class="mathbook-content"><p id="p-2410">One interesting application of the Chinese Remainder Theorem in the design of computer software is that the theorem allows us to break up a calculation involving large integers into several less formidable calculations. A computer will handle integer calculations only up to a certain size due to the size of its processor chip, which is usually a 32 or 64-bit processor chip. For example, the largest integer available on a computer with a 64-bit processor chip is</p><div class="displaymath">
\begin{equation*}
2^{63} - 1 = \text{9,223,372,036,854,775,807}.
\end{equation*}
</div><p>Larger processors such as 128 or 256-bit have been proposed or are under development.  There is even talk of a 512-bit processor chip.  The largest integer that such a chip could store with be $2^{511} - 1\text{,}$ which would be a 154 digit number.  However, we would need to deal with much larger numbers to break sophisticated encryption schemes.</p></div>

<div class="mathbook-content"><p id="p-2411">Special software is required for calculations involving larger integers which cannot be added directly by the machine.  By using the Chinese Remainder Theorem we can break down large integer additions and multiplications into calculations that the computer can handle directly. This is especially useful on parallel processing computers which have the ability to run several programs concurrently.</p></div>

<div class="mathbook-content"><p id="p-2412">Most computers have a single central processing unit (CPU) containing one processor chip and can only add two numbers at a time. To add  a list of ten numbers, the CPU must do nine additions in sequence. However, a parallel processing computer has more than one CPU. A computer with 10 CPUs, for example,  can perform 10 different additions at the same time. If we can take a large integer  and break it down into parts, sending each part to a different CPU, then by performing several additions or multiplications simultaneously on those parts, we can work with an integer that the computer would not be able to handle as a whole.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-2134x1531"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.45</span></h6><p id="p-2413">Suppose that we wish to multiply 2134 by 1531. We will use the integers 95, 97, 98, and 99 because they are relatively prime. We can break down each integer into four parts:</p><div class="displaymath">
\begin{align*}
2134 & \equiv  44 \pmod{95}\\
2134 & \equiv  0 \pmod{97}\\
2134 & \equiv  76 \pmod{98}\\
2134 & \equiv  55 \pmod{99}
\end{align*}
</div><p>and</p><div class="displaymath">
\begin{align*}
1531 & \equiv  11 \pmod{95}\\
1531 & \equiv  76 \pmod{97}\\
1531 & \equiv  61 \pmod{98}\\
1531 & \equiv  46 \pmod{99}.
\end{align*}
</div><p>Multiplying the corresponding equations, we obtain</p><div class="displaymath">
\begin{align*}
2134 \cdot 1531 & \equiv  44 \cdot 11  \equiv  9 \pmod{95}\\
2134 \cdot 1531 & \equiv  0  \cdot 76  \equiv  0 \pmod{97}\\
2134 \cdot 1531 & \equiv  76 \cdot 61  \equiv  30 \pmod{98}\\
2134 \cdot 1531 & \equiv  55 \cdot 46  \equiv 55 \pmod{99}.
\end{align*}
</div><p>Each of these four computations can be sent to a different processor if our computer has several CPUs. By the above calculation, we know that $2134 \cdot 1531$ is a solution of the system</p><div class="displaymath">
\begin{align*}
x & \equiv  9 \pmod{95}\\
x & \equiv  0 \pmod{97}\\
x & \equiv  30 \pmod{98}\\
x & \equiv  55 \pmod{99}.
\end{align*}
</div><p>The Chinese Remainder Theorem tells us that solutions are unique up to modulo $95 \cdot 97 \cdot 98 \cdot 99 = \text{89,403,930}\text{.}$ Solving this system of congruences for $x$ tells us that $2134 \cdot 1531 = \text{3,267,154}\text{.}$</p><p id="p-2414">The conversion of the computation into the four subcomputations will take some computing time.  In addition, solving the system of congruences can also take considerable time.  However, if we have many computations to be performed on a particular set of numbers, it makes sense to transform the problem as we have done above and to perform the necessary calculations simultaneously.</p></article></div>
