<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Partially Ordered Sets"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Partially Ordered Sets</span></h3><a href="section-boolean-lattices.ipynb#boolean-subsection-poset" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-2899">We begin the study of lattices and Boolean algebras by generalizing the idea of inequality. Recall that a <dfn class="terminology">relation</dfn> on a set $X$ is a subset of $X \times X\text{.}$  A relation $P$ on $X$ is called a <dfn class="terminology">partial order</dfn> of $X$ if it satisfies the following axioms. </p><ol class="decimal"><li id="li-648"><p id="p-2900">The relation is <dfn class="terminology">reflexive</dfn>: $(a, a) \in P$ for all $a \in X\text{.}$</p></li><li id="li-649"><p id="p-2901">The relation is <dfn class="terminology">antisymmetric</dfn>: if $(a,b) \in P$ and $(b,a) \in P\text{,}$ then $a = b\text{.}$</p></li><li id="li-650"><p id="p-2902">The relation is <dfn class="terminology">transitive</dfn>: if $(a, b) \in P$ and $(b, c) \in P\text{,}$ then $(a, c) \in P\text{.}$</p></li></ol><p> We will usually write $a \preceq b$ to mean $(a, b) \in P$ unless some symbol is naturally associated with a particular partial order, such as $a \leq b$ with integers $a$ and $b\text{,}$ or $A \subset B$ with sets $A$ and $B\text{.}$ A set $X$ together with a partial order $\preceq$ is called a <dfn class="terminology">partially ordered set</dfn>, or <dfn class="terminology">poset</dfn>. </p></div>

<div class="mathbook-content"><article class="example-like" id="example-boolean-z-less-equal"><h6 class="heading"><span class="type">Example</span><span class="codenumber">19.1</span></h6><p id="p-2903">The set of integers (or rationals  or reals) is a poset where $a \leq  b$ has the usual meaning for two integers $a$ and $b$ in ${\mathbb Z}\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-boolean-power-set-subset"><h6 class="heading"><span class="type">Example</span><span class="codenumber">19.2</span></h6><p id="p-2904">Let $X$ be any set.  We will define the <dfn class="terminology">power set</dfn> of $X$ to be the set of all subsets of $X\text{.}$ We denote the power set of $X$ by ${\mathcal P}(X)\text{.}$ For example, let $X = \{ a, b, c \}\text{.}$  Then ${\mathcal P}(X)$ is  the set of all subsets of the set  $\{ a, b, c \}\text{:}$</p><div class="displaymath">
\begin{align*}
& \emptyset & & \{ a \} & & \{ b \} & & \{ c \} &\\
& \{ a, b \} & & \{ a, c\} & &\{ b, c\} & & \{ a, b, c \}. &
\end{align*}
</div><p>On any power set of a set $X\text{,}$ set inclusion, $\subset\text{,}$ is a partial order.  We can represent the order on $\{ a, b, c \}$ schematically by a diagram such as the one in  Figure <a href="section-boolean-lattices.ipynb#figure-partial-order" class="xref" alt="Figure 19.3 " title="Figure 19.3 ">19.3</a>.</p></article></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-partial-order"><img src="images/boolean-order-abc.svg" width="40%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">19.3</span>Partial order on $\mathcal P( \{ a, b, c \})$</figcaption></figure></div>

<div class="mathbook-content"><article class="example-like" id="example-boolean-subgroup-poset"><h6 class="heading"><span class="type">Example</span><span class="codenumber">19.4</span></h6><p id="p-2905">Let $G$ be a group. The set of subgroups of $G$ is a poset, where the partial order is set inclusion.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-boolean-p-order-not-unique"><h6 class="heading"><span class="type">Example</span><span class="codenumber">19.5</span></h6><p id="p-2906">There can be more than one partial order on a particular set.  We can form a partial order on ${\mathbb N}$ by $a \preceq b$ if $a \mid b\text{.}$  The relation is certainly reflexive since $a \mid a$ for all $a \in {\mathbb N}\text{.}$  If $m \mid n$ and $n \mid m\text{,}$ then $m = n\text{;}$ hence, the relation is also antisymmetric.  The relation is transitive, because if $m \mid n$ and $n \mid p\text{,}$ then $m \mid p\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-boolean-poset-divisors-24"><h6 class="heading"><span class="type">Example</span><span class="codenumber">19.6</span></h6><p id="p-2907">Let $X = \{ 1, 2, 3, 4, 6, 8, 12, 24 \}$ be the set of divisors of 24 with the partial order defined in Example <a href="section-boolean-lattices.ipynb#example-boolean-p-order-not-unique" class="xref" alt="Example 19.5 " title="Example 19.5 ">19.5</a>. Figure <a href="section-boolean-lattices.ipynb#figure-poset-divisors-24" class="xref" alt="Figure 19.7 " title="Figure 19.7 ">19.7</a> shows the partial order on $X\text{.}$</p></article></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-poset-divisors-24"><img src="images/boolean-order-24.svg" width="25%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">19.7</span>A partial order on the divisors of 24</figcaption></figure></div>

<div class="mathbook-content"><p id="p-2908">Let $Y$ be a subset of a poset $X\text{.}$ An element $u$ in $X$ is an <dfn class="terminology">upper bound</dfn> of $Y$ if $a \preceq u$ for every element $a \in Y\text{.}$ If $u$ is an upper bound of $Y$ such that $u \preceq v$ for every other upper bound $v$ of $Y\text{,}$ then $u$ is called a <dfn class="terminology">least upper bound</dfn> or <dfn class="terminology">supremum</dfn> of $Y\text{.}$ An element $l$ in $X$ is said to be a <dfn class="terminology">lower bound</dfn> of $Y$ if $l \preceq a$ for all $a \in Y\text{.}$ If $l$ is a lower bound of $Y$ such that $k \preceq l$ for every other lower bound $k$ of $Y\text{,}$ then $l$ is called a <dfn class="terminology">greatest lower bound</dfn> or <dfn class="terminology">infimum</dfn> of $Y\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-boolean-poset-gcd"><h6 class="heading"><span class="type">Example</span><span class="codenumber">19.8</span></h6><p id="p-2909">Let $Y = \{  2, 3, 4, 6 \}$ be contained in the set $X$ of Example <a href="section-boolean-lattices.ipynb#example-boolean-poset-divisors-24" class="xref" alt="Example 19.6 " title="Example 19.6 ">19.6</a>.  Then $Y$ has upper bounds 12 and 24, with 12 as a least upper bound.  The only lower bound is 1; hence, it must be a greatest lower bound.</p></article></div>

<div class="mathbook-content"><p id="p-2910">As it turns out, least upper bounds and greatest lower bounds are unique if they exist.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-101"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">19.9</span></h6><p id="p-2911">Let $Y$ be a nonempty subset of a poset $X\text{.}$ If $Y$ has a least upper bound, then $Y$ has a unique least upper bound. If $Y$ has a greatest lower bound, then $Y$ has a unique greatest lower bound.</p></article><article class="proof" id="proof-143"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2912">Let $u_1$ and $u_2$ be least upper bounds for $Y\text{.}$ By the definition of the least upper bound, $u_1 \preceq u$ for all upper bounds $u$ of $Y\text{.}$ In particular, $u_1 \preceq u_2\text{.}$ Similarly, $u_2 \preceq u_1\text{.}$ Therefore, $u_1 = u_2$ by antisymmetry.  A similar argument show that the greatest lower bound is unique.</p></article></div>

<div class="mathbook-content"><p id="p-2913">On many posets it is possible to define binary operations by using the greatest lower bound and the least upper bound of two elements. A <dfn class="terminology">lattice</dfn> is a poset $L$ such that every pair of elements in $L$ has a least upper bound and a greatest lower bound. The least upper bound of $a, b \in L$ is called the <dfn class="terminology">join</dfn> of $a$ and $b$ and is denoted  by $a \vee b\text{.}$   The greatest lower bound of $a, b \in L$ is called the <dfn class="terminology">meet</dfn> of $a$ and $b$ and is denoted by $a \wedge b\text{.}$  </p></div>

<div class="mathbook-content"><article class="example-like" id="example-boolean-lub-glb"><h6 class="heading"><span class="type">Example</span><span class="codenumber">19.10</span></h6><p id="p-2914">Let $X$ be a set. Then the power set of $X\text{,}$ ${\mathcal P}(X)\text{,}$ is a lattice. For two sets $A$ and $B$ in ${\mathcal P}(X)\text{,}$ the least upper bound of $A$ and $B$ is $A \cup B\text{.}$ Certainly $A \cup B$ is an upper bound of $A$ and $B\text{,}$ since $A \subset A \cup B$ and $B \subset A \cup B\text{.}$  If $C$ is some other set containing both $A$ and $B\text{,}$ then $C$ must contain $A \cup B\text{;}$ hence, $A \cup B$ is the least upper bound of $A$ and $B\text{.}$ Similarly,  the greatest lower bound of $A$ and $B$ is $A \cap B\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-boolean-subgroup-lattice"><h6 class="heading"><span class="type">Example</span><span class="codenumber">19.11</span></h6><p id="p-2915">Let $G$ be a group and suppose that $X$ is the set of subgroups of $G\text{.}$  Then $X$ is a poset ordered by set-theoretic inclusion, $\subset\text{.}$  The set of subgroups of $G$ is also a lattice.  If $H$ and $K$ are subgroups of $G\text{,}$ the greatest lower bound of $H$ and $K$ is $H \cap K\text{.}$ The set $H \cup K$ may not be a subgroup of $G\text{.}$  We leave it as an exercise to show that the least upper bound of $H$ and $K$ is the subgroup generated by $H \cup K\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-2916">In set theory we have certain duality conditions. For example, by De Morgan's laws, any statement about sets that is true about $(A \cup B)'$ must also be true about $A' \cap B'\text{.}$ We also have a duality principle for lattices.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="axiom-1"><h6 class="heading"><span class="type">Axiom</span><span class="codenumber">19.12</span><span class="title">Principle of Duality</span></h6><p id="p-2917">Any statement that is true for all lattices remains true when $\preceq$ is replaced by $\succeq$ and $\vee$ and $\wedge$ are interchanged throughout the statement.</p></article></div>

<div class="mathbook-content"><p id="p-2918">The following theorem tells us that a lattice is an algebraic structure with two binary operations that satisfy certain axioms.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-102"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">19.13</span></h6><p id="p-2919">If $L$ is a lattice, then the binary operations $\vee$ and $\wedge$ satisfy the following properties for $a, b, c \in L\text{.}$ </p><ol class="decimal"><li id="li-651"><p id="p-2920">Commutative laws: $a \vee b = b \vee a$ and $a \wedge b = b \wedge a\text{.}$</p></li><li id="li-652"><p id="p-2921">Associative laws: $a \vee ( b \vee c) = (a \vee b) \vee c$ and $a \wedge (b \wedge c) = (a \wedge b) \wedge c\text{.}$</p></li><li id="li-653"><p id="p-2922">Idempotent laws: $a \vee a = a$ and $a \wedge a = a\text{.}$</p></li><li id="li-654"><p id="p-2923">Absorption laws: $a \vee (a \wedge b) = a$ and $a \wedge ( a \vee b ) =a\text{.}$</p></li></ol></article><article class="proof" id="proof-144"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2924">By the Principle of Duality, we need only prove the first statement in each part.</p><p id="p-2925">(1)   By definition $a \vee b$ is the least upper bound of $\{ a, b\}\text{,}$ and $b \vee a$ is the least upper bound of $\{ b, a \}\text{;}$ however, $\{ a, b\}  = \{ b, a \}\text{.}$</p><p id="p-2926">(2) We will show that $a \vee ( b \vee c)$ and $(a \vee b) \vee c$ are both least upper bounds of $\{ a, b, c \}\text{.}$ Let $d =  a \vee b\text{.}$ Then $c \preceq  d \vee c = (a \vee b) \vee c\text{.}$ We also know that</p><div class="displaymath">
\begin{equation*}
a \preceq  a \vee b =d \preceq  d \vee c = (a \vee b) \vee c.
\end{equation*}
</div><p>A similar argument demonstrates that $b \preceq (a \vee b) \vee c\text{.}$ Therefore, $(a \vee b) \vee c$ is an upper bound of $\{ a, b, c \}\text{.}$ We now need to show that $(a \vee b) \vee c$ is the least upper bound of $\{ a, b, c\}\text{.}$ Let $u$ be some other upper bound of $\{ a, b, c \}\text{.}$  Then $a \preceq u$ and $b \preceq u\text{;}$ hence, $d = a \vee b \preceq u\text{.}$ Since $c \preceq u\text{,}$ it follows that $(a \vee b) \vee c = d \vee c \preceq u\text{.}$ Therefore, $(a \vee b) \vee c$ must be the least upper  bound of $\{ a, b, c\}\text{.}$ The argument that shows  $a \vee ( b \vee c)$  is the least upper bound of $\{ a, b, c \}$ is the same.  Consequently, $a \vee ( b \vee c) = (a \vee b) \vee c\text{.}$</p><p id="p-2927">(3) The join of $a$ and $a$ is the least upper bound of $\{ a \}\text{;}$ hence, $a \vee a = a\text{.}$</p><p id="p-2928">(4) Let $d = a \wedge b\text{.}$ Then $a \preceq a \vee d\text{.}$ On the other hand, $d = a \wedge b \preceq a\text{,}$ and so $a \vee d \preceq a\text{.}$ Therefore, $a \vee ( a \wedge b) = a\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-2929">Given any arbitrary set $L$ with operations $\vee$ and $\wedge\text{,}$ satisfying the conditions of the previous theorem, it is natural to ask whether or not this set comes from some lattice. The following theorem says that this is always the case.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-partial-order"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">19.14</span></h6><p id="p-2930">Let $L$ be a nonempty set with two binary operations $\vee$ and $\wedge$ satisfying the commutative, associative, idempotent, and  absorption laws.  We can define a partial order on $L$ by $a \preceq b$ if $a \vee b = b\text{.}$ Furthermore, $L$ is a lattice with respect to $\preceq$ if for all $a, b \in L\text{,}$ we define the least upper bound and greatest lower bound of $a$ and $b$ by $a \vee b$ and $a \wedge b\text{,}$ respectively.</p></article><article class="proof" id="proof-145"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2931">We first show that $L$ is a poset under $\preceq\text{.}$ Since $a \vee a = a\text{,}$ $a \preceq a$ and $\preceq$ is reflexive. To show that $\preceq$ is antisymmetric, let $a \preceq b$ and $b \preceq a\text{.}$ Then $a \vee b = b$ and $b \vee a = a\text{.}$By the commutative law, $b = a \vee b = b \vee a = a\text{.}$   Finally, we must show that $\preceq$ is transitive. Let $a \preceq b$ and $b \preceq c\text{.}$ Then $a \vee b = b$ and $b \vee c = c\text{.}$  Thus,</p><div class="displaymath">
\begin{equation*}
a \vee c = a \vee (b \vee c ) = ( a \vee b) \vee c = b \vee c = c,
\end{equation*}
</div><p>or $a \preceq c\text{.}$</p><p id="p-2932">To show that $L$ is a lattice, we must prove that $a \vee b$ and $a \wedge b$ are, respectively, the least upper and greatest lower bounds of $a$ and $b\text{.}$ Since $a=(a \vee b) \wedge a = a \wedge (a \vee b)\text{,}$ it follows that $a \preceq a \vee b\text{.}$  Similarly, $b \preceq a \vee b\text{.}$ Therefore, $a \vee b$ is an upper bound for $a$ and $b\text{.}$ Let $u$ be any other upper bound of both $a$ and $b\text{.}$ Then $a \preceq u$ and $b \preceq u\text{.}$ But $a \vee b \preceq u$ since</p><div class="displaymath">
\begin{equation*}
(a \vee b) \vee u = a \vee (b \vee u) = a \vee u = u.
\end{equation*}
</div><p>The proof that $a \wedge b$ is the greatest lower bound of $a$ and $b$ is left as an exercise.</p></article></div>
