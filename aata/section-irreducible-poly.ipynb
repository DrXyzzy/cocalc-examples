<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 17.3 Irreducible Polynomials"><span class="type">Section</span><span class="codenumber">17.3</span><span class="title">Irreducible Polynomials</span></h2><a href="section-irreducible-poly.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-2596">A nonconstant polynomial $f(x) \in F[x]$ is <dfn class="terminology">irreducible</dfn> over a field $F$ if $f(x)$ cannot be expressed as a product of two polynomials $g(x)$ and $h(x)$ in $F[x]\text{,}$ where the degrees of $g(x)$ and $h(x)$ are both smaller than the degree of $f(x)\text{.}$  Irreducible polynomials function as the “prime numbers” of polynomial rings.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-poly-irreducible"><h6 class="heading"><span class="type">Example</span><span class="codenumber">17.11</span></h6><p id="p-2597">The polynomial $x^2 - 2 \in {\mathbb Q}[x]$ is irreducible since it cannot be factored any further over the rational numbers. Similarly, $x^2 + 1$ is  irreducible over the real numbers.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-poly-finite"><h6 class="heading"><span class="type">Example</span><span class="codenumber">17.12</span></h6><p id="p-2598">The polynomial $p(x) = x^3 + x^2 + 2$ is irreducible over ${\mathbb Z}_3[x]\text{.}$ Suppose that this polynomial was reducible over ${\mathbb Z}_3[x]\text{.}$  By the division algorithm there would have to be a factor of the form $x - a\text{,}$ where $a$ is some element in ${\mathbb Z}_3[x]\text{.}$ Hence, it would have to be true that $p(a) = 0\text{.}$  However,</p><div class="displaymath">
\begin{align*}
p(0) & = 2\\
p(1) & = 1\\
p(2) & = 2.
\end{align*}
</div><p>Therefore, $p(x)$ has no zeros in ${\mathbb Z}_3$ and must be irreducible.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-poly-integer-coef"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">17.13</span></h6><p id="p-2599">Let $p(x) \in {\mathbb Q}[x]\text{.}$  Then</p><div class="displaymath">
\begin{equation*}
p(x) = \frac{r}{s}(a_0 + a_1 x + \cdots + a_n x^n),
\end{equation*}
</div><p>where $r, s, a_0, \ldots, a_n$ are integers, the $a_i$'s are relatively prime, and $r$ and $s$ are relatively prime.</p></article><article class="proof" id="proof-123"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2600">Suppose that</p><div class="displaymath">
\begin{equation*}
p(x) = \frac{b_0}{c_0} + \frac{b_1}{c_1} x + \cdots + \frac{b_n}{c_n} x^n,
\end{equation*}
</div><p>where the $b_i$'s and the $c_i$'s are integers. We can rewrite $p(x)$ as</p><div class="displaymath">
\begin{equation*}
p(x) = \frac{1}{c_0 \cdots c_n} (d_0 + d_1 x + \cdots + d_n x^n),
\end{equation*}
</div><p>where $d_0, \ldots, d_n$ are integers. Let $d$ be the greatest common divisor of $d_0, \ldots, d_n\text{.}$  Then</p><div class="displaymath">
\begin{equation*}
p(x) = \frac{d}{c_0 \cdots c_n} (a_0 + a_1 x + \cdots + a_n x^n),
\end{equation*}
</div><p>where $d_i = d a_i$ and the $a_i$'s are relatively prime. Reducing $d /(c_0 \cdots c_n)$ to its lowest terms, we can write</p><div class="displaymath">
\begin{equation*}
p(x) = \frac{r}{s}(a_0 + a_1 x + \cdots + a_n x^n),
\end{equation*}
</div><p>where $\gcd(r,s) = 1\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-gauss-lemma"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">17.14</span><span class="title">Gauss's Lemma</span></h6><p id="p-2601">Let $p(x) \in {\mathbb Z}[x]$ be a monic polynomial such that $p(x)$ factors into a product of two polynomials $\alpha(x)$ and $\beta(x)$ in ${\mathbb Q}[x]\text{,}$ where the degrees of both $\alpha(x)$ and $\beta(x)$ are less than the degree of $p(x)\text{.}$ Then $p(x) = a(x) b(x)\text{,}$ where $a(x)$ and $b(x)$ are monic polynomials in ${\mathbb Z}[x]$ with $\deg \alpha(x) = \deg a(x)$ and $\deg \beta(x) = \deg b(x)\text{.}$</p></article><article class="proof" id="proof-124"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2602">By Lemma <a href="section-irreducible-poly.ipynb#lemma-poly-integer-coef" class="xref" alt="Lemma 17.13 " title="Lemma 17.13 ">17.13</a>, we can assume that</p><div class="displaymath">
\begin{align*}
\alpha(x)  & =  \frac{c_1}{d_1} (a_0 + a_1 x + \cdots + a_m x^m ) = \frac{c_1}{d_1} \alpha_1(x)\\
\beta(x)  & =  \frac{c_2}{d_2} (b_0 + b_1 x + \cdots + b_n x^n)  = \frac{c_2}{d_2} \beta_1(x),
\end{align*}
</div><p>where the $a_i$'s are relatively prime and the $b_i$'s are relatively prime. Consequently,</p><div class="displaymath">
\begin{equation*}
p(x) = \alpha(x) \beta(x) = \frac{c_1 c_2}{d_1 d_2} \alpha_1(x) \beta_1(x) = \frac{c}{d} \alpha_1(x) \beta_1(x),
\end{equation*}
</div><p>where  $c/d$ is the product of $c_1/d_1$ and $c_2/d_2$ expressed in lowest terms. Hence, $d p(x) = c \alpha_1(x) \beta_1(x)\text{.}$</p><p id="p-2603">If $d = 1\text{,}$ then $c a_m b_n = 1$ since $p(x)$ is a monic polynomial. Hence, either $c=1$ or $c = -1\text{.}$ If $c=1\text{,}$ then either $a_m = b_n = 1$ or $a_m = b_n = -1\text{.}$ In the first case $p(x) = \alpha_1(x) \beta_1(x)\text{,}$ where $\alpha_1(x)$ and $\beta_1(x)$ are monic polynomials with $\deg \alpha(x) = \deg \alpha_1(x)$ and $\deg \beta(x) = \deg \beta_1(x)\text{.}$ In the second case $a(x) = -\alpha_1(x)$ and $b(x) = -\beta_1(x)$ are the correct monic polynomials since $p(x) = (-\alpha_1(x))(- \beta_1(x)) = a(x) b(x)\text{.}$ The case in which $c = -1$ can be handled similarly.</p><p id="p-2604">Now suppose that $d \neq 1\text{.}$ Since $\gcd(c, d) = 1\text{,}$ there exists a prime $p$ such that $p \mid d$ and $p \notdivide c\text{.}$ Also, since the coefficients of $\alpha_1(x)$ are relatively prime, there exists a coefficient $a_i$ such that $p \notdivide a_i\text{.}$  Similarly, there exists a coefficient $b_j$ of $\beta_1(x)$ such that $p \notdivide b_j\text{.}$ Let  $\alpha_1'(x)$ and $\beta_1'(x)$ be the polynomials in ${\mathbb Z}_p[x]$ obtained by reducing the coefficients of $\alpha_1(x)$ and $\beta_1(x)$ modulo $p\text{.}$ Since $p \mid d\text{,}$ $\alpha_1'(x) \beta_1'(x) = 0$ in ${\mathbb Z}_p[x]\text{.}$ However, this is impossible since neither $\alpha_1'(x)$ nor $\beta_1'(x)$ is the zero polynomial and ${\mathbb Z}_p[x]$ is an integral domain.  Therefore, $d=1$ and the theorem is proven.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-rational-roots-theorem"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">17.15</span></h6><p id="p-2605">Let $p(x) = x^n + a_{n - 1} x^{n - 1} + \cdots + a_0$ be  a polynomial with coefficients in ${\mathbb Z}$ and $a_0 \neq 0\text{.}$ If $p(x)$ has a zero in ${\mathbb Q}\text{,}$ then $p(x)$ also has a zero $\alpha$ in ${\mathbb Z}\text{.}$ Furthermore, $\alpha$ divides $a_0\text{.}$</p></article><article class="proof" id="proof-125"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2606">Let $p(x)$ have a zero $a \in {\mathbb Q}\text{.}$ Then $p(x)$ must have a linear factor $x-a\text{.}$ By Gauss's Lemma, $p(x)$ has a factorization with a linear factor in ${\mathbb Z}[x]\text{.}$ Hence, for some $\alpha \in {\mathbb Z}$</p><div class="displaymath">
\begin{equation*}
p(x) = (x - \alpha)( x^{n - 1} + \cdots - a_0 / \alpha ).
\end{equation*}
</div><p>Thus $a_0 /\alpha \in {\mathbb Z}$ and so $\alpha \mid a_0\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-poly-factor"><h6 class="heading"><span class="type">Example</span><span class="codenumber">17.16</span></h6><p id="p-2607">Let $p(x) = x^4 - 2 x^3 + x + 1\text{.}$ We shall show that $p(x)$ is irreducible over ${\mathbb Q}[x]\text{.}$  Assume that $p(x)$ is reducible. Then either $p(x)$ has a linear factor, say $p(x) = (x - \alpha) q(x)\text{,}$ where $q(x)$ is a polynomial of degree three, or $p(x)$ has two  quadratic factors.</p><p id="p-2608">If $p(x)$ has a linear factor in ${\mathbb Q}[x]\text{,}$ then it has a zero in ${\mathbb Z}\text{.}$  By Corollary <a href="section-irreducible-poly.ipynb#corollary-rational-roots-theorem" class="xref" alt="Corollary 17.15 " title="Corollary 17.15 ">17.15</a>, any zero must divide 1 and therefore must be $\pm 1\text{;}$ however, $p(1) = 1$ and $p(-1)= 3\text{.}$ Consequently, we have eliminated the possibility that $p(x)$ has any linear factors.</p><p id="p-2609">Therefore, if $p(x)$ is reducible it must factor into two quadratic polynomials, say</p><div class="displaymath">
\begin{align*}
p(x) & = (x^2 + ax + b )( x^2 + cx + d )\\
& = x^4 + (a + c)x^3 + (ac + b + d)x^2 + (ad + bc)x + bd,
\end{align*}
</div><p>where each factor is in ${\mathbb Z}[x]$ by Gauss's Lemma. Hence,</p><div class="displaymath">
\begin{align*}
a + c & = - 2\\
ac + b + d & = 0\\
ad + bc & = 1\\
bd & = 1.
\end{align*}
</div><p>Since $bd = 1\text{,}$ either $b = d = 1$ or $b = d = -1\text{.}$ In either case $b = d$ and so</p><div class="displaymath">
\begin{equation*}
ad + bc  = b( a + c ) = 1.
\end{equation*}
</div><p>Since $a + c = -2\text{,}$ we know that $-2b = 1\text{.}$ This is impossible since $b$ is an integer. Therefore, $p(x)$ must be irreducible over ${\mathbb Q}\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-eisenstein"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">17.17</span><span class="title">Eisenstein's Criterion</span></h6><p id="p-2610">Let $p$ be a prime and suppose that</p><div class="displaymath">
\begin{equation*}
f(x) = a_n x^n + \cdots + a_0 \in {\mathbb Z}[x].
\end{equation*}
</div><p>If $p \mid a_i$ for $i = 0, 1, \ldots, n-1\text{,}$ but $p \notdivide a_n$ and $p^2 \notdivide a_0\text{,}$ then $f(x)$ is irreducible over ${\mathbb Q}\text{.}$</p></article><article class="proof" id="proof-126"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2611">By Gauss's Lemma, we need only show that $f(x)$ does not factor into polynomials of lower degree in ${\mathbb Z}[x]\text{.}$ Let</p><div class="displaymath">
\begin{equation*}
f(x) = (b_rx^r + \cdots + b_0)(c_s x^s + \cdots + c_0 )
\end{equation*}
</div><p>be a factorization in ${\mathbb Z}[x]\text{,}$ with $b_r$ and $c_s$ not equal to zero and $r, s \lt n\text{.}$ Since $p^2$ does not divide $a_0 = b_0 c_0\text{,}$ either $b_0$ or $c_0$ is not divisible by $p\text{.}$ Suppose that $p \notdivide b_0$ and $p \mid c_0\text{.}$ Since $p \notdivide a_n$ and $a_n = b_r c_s\text{,}$ neither $b_r$ nor $c_s$ is divisible by $p\text{.}$ Let $m$ be the smallest value of $k$ such that $p \notdivide c_k\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
a_m = b_0 c_m + b_1 c_{m - 1} + \cdots + b_m c_0
\end{equation*}
</div><p>is not divisible by $p\text{,}$ since each term on the right-hand side of the equation is divisible by $p$ except for $b_0 c_m\text{.}$ Therefore, $m = n$ since $a_i$ is divisible by $p$ for $m \lt n\text{.}$  Hence, $f(x)$ cannot be factored into polynomials of lower degree and therefore must be irreducible.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-poly-eisenstein"><h6 class="heading"><span class="type">Example</span><span class="codenumber">17.18</span></h6><p id="p-2612">The polynomial</p><div class="displaymath">
\begin{equation*}
f(x) = 16 x^5  - 9 x^4 + 3x^2 + 6 x - 21
\end{equation*}
</div><p>is easily seen to be irreducible over ${\mathbb Q}$ by Eisenstein's Criterion if we let $p = 3\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-2613">Eisenstein's Criterion is more useful in constructing irreducible polynomials of a certain degree over ${\mathbb Q}$ than in determining the irreducibility of an arbitrary polynomial in ${\mathbb Q}[x]\text{:}$ given an arbitrary polynomial, it is not very likely that we can apply Eisenstein's Criterion.  The real value of Theorem <a href="section-irreducible-poly.ipynb#theorem-eisenstein" class="xref" alt="Theorem 17.17 Eisenstein's Criterion" title="Theorem 17.17 Eisenstein's Criterion">17.17</a> is that we now have an easy method of generating irreducible polynomials of any degree.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Ideals in $F\lbrack x \rbrack$"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Ideals in $F\lbrack x \rbrack$</span></h3><a href="section-irreducible-poly.ipynb#poly-subsection-ideals" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-2614">Let $F$ be a field. Recall that a principal ideal in $F[x]$ is an ideal $\langle p(x) \rangle$ generated by some polynomial $p(x)\text{;}$ that is,</p><div class="displaymath">
\begin{equation*}
\langle p(x) \rangle = \{ p(x) q(x) : q(x) \in F[x] \}.
\end{equation*}
</div></div>

<div class="mathbook-content"><article class="example-like" id="example-poly-ideal"><h6 class="heading"><span class="type">Example</span><span class="codenumber">17.19</span></h6><p id="p-2615">The polynomial $x^2$ in $F[x]$ generates the ideal $\langle x^2 \rangle$ consisting of all polynomials with no constant term or term of degree 1.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-poly-principal-ideal"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">17.20</span></h6><p id="p-2616">If $F$ is a field, then every ideal in $F[x]$ is a principal ideal.</p></article><article class="proof" id="proof-127"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2617">Let $I$ be an ideal of $F[x]\text{.}$  If $I$ is the zero ideal, the theorem is easily true. Suppose that $I$ is a nontrivial ideal in $F[x]\text{,}$ and let $p(x) \in I$ be a nonzero element of minimal degree. If $\deg p(x)= 0\text{,}$ then $p(x)$ is a nonzero constant and 1 must be in $I\text{.}$ Since 1 generates all of $F[x]\text{,}$ $\langle 1 \rangle = I = F[x]$ and $I$ is again a principal ideal.</p><p id="p-2618">Now assume that $\deg p(x) \geq 1$ and let $f(x)$ be any element in $I\text{.}$  By the division algorithm there exist $q(x)$ and $r(x)$ in $F[x]$ such that $f(x) = p(x) q(x) + r(x)$ and $\deg r(x) \lt \deg  p(x)\text{.}$ Since $f(x), p(x) \in I$ and $I$ is an ideal, $r(x) = f(x) - p(x) q(x)$ is also in $I\text{.}$ However, since we chose $p(x)$ to be of minimal degree, $r(x)$ must be the zero polynomial. Since we can write any element $f(x)$ in $I$ as $p(x) q(x)$ for some $q(x) \in F[x]\text{,}$ it must be the case that $I = \langle p(x) \rangle\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-poly-xy"><h6 class="heading"><span class="type">Example</span><span class="codenumber">17.21</span></h6><p id="p-2619">It is not the case that every ideal in the ring $F[x,y]$ is a principal ideal. Consider the ideal of $F[x, y]$ generated by the polynomials $x$ and $y\text{.}$  This is the ideal of $F[x, y]$ consisting of all polynomials with no constant term. Since both $x$ and $y$ are in the ideal, no single polynomial can generate the entire ideal.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-max-ideal-poly"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">17.22</span></h6><p id="p-2620">Let $F$ be a field and suppose that $p(x) \in F[x]\text{.}$ Then the ideal generated by $p(x)$ is maximal if and only if $p(x)$ is irreducible.</p></article><article class="proof" id="proof-128"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2621">Suppose that $p(x)$ generates a maximal ideal of $F[x]\text{.}$ Then $\langle p(x) \rangle$ is also a prime ideal of $F[x]\text{.}$ Since a maximal ideal must be properly contained inside $F[x]\text{,}$ $p(x)$ cannot be a constant polynomial. Let us assume that $p(x)$ factors into two polynomials of lesser degree, say $p(x) = f(x) g(x)\text{.}$ Since $\langle p(x) \rangle$ is a prime ideal one of these factors, say $f(x)\text{,}$ is in $\langle p(x) \rangle$ and therefore be a multiple of $p(x)\text{.}$ But this would imply that $\langle p(x) \rangle \subset \langle f(x) \rangle\text{,}$ which is impossible since $\langle p(x) \rangle$ is maximal.</p><p id="p-2622">Conversely, suppose that $p(x)$ is irreducible over $F[x]\text{.}$ Let $I$ be an ideal in $F[x]$ containing $\langle p(x) \rangle\text{.}$ By Theorem <a href="section-irreducible-poly.ipynb#theorem-poly-principal-ideal" class="xref" alt="Theorem 17.20 " title="Theorem 17.20 ">17.20</a>, $I$ is a principal ideal; hence, $I = \langle f(x) \rangle$ for some $f(x) \in F[x]\text{.}$ Since $p(x) \in I\text{,}$ it must be the case that $p(x) = f(x) g(x)$ for some $g(x) \in F[x]\text{.}$ However, $p(x)$ is irreducible; hence, either $f(x)$ or $g(x)$ is a constant polynomial. If $f(x)$ is constant, then $I = F[x]$ and we are done.  If $g(x)$ is constant, then $f(x)$ is a constant multiple of $I$ and $I = \langle p(x) \rangle\text{.}$ Thus, there are no proper ideals of $F[x]$ that properly contain $\langle p(x)\rangle\text{.}$</p></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Historical Note"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Historical Note</span></h3><a href="section-irreducible-poly.ipynb#poly-subsection-historical-note" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-2624">Throughout history, the solution of polynomial equations has been a challenging problem.  The Babylonians knew how to solve the equation $ax^2 + bx + c = 0\text{.}$  Omar Khayyam (1048–1131) devised methods of solving cubic equations through the use of  geometric constructions and conic sections.  The algebraic solution of the general cubic equation $ax^3 + bx^2 + cx + d = 0$ was not discovered until the sixteenth century.  An Italian mathematician, Luca Pacioli (ca. 1445–1509), wrote  in <span class="booktitle">Summa de Arithmetica</span> that the solution of the cubic was impossible.  This was taken as a challenge by the rest of the mathematical community.</p></div>

<div class="mathbook-content"><p id="p-2625">Scipione del Ferro (1465–1526), of the University of Bologna, solved the “depressed cubic,”</p><div class="displaymath">
\begin{equation*}
ax^3 + cx + d = 0.
\end{equation*}
</div><p>He kept his solution an absolute secret.  This may seem surprising today, when mathematicians are usually very eager to publish their results, but in the days of the Italian Renaissance secrecy was customary. Academic appointments were not easy to secure and depended on the ability to prevail in public contests.  Such challenges could be issued at any time.  Consequently, any major new discovery was a valuable weapon in such a contest. If an opponent presented a list of problems to be solved, del Ferro could in turn present a list of depressed cubics.  He kept the secret of his discovery throughout his life, passing it on only on his deathbed to his student Antonio Fior (ca. 1506–?).</p></div>

<div class="mathbook-content"><p id="p-2626">Although Fior was not the equal of his teacher, he immediately issued a challenge to Niccolo Fontana (1499–1557).  Fontana was known as Tartaglia (the Stammerer).  As a youth he had suffered a blow from the sword of a French soldier during an attack on his village. He survived the savage wound, but his speech was permanently impaired.  Tartaglia sent Fior a list of 30 various mathematical problems; Fior countered by sending Tartaglia a list of 30 depressed cubics.  Tartaglia would either solve all 30 of the problems or absolutely fail.  After much effort Tartaglia finally succeeded in solving the depressed cubic and defeated Fior, who faded into obscurity.</p></div>

<div class="mathbook-content"><p id="p-2627">At this point another mathematician, Gerolamo Cardano (1501–1576), entered the story.  Cardano wrote to Tartaglia, begging him for the solution to the depressed cubic.  Tartaglia refused several of his requests, then finally revealed the solution to Cardano after the latter swore an oath not to publish the secret or to pass it on to anyone else. Using the knowledge that he had obtained from Tartaglia, Cardano eventually solved the general cubic</p><div class="displaymath">
\begin{equation*}
a x^3 + bx^2 +cx +d = 0.
\end{equation*}
</div><p>Cardano shared the secret with his student, Ludovico Ferrari (1522–1565), who solved the general quartic equation,</p><div class="displaymath">
\begin{equation*}
a x^4 + b x^3 + cx^2 + d x + e = 0.
\end{equation*}
</div><p>In 1543, Cardano and Ferrari examined del Ferro's papers and discovered that he had also solved the depressed cubic.  Cardano felt that this relieved him of his obligation to Tartaglia, so he proceeded to publish the solutions in <span class="booktitle">Ars Magna</span> (1545), in which he gave credit to del Ferro for solving the special case of the cubic.  This resulted in a bitter dispute between Cardano and Tartaglia, who published the story of the oath a year later.</p></div>
