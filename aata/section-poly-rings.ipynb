<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 17.1 Polynomial Rings"><span class="type">Section</span><span class="codenumber">17.1</span><span class="title">Polynomial Rings</span></h2><a href="section-poly-rings.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-2564">Throughout this chapter we shall assume that $R$ is a commutative ring with identity.  Any expression of the form</p><div class="displaymath">
\begin{equation*}
f(x) = \sum^{n}_{i=0} a_i x^i = a_0 + a_1 x +a_2 x^2 + \cdots + a_n x^n,
\end{equation*}
</div><p>where $a_i \in R$ and $a_n \neq 0\text{,}$ is called a <dfn class="terminology">polynomial over $R$</dfn> with <dfn class="terminology">indeterminate</dfn> $x\text{.}$  The elements $a_0, a_1, \ldots, a_n$ are called the <dfn class="terminology">coefficients</dfn> of $f\text{.}$  The coefficient $a_n$ is called the <dfn class="terminology">leading coefficient</dfn>.  A polynomial is called <dfn class="terminology">monic</dfn> if the leading coefficient is 1.  If $n$ is the largest nonnegative number for which $a_n \neq 0\text{,}$ we say that the <dfn class="terminology">degree</dfn> of $f$ is $n$ and write $\deg f(x) = n\text{.}$    If no such $n$ exists—that is, if $f=0$ is the zero polynomial—then the degree of $f$ is defined to be $-\infty\text{.}$  We will denote the set of all polynomials with coefficients in a ring $R$ by $R[x]\text{.}$    Two polynomials are equal exactly when their corresponding coefficients are equal; that is, if we let</p><div class="displaymath">
\begin{align*}
p(x) & = a_0 + a_1 x + \cdots + a_n x^n\\
q(x) & = b_0 + b_1 x + \cdots + b_m x^m,
\end{align*}
</div><p>then $p(x) = q(x)$ if and only if $a_i = b_i$ for all $i \geq 0\text{.}$</p></div>

<div class="mathbook-content"><p id="p-2565">To show that the set of all polynomials forms a ring, we must first define addition and multiplication.  We define the sum of two polynomials as follows.  Let</p><div class="displaymath">
\begin{align*}
p(x) & = a_0 + a_1 x + \cdots + a_n x^n\\
q(x) & = b_0 + b_1 x + \cdots + b_m x^m.
\end{align*}
</div><p>Then the sum of $p(x)$ and $q(x)$ is</p><div class="displaymath">
\begin{equation*}
p(x) + q(x) = c_0 + c_1 x + \cdots + c_k x^k,
\end{equation*}
</div><p>where $c_i = a_i + b_i$ for each $i\text{.}$  We define the product of $p(x)$ and $q(x)$ to be</p><div class="displaymath">
\begin{equation*}
p(x) q(x) = c_0 + c_1 x + \cdots + c_{m + n} x^{m + n},
\end{equation*}
</div><p>where</p><div class="displaymath">
\begin{equation*}
c_i = \sum_{k = 0}^i a_k b_{i - k} = a_0  b_i + a_1 b_{i -1} + \cdots + a_{i -1} b _1 + a_i b_0
\end{equation*}
</div><p>for each $i\text{.}$  Notice that in each case some of the coefficients may be zero.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-poly-operations"><h6 class="heading"><span class="type">Example</span><span class="codenumber">17.1</span></h6><p id="p-2566">Suppose that</p><div class="displaymath">
\begin{equation*}
p(x) = 3 + 0 x + 0 x^2 + 2 x^3 + 0 x^4
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
q(x) = 2 + 0 x - x^2 + 0 x^3 + 4 x^4
\end{equation*}
</div><p>are polynomials in ${\mathbb Z}[x]\text{.}$  If the coefficient of some term in a polynomial is zero, then we usually just omit that term.  In this case we  would write $p(x) =  3 + 2 x^3$ and $q(x) = 2 - x^2 + 4 x^4\text{.}$  The sum of these two polynomials is</p><div class="displaymath">
\begin{equation*}
p(x) + q(x)= 5 - x^2 + 2 x^3 + 4 x^4.
\end{equation*}
</div><p>The product,</p><div class="displaymath">
\begin{equation*}
p(x) q(x) = (3 + 2 x^3)( 2 - x^2 + 4 x^4 ) =  6 - 3x^2 + 4 x^3 + 12 x^4  - 2 x^5 + 8 x^7,
\end{equation*}
</div><p>can be calculated either by determining the $c_i$'s in the definition or by simply multiplying polynomials in the same way as we have always done.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-poly-integral-domains"><h6 class="heading"><span class="type">Example</span><span class="codenumber">17.2</span></h6><p id="p-2567">Let</p><div class="displaymath">
\begin{equation*}
p(x) = 3 + 3 x^3 \qquad \text{and} \qquad q(x) = 4 + 4 x^2 + 4 x^4
\end{equation*}
</div><p>be polynomials in ${\mathbb Z}_{12}[x]\text{.}$ The sum of $p(x)$ and $q(x)$ is $7 + 4 x^2 + 3 x^3 + 4 x^4\text{.}$  The product of the two polynomials is  the zero polynomial.  This example tells us that we can not expect $R[x]$ to be an integral domain if $R$ is not an integral domain.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-poly-1"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">17.3</span></h6><p id="p-2568">Let $R$ be a commutative ring with identity.  Then $R[x]$ is a commutative ring with identity.</p></article><article class="proof" id="proof-116"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2569">Our first task is to show that $R[x]$ is an abelian group under polynomial addition.  The zero polynomial, $f(x) = 0\text{,}$ is the additive identity.  Given a polynomial $p(x) = \sum_{i = 0}^{n} a_i x^i\text{,}$ the inverse of $p(x)$ is easily verified to be $-p(x) = \sum_{i = 0}^{n} (-a_i) x^i = -\sum_{i = 0}^{n} a_i x^i\text{.}$  Commutativity and associativity follow immediately from the definition of polynomial addition and from the fact that addition in $R$ is both commutative and associative.</p><p id="p-2570">To show that polynomial multiplication is associative, let</p><div class="displaymath">
\begin{align*}
p(x) & = \sum_{i = 0}^{m} a_i x^i,\\
q(x) & = \sum_{i = 0}^{n} b_i x^i,\\
r(x) & = \sum_{i = 0}^{p} c_i x^i.
\end{align*}
</div><p>Then</p><div class="displaymath">
\begin{align*}
[p(x) q(x)] r(x) & = \left[ \left( \sum_{i=0}^{m} a_i x^i  \right) \left(  \sum_{i=0}^{n} b_i x^i \right) \right] \left( \sum_{i = 0}^{p} c_i x^i \right)\\
& = \left[ \sum_{i = 0}^{m+n} \left(  \sum_{j = 0}^{i} a_j b_{i - j} \right) x^i \right] \left( \sum_{i = 0}^{p} c_i x^i \right)\\
& = \sum_{i = 0}^{m + n + p} \left[ \sum_{j = 0}^{i} \left( \sum_{k=0}^j a_k b_{j-k} \right) c_{i-j} \right] x^i\\
& = \sum_{i = 0}^{m + n + p} \left(\sum_{j + k + l = i} a_j b_k c_l \right) x^i\\
& = \sum_{i = 0}^{m+n+p} \left[ \sum_{j = 0}^{i} a_j  \left( \sum_{k = 0}^{i - j} b_k c_{i - j - k} \right) \right]  x^i\\
& = \left( \sum_{i = 0}^{m} a_i x^i  \right) \left[ \sum_{i = 0}^{n + p}  \left( \sum_{j = 0}^{i} b_j c_{i - j} \right) x^i \right]\\
& = \left( \sum_{i = 0}^{m} a_i x^i \right) \left[ \left(  \sum_{i = 0}^{n} b_i x^i \right) \left( \sum_{i = 0}^{p} c_i x^i \right) \right]\\
& = p(x) [ q(x) r(x) ]
\end{align*}
</div><p>The commutativity and distribution properties of polynomial multiplication are proved in a similar manner.  We shall leave the proofs of these properties as an exercise.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-poly-integral-domain"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">17.4</span></h6><p id="p-2571">Let $p(x)$ and $q(x)$ be polynomials in $R[x]\text{,}$ where $R$ is an integral domain.  Then $\deg p(x) + \deg q(x) = \deg( p(x) q(x) )\text{.}$  Furthermore, $R[x]$ is an integral domain.</p></article><article class="proof" id="proof-117"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2572">Suppose that we have two nonzero polynomials</p><div class="displaymath">
\begin{equation*}
p(x) = a_m x^m + \cdots + a_1 x + a_0
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
q(x) = b_n x^n + \cdots + b_1 x + b_0
\end{equation*}
</div><p>with $a_m \neq 0$ and $b_n \neq 0\text{.}$ The degrees of $p(x)$ and $q(x)$ are $m$ and $n\text{,}$ respectively.  The leading term of $p(x) q(x)$ is $a_m b_n x^{m + n}\text{,}$ which cannot be zero since $R$ is an integral domain; hence, the degree of $p(x) q(x)$ is $m + n\text{,}$ and $p(x)q(x) \neq 0\text{.}$  Since $p(x) \neq 0$ and $q(x) \neq 0$ imply that $p(x)q(x) \neq 0\text{,}$ we know that $R[x]$ must also be an integral domain.</p></article></div>

<div class="mathbook-content"><p id="p-2573">We also want to consider polynomials in two or more variables, such as $x^2 - 3 x y + 2 y^3\text{.}$  Let $R$ be a ring and suppose that we are given two indeterminates $x$ and $y\text{.}$  Certainly we can form the ring $(R[x])[y]\text{.}$  It is straightforward but perhaps tedious to show that $(R[x])[y] \cong R([y])[x]\text{.}$  We shall identify these two rings by this isomorphism and simply write $R[x,y]\text{.}$  The ring $R[x, y]$ is called the <dfn class="terminology">ring of polynomials in two indeterminates $x$ and $y$ with coefficients in</dfn> $R\text{.}$  We can define the <dfn class="terminology">ring of polynomials in</dfn> $n$ <dfn class="terminology">indeterminates with coefficients in</dfn> $R$ similarly.  We shall denote this ring by $R[x_1, x_2, \ldots, x_n]\text{.}$ </p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-poly-3"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">17.5</span></h6><p id="p-2574">Let $R$ be a commutative ring with identity and $\alpha \in R\text{.}$  Then we have a ring homomorphism  $\phi_{\alpha} : R[x] \rightarrow R$ defined by </p><div class="displaymath">
\begin{equation*}
\phi_{\alpha} (p(x) ) = p( \alpha ) = a_n \alpha^n + \cdots + a_1 \alpha + a_0,
\end{equation*}
</div><p>where $p( x ) = a_n x^n + \cdots + a_1 x + a_0\text{.}$</p></article><article class="proof" id="proof-118"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2575">Let $p(x) = \sum_{i = 0}^n a_i x^i$ and $q(x) = \sum_{i = 0}^m b_i x^i\text{.}$  It is easy to show that $\phi_{\alpha}(p(x) + q(x)) = \phi_{\alpha}(p(x))  + \phi_{\alpha}(q(x))\text{.}$  To show that multiplication is preserved under the map $\phi_{\alpha}\text{,}$ observe that</p><div class="displaymath">
\begin{align*}
\phi_{\alpha} (p(x) ) \phi_{\alpha} (q(x)) & = p( \alpha ) q(\alpha)\\
& = \left( \sum_{i = 0}^n a_i \alpha^i \right) \left( \sum_{i = 0}^m b_i \alpha^i \right)\\
& = \sum_{i = 0}^{m + n} \left( \sum_{k = 0}^i a_k b_{i - k} \right) \alpha^i\\
& = \phi_{\alpha} (p(x) q(x)).
\end{align*}
</div></article></div>

<div class="mathbook-content"><p id="p-2576">The map  $\phi_{\alpha} : R[x] \rightarrow R$ is called the <dfn class="terminology">evaluation homomorphism</dfn> at $\alpha\text{.}$</p></div>
