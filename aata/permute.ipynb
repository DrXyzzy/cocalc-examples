<div class="mathbook-content"><p id="p-781">Permutation groups are central to the study of geometric symmetries and to Galois theory, the study of finding solutions of polynomial equations.  They also provide abundant examples of nonabelian groups.</p></div>

<div class="mathbook-content"><p id="p-782">Let us recall for a moment the symmetries of the equilateral triangle $\bigtriangleup ABC$ from ChapterÂ <a href="groups.ipynb" class="xref" alt="Chapter 3 Groups" title="Chapter 3 Groups">3</a>. The symmetries actually consist of permutations of the three vertices, where a <dfn class="terminology">permutation</dfn> of the set $S = \{ A, B, C \}$ is a one-to-one and onto map $\pi :S \rightarrow S\text{.}$ The three vertices have the following six permutations.</p><div class="displaymath">
\begin{align*}
\begin{pmatrix}
A & B & C \\
A & B & C
\end{pmatrix}
\qquad
\begin{pmatrix}
A & B & C \\
C & A & B
\end{pmatrix}
\qquad
\begin{pmatrix}
A & B & C \\
B & C & A
\end{pmatrix}\\
\begin{pmatrix}
A & B & C \\
A & C & B
\end{pmatrix}
\qquad
\begin{pmatrix}
A & B & C \\
C & B & A
\end{pmatrix}
\qquad
\begin{pmatrix}
A & B & C \\
B & A & C
\end{pmatrix}
\end{align*}
</div></div>

<div class="mathbook-content"><p id="p-783">We have used the array</p><div class="displaymath">
\begin{equation*}
\begin{pmatrix}
A & B & C \\
B & C & A
\end{pmatrix}
\end{equation*}
</div><p>to denote the permutation that sends $A$ to $B\text{,}$ $B$ to $C\text{,}$ and $C$ to $A\text{.}$ That is,</p><div class="displaymath">
\begin{align*}
A & \mapsto  B\\
B & \mapsto  C\\
C & \mapsto  A.
\end{align*}
</div><p>The symmetries of a triangle form a group. In this chapter we will study groups of this type.</p></div>
