<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 6.1 Cosets"><span class="type">Section</span><span class="codenumber">6.1</span><span class="title">Cosets</span></h2><a href="section-cosets.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-968">Let $G$ be a group and $H$ a subgroup of $G\text{.}$  Define a <dfn class="terminology">left coset</dfn> of $H$ with <dfn class="terminology">representative</dfn> $g \in G$ to be the set</p><div class="displaymath">
\begin{equation*}
gH = \{ gh : h \in H \}.
\end{equation*}
</div><p><dfn class="terminology">Right cosets</dfn> can be defined similarly by</p><div class="displaymath">
\begin{equation*}
Hg = \{ hg : h \in H \}.
\end{equation*}
</div><p>If left and right cosets coincide or if it is clear from the context to which type of coset that we are referring, we will use the word <em class="emphasis">coset</em> without specifying left or right.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-cosets-z6-cosets"><h6 class="heading"><span class="type">Example</span><span class="codenumber">6.1</span></h6><p id="p-969">Let $H$ be the subgroup of ${\mathbb Z}_6$ consisting of the elements 0 and 3.  The cosets are</p><div class="displaymath">
\begin{gather*}
0 + H = 3 + H = \{ 0, 3 \}\\
1 + H = 4 + H = \{ 1, 4 \}\\
2 + H = 5 + H = \{ 2, 5 \}.
\end{gather*}
</div><p>We will always write the cosets of subgroups of ${\mathbb Z}$ and ${\mathbb Z}_n$ with the additive notation we have used for cosets here.  In a commutative group, left and right cosets are always identical.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-cosets-s3-cosets"><h6 class="heading"><span class="type">Example</span><span class="codenumber">6.2</span></h6><p id="p-970">Let $H$ be the subgroup of $S_3$ defined by the permutations $\{(1), (123), (132) \}\text{.}$  The left cosets of $H$ are</p><div class="displaymath">
\begin{gather*}
(1)H = (1 2 3)H =  (132)H = \{(1), (1 23), (132) \}\\
(1 2)H = (1 3)H = (2 3)H =  \{ (1 2), (1 3), (2 3)  \}.
\end{gather*}
</div><p>The right cosets of $H$ are exactly the same as the left cosets:</p><div class="displaymath">
\begin{gather*}
H(1) = H(1 2 3) =  H(132) = \{(1), (1 23), (132) \}\\
H(1 2) = H(1 3) = H(2 3) =  \{ (1 2), (1 3), (2 3)  \}.
\end{gather*}
</div><p id="p-971">It is not always the case that a left coset is the same as a right coset.  Let $K$ be the subgroup of $S_3$ defined by the permutations $\{(1), (1 2)\}\text{.}$  Then the left cosets of $K$ are</p><div class="displaymath">
\begin{gather*}
(1)K = (1 2)K = \{(1), (1 2)\}\\
(1 3)K = (1 2 3)K = \{(1 3), (1 2 3)\}\\
(2 3)K = (1 3 2)K = \{(2 3), (1 3 2)\};
\end{gather*}
</div><p>however, the right cosets of $K$ are</p><div class="displaymath">
\begin{gather*}
K(1) = K(1 2) = \{(1), (1 2)\}\\
K(1 3) = K(1 3 2) = \{(1 3), (1 3 2)\}\\
K(2 3) = K(1 2 3) = \{(2 3), (1 2 3)\}.
\end{gather*}
</div></article></div>

<div class="mathbook-content"><p id="p-972">The following lemma is quite useful when dealing with cosets.  (We leave its proof as an exercise.)</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-cosets-theorem-1"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">6.3</span></h6><p id="p-973">Let $H$ be a subgroup of a group $G$ and suppose that $g_1, g_2 \in G\text{.}$  The following conditions are equivalent. </p><ol class="decimal"><li id="li-253"><p id="p-974">$g_1 H = g_2 H\text{;}$</p></li><li id="li-254"><p id="p-975">$H g_1^{-1}  = H g_2^{-1}\text{;}$</p></li><li id="li-255"><p id="p-976">$g_1 H \subset g_2 H\text{;}$</p></li><li id="li-256"><p id="p-977">$g_2 \in g_1 H\text{;}$</p></li><li id="li-257"><p id="p-978">$g_1^{-1} g_2 \in H\text{.}$</p></li></ol></article></div>

<div class="mathbook-content"><p id="p-979">In all of our  examples the cosets of a subgroup $H$ partition the larger group $G\text{.}$  The following theorem proclaims that this will always be the case.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-cosets-theorem-2"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">6.4</span></h6><p id="p-980">Let $H$ be a subgroup of a group $G\text{.}$  Then the left cosets of $H$ in $G$ partition $G\text{.}$  That is, the group $G$ is the disjoint union of the left cosets of $H$ in $G\text{.}$</p></article><article class="proof" id="proof-36"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-981">Let $g_1 H$ and $g_2 H$ be two cosets of $H$ in $G\text{.}$  We must show that either $g_1 H \cap g_2 H = \emptyset$ or $g_1 H = g_2 H\text{.}$  Suppose that $g_1 H \cap g_2 H \neq \emptyset$ and $a \in g_1 H \cap g_2 H\text{.}$  Then by the definition of a left coset, $a = g_1 h_1 = g_2 h_2$ for some elements $h_1$ and $h_2$ in $H\text{.}$  Hence, $g_1 = g_2 h_2 h_1^{-1}$ or $g_1 \in g_2 H\text{.}$  By Lemma <a href="section-cosets.ipynb#lemma-cosets-theorem-1" class="xref" alt="Lemma 6.3 " title="Lemma 6.3 ">6.3</a>, $g_1 H = g_2 H\text{.}$</p></article></div>

<div class="mathbook-content"><article class="remark-like" id="remark-4"><h6 class="heading"><span class="type">Remark</span><span class="codenumber">6.5</span></h6><p id="p-982">There is nothing special in this theorem about left cosets.  Right cosets also partition $G\text{;}$ the proof of this fact is exactly the same as the proof for left cosets except that all group multiplications are done on the opposite side of $H\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-983">Let $G$ be a group and $H$ be a subgroup of $G\text{.}$  Define the <dfn class="terminology">index</dfn> of $H$ in $G$ to be the number of left cosets of $H$ in $G\text{.}$  We will denote the index by $[G:H]\text{.}$ </p></div>

<div class="mathbook-content"><article class="example-like" id="example-cosets-z6-index"><h6 class="heading"><span class="type">Example</span><span class="codenumber">6.6</span></h6><p id="p-984">Let $G= {\mathbb Z}_6$ and $H = \{ 0, 3 \}\text{.}$ Then $[G:H] = 3\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-cosets-s3-index"><h6 class="heading"><span class="type">Example</span><span class="codenumber">6.7</span></h6><p id="p-985">Suppose that $G= S_3\text{,}$ $H = \{ (1),(123), (132) \}\text{,}$ and $K= \{ (1), (12) \}\text{.}$  Then $[G:H] = 2$ and $[G:K] = 3\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-cosets-theorem-3"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">6.8</span></h6><p id="p-986">Let $H$ be a subgroup of a group $G\text{.}$  The number of left cosets of $H$ in $G$ is the same as the number of right cosets of $H$ in $G\text{.}$</p></article><article class="proof" id="proof-37"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-987">Let ${\mathcal L}_H$ and  ${\mathcal R}_H$ denote the set of left and right cosets of $H$ in $G\text{,}$ respectively.   If we can define a bijective map $\phi :  {\mathcal L}_H \rightarrow {\mathcal R}_H\text{,}$ then the theorem will be proved.  If $gH \in {\mathcal L}_H\text{,}$ let $\phi( gH ) = Hg^{-1}\text{.}$  By Lemma <a href="section-cosets.ipynb#lemma-cosets-theorem-1" class="xref" alt="Lemma 6.3 " title="Lemma 6.3 ">6.3</a>, the map $\phi$ is well-defined; that is, if $g_1 H = g_2 H\text{,}$ then $H g_1^{-1} = H g_2^{-1}\text{.}$  To show that $\phi$ is one-to-one, suppose that</p><div class="displaymath">
\begin{equation*}
H g_1^{-1} = \phi( g_1 H ) = \phi( g_2 H ) = H g_2^{-1}.
\end{equation*}
</div><p>Again by Lemma <a href="section-cosets.ipynb#lemma-cosets-theorem-1" class="xref" alt="Lemma 6.3 " title="Lemma 6.3 ">6.3</a>, $g_1 H = g_2 H\text{.}$  The map $\phi$ is onto since $\phi(g^{-1} H ) = H g\text{.}$</p></article></div>
