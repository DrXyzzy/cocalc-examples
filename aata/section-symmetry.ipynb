<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 12.2 Symmetry"><span class="type">Section</span><span class="codenumber">12.2</span><span class="title">Symmetry</span></h2><a href="section-symmetry.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-1850">An <dfn class="terminology">isometry</dfn> or <dfn class="terminology">rigid motion</dfn> in ${\mathbb R}^n$ is a distance-preserving function $f$ from ${\mathbb R}^n$ to ${\mathbb R}^n\text{.}$ This means that $f$ must satisfy</p><div class="displaymath">
\begin{equation*}
\| f({\mathbf x}) - f({\mathbf y}) \| =\|{\mathbf x} - {\mathbf y} \|
\end{equation*}
</div><p>for all ${\mathbf x}, {\mathbf y} \in {\mathbb R}^n\text{.}$ It is not difficult to show that $f$ must be a one-to-one map. By Theorem <a href="section-matrix-groups.ipynb#theorem-orthnormal-matrix" class="xref" alt="Theorem 12.8 " title="Theorem 12.8 ">12.8</a>, any element in $O(n)$ is an isometry on ${\mathbb R}^n\text{;}$ however, $O(n)$ does not include all possible isometries on ${\mathbb R}^n\text{.}$ Translation by a vector ${\mathbf x}\text{,}$ $T_{\mathbf y}({\mathbf x}) = {\mathbf x} + {\mathbf y}$ is also an isometry (Figure <a href="section-matrix-groups.ipynb#figure-isometries" class="xref" alt="Figure 12.11 " title="Figure 12.11 ">12.11</a>); however, $T$ cannot be in $O(n)$ since it is not a linear map.</p></div>

<div class="mathbook-content"><p id="p-1851">We are mostly interested in isometries in ${\mathbb R}^2\text{.}$ In fact, the only isometries in ${\mathbb R}^2$ are rotations and reflections  about the origin, translations, and combinations of the two. For example, a <dfn class="terminology">glide reflection</dfn> is a translation followed by a reflection (Figure <a href="section-symmetry.ipynb#figure-glide" class="xref" alt="Figure 12.12 " title="Figure 12.12 ">12.12</a>). In ${\mathbb R}^n$ all isometries are given in the same manner. The proof is very easy to generalize.</p></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-glide"><img src="images/matrix-glide-reflections.svg" width="100%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">12.12</span>Glide reflections</figcaption></figure></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-10"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">12.13</span></h6><p id="p-1852">An isometry $f$ that fixes the origin in ${\mathbb R}^2$ is a linear transformation.  In particular, $f$ is given by an element in $O(2)\text{.}$</p></article><article class="proof" id="proof-78"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1853">Let $f$ be an isometry in ${\mathbb R}^2$ fixing the origin. We will first show that $f$ preserves inner products. Since $f(0) = 0\text{,}$ $\| f({\mathbf x})\| = \| {\mathbf x} \|\text{;}$ therefore,</p><div class="displaymath">
\begin{align*}
\| {\mathbf x} \|^2 - 2 \langle f({\mathbf x}), f({\mathbf y}) \rangle + \| {\mathbf y} \|^2  & = \| f({\mathbf x}) \|^2 - 2 \langle f({\mathbf x}), f({\mathbf y}) \rangle + \| f({\mathbf y}) \|^2\\
& = \langle f({\mathbf x}) -  f({\mathbf y}), f({\mathbf x}) -  f({\mathbf y}) \rangle\\
& = \| f({\mathbf x}) -  f({\mathbf y}) \|^2\\
& = \| {\mathbf x} -  {\mathbf y} \|^2\\
& = \langle {\mathbf x} -  {\mathbf y}, {\mathbf x} -  {\mathbf y} \rangle\\
& = \| {\mathbf x} \|^2 - 2 \langle {\mathbf x}, {\mathbf y} \rangle + \| {\mathbf y} \|^2.
\end{align*}
</div><p>Consequently,</p><div class="displaymath">
\begin{equation*}
\langle f({\mathbf x}), f({\mathbf y}) \rangle = \langle {\mathbf x}, {\mathbf y} \rangle.
\end{equation*}
</div><p>Now let ${\mathbf e}_1$ and ${\mathbf e_2}$ be $(1, 0)^{\rm t}$ and $(0, 1)^{\rm t}\text{,}$ respectively. If</p><div class="displaymath">
\begin{equation*}
{\mathbf x} = (x_1, x_2) = x_1 {\mathbf e}_1 + x_2 {\mathbf e}_2,
\end{equation*}
</div><p>then</p><div class="displaymath">
\begin{equation*}
f({\mathbf x}) = \langle f({\mathbf x}), f({\mathbf e}_1) \rangle f({\mathbf e}_1) + \langle f({\mathbf x}), f({\mathbf e}_2) \rangle f({\mathbf e}_2) = x_1 f({\mathbf e}_1)+x_2 f({\mathbf e}_2).
\end{equation*}
</div><p>The linearity of $f$ easily follows.</p></article></div>

<div class="mathbook-content"><p id="p-1854">For any arbitrary isometry, $f\text{,}$  $T_{\mathbf x} f$ will fix the origin for some vector ${\mathbf x}$ in ${\mathbb R}^2\text{;}$ hence, $T_{\mathbf x} f({\mathbf y}) = A {\mathbf y}$ for some matrix $A \in O(2)\text{.}$ Consequently, $f({\mathbf y}) = A {\mathbf y} + {\mathbf x}\text{.}$  Given the isometries</p><div class="displaymath">
\begin{align*}
f({\mathbf y}) & = A {\mathbf y} + {\mathbf x}_1\\
g({\mathbf y}) & = B {\mathbf y} + {\mathbf x}_2,
\end{align*}
</div><p>their composition is</p><div class="displaymath">
\begin{equation*}
f(g({\mathbf y})) = f(B {\mathbf y} + {\mathbf x}_2) = AB {\mathbf y} + A{\mathbf x}_2 + {\mathbf x}_1.
\end{equation*}
</div><p>This last computation allows us to identify the group of isometries on ${\mathbb R}^2$ with $E(2)\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-59"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">12.14</span></h6><p id="p-1855">The group of isometries on ${\mathbb R}^2$ is the Euclidean group, $E(2)\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1856">A <dfn class="terminology">symmetry group</dfn> in ${\mathbb R}^n$ is a subgroup of the group of isometries on ${\mathbb R}^n$ that fixes a set of points $X \subset {\mathbb R}^2\text{.}$  It is important to realize that the symmetry group of $X$ depends <em class="emphasis">both</em> on ${\mathbb R}^n$ and on $X\text{.}$ For example, the symmetry group of the origin in ${\mathbb R}^1$ is ${\mathbb Z}_2\text{,}$ but the symmetry group of the origin in ${\mathbb R}^2$ is $O(2)\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-60"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">12.15</span></h6><p id="p-1857">The only finite symmetry groups in ${\mathbb R}^2$ are ${\mathbb Z}_n$ and $D_n\text{.}$</p></article><article class="proof" id="proof-79"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1858">Let $G = \{ f_1, f_2, \ldots, f_n \}$ be a finite symmetry group that fixes a set of points in $X \subset \mathbb R^2\text{.}$  Choose a point $\mathbf x \in X\text{.}$  This point may not be a fixed point—it could be moved by $G$ to another point in $X\text{.}$ Define a set $S =\{ \mathbf y_1, \mathbf y_2, \ldots \mathbf y_n\}\text{,}$ where $\mathbf y_i = f_i(\mathbf x)\text{.}$  Now, let</p><div class="displaymath">
\begin{equation*}
\mathbf z = \frac{1}{n} \sum_{i=1}^n \mathbf x_i.
\end{equation*}
</div><p>While the point $\mathbf z$ is not necessarily in the set $X\text{,}$ it is fixed by every element in the symetry group. Without loss of generality, we may now assume that $\mathbf z$ is the origin.</p><p id="p-1859">Any finite symmetry group $G$ in ${\mathbb R}^2$ that fixes the origin must be a finite subgroup of $O(2)\text{,}$ since translations and glide reflections have infinite order.  By Example <a href="section-matrix-groups.ipynb#example-o2" class="xref" alt="Example 12.10 " title="Example 12.10 ">12.10</a>, elements in $O(2)$ are either rotations of the form</p><div class="displaymath">
\begin{equation*}
R_{\theta}
=
\begin{pmatrix}
\cos \theta & - \sin \theta \\
\sin \theta & \cos \theta
\end{pmatrix}
\end{equation*}
</div><p>or reflections of the form</p><div class="displaymath">
\begin{equation*}
T_{\phi}
=
\begin{pmatrix}
\cos \phi &  - \sin \phi \\
\sin \phi & \cos \phi
\end{pmatrix}
\begin{pmatrix}
1 & 0 \\
0 & -1
\end{pmatrix}
=
\begin{pmatrix}
\cos \phi &  \sin \phi \\
\sin \phi & - \cos \phi
\end{pmatrix}.
\end{equation*}
</div><p>Notice that $\det(R_{\theta})=1\text{,}$  $\det(T_{\phi})=-1\text{,}$ and $T_{\phi}^2=I\text{.}$ We can divide the proof up into two cases. In the first case, all of the elements in $G$ have determinant one. In the second case, there exists at least one element in $G$ with determinant $-1\text{.}$</p><article class="case" id="case-1"><h6 class="heading">Case 1.</h6><p id="p-1860">The determinant of every element in $G$ is one. In this case every element in $G$ must be a rotation. Since $G$ is finite, there is a smallest angle, say $\theta_0\text{,}$ such that the corresponding element $R_{\theta_0}$ is the smallest rotation in the positive direction.  We claim that $R_{\theta_0}$ generates $G\text{.}$  If not, then for some positive integer $n$ there is an angle $\theta_1$ between $n \theta_0$ and $(n+1) \theta_0\text{.}$ If so, then $(n+1) \theta_0 - \theta_1$ corresponds to a rotation smaller than $\theta_0\text{,}$ which contradicts the minimality of $\theta_0\text{.}$</p></article><article class="case" id="case-2"><h6 class="heading">Case 2.</h6><p id="p-1861">The group $G$ contains a reflection $T\text{.}$  The kernel of the homomorphism $\phi : G \rightarrow \{-1, 1\}$ given by $A \mapsto \det(A)$ consists of elements whose determinant is 1.  Therefore, $|G/ \ker \phi|=2\text{.}$  We know that the kernel is cyclic by the first case and is a subgroup of $G$ of, say, order $n\text{.}$ Hence, $|G| = 2n\text{.}$ The elements of $G$ are</p><div class="displaymath">
\begin{equation*}
R_{\theta}, \ldots, R_{\theta}^{n-1},  TR_{\theta}, \ldots, TR_{\theta}^{n-1}.
\end{equation*}
</div><p>These elements satisfy the relation</p><div class="displaymath">
\begin{equation*}
TR_{\theta}T = R_{\theta}^{-1}.
\end{equation*}
</div><p>Consequently, $G$ must be isomorphic to $D_n$ in this case.</p></article></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  The Wallpaper Groups"><span class="type">Subsection</span><span class="codenumber" /><span class="title">The Wallpaper Groups</span></h3><a href="section-symmetry.ipynb#matrix-subsection-wallpaper" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-1862">Suppose that we wish to study wallpaper patterns in the plane or crystals in three dimensions. Wallpaper patterns are simply repeating patterns in the plane (Figure <a href="section-symmetry.ipynb#figure-wallpaper" class="xref" alt="Figure 12.16 " title="Figure 12.16 ">12.16</a>). The analogs of wallpaper patterns in ${\mathbb R}^3$ are crystals, which we can think of as repeating patterns of molecules in three dimensions (Figure <a href="section-symmetry.ipynb#figure-crystals" class="xref" alt="Figure 12.17 " title="Figure 12.17 ">12.17</a>). The mathematical equivalent of a wallpaper or crystal pattern is called a  lattice.</p></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-wallpaper"><img src="images/matrix-wallpaper-r2.svg" width="100%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">12.16</span>A wallpaper pattern in $\mathbb R^2$</figcaption></figure></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-crystals"><img src="images/matrix-crystal-r3.svg" width="50%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">12.17</span>A crystal structure in $\mathbb R^3$</figcaption></figure></div>

<div class="mathbook-content"><p id="p-1863">Let us examine wallpaper patterns in the plane a little more closely. Suppose that ${\mathbf x}$ and ${\mathbf y}$ are linearly independent vectors in ${\mathbb R}^2\text{;}$ that is, one vector cannot be a scalar multiple of the other. A <dfn class="terminology">lattice</dfn> of ${\mathbf x}$ and ${\mathbf y}$ is the set of all linear combinations $m {\mathbf x} + n {\mathbf y}\text{,}$ where $m$ and $n$ are integers. The vectors ${\mathbf x}$ and ${\mathbf y}$ are said to be a <dfn class="terminology">basis</dfn> for the lattice.</p></div>

<div class="mathbook-content"><p id="p-1864">Notice that a lattice can have several bases. For example, the vectors $(1,1)^{\rm t}$ and $(2,0)^{\rm t}$ have the  same lattice as the vectors $(-1, 1)^{\rm t}$ and $(-1, -1)^{\rm t}$ (Figure <a href="section-symmetry.ipynb#figure-lattice" class="xref" alt="Figure 12.18 " title="Figure 12.18 ">12.18</a>). However, any lattice is completely determined by a basis. Given two bases for the same lattice, say $\{ {\mathbf x}_1, {\mathbf x}_2 \}$ and $\{ {\mathbf y}_1, {\mathbf y}_2 \}\text{,}$ we can write</p><div class="displaymath">
\begin{align*}
{\mathbf y}_1 & = \alpha_1  {\mathbf x}_1 + \alpha_2 {\mathbf x}_2\\
{\mathbf y}_2 & = \beta_1  {\mathbf x}_1 + \beta_2 {\mathbf x}_2,
\end{align*}
</div><p>where $\alpha_1\text{,}$ $\alpha_2\text{,}$ $\beta_1\text{,}$ and $\beta_2$ are integers. The matrix corresponding to this transformation is</p><div class="displaymath">
\begin{equation*}
U
=
\begin{pmatrix}
\alpha_1 & \alpha_2 \\
\beta_1 & \beta_2
\end{pmatrix}.
\end{equation*}
</div><p>If we wish to give ${\mathbf x}_1$ and ${\mathbf x}_2$ in terms of ${\mathbf y}_1$ and ${\mathbf y}_2\text{,}$ we need only calculate $U^{-1}\text{;}$ that is,</p><div class="displaymath">
\begin{equation*}
U^{-1}
\begin{pmatrix}
{\mathbf y}_1 \\ {\mathbf y}_2
\end{pmatrix}
=
\begin{pmatrix}
{\mathbf x}_1 \\ {\mathbf x}_2
\end{pmatrix}.
\end{equation*}
</div><p>Since $U$ has integer entries, $U^{-1}$ must also have integer entries; hence the determinants of both $U$ and $U^{-1}$ must be integers. Because $U U^{-1} = I\text{,}$</p><div class="displaymath">
\begin{equation*}
\det(U U^{-1}) =\det(U) \det( U^{-1}) = 1;
\end{equation*}
</div><p>consequently, $\det(U) = \pm 1\text{.}$ A matrix with determinant $\pm 1$ and integer entries is called <dfn class="terminology">unimodular</dfn>. For example, the matrix</p><div class="displaymath">
\begin{equation*}
\begin{pmatrix}
3 & 1 \\
5 & 2
\end{pmatrix}
\end{equation*}
</div><p>is unimodular. It should be clear that there is a minimum length for vectors in a lattice.</p></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-lattice"><img src="images/matrix-lattice-r2.svg" width="100%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">12.18</span>A lattice in  $\mathbb R^2$</figcaption></figure></div>

<div class="mathbook-content"><p id="p-1865">We can classify lattices by studying their symmetry groups. The symmetry group of a lattice is the subgroup of $E(2)$ that maps the lattice to itself. We consider two lattices in ${\mathbb R}^2$ to be equivalent if they have the same symmetry group.  Similarly, classification of crystals in ${\mathbb R}^3$ is accomplished by associating a symmetry group, called a <dfn class="terminology">space group</dfn>, with each type of crystal. Two lattices are considered different if their space groups are not the same.  The natural question that now arises is how many space groups exist.</p></div>

<div class="mathbook-content"><p id="p-1866">A space group is composed of two parts: a <dfn class="terminology">translation subgroup</dfn> and a <dfn class="terminology">point</dfn>.  The translation subgroup is an infinite abelian subgroup of the space group made up of the translational symmetries of the crystal; the point group is a finite group  consisting  of rotations and reflections of the crystal about a point. More specifically, a space group is a subgroup of $G \subset E(2)$ whose translations are a set of the form $\{ (I, t) : t \in L \}\text{,}$ where $L$ is a lattice. Space groups are, of course, infinite. Using geometric arguments, we can prove the following theorem (see [5] or [6]).</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-61"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">12.19</span></h6><p id="p-1867">Every translation group in ${\mathbb R}^2$ is isomorphic to ${\mathbb Z} \times {\mathbb Z}\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1868">The point group of $G$ is $G_0 = \{A : (A,b) \in G \text{ for some } b \}\text{.}$ In particular, $G_0$ must be a subgroup of $O(2)\text{.}$ Suppose that ${\mathbf x}$ is a vector in a lattice $L$ with space group $G\text{,}$ translation group $H\text{,}$ and point group $G_0\text{.}$ For any element $(A, {\mathbf y})$ in $G\text{,}$</p><div class="displaymath">
\begin{align*}
(A, {\mathbf y}) (I, {\mathbf x}) (A, {\mathbf y})^{-1} & = (A,A {\mathbf x} + {\mathbf y}) (A^{-1},-A^{-1} {\mathbf y})\\
& = (A A^{-1},-A A^{-1} {\mathbf y} + A {\mathbf x} + {\mathbf y})\\
& = (I, A {\mathbf x});
\end{align*}
</div><p>hence, $(I, A {\mathbf x})$ is in the translation group of $G\text{.}$ More specifically, $A {\mathbf x}$ must be in the lattice $L\text{.}$ It is important to note that $G_0$ is not usually a subgroup of the space group $G\text{;}$ however, if $T$ is the translation subgroup of $G\text{,}$ then $G/T \cong G_0\text{.}$ The proof of the following theorem can be found in [2], [5], or [6].</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-62"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">12.20</span></h6><p id="p-1869">The point group in the wallpaper groups is isomorphic to ${\mathbb Z}_n$ or $D_n\text{,}$ where $n = 1, 2, 3, 4, 6\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1870">To answer the question of how the point groups and the translation groups can be combined, we must look at the different types of lattices. Lattices can be classified by the structure of a single lattice cell. The possible cell shapes are parallelogram, rectangular, square, rhombic, and hexagonal (Figure <a href="section-symmetry.ipynb#figure-lattice-types" class="xref" alt="Figure 12.21 " title="Figure 12.21 ">12.21</a>). The wallpaper groups can now be classified according to the types of reflections that occur in each group: these are ordinarily reflections, glide reflections, both, or none.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Historical Note"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Historical Note</span></h3><a href="section-symmetry.ipynb#matrix-subsection-historical-note" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-1874">Symmetry groups have intrigued mathematicians for a long time. Leonardo da Vinci was probably the first person to know all of the point groups.  At the International Congress of Mathematicians in 1900, David Hilbert gave a now-famous address outlining 23 problems to guide mathematics in the twentieth century.  Hilbert's eighteenth problem asked whether or not crystallographic groups in $n$ dimensions were always finite.  In 1910, L. Bieberbach proved that crystallographic groups are finite in every dimension.  Finding out how many of these groups there are in each dimension is another matter. In ${\mathbb R}^3$ there are 230 different space groups; in ${\mathbb R}^4$ there are 4783. No one has been able to compute the number of space groups for ${\mathbb R}^5$ and beyond. It is interesting to note that the crystallographic groups were found mathematically for ${\mathbb R}^3$ before the 230 different types of crystals were actually discovered in nature.</p></div>
