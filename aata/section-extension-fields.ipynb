<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 21.1 Extension Fields"><span class="type">Section</span><span class="codenumber">21.1</span><span class="title">Extension Fields</span></h2><a href="section-extension-fields.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-3216">A field $E$ is an <dfn class="terminology">extension field</dfn> of a field $F$ if $F$ is a subfield of $E\text{.}$ The field $F$ is called the <dfn class="terminology">base field</dfn>.  We write $F \subset E\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-fields-q-sqrt2"><h6 class="heading"><span class="type">Example</span><span class="codenumber">21.1</span></h6><p id="p-3217">For example, let</p><div class="displaymath">
\begin{equation*}
F = {\mathbb Q}( \sqrt{2}\,) = \{ a + b \sqrt{2} : a, b \in {\mathbb Q} \}
\end{equation*}
</div><p>and let $E =  {\mathbb Q }( \sqrt{2} +  \sqrt{3}\,)$ be the smallest field containing both ${\mathbb Q}$ and $\sqrt{2} + \sqrt{3}\text{.}$ Both $E$ and $F$ are extension fields of the rational numbers. We claim that $E$ is an extension field of $F\text{.}$ To see this, we need only show that $\sqrt{2}$ is in $E\text{.}$ Since $\sqrt{2} + \sqrt{3}$ is in $E\text{,}$ $1 / (\sqrt{2} + \sqrt{3}\,) = \sqrt{3} - \sqrt{2}$ must also be in $E\text{.}$ Taking linear combinations of $\sqrt{2} + \sqrt{3}$ and $\sqrt{3} - \sqrt{2}\text{,}$ we find that $\sqrt{2}$ and $\sqrt{3}$ must both be in $E\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-fields-finite-f4"><h6 class="heading"><span class="type">Example</span><span class="codenumber">21.2</span></h6><p id="p-3218">Let $p(x) = x^2 + x + 1 \in {\mathbb Z}_2[x]\text{.}$ Since neither 0 nor 1 is a root of this polynomial, we know that $p(x)$ is irreducible over ${\mathbb Z}_2\text{.}$ We will construct a field extension of ${\mathbb Z}_2$ containing an element $\alpha$ such that $p(\alpha) = 0\text{.}$ By Theorem <a href="section-irreducible-poly.ipynb#theorem-max-ideal-poly" class="xref" alt="Theorem 17.22 " title="Theorem 17.22 ">17.22</a>, the ideal $\langle p(x) \rangle$ generated by $p(x)$ is maximal; hence, ${\mathbb Z}_2[x] / \langle p(x) \rangle$ is a field.  Let $f(x) + \langle p(x) \rangle$ be an arbitrary element of ${\mathbb Z}_2[x] / \langle p(x) \rangle\text{.}$ By the division algorithm,</p><div class="displaymath">
\begin{equation*}
f(x) = (x^2 + x + 1) q(x) + r(x),
\end{equation*}
</div><p>where the degree of $r(x)$ is less than the degree of $x^2 + x + 1\text{.}$ Therefore,</p><div class="displaymath">
\begin{equation*}
f(x) + \langle x^2 + x + 1 \rangle = r(x) + \langle x^2 + x + 1 \rangle.
\end{equation*}
</div><p>The only possibilities for $r(x)$ are then $0\text{,}$ $1\text{,}$ $x\text{,}$ and $1 + x\text{.}$ Consequently, $E = {\mathbb Z}_2[x] / \langle x^2 + x + 1 \rangle$ is a field with four elements and must be a field extension of ${\mathbb Z}_2\text{,}$ containing a zero $\alpha$ of $p(x)\text{.}$ The field ${\mathbb Z}_2(  \alpha)$ consists of elements</p><div class="displaymath">
\begin{align*}
0 + 0 \alpha & = 0\\
1 + 0 \alpha & = 1\\
0 + 1 \alpha & = \alpha\\
1 + 1 \alpha & = 1 + \alpha.
\end{align*}
</div><p>Notice that ${\alpha}^2 + {\alpha} + 1 = 0\text{;}$ hence, if we compute $(1 + \alpha)^2\text{,}$</p><div class="displaymath">
\begin{equation*}
(1 + \alpha)(1 + \alpha)= 1 + \alpha + \alpha + (\alpha)^2 = \alpha.
\end{equation*}
</div><p>Other calculations are accomplished in a similar manner.  We summarize these computations in the following tables, which tell us how to add and multiply elements in $E\text{.}$</p></article></div>

<div class="mathbook-content"><figure class="figure-like" id="table-addition-z2-alpha"><div class="displaymath" id="p-3219">
\begin{equation*}
\begin{array}{c|cccc}
+ & 0 & 1 & \alpha & 1 + \alpha \\
\hline
0 & 0 & 1 & \alpha & 1 + \alpha \\
1 & 1 & 0 & 1 + \alpha & \alpha \\
\alpha & \alpha & 1 + \alpha & 0 & 1  \\
1 + \alpha & 1 + \alpha & \alpha & 1 & 0
\end{array}
\end{equation*}
</div><figcaption><span class="type">Table</span><span class="codenumber">21.3</span>Addition Table for ${\mathbb Z}_2(\alpha)$</figcaption></figure></div>

<div class="mathbook-content"><figure class="figure-like" id="table-multiplication-z2-alpha"><div class="displaymath" id="p-3220">
\begin{equation*}
\begin{array}{c|cccc}
\cdot & 0 & 1 & \alpha & 1 + \alpha \\
\hline
0 & 0 & 0 & 0 & 0 \\
1 & 0 & 1 & \alpha & 1 + \alpha \\
\alpha & 0 & \alpha & 1 + \alpha & 1  \\
1 + \alpha & 0 & 1 + \alpha & 1 & \alpha
\end{array}
\end{equation*}
</div><figcaption><span class="type">Table</span><span class="codenumber">21.4</span>Multiplication Table for ${\mathbb Z}_2(\alpha)$</figcaption></figure></div>

<div class="mathbook-content"><p id="p-3221">The following theorem, due to Kronecker, is so important and so basic  to our understanding of fields that it is often known as the Fundamental Theorem of Field Theory.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-fundamental-of-field-theory"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.5</span></h6><p id="p-3222">Let $F$ be a field and let $p(x)$ be a nonconstant polynomial in $F[x]\text{.}$  Then there exists an extension field $E$ of $F$ and an  element $\alpha \in E$ such that $p(\alpha) = 0\text{.}$</p></article><article class="proof" id="proof-160"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3223">To prove this theorem, we will employ the method that we used to construct Example <a href="section-extension-fields.ipynb#example-fields-finite-f4" class="xref" alt="Example 21.2 " title="Example 21.2 ">21.2</a>. Clearly, we can assume that $p(x)$ is an irreducible polynomial. We wish to find an extension field $E$ of $F$ containing an element $\alpha$ such that $p(\alpha) = 0\text{.}$ The ideal $\langle p(x) \rangle$ generated by $p(x)$ is a maximal ideal in $F[x]$ by Theorem <a href="section-irreducible-poly.ipynb#theorem-max-ideal-poly" class="xref" alt="Theorem 17.22 " title="Theorem 17.22 ">17.22</a>; hence, $F[x]/\langle p(x) \rangle$ is a field. We claim that $E = F[x]/\langle p(x) \rangle$ is the desired field.</p><p id="p-3224">We first show that $E$ is a field extension of $F\text{.}$ We can define a homomorphism of commutative rings by the map $\psi:F \rightarrow F[x]/\langle p(x) \rangle\text{,}$ where $\psi(a) = a + \langle p(x)\rangle$ for $a \in F\text{.}$ It is easy to check that $\psi$ is indeed a ring homomorphism. Observe that</p><div class="displaymath">
\begin{equation*}
\psi( a ) + \psi( b ) = (a + \langle p(x) \rangle) + (b + \langle p(x) \rangle) = (a + b) + \langle p(x) \rangle = \psi( a + b )
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
\psi( a )  \psi( b ) = (a  + \langle p(x) \rangle)  (b + \langle p(x) \rangle) = a  b  + \langle p(x) \rangle = \psi( a  b ).
\end{equation*}
</div><p>To prove that $\psi$ is one-to-one, assume that</p><div class="displaymath">
\begin{equation*}
a + \langle p(x) \rangle = \psi(a) = \psi(b) = b + \langle p(x) \rangle.
\end{equation*}
</div><p>Then $a - b$ is a multiple of $p(x)\text{,}$ since it lives in the ideal $\langle p(x) \rangle\text{.}$ Since $p(x)$ is a nonconstant polynomial, the only possibility is that $a - b = 0\text{.}$ Consequently, $a = b$ and $\psi$ is injective. Since $\psi$ is one-to-one, we can identify $F$ with the subfield $\{ a + \langle p(x) \rangle : a \in F \}$ of $E$ and view $E$ as an extension field of $F\text{.}$</p><p id="p-3225">It remains for us to prove that $p(x)$ has a zero $\alpha \in E\text{.}$ Set $\alpha = x + \langle p(x) \rangle\text{.}$ Then $\alpha$ is in $E\text{.}$ If $p(x) = a_0 + a_1 x + \cdots + a_n x^n\text{,}$ then</p><div class="displaymath">
\begin{align*}
p( \alpha ) & =  a_0 + a_1( x + \langle p(x) \rangle) + \cdots + a_n ( x + \langle p(x) \rangle)^n\\
& = a_0 + ( a_1 x + \langle p(x) \rangle) + \cdots + (a_n  x^n + \langle p(x) \rangle)\\
& = a_0 + a_1 x + \cdots + a_n x^n + \langle p(x) \rangle\\
& = 0 + \langle p(x) \rangle.
\end{align*}
</div><p>Therefore, we have found an element $\alpha \in E = F[x]/\langle p(x) \rangle$ such that $\alpha$ is a zero of $p(x)\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-fields-finite-f8"><h6 class="heading"><span class="type">Example</span><span class="codenumber">21.6</span></h6><p id="p-3226">Let $p(x) = x^5 + x^4 + 1 \in {\mathbb Z}_2[x]\text{.}$ Then $p(x)$ has irreducible factors $x^2 + x + 1$ and $x^3 + x + 1\text{.}$ For a field extension $E$ of ${\mathbb Z}_2$ such that $p(x)$ has a root in $E\text{,}$ we can let $E$ be either ${\mathbb Z}_2[x] / \langle x^2 + x + 1 \rangle$ or ${\mathbb Z}_2[x] / \langle x^3 + x + 1 \rangle\text{.}$  We will leave it as an exercise to show that ${\mathbb Z}_2[x] / \langle x^3 + x + 1 \rangle$ is a field with $2^3=8$ elements.</p></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Algebraic Elements"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Algebraic Elements</span></h3><a href="section-extension-fields.ipynb#fields-subsection-algebraic-elements" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-3227">An element $\alpha$ in an extension field $E$ over $F$ is <dfn class="terminology">algebraic</dfn> over $F$ if $f(\alpha)=0$ for some nonzero polynomial $f(x) \in F[x]\text{.}$ An element in $E$ that is not algebraic over $F$ is <dfn class="terminology">transcendental</dfn> over $F\text{.}$ An extension field $E$ of a field $F$ is an <dfn class="terminology">algebraic  extension</dfn> of $F$ if every element in $E$ is algebraic over $F\text{.}$ If $E$ is a field extension of $F$ and $\alpha_1, \ldots, \alpha_n$ are contained in $E\text{,}$ we denote the smallest field containing $F$ and $\alpha_1, \ldots, \alpha_n$ by $F( \alpha_1, \ldots, \alpha_n)\text{.}$  If $E = F( \alpha )$ for some $\alpha \in E\text{,}$ then $E$ is a <dfn class="terminology">simple extension</dfn> of $F\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-fields-extension-pi"><h6 class="heading"><span class="type">Example</span><span class="codenumber">21.7</span></h6><p id="p-3228">Both $\sqrt{2}$ and $i$ are algebraic over ${\mathbb Q}$ since they are zeros of the polynomials $x^2 -2$ and $x^2 + 1\text{,}$ respectively. Clearly $\pi$ and $e$ are algebraic over the real numbers; however, it is a nontrivial fact that they are transcendental over ${\mathbb Q}\text{.}$ Numbers in ${\mathbb R}$ that are algebraic over ${\mathbb Q}$ are in fact quite rare. Almost all real numbers are transcendental over ${\mathbb Q}\text{.}$<span class="footnote"><a knowl="" class="id-ref" refid="hk-fn-6" id="fn-6"><sup> 6 </sup></a></span><span id="hk-fn-6" class="hidden-content tex2jax_ignore"><span class="footnote">The probability that a real number chosen at random from the interval $[0, 1]$ will be transcendental over the rational numbers is one.</span></span> (In many cases we do not know whether or not a particular number is transcendental; for example, it is still not known whether $\pi + e$ is transcendental or algebraic.)</p></article></div>

<div class="mathbook-content"><p id="p-3229">A complex number that is algebraic over ${\mathbb Q}$ is an <dfn class="terminology">algebraic number</dfn>. A <dfn class="terminology">transcendental number</dfn> is an element of ${\mathbb C}$ that is transcendental over ${\mathbb Q}\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-fields-sqrt2-sqrt3"><h6 class="heading"><span class="type">Example</span><span class="codenumber">21.8</span></h6><p id="p-3230">We will show that $\sqrt{2 + \sqrt{3} }$ is algebraic over ${\mathbb Q}\text{.}$ If $\alpha = \sqrt{2 + \sqrt{3} }\text{,}$ then $\alpha^2 = 2 + \sqrt{3}\text{.}$ Hence, $\alpha^2 -  2 = \sqrt{3}$ and $( \alpha^2 -  2)^2 = 3\text{.}$ Since $\alpha^4 - 4 \alpha^2 + 1 = 0\text{,}$ it must be true that $\alpha$ is a zero of the polynomial $x^4 - 4 x^2 + 1 \in {\mathbb Q}[x]\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3231">It is very easy to give an example of an extension field $E$ over a field $F\text{,}$ where $E$ contains an element transcendental over $F\text{.}$ The following theorem characterizes transcendental extensions.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-111"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.9</span></h6><p id="p-3232">Let $E$ be an extension field of $F$ and $\alpha \in E\text{.}$ Then $\alpha$ is transcendental over $F$ if and only if $F( \alpha )$ is isomorphic to $F(x)\text{,}$ the field of fractions of $F[x]\text{.}$</p></article><article class="proof" id="proof-161"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3233">Let $\phi_{\alpha} : F[x] \rightarrow E$ be the evaluation homomorphism for $\alpha\text{.}$ Then $\alpha$ is transcendental over  $F$ if and only if $\phi_{\alpha} (p(x)) = p(\alpha) \neq 0$ for all  nonconstant polynomials $p(x) \in F[x]\text{.}$  This is true if and only  if $\ker \phi_{\alpha} = \{ 0 \}\text{;}$ that is, it is true exactly when  $\phi_{\alpha}$ is one-to-one. Hence, $E$ must contain a copy of $F[x]\text{.}$  The smallest field containing $F[x]$ is the field of fractions $F(x)\text{.}$  By Theorem <a href="section-field-of-fractions.ipynb#theorem-field-of-quotients" class="xref" alt="Theorem 18.4 " title="Theorem 18.4 ">18.4</a>, $E$ must contain a copy of this field.</p></article></div>

<div class="mathbook-content"><p id="p-3234">We have a more interesting situation in the case of algebraic extensions.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-112"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.10</span></h6><p id="p-3235">Let $E$ be an extension field of a field $F$ and $\alpha \in E$ with $\alpha$ algebraic over $F\text{.}$ Then there is a unique irreducible monic polynomial $p(x) \in F[x]$ of smallest degree such that $p( \alpha ) = 0\text{.}$ If $f(x)$ is another polynomial in $F[x]$ such that  $f(\alpha) = 0\text{,}$ then $p(x)$ divides $f(x)\text{.}$</p></article><article class="proof" id="proof-162"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3236">Let $\phi_{\alpha} : F[x] \rightarrow E$ be the evaluation homomorphism. The kernel of $\phi_{\alpha}$ is a principal ideal generated by some $p(x) \in F[x]$ with $\deg p(x) \geq 1\text{.}$ We know that such a polynomial exists, since $F[x]$ is a principal ideal domain and $\alpha$ is algebraic. The ideal $\langle p(x) \rangle$ consists exactly of those elements of $F[x]$ having $\alpha$ as a zero. If $f( \alpha ) = 0$ and $f(x)$ is not the zero polynomial, then $f(x) \in \langle p(x) \rangle$ and $p(x)$ divides $f(x)\text{.}$ So $p(x)$ is a polynomial of minimal degree having $\alpha$ as a zero. Any other polynomial of the same degree having $\alpha$ as a zero must have the form $\beta p( x)$ for some $\beta \in F\text{.}$</p><p id="p-3237">Suppose now that $p(x) = r(x) s(x)$ is a factorization of $p(x)$ into polynomials of lower degree. Since $p( \alpha ) = 0\text{,}$ $r( \alpha ) s( \alpha ) = 0\text{;}$ consequently, either $r( \alpha )=0$ or $s( \alpha ) = 0\text{,}$ which contradicts the fact that $p$ is of minimal degree. Therefore, $p(x)$ must be irreducible.</p></article></div>

<div class="mathbook-content"><p id="p-3238">Let $E$ be an extension field of $F$ and $\alpha \in E$ be algebraic over $F\text{.}$ The unique monic polynomial $p(x)$ of the last theorem is called the <dfn class="terminology">minimal polynomial</dfn> for $\alpha$ over $F\text{.}$ The degree of $p(x)$ is the <dfn class="terminology">degree of $\alpha$ over $F$</dfn>.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-fields-minimal-poly"><h6 class="heading"><span class="type">Example</span><span class="codenumber">21.11</span></h6><p id="p-3239">Let $f(x) = x^2 - 2$ and $g(x) = x^4 - 4 x^2 + 1\text{.}$ These polynomials are the minimal polynomials of $\sqrt{2}$ and $\sqrt{2 + \sqrt{3} }\text{,}$ respectively.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-minimal-poly"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">21.12</span></h6><p id="p-3240">Let $E$ be a field extension of $F$ and $\alpha \in E$ be algebraic over $F\text{.}$  Then $F( \alpha ) \cong F[x] / \langle p(x) \rangle\text{,}$ where $p(x)$ is the minimal polynomial of $\alpha$ over $F\text{.}$</p></article><article class="proof" id="proof-163"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3241">Let $\phi_{\alpha} : F[x] \rightarrow E$ be the evaluation homomorphism. The kernel of this map is $\langle p(x) \rangle\text{,}$ where $p(x)$ is the minimal polynomial  of $\alpha\text{.}$ By the First Isomorphism Theorem for rings, the image of $\phi_{\alpha}$ in $E$ is isomorphic to $F( \alpha )$ since it contains both $F$ and $\alpha\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-simple_extension"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.13</span></h6><p id="p-3242">Let $E = F( \alpha )$ be a simple extension of $F\text{,}$ where $\alpha \in E$ is algebraic over $F\text{.}$  Suppose that the degree of $\alpha$ over $F$ is $n\text{.}$ Then every element $\beta \in E$ can be expressed uniquely in the form</p><div class="displaymath">
\begin{equation*}
\beta = b_0 + b_1 \alpha + \cdots + b_{n-1} \alpha^{n - 1}
\end{equation*}
</div><p>for $b_i \in F\text{.}$</p></article><article class="proof" id="proof-164"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3243">Since $\phi_{\alpha} ( F[x] ) \cong F( \alpha )\text{,}$  every element in $E = F( \alpha )$ must be of the form $\phi_{\alpha} ( f(x) ) = f( \alpha )\text{,}$ where $f(\alpha)$ is a polynomial in $\alpha$ with coefficients in $F\text{.}$ Let</p><div class="displaymath">
\begin{equation*}
p(x) = x^n + a_{n - 1} x^{n - 1} + \cdots + a_0
\end{equation*}
</div><p>be the minimal polynomial of $\alpha\text{.}$ Then $p( \alpha ) = 0\text{;}$ hence,</p><div class="displaymath">
\begin{equation*}
{\alpha}^n = - a_{n - 1} {\alpha}^{n - 1} - \cdots - a_0.
\end{equation*}
</div><p>Similarly,</p><div class="displaymath">
\begin{align*}
{\alpha}^{n + 1} & = {\alpha} {\alpha}^n\\
& = - a_{n - 1} {\alpha}^n - a_{n - 2} {\alpha}^{n - 1} - \cdots - a_0 {\alpha}\\
& = - a_{n - 1}( - a_{n - 1} {\alpha}^{n - 1} - \cdots - a_0) - a_{n - 2} {\alpha}^{n - 1} - \cdots - a_0 {\alpha}.
\end{align*}
</div><p>Continuing in this manner, we can express every monomial ${\alpha}^m\text{,}$ $m \geq n\text{,}$ as a linear combination of powers of ${\alpha}$ that are less than $n\text{.}$ Hence, any $\beta \in F( \alpha )$ can be written as</p><div class="displaymath">
\begin{equation*}
\beta = b_0 + b_1 \alpha + \cdots + b_{n - 1} \alpha^{n - 1}.
\end{equation*}
</div><p id="p-3244">To show uniqueness, suppose that</p><div class="displaymath">
\begin{equation*}
\beta =  b_0 + b_1 \alpha + \cdots + b_{n-1} \alpha^{n-1} = c_0 + c_1 \alpha + \cdots + c_{n - 1} \alpha^{n - 1}
\end{equation*}
</div><p>for $b_i$ and $c_i$ in $F\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
g(x) = (b_0 - c_0) + (b_1 - c_1) x + \cdots + (b_{n - 1} - c_{n - 1})x^{n - 1}
\end{equation*}
</div><p>is in $F[x]$ and $g( \alpha ) = 0\text{.}$ Since the degree of $g(x)$ is less than the degree of $p( x )\text{,}$ the irreducible polynomial of $\alpha\text{,}$ $g(x)$ must be the zero polynomial. Consequently,</p><div class="displaymath">
\begin{equation*}
b_0 - c_0 = b_1 - c_1 = \cdots = b_{n - 1} - c_{n - 1} = 0,
\end{equation*}
</div><p>or $b_i = c_i$ for $i = 0, 1, \ldots, n-1\text{.}$  Therefore, we have shown uniqueness.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-fields-isomorph"><h6 class="heading"><span class="type">Example</span><span class="codenumber">21.14</span></h6><p id="p-3245">Since $x^2 + 1$ is irreducible over ${\mathbb R}\text{,}$ $\langle x^2 + 1 \rangle$ is a maximal ideal in ${\mathbb R}[x]\text{.}$ So $E = {\mathbb R}[x]/\langle x^2 + 1 \rangle$ is a field extension of ${\mathbb R}$ that contains a root of $x^2 + 1\text{.}$ Let $\alpha = x + \langle x^2 + 1 \rangle\text{.}$ We can identify $E$ with the complex numbers. By Proposition <a href="section-extension-fields.ipynb#proposition-minimal-poly" class="xref" alt="Proposition 21.12 " title="Proposition 21.12 ">21.12</a>, $E$ is isomorphic to ${\mathbb R}( \alpha ) = \{ a + b \alpha : a, b \in {\mathbb R} \}\text{.}$ We know that $\alpha^2 = -1$ in $E\text{,}$ since</p><div class="displaymath">
\begin{align*}
\alpha^2 + 1 & = (x + \langle  x^2 + 1  \rangle)^2 + (1 + \langle  x^2 + 1  \rangle)\\
& = (x^2 + 1) +  \langle  x^2 + 1  \rangle\\
& = 0.
\end{align*}
</div><p>Hence, we have an isomorphism of ${\mathbb R}( \alpha )$ with ${\mathbb C}$ defined by the map that takes $a + b \alpha$ to $a + bi\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3246">Let $E$ be a field extension of a field $F\text{.}$  If we regard $E$ as a vector space over $F\text{,}$ then we can bring the machinery of linear algebra to bear on the problems that we will encounter in our study of fields. The elements in the field $E$ are vectors; the elements in the field $F$ are scalars. We can think of addition in $E$ as adding vectors.  When we multiply an element in $E$ by an element of $F\text{,}$ we are multiplying a vector by a scalar.  This view of field  extensions is especially fruitful if a field extension $E$ of $F$  is a finite dimensional vector space over $F\text{,}$ and Theorem <a href="section-extension-fields.ipynb#theorem-simple_extension" class="xref" alt="Theorem 21.13 " title="Theorem 21.13 ">21.13</a> states that $E = F(\alpha )$ is finite dimensional vector space over $F$ with basis $\{ 1, \alpha, {\alpha}^2, \ldots, {\alpha}^{n - 1} \}\text{.}$</p></div>

<div class="mathbook-content"><p id="p-3247">If an extension field $E$ of a field $F$ is a finite dimensional vector space over $F$ of dimension $n\text{,}$ then we say that $E$ is a  <dfn class="terminology">finite extension of degree $n$ over $F$</dfn>.  We write</p><div class="displaymath">
\begin{equation*}
[E:F]= n.
\end{equation*}
</div><p>to indicate the dimension of $E$ over $F\text{.}$ </p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-finite-extension"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.15</span></h6><p id="p-3248">Every finite extension field $E$ of a field $F$ is an algebraic extension.</p></article><article class="proof" id="proof-165"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3249">Let $\alpha \in E\text{.}$ Since $[E:F] = n\text{,}$ the elements</p><div class="displaymath">
\begin{equation*}
1, \alpha, \ldots, {\alpha}^n
\end{equation*}
</div><p>cannot be linearly independent.  Hence, there exist $a_i \in F\text{,}$ not all zero, such that</p><div class="displaymath">
\begin{equation*}
a_n {\alpha}^n + a_{n - 1} {\alpha}^{n - 1} + \cdots + a_1 \alpha + a_0 = 0.
\end{equation*}
</div><p>Therefore,</p><div class="displaymath">
\begin{equation*}
p(x) = a_n x^n + \cdots + a_0 \in F[x]
\end{equation*}
</div><p>is a nonzero polynomial with $p( \alpha ) = 0\text{.}$</p></article></div>

<div class="mathbook-content"><article class="remark-like" id="remark-9"><h6 class="heading"><span class="type">Remark</span><span class="codenumber">21.16</span></h6><p id="p-3250">Theorem <a href="section-extension-fields.ipynb#theorem-finite-extension" class="xref" alt="Theorem 21.15 " title="Theorem 21.15 ">21.15</a> says that every finite extension of a field $F$ is an algebraic extension. The converse is false, however. We will leave it as an exercise to show that the set of all elements in ${\mathbb R}$ that are algebraic over ${\mathbb Q}$ forms an infinite field extension of ${\mathbb Q}\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3251">The next theorem is a counting theorem, similar to Lagrange's Theorem in group theory. Theorem <a href="section-extension-fields.ipynb#theorem-tower-indices" class="xref" alt="Theorem 21.17 " title="Theorem 21.17 ">21.17</a> will prove to be an extremely useful tool in our investigation of finite field extensions.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-tower-indices"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.17</span></h6><p id="p-3252">If $E$ is a finite extension of $F$ and $K$ is a finite extension of $E\text{,}$ then $K$ is a finite extension of $F$ and</p><div class="displaymath">
\begin{equation*}
[K:F]= [K:E] [E:F].
\end{equation*}
</div></article><article class="proof" id="proof-166"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3253">Let $\{ \alpha_1, \ldots, \alpha_n \}$ be a basis for $E$ as a vector space over $F$ and $\{ \beta_1, \ldots, \beta_m \}$ be a basis for $K$ as a vector space over $E\text{.}$ We claim that $\{ \alpha_i \beta_j \}$ is a basis for $K$ over $F\text{.}$  We will first show that these vectors span $K\text{.}$ Let $u \in K\text{.}$ Then $u = \sum_{j = 1}^{m} b_j \beta_j$ and $b_j = \sum_{i = 1}^{n} a_{ij} \alpha_i\text{,}$ where $b_j \in E$ and $a_{ij} \in F\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
u = \sum_{j = 1}^{m} \left( \sum_{i = 1}^{n} a_{ij} \alpha_i  \right) \beta_j = \sum_{i,j} a_{ij} ( \alpha_i \beta_j ).
\end{equation*}
</div><p>So the $mn$ vectors $\alpha_i \beta_j$ must span $K$ over $F\text{.}$</p><p id="p-3254">We must show that $\{ \alpha_i \beta_j \}$ are linearly independent. Recall that a set of vectors $v_1, v_2, \ldots, v_n$ in a vector space $V$ are linearly independent if</p><div class="displaymath">
\begin{equation*}
c_1 v_1 + c_2 v_2 + \cdots + c_n v_n = 0
\end{equation*}
</div><p>implies that</p><div class="displaymath">
\begin{equation*}
c_1 = c_2 = \cdots = c_n = 0.
\end{equation*}
</div><p>Let</p><div class="displaymath">
\begin{equation*}
u = \sum_{i,j} c_{ij} ( \alpha_i \beta_j ) = 0
\end{equation*}
</div><p>for $c_{ij} \in F\text{.}$ We need to prove that all of the $c_{ij}$'s are zero. We can rewrite $u$ as</p><div class="displaymath">
\begin{equation*}
\sum_{j = 1}^{m} \left( \sum_{i = 1}^{n} c_{ij} \alpha_i \right) \beta_j = 0,
\end{equation*}
</div><p>where $\sum_i c_{ij} \alpha_i \in E\text{.}$  Since the $\beta_j$'s are linearly independent over $E\text{,}$ it must be the case that</p><div class="displaymath">
\begin{equation*}
\sum_{i = 1}^n c_{ij} \alpha_i = 0
\end{equation*}
</div><p>for all $j\text{.}$ However, the $\alpha_j$ are also linearly independent  over $F\text{.}$  Therefore, $c_{ij} = 0$ for all $i$ and $j\text{,}$ which completes the proof.</p></article></div>

<div class="mathbook-content"><p id="p-3255">The following corollary is easily proved using mathematical induction.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-29"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">21.18</span></h6><p id="p-3256">If $F_i$ is a field for $i = 1, \dots, k$ and $F_{i+1}$ is a finite extension of $F_i\text{,}$ then $F_k$ is a finite extension of $F_1$ and</p><div class="displaymath">
\begin{equation*}
[F_k : F_1] = [F_k : F_{k-1} ] \cdots [F_2 : F_1 ].
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-30"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">21.19</span></h6><p id="p-3257">Let $E$ be an extension field of $F\text{.}$ If $\alpha \in E$ is algebraic over $F$ with minimal polynomial $p(x)$ and $\beta \in F( \alpha )$ with minimal polynomial $q(x)\text{,}$ then $\deg q(x)$ divides $\deg p(x)\text{.}$</p></article><article class="proof" id="proof-167"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3258">We know that $\deg p(x) = [F( \alpha ) : F ]$ and $\deg q(x) = [F( \beta ) : F ]\text{.}$ Since $F \subset F( \beta ) \subset F( \alpha )\text{,}$</p><div class="displaymath">
\begin{equation*}
[F( \alpha ) : F ]= [ F( \alpha ) : F( \beta ) ] [ F( \beta ) : F ].
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="example-like" id="example-fields-sqrt3-sqrt5"><h6 class="heading"><span class="type">Example</span><span class="codenumber">21.20</span></h6><p id="p-3259">Let us determine an extension field of ${\mathbb Q}$ containing $\sqrt{3} + \sqrt{5}\text{.}$ It is easy to determine that the minimal  polynomial of $\sqrt{3} + \sqrt{5}$ is $x^4 - 16 x^2 + 4\text{.}$ It follows that</p><div class="displaymath">
\begin{equation*}
[{\mathbb Q}( \sqrt{3} + \sqrt{5}\, ) : {\mathbb Q} ] = 4.
\end{equation*}
</div><p>We know that $\{ 1, \sqrt{3}\, \}$ is a basis for  ${\mathbb Q}( \sqrt{3}\, )$ over ${\mathbb Q}\text{.}$ Hence, $\sqrt{3} + \sqrt{5}$ cannot be in ${\mathbb Q}( \sqrt{3}\, )\text{.}$ It follows that $\sqrt{5}$ cannot be in ${\mathbb Q}( \sqrt{3}\, )$ either. Therefore, $\{ 1, \sqrt{5}\, \}$ is a basis for ${\mathbb Q}( \sqrt{3}, \sqrt{5}\, ) = ( {\mathbb Q}(\sqrt{3}\, ))( \sqrt{5}\, )$ over ${\mathbb Q}( \sqrt{3}\, )$ and $\{ 1, \sqrt{3}, \sqrt{5}, \sqrt{3} \sqrt{5} = \sqrt{15}\,  \}$ is a basis for ${\mathbb Q}( \sqrt{3}, \sqrt{5}\, ) = {\mathbb Q}( \sqrt{3} + \sqrt{5}\, )$ over ${\mathbb Q}\text{.}$  This example shows that it is possible that some extension $F( \alpha_1, \ldots, \alpha_n )$ is actually a simple extension of $F$ even though $n \gt 1\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-fields-cubert5-sqrt5-i"><h6 class="heading"><span class="type">Example</span><span class="codenumber">21.21</span></h6><p id="p-3260">Let us compute a basis for ${\mathbb Q}( \sqrt[3]{5}, \sqrt{5} \, i )\text{,}$ where $\sqrt{5}$ is the positive square root of 5 and $\sqrt[3]{5}$ is the real cube root of 5.  We know that $\sqrt{5} \, i \notin {\mathbb Q}(\sqrt[3]{5}\, )\text{,}$ so</p><div class="displaymath">
\begin{equation*}
[ {\mathbb Q}(\sqrt[3]{5}, \sqrt{5}\, i) : {\mathbb Q}(\sqrt[3]{5}\, )] = 2.
\end{equation*}
</div><p>It is easy to determine that $\{ 1, \sqrt{5}i\,  \}$ is a basis for ${\mathbb Q}( \sqrt[3]{5}, \sqrt{5}\, i )$ over ${\mathbb Q}( \sqrt[3]{5}\, )\text{.}$  We also know that $\{ 1, \sqrt[3]{5}, (\sqrt[3]{5}\, )^2  \}$ is a basis for ${\mathbb Q}(\sqrt[3]{5}\, )$ over ${\mathbb Q}\text{.}$ Hence, a basis for ${\mathbb Q}(\sqrt[3]{5}, \sqrt{5}\, i )$ over ${\mathbb Q}$ is</p><div class="displaymath">
\begin{equation*}
\{ 1, \sqrt{5}\, i, \sqrt[3]{5}, (\sqrt[3]{5}\, )^2, (\sqrt[6]{5}\, )^5 i, (\sqrt[6]{5}\, )^7 i = 5 \sqrt[6]{5}\, i \text{ or } \sqrt[6]{5}\, i \}.
\end{equation*}
</div><p>Notice that $\sqrt[6]{5}\, i$ is a zero of $x^6 + 5\text{.}$ We can show that this polynomial is irreducible over ${\mathbb Q}$ using Eisenstein's Criterion, where we let $p = 5\text{.}$ Consequently,</p><div class="displaymath">
\begin{equation*}
{\mathbb Q} \subset {\mathbb Q}( \sqrt[6]{5}\, i) \subset {\mathbb Q}( \sqrt[3]{5}, \sqrt{5}\, i ).
\end{equation*}
</div><p>But it must be the case that ${\mathbb Q}( \sqrt[6]{5}\, i) = {\mathbb Q}( \sqrt[3]{5}, \sqrt{5}\, i )\text{,}$ since the degree of both of these extensions is 6.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-116"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.22</span></h6><p id="p-3261">Let $E$ be a field extension of $F\text{.}$ Then  the following statements are equivalent. </p><ol class="decimal"><li id="li-726"><p id="p-3262">$E$ is a finite extension of $F\text{.}$</p></li><li id="li-727"><p id="p-3263">There exists a finite number of algebraic elements $\alpha_1,  \ldots, \alpha_n \in E$ such that  $E = F(\alpha_1, \ldots, \alpha_n)\text{.}$</p></li><li id="li-728"><p id="p-3264">There exists a sequence of fields</p><div class="displaymath">
\begin{equation*}
E = F(\alpha_1, \ldots, \alpha_n) \supset F(\alpha_1, \ldots, \alpha_{n-1} ) \supset \cdots \supset F( \alpha_1 ) \supset F,
\end{equation*}
</div><p>where each field $F(\alpha_1, \ldots, \alpha_i)$ is algebraic over $F(\alpha_1, \ldots, \alpha_{i-1})\text{.}$</p></li></ol></article><article class="proof" id="proof-168"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3265">(1) $\Rightarrow$ (2). Let $E$ be a finite algebraic extension of $F\text{.}$  Then $E$ is a finite dimensional vector space over $F$ and there exists a basis consisting  of elements $\alpha_1, \ldots, \alpha_n$ in $E$ such that $E = F(\alpha_1, \ldots, \alpha_n)\text{.}$ Each $\alpha_i$ is algebraic over $F$ by Theorem <a href="section-extension-fields.ipynb#theorem-finite-extension" class="xref" alt="Theorem 21.15 " title="Theorem 21.15 ">21.15</a>.</p><p id="p-3266">(2) $\Rightarrow$ (3). Suppose that  $E = F(\alpha_1, \ldots, \alpha_n)\text{,}$ where every $\alpha_i$ is algebraic over $F\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
E = F(\alpha_1, \ldots, \alpha_n) \supset F(\alpha_1, \ldots, \alpha_{n - 1} ) \supset \cdots \supset F( \alpha_1 ) \supset F,
\end{equation*}
</div><p>where each field $F(\alpha_1, \ldots, \alpha_i)$ is algebraic over $F(\alpha_1, \ldots, \alpha_{i - 1})\text{.}$</p><p id="p-3267">(3) $\Rightarrow$ (1). Let</p><div class="displaymath">
\begin{equation*}
E = F(\alpha_1, \ldots, \alpha_n) \supset F(\alpha_1, \ldots, \alpha_{n - 1} ) \supset \cdots \supset F( \alpha_1 ) \supset F,
\end{equation*}
</div><p>where each field $F(\alpha_1, \ldots, \alpha_i)$ is algebraic over  $F(\alpha_1, \ldots, \alpha_{i - 1})\text{.}$ Since</p><div class="displaymath">
\begin{equation*}
F(\alpha_1, \ldots, \alpha_i) =  F(\alpha_1, \ldots, \alpha_{i - 1} )(\alpha_i)
\end{equation*}
</div><p>is simple extension and $\alpha_i$ is algebraic over $F(\alpha_1, \ldots, \alpha_{i - 1})\text{,}$ it follows that</p><div class="displaymath">
\begin{equation*}
[ F(\alpha_1, \ldots, \alpha_i) : F(\alpha_1, \ldots, \alpha_{i - 1} )]
\end{equation*}
</div><p>is finite for each $i\text{.}$ Therefore, $[E : F]$ is finite.</p></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Algebraic Closure"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Algebraic Closure</span></h3><a href="section-extension-fields.ipynb#fields-subsection-algebraic-closure" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-3268">Given a field $F\text{,}$ the question arises as to whether or not we can find a field $E$ such that every polynomial $p(x)$ has a root in $E\text{.}$ This leads us to the following theorem.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-117"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.23</span></h6><p id="p-3269">Let $E$ be an extension field of $F\text{.}$ The set of elements in $E$ that are algebraic over $F$ form a field.</p></article><article class="proof" id="proof-169"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3270">Let $\alpha, \beta \in E$ be algebraic over $F\text{.}$ Then $F( \alpha, \beta )$ is a finite extension of $F\text{.}$  Since every element of $F( \alpha, \beta )$ is algebraic over $F\text{,}$ $\alpha \pm \beta\text{,}$ $\alpha  \beta\text{,}$ and $\alpha / \beta$ ($\beta \neq 0$) are all algebraic over $F\text{.}$ Consequently, the set of elements in $E$ that are algebraic over $F$ form a field.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-algebraic-numbers-field"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">21.24</span></h6><p id="p-3271">The set of all algebraic numbers forms a field; that is, the set of all complex numbers that are algebraic over ${\mathbb Q}$ makes up a field.</p></article></div>

<div class="mathbook-content"><p id="p-3272">Let $E$ be a field extension of a field $F\text{.}$ We define the <dfn class="terminology">algebraic closure</dfn>  of a field $F$ in $E$ to be the field consisting of all elements in $E$ that are algebraic over $F\text{.}$ A field $F$ is <dfn class="terminology">algebraically closed</dfn> if every nonconstant polynomial in  $F[x]$ has a root in $F\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-118"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.25</span></h6><p id="p-3273">A field $F$ is algebraically closed if and only if every nonconstant polynomial in $F[x]$ factors into linear factors over $F[x]\text{.}$</p></article><article class="proof" id="proof-170"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3274">Let $F$ be an algebraically closed field. If $p(x) \in F[x]$ is a nonconstant polynomial, then $p(x)$ has a zero in $F\text{,}$ say $\alpha\text{.}$ Therefore, $x-\alpha$ must be a factor of $p(x)$ and so $p(x) = (x - \alpha) q_1(x)\text{,}$ where $\deg q_1(x) = \deg p(x) - 1\text{.}$ Continue this process with $q_1(x)$ to find a factorization</p><div class="displaymath">
\begin{equation*}
p(x) = (x - \alpha)(x - \beta)q_2(x),
\end{equation*}
</div><p>where $\deg q_2(x) = \deg p(x) -2\text{.}$ The process must eventually stop since the degree of $p(x)$ is finite.</p><p id="p-3275">Conversely, suppose that every nonconstant polynomial $p(x)$ in $F[x]$ factors into linear factors. Let $ax - b$ be such a factor. Then $p( b/a) = 0\text{.}$ Consequently, $F$ is algebraically closed.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-32"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">21.26</span></h6><p id="p-3276">An algebraically closed field $F$ has no proper algebraic extension $E\text{.}$</p></article><article class="proof" id="proof-171"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3277">Let $E$ be an algebraic extension of $F\text{;}$ then $F \subset E\text{.}$ For $\alpha \in E\text{,}$ the minimal polynomial of $\alpha$ is $x - \alpha\text{.}$ Therefore, $\alpha \in F$ and $F = E\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-119"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.27</span></h6><p id="p-3278">Every field $F$ has a unique algebraic closure.</p></article></div>

<div class="mathbook-content"><p id="p-3279">It is a nontrivial fact that every field has a unique algebraic closure. The proof is not extremely difficult, but requires some rather sophisticated set theory. We refer the reader to [3], [4], or [8] for a proof of this result.</p></div>

<div class="mathbook-content"><p id="p-3280">We now state the  Fundamental Theorem of Algebra, first proven by Gauss at the age of 22 in his doctoral thesis. This theorem states that every polynomial with coefficients in the complex numbers has a root in the complex numbers.  The proof of this theorem will be given in Chapter <a href="galois.ipynb" class="xref" alt="Chapter 23 Galois Theory" title="Chapter 23 Galois Theory">23</a>.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-120"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.28</span><span class="title">Fundamental Theorem of Algebra</span></h6><p id="p-3281">The field of complex numbers is algebraically closed.</p></article></div>
