<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 6.3 Fermat's and Euler's Theorems"><span class="type">Section</span><span class="codenumber">6.3</span><span class="title">Fermat's and Euler's Theorems</span></h2><a href="section-fermat-euler.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-1005">The <dfn class="terminology">Euler</dfn> $\phi$-<dfn class="terminology">function</dfn> is the map $\phi : {\mathbb N } \rightarrow {\mathbb N}$ defined by $\phi(n) = 1$ for $n=1\text{,}$ and, for $n \gt 1\text{,}$  $\phi(n)$ is the number of positive integers $m$ with $1 \leq m \lt n$ and $\gcd(m,n) = 1\text{.}$</p></div>

<div class="mathbook-content"><p id="p-1006">From Proposition <a href="section-mod-n-sym.ipynb#proposition-zn-equiv-classes" class="xref" alt="Proposition 3.4 " title="Proposition 3.4 ">3.4</a>, we know that the order of $U(n)\text{,}$ the group of units in ${\mathbb Z}_n\text{,}$ is $\phi(n)\text{.}$ For example, $|U(12)| = \phi(12)  = 4$ since the numbers that are relatively prime to 12 are 1, 5, 7, and 11. For any prime $p\text{,}$ $\phi(p) = p-1\text{.}$  We state these results in the following theorem.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-cosets-theorem-11"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">6.17</span></h6><p id="p-1007">Let $U(n)$ be the group of units in ${\mathbb Z}_n\text{.}$  Then $|U(n)| = \phi(n)\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1008">The following theorem is an important result in number theory, due to Leonhard Euler.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-eulers_theorem"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">6.18</span><span class="title">Euler's Theorem</span></h6><p id="p-1009">Let $a$ and $n$ be integers such that $n \gt 0$ and $\gcd(a, n) = 1\text{.}$  Then $a^{\phi(n)} \equiv 1 \pmod{n}\text{.}$</p></article><article class="proof" id="proof-44"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1010">By Theorem <a href="section-fermat-euler.ipynb#theorem-cosets-theorem-11" class="xref" alt="Theorem 6.17 " title="Theorem 6.17 ">6.17</a> the order of $U(n)$ is $\phi(n)\text{.}$  Consequently, $a^{\phi(n)} = 1$ for all $a \in U(n)\text{;}$ or $a^{\phi(n)} - 1$ is divisible by $n\text{.}$  Therefore, $a^{\phi(n)} \equiv 1 \pmod{n}\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1011">If we consider the special case of Euler's Theorem in which $n = p$ is prime and recall that $\phi(p) = p - 1\text{,}$ we obtain the following result, due to Pierre de Fermat.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-cosets-theorem-13"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">6.19</span><span class="title">Fermat's Little Theorem</span></h6><p id="p-1012">Let $p$ be any prime number and suppose that $p \notdivide a$ ($p$ does not divide $a$).   Then</p><div class="displaymath">
\begin{equation*}
a^{p-1} \equiv 1 \pmod{ p }.
\end{equation*}
</div><p>Furthermore, for any integer $b\text{,}$ $b^p \equiv b \pmod{ p}\text{.}$</p></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Historical Note"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Historical Note</span></h3><a href="section-fermat-euler.ipynb#cosets-subsection-historical-note" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-1014">Joseph-Louis Lagrange (1736–1813), born in Turin, Italy, was of French and Italian descent. His talent for mathematics became apparent at an early age. Leonhard Euler recognized Lagrange's abilities when Lagrange, who was only 19, communicated to Euler some work that he had done in the calculus of variations. That year he was also named a professor at the Royal Artillery School in Turin.  At the age of 23 he joined the Berlin Academy. Frederick the Great had written to Lagrange proclaiming that the “greatest king in Europe” should have the “greatest mathematician in Europe” at his court. For 20 years Lagrange held the position vacated by his mentor, Euler. His works include contributions to number theory, group theory, physics and mechanics, the calculus of variations, the theory of equations, and differential equations. Along with Laplace and Lavoisier, Lagrange was one of the people responsible for designing the metric system. During his life Lagrange profoundly influenced the development of mathematics, leaving much to the next generation of mathematicians in the form of examples and new problems to be solved.</p></div>
