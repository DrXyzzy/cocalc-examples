<div class="mathbook-content"><article class="exercise-like" id="exercise-608"><h6 class="heading"><span class="codenumber">1</span></h6><p id="p-2695">Solve the general quadratic equation</p><div class="displaymath">
\begin{equation*}
ax^2 + bx + c = 0
\end{equation*}
</div><p>to obtain</p><div class="displaymath">
\begin{equation*}
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}.
\end{equation*}
</div><p>The <dfn class="terminology">discriminant</dfn> of the quadratic equation $\Delta = b^2 - 4ac$ determines the nature of the solutions of the equation.  If $\Delta \gt 0\text{,}$ the equation has two distinct real solutions.  If $\Delta = 0\text{,}$ the equation has a single repeated real root. If $\Delta \lt 0\text{,}$ there are two distinct imaginary solutions.</p></article></div>
