<div class="mathbook-content"><p id="p-2028">Group actions generalize group multiplication. If $G$ is a group and $X$ is an arbitrary set, a group action of an element $g \in G$ and $x \in X$ is a product, $gx\text{,}$  living in $X\text{.}$  Many problems in algebra are best be attacked via group actions.
