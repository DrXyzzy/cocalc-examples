Show that $yP = Py\text{.}$</p></li><li id="li-375"><p id="p-1562">Suppose that $G$ does not contain an element of order $2p$ and $P = \langle z \rangle$ is a subgroup of order $p$ generated by $z\text{.}$  If $y$ is an element of order 2, then $yz = z^ky$ for some $2 \leq k \lt p\text{.}$</p></li><li id="li-376"><p id="p-1563">Suppose that $G$ does not contain an element of order $2p\text{.}$  Prove that $G$ is not abelian.</p></li><li id="li-377"><p id="p-1564">Suppose that $G$ does not contain an element of order $2p$ and $P = \langle z \rangle$ is a subgroup of order $p$ generated by $z$ and $y$ is an element of order 2. Prove that the product $(z^iy^j)(z^ry^s)$ can be expressed as a uniquely as $z^m y^n$ for some non negative integers $m, n\text{.}$  Thus, conclude that there is only one possibility for a non-abelian group of order $2p\text{,}$ it must therefore be the one we have seen already, the dihedral group.</p></li></ol></article></div>
