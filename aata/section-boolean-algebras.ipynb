<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 19.2 Boolean Algebras"><span class="type">Section</span><span class="codenumber">19.2</span><span class="title">Boolean Algebras</span></h2><a href="section-boolean-algebras.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-2933">Let us investigate the example of the power set, ${\mathcal P}(X)\text{,}$ of a set $X$ more  closely. The power set is a lattice that is  ordered by inclusion. By the definition of the power set, the largest element in ${\mathcal P}(X)$ is $X$ itself and the smallest element is $\emptyset\text{,}$ the empty set. For any set $A$ in ${\mathcal P}(X)\text{,}$ we know that $A \cap X = A$ and $A \cup \emptyset = A\text{.}$ This suggests the following definition for lattices. An element $I$ in a poset $X$ is a <dfn class="terminology">largest element</dfn> if $a \preceq I$ for all $a \in X\text{.}$   An element $O$ is a  <dfn class="terminology">smallest element</dfn> of $X$ if $O \preceq a$ for all $a \in X\text{.}$ </p></div>

<div class="mathbook-content"><p id="p-2934">Let $A$ be in ${\mathcal P}(X)\text{.}$ Recall that the complement of $A$ is</p><div class="displaymath">
\begin{equation*}
A' = X \setminus A = \{ x : x \in X \text{ and } x \notin A \}.
\end{equation*}
</div><p>We know that $A \cup A' = X$ and $A \cap A' = \emptyset\text{.}$ We can generalize this example for lattices. A lattice $L$ with a largest element $I$ and a smallest element $O$ is <dfn class="terminology">complemented</dfn> if for each $a \in L\text{,}$ there exists an $a'$ such that $a \vee a' = I$ and $a \wedge a' = O\text{.}$ </p></div>

<div class="mathbook-content"><p id="p-2935">In a lattice $L\text{,}$ the binary operations $\vee$ and $\wedge$ satisfy commutative and associative laws; however, they need not satisfy the distributive law</p><div class="displaymath">
\begin{equation*}
a \wedge ( b \vee c ) = (a \wedge b ) \vee ( a \wedge c );
\end{equation*}
</div><p>however, in ${\mathcal P}(X)$ the distributive law is satisfied since</p><div class="displaymath">
\begin{equation*}
A \cap ( B \cup C ) = (A \cap B ) \cup ( A \cap C )
\end{equation*}
</div><p>for $A, B, C \in {\mathcal P}(X)\text{.}$ We will say that a lattice $L$ is <dfn class="terminology">distributive</dfn> if the following distributive law holds:</p><div class="displaymath">
\begin{equation*}
a \wedge ( b \vee c ) = (a \wedge b ) \vee ( a \wedge c )
\end{equation*}
</div><p>for all $a, b, c \in L\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-distributive-lattice"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">19.15</span></h6><p id="p-2936">A lattice $L$ is distributive if and only if</p><div class="displaymath">
\begin{equation*}
a \vee ( b \wedge c ) = ( a \vee b ) \wedge ( a \vee c )
\end{equation*}
</div><p>for all $a, b, c \in L\text{.}$</p></article><article class="proof" id="proof-146"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2937">Let us assume that $L$ is a distributive lattice.</p><div class="displaymath">
\begin{align*}
a \vee ( b \wedge c ) & = [a \vee (a \wedge c) ] \vee ( b \wedge c )\\
& = a \vee [(a \wedge c)  \vee ( b \wedge c )]\\
& = a \vee [(c \wedge a)  \vee ( c \wedge b )]\\
& = a \vee [c \wedge ( a  \vee b )]\\
& = a \vee [( a  \vee b ) \wedge c ]\\
& = [( a  \vee b ) \wedge a ] \vee [(a \vee b) \wedge c ]\\
& = ( a \vee b ) \wedge ( a \vee c ).
\end{align*}
</div><p>The converse follows directly from the Duality Principle.</p></article></div>

<div class="mathbook-content"><p id="p-2938">A <dfn class="terminology">Boolean algebra</dfn> is a lattice $B$ with a greatest element $I$ and a smallest element $O$ such that $B$ is both distributive and complemented. The power set of $X\text{,}$ ${\mathcal P}(X)\text{,}$ is our prototype for a Boolean algebra.  As it turns out, it is also one of the most important Boolean algebras. The following theorem allows us to characterize Boolean algebras in terms of the binary relations $\vee$ and $\wedge$ without mention of the fact that a Boolean algebra is a poset.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-105"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">19.16</span></h6><p id="p-2939">A set $B$ is a Boolean algebra if and only if there exist binary operations $\vee$ and $\wedge$ on $B$ satisfying the following axioms. </p><ol class="decimal"><li id="li-655"><p id="p-2940">$a \vee b = b \vee a$ and $a \wedge b = b \wedge a$ for $a, b \in B\text{.}$</p></li><li id="li-656"><p id="p-2941">$a \vee ( b \vee c) = (a \vee b) \vee c$ and $a \wedge ( b \wedge c) = (a \wedge b) \wedge c$ for $a, b, c \in B\text{.}$</p></li><li id="li-657"><p id="p-2942">$a \wedge ( b \vee c ) = (a \wedge b ) \vee ( a \wedge c )$ and $a \vee ( b \wedge c ) = (a \vee b ) \wedge ( a \vee c )$  for $a, b, c \in B\text{.}$</p></li><li id="li-658"><p id="p-2943">There exist elements $I$ and $O$ such that $a \vee O = a$ and $a \wedge I = a$ for all $a \in B\text{.}$</p></li><li id="li-659"><p id="p-2944">For every $a \in B$ there exists an $a' \in B$ such that $a \vee a' = I$ and $a \wedge a' = O\text{.}$</p></li></ol></article><article class="proof" id="proof-147"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2945">Let $B$ be a set satisfying (1)–(5) in the theorem. One of the idempotent laws is satisfied since</p><div class="displaymath">
\begin{align*}
a & = a \vee O\\
& = a \vee (a \wedge a')\\
& = (a \vee a) \wedge (a \vee a')\\
& = (a \vee a ) \wedge I\\
& = a \vee a.
\end{align*}
</div><p>Observe that</p><div class="displaymath">
\begin{equation*}
I \vee b = (b \vee b' ) \vee b = (b' \vee b ) \vee b = b' \vee (b \vee b) =  b' \vee b = I.
\end{equation*}
</div><p>Consequently, the first of the two absorption laws holds, since</p><div class="displaymath">
\begin{align*}
a \vee (a \wedge b) & = (a \wedge I) \vee (a \wedge b)\\
& = a \wedge (I \vee b)\\
& = a  \wedge I\\
& = a.
\end{align*}
</div><p>The other idempotent and absorption laws are proven similarly. Since $B$ also satisfies (1)–(3), the conditions of Theorem <a href="section-boolean-lattices.ipynb#theorem-partial-order" class="xref" alt="Theorem 19.14 " title="Theorem 19.14 ">19.14</a> are met; therefore, $B$ must be a lattice.  Condition (4) tells us that $B$ is a distributive lattice.</p><p id="p-2946">For $a \in B\text{,}$ $O \vee a = a\text{;}$ hence, $O \preceq a$ and $O$ is the smallest element in $B\text{.}$ To show that $I$ is the largest element in $B\text{,}$ we will first show that $a \vee b = b$ is equivalent to $a \wedge b = a\text{.}$  Since $a \vee I = a$ for all $a \in B\text{,}$ using the absorption laws we can determine that</p><div class="displaymath">
\begin{equation*}
a \vee I =(a \wedge I) \vee I = I \vee ( I \wedge a) = I
\end{equation*}
</div><p>or $a \preceq I$ for all $a$ in $B\text{.}$ Finally, since we know that $B$ is complemented by (5), $B$ must be a Boolean algebra.</p><p id="p-2947">Conversely, suppose that $B$ is a Boolean algebra. Let $I$ and $O$ be the greatest and least elements in $B\text{,}$ respectively.  If we define $a \vee b$ and $a \wedge b$ as least upper and greatest lower bounds of $\{ a, b\}\text{,}$ then $B$ is a Boolean algebra by Theorem <a href="section-boolean-lattices.ipynb#theorem-partial-order" class="xref" alt="Theorem 19.14 " title="Theorem 19.14 ">19.14</a>, Theorem <a href="section-boolean-algebras.ipynb#theorem-distributive-lattice" class="xref" alt="Theorem 19.15 " title="Theorem 19.15 ">19.15</a>, and our hypothesis.</p></article></div>

<div class="mathbook-content"><p id="p-2948">Many other identities hold in Boolean algebras. Some of these identities are listed in the following theorem.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-106"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">19.17</span></h6><p id="p-2949">Let $B$ be a Boolean algebra. Then </p><ol class="decimal"><li id="li-660"><p id="p-2950">$a \vee I = I$ and $a \wedge O = O$ for all $a \in B\text{.}$</p></li><li id="li-661"><p id="p-2951">If $a \vee b = a \vee c$ and $a \wedge b = a \wedge c$ for $a, b, c \in B\text{,}$ then $b = c\text{.}$</p></li><li id="li-662"><p id="p-2952">If $a \vee b = I$ and $a \wedge b = O\text{,}$ then $b = a'\text{.}$</p></li><li id="li-663"><p id="p-2953">$(a')'=a$ for all $a \in B\text{.}$</p></li><li id="li-664"><p id="p-2954">$I' = O$ and $O' = I\text{.}$</p></li><li id="li-665"><p id="p-2955">$(a \vee b)' = a' \wedge b'$ and $(a \wedge b)' = a' \vee b'$ (De Morgan's Laws).</p></li></ol></article><article class="proof" id="proof-148"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2956">We will prove only (2). The rest of the identities are left as exercises. For $a \vee b = a \vee c$ and $a \wedge b = a \wedge c\text{,}$ we have</p><div class="displaymath">
\begin{align*}
b & = b \vee (b \wedge a) \\
& =  b \vee (a \wedge b) \\
& =  b \vee (a \wedge c) \\
& =  ( b \vee a) \wedge ( b \vee c) \\
& =  ( a \vee b) \wedge ( b \vee c) \\
& =  ( a \vee c) \wedge ( b \vee c) \\
& =  ( c \vee a ) \wedge ( c\vee b ) \\
& = c \vee (a \wedge b)\\
& = c \vee ( a \wedge c )\\
& = c \vee ( c \wedge a )\\
& = c.
\end{align*}
</div></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Finite Boolean Algebras"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Finite Boolean Algebras</span></h3><a href="section-boolean-algebras.ipynb#boolean-subsection-finite-algebras" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-2957">A Boolean algebra is a <dfn class="terminology">finite Boolean algebra</dfn> if it contains a finite number of elements as a set. Finite Boolean algebras are particularly nice since we can classify them  up to isomorphism.</p></div>

<div class="mathbook-content"><p id="p-2958">Let $B$ and $C$ be Boolean algebras. A bijective map $\phi : B \rightarrow C$ is an <dfn class="terminology">isomorphism</dfn> of Boolean algebras  if</p><div class="displaymath">
\begin{align*}
\phi( a \vee b )  & = \phi(a) \vee \phi(b)\\
\phi( a \wedge b )  & = \phi(a) \wedge \phi(b)
\end{align*}
</div><p>for all $a$ and $b$ in $B\text{.}$</p></div>

<div class="mathbook-content"><p id="p-2959">We will show that any finite Boolean algebra is isomorphic to the Boolean algebra obtained by taking the power set of some finite set $X\text{.}$ We will need a few lemmas and definitions before we prove this result. Let $B$ be a finite Boolean algebra. An element $a \in B$ is  an <dfn class="terminology">atom</dfn> of $B$ if $a \neq O$ and $a \wedge b = a$ for all nonzero $b \in B\text{.}$ Equivalently, $a$ is an atom of $B$ if there is no nonzero $b \in B$ distinct from $a$ such that $O \preceq b \preceq a\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-29"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">19.18</span></h6><p id="p-2960">Let $B$ be a finite Boolean algebra. If $b$ is a nonzero element of $B\text{,}$ then there is an atom $a$ in $B$ such that $a \preceq b\text{.}$</p></article><article class="proof" id="proof-149"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2961">If $b$ is an atom, let $a =b\text{.}$ Otherwise, choose an element $b_1\text{,}$ not equal to $O$ or $b\text{,}$ such that $b_1 \preceq b\text{.}$ We are guaranteed that this is possible since $b$ is not an atom. If $b_1$ is an atom, then we are done.  If not, choose $b_2\text{,}$ not equal to $O$ or $b_1\text{,}$ such that  $b_2 \preceq b_1\text{.}$ Again, if $b_2$ is an atom, let $a = b_2\text{.}$ Continuing this process, we can obtain a chain</p><div class="displaymath">
\begin{equation*}
O \preceq \cdots \preceq b_3 \preceq b_2 \preceq b_1 \preceq b.
\end{equation*}
</div><p>Since $B$ is a finite Boolean algebra, this chain must be finite.  That is, for some $k\text{,}$ $b_k$ is an atom. Let $a = b_k\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-zero-vee"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">19.19</span></h6><p id="p-2962">Let $a$ and $b$ be atoms in a finite Boolean algebra $B$ such that $a \neq b\text{.}$ Then $a \wedge b = O\text{.}$</p></article><article class="proof" id="proof-150"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2963">Since $a \wedge b$ is the greatest lower bound of $a$ and $b\text{,}$ we know that $a \wedge b \preceq a\text{.}$  Hence, either $a \wedge b = a$ or $a \wedge b = O\text{.}$ However, if $a \wedge b = a\text{,}$ then either $a \preceq b$ or $a = O\text{.}$  In either case we have a contradiction because $a$ and $b$ are both atoms; therefore, $a \wedge b = O\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-partial-order-equivalence"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">19.20</span></h6><p id="p-2964">Let $B$ be a Boolean algebra and $a, b \in B\text{.}$ The following statements are equivalent. </p><ol class="decimal"><li id="li-666"><p id="p-2965">$a \preceq b\text{.}$</p></li><li id="li-667"><p id="p-2966">$a \wedge b' = O\text{.}$</p></li><li id="li-668"><p id="p-2967">$a' \vee b = I\text{.}$</p></li></ol></article><article class="proof" id="proof-151"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2968">(1) $\Rightarrow$ (2). If $a \preceq b\text{,}$ then $a \vee b = b\text{.}$ Therefore,</p><div class="displaymath">
\begin{align*}
a \wedge b' & = a \wedge (a \vee b)'\\
& = a \wedge ( a' \wedge b')\\
& = ( a \wedge a') \wedge b'\\
& = O \wedge b'\\
& = O.
\end{align*}
</div><p id="p-2969">(2) $\Rightarrow$ (3). If $a \wedge b' = O\text{,}$ then $a' \vee b = (a \wedge b')' = O' = I\text{.}$</p><p id="p-2970">(3) $\Rightarrow$ (1). If $a' \vee b = I\text{,}$ then</p><div class="displaymath">
\begin{align*}
a & =  a \wedge (a' \vee b) \\
& =  (a \wedge a') \vee (a  \wedge  b)\\
& = O \vee (a  \wedge  b)\\
& = a \wedge b.
\end{align*}
</div><p>Thus, $a \preceq b\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-32"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">19.21</span></h6><p id="p-2971">Let $B$ be a Boolean algebra and $b$ and $c$ be elements in $B$ such that $b \not\preceq c\text{.}$ Then there exists an atom $a \in B$ such that $a \preceq b$ and $a \not\preceq c\text{.}$</p></article><article class="proof" id="proof-152"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2972">By  Lemma <a href="section-boolean-algebras.ipynb#lemma-partial-order-equivalence" class="xref" alt="Lemma 19.20 " title="Lemma 19.20 ">19.20</a>, $b \wedge c' \neq O\text{.}$ Hence, there exists an atom $a$ such that $a \preceq b \wedge c'\text{.}$ Consequently, $a \preceq b$ and $a \not\preceq c\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-atoms"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">19.22</span></h6><p id="p-2973">Let $b \in B$ and $a_1, \ldots, a_n$ be the atoms of $B$ such that $a_i \preceq b\text{.}$ Then $b = a_1 \vee \cdots \vee a_n\text{.}$ Furthermore, if $a, a_1, \ldots, a_n$ are atoms of $B$ such that $a \preceq b\text{,}$ $a_i \preceq b\text{,}$ and $b = a \vee a_1 \vee \cdots \vee a_n\text{,}$ then $a = a_i$ for some $i = 1, \ldots, n\text{.}$</p></article><article class="proof" id="proof-153"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2974">Let $b_1 =   a_1 \vee \cdots \vee a_n\text{.}$ Since $a_i \preceq b$ for each $i\text{,}$ we know that $b_1 \preceq b\text{.}$  If we can show that $b \preceq b_1\text{,}$ then the lemma is true by antisymmetry.  Assume $b \not\preceq b_1\text{.}$ Then there exists an atom $a$ such that $a \preceq b$ and $a \not\preceq b_1\text{.}$  Since $a$ is an atom and $a \preceq b\text{,}$ we can deduce that $a = a_i$ for  some $a_i\text{.}$ However, this is impossible since $a \preceq b_1\text{.}$ Therefore, $b \preceq b_1\text{.}$</p><p id="p-2975">Now suppose that $b = a_1 \vee \cdots \vee a_n\text{.}$ If $a$ is an atom less than $b\text{,}$</p><div class="displaymath">
\begin{equation*}
a = a \wedge b  = a \wedge( a_1 \vee \cdots \vee a_n ) = (a \wedge a_1) \vee \cdots \vee ( a \wedge a_n ).
\end{equation*}
</div><p>But each term is $O$ or $a$ with $a \wedge a_i$ occurring for only one $a_i\text{.}$ Hence, by Lemma <a href="section-boolean-algebras.ipynb#lemma-zero-vee" class="xref" alt="Lemma 19.19 " title="Lemma 19.19 ">19.19</a>, $a = a_i$ for some $i\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-classification-boolean-algebra"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">19.23</span></h6><p id="p-2976">Let $B$ be a finite Boolean algebra.  Then there exists a set $X$ such that $B$ is isomorphic to ${\mathcal P}(X)\text{.}$</p></article><article class="proof" id="proof-154"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2977">We will show that $B$ is isomorphic to ${\mathcal P}(X)\text{,}$ where $X$ is the set of atoms of $B\text{.}$ Let $a \in B\text{.}$ By Lemma <a href="section-boolean-algebras.ipynb#lemma-atoms" class="xref" alt="Lemma 19.22 " title="Lemma 19.22 ">19.22</a>, we can write $a$ uniquely as $a = a_1 \vee \cdots \vee a_n$ for $a_1, \ldots, a_n \in X\text{.}$ Consequently, we can define a  map $\phi : B \rightarrow {\mathcal P}(X)$ by</p><div class="displaymath">
\begin{equation*}
\phi(a) = \phi(  a_1 \vee \cdots \vee a_n ) = \{a_1, \ldots, a_n \}.
\end{equation*}
</div><p>Clearly, $\phi$ is onto.</p><p id="p-2978">Now let $a = a_1 \vee \cdots \vee a_n$ and $b = b_1 \vee \cdots \vee b_m$ be elements in $B\text{,}$ where each $a_i$ and each $b_i$ is an atom. If $\phi(a) = \phi(b)\text{,}$ then $\{a_1, \ldots, a_n \} = \{b_1, \ldots, b_m \}$ and $a = b\text{.}$ Consequently, $\phi$ is injective.</p><p id="p-2979">The join of $a$ and $b$ is preserved by $\phi$ since</p><div class="displaymath">
\begin{align*}
\phi(a \vee b) & = \phi( a_1 \vee \cdots \vee a_n \vee b_1 \vee \cdots \vee b_m )\\
& = \{ a_1, \ldots, a_n, b_1, \ldots, b_m \}\\
& = \{ a_1, \ldots, a_n \} \cup \{ b_1, \ldots, b_m \}\\
& = \phi( a_1 \vee \cdots \vee a_n ) \cup \phi( b_1 \wedge \cdots \vee b_m )\\
& = \phi(a) \cup \phi(b).
\end{align*}
</div><p>Similarly, $\phi( a \wedge b ) = \phi(a) \cap \phi(b)\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-2980">We leave the proof of the following corollary as an exercise.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-28"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">19.24</span></h6><p id="p-2981">The order of any finite Boolean algebra must be $2^n$ for some positive integer $n\text{.}$</p></article></div>
