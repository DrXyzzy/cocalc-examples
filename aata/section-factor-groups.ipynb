<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Normal Subgroups"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Normal Subgroups</span></h3><a href="section-factor-groups.ipynb#normal-subsection-normal-subgroups" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-1599">A subgroup $H$ of a group $G$ is <dfn class="terminology">normal</dfn> in G if $gH = Hg$ for all $g \in G\text{.}$ That is, a normal subgroup of a group $G$ is one in which the right and left cosets are precisely the same.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-normal-abelian"><h6 class="heading"><span class="type">Example</span><span class="codenumber">10.1</span></h6><p id="p-1600">Let $G$ be an abelian group. Every subgroup $H$ of $G$ is a normal subgroup.  Since $gh = hg$ for all $g \in G$ and $h \in H\text{,}$ it will always be the case that $gH = Hg\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-normal-s3"><h6 class="heading"><span class="type">Example</span><span class="codenumber">10.2</span></h6><p id="p-1601">Let $H$ be the subgroup of $S_3$ consisting of elements $(1)$ and $(12)\text{.}$ Since</p><div class="displaymath">
\begin{equation*}
(123) H = \{ (123), (13) \} \quad \text{and} \quad H (123) = \{ (123), (23) \},
\end{equation*}
</div><p>$H$ cannot be a normal subgroup of $S_3\text{.}$  However, the subgroup $N\text{,}$ consisting of the permutations $(1)\text{,}$ $(123)\text{,}$ and $(132)\text{,}$ is normal since the cosets of $N$ are</p><div class="displaymath">
\begin{gather*}
N  =   \{ (1), (123), (132) \}\\
(12) N =  N (12)  =  \{ (12), (13), (23) \}.
\end{gather*}
</div></article></div>

<div class="mathbook-content"><p id="p-1602">The following theorem is fundamental to our understanding of normal subgroups.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-normal-equivalents"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">10.3</span></h6><p id="p-1603">Let $G$ be a group and $N$ be a subgroup of $G\text{.}$ Then the following statements are equivalent. </p><ol class="decimal"><li id="li-381"><p id="p-1604">The subgroup $N$ is normal in $G\text{.}$</p></li><li id="li-382"><p id="p-1605">For all $g \in G\text{,}$ $gNg^{-1} \subset N\text{.}$</p></li><li id="li-383"><p id="p-1606">For all $g \in G\text{,}$ $gNg^{-1} = N\text{.}$</p></li></ol></article><article class="proof" id="proof-67"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1607">(1) $\Rightarrow$ (2). Since $N$ is normal in $G\text{,}$ $gN = Ng$ for all $g \in G\text{.}$ Hence, for a given $g \in G$ and $n \in N\text{,}$ there exists an $n'$ in $N$ such that $g n = n' g\text{.}$ Therefore, $gng^{-1} = n' \in N$ or $gNg^{-1} \subset N\text{.}$</p><p id="p-1608">(2)  $\Rightarrow$ (3).  Let $g \in G\text{.}$ Since $gNg^{-1} \subset N\text{,}$ we need only show $N \subset gNg^{-1}\text{.}$ For $n \in N\text{,}$  $g^{-1}ng=g^{-1}n(g^{-1})^{-1} \in N\text{.}$  Hence, $g^{-1}ng = n'$ for some $n' \in N\text{.}$ Therefore, $n = g n' g^{-1}$ is in $g N g^{-1}\text{.}$</p><p id="p-1609">(3) $\Rightarrow$ (1). Suppose that $gNg^{-1} = N$ for all $g \in G\text{.}$ Then for any $n \in N$ there exists an $n' \in N$ such that $gng^{-1} = n'\text{.}$  Consequently, $gn = n' g$ or $gN \subset Ng\text{.}$ Similarly, $Ng \subset gN\text{.}$</p></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Factor Groups"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Factor Groups</span></h3><a href="section-factor-groups.ipynb#normal-subsection-factor-groups" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-1610">If $N$ is a normal subgroup of a group $G\text{,}$ then the cosets of $N$ in $G$ form a group $G/N$ under the operation $(aN) (bN) = abN\text{.}$ This group is called the <dfn class="terminology">factor</dfn> or <dfn class="terminology">quotient group</dfn> of $G$ and $N\text{.}$   Our first task is to prove that $G/N$ is indeed a group.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-51"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">10.4</span></h6><p id="p-1611">Let $N$ be a normal subgroup of a group $G\text{.}$ The cosets of $N$ in $G$ form a group $G/N$ of order $[G:N]\text{.}$</p></article><article class="proof" id="proof-68"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1612">The group operation on $G/N$ is $(a N ) (b N)= a b N\text{.}$  This operation must be shown to be well-defined; that is, group multiplication must be independent of the choice of  coset representative. Let $aN = bN$ and $cN = dN\text{.}$ We must show that</p><div class="displaymath">
\begin{equation*}
(aN) (cN) = acN = bd N = (b N)(d N).
\end{equation*}
</div><p>Then $a = b n_1$ and $c = d n_2$ for some $n_1$ and $n_2$ in $N\text{.}$ Hence,</p><div class="displaymath">
\begin{align*}
acN & = b n_1 d n_2 N\\
& = b n_1 d N\\
& = b n_1 N d\\
& = b N d\\
& = b d N.
\end{align*}
</div><p>The remainder of the theorem is easy: $eN = N$ is the identity and $g^{-1} N$ is the inverse of $gN\text{.}$ The order of $G/N$ is, of course, the number of cosets of $N$ in $G\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1613">It is very important to remember that the elements in a factor group are <em class="emphasis">sets of elements</em> in the original group.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-normal-factor-s3"><h6 class="heading"><span class="type">Example</span><span class="codenumber">10.5</span></h6><p id="p-1614">Consider the normal subgroup of $S_3\text{,}$ $N = \{ (1), (123), (132)  \}\text{.}$ The cosets of $N$ in $S_3$ are $N$ and $(12) N\text{.}$ The factor group $S_3 / N$ has the following multiplication table.</p><div class="sidebyside" id="sidebyside-6"><div class="sbsrow" style="margin-left:0%;margin-right:0%;"><div class="sbspanel" style="width:100%;justify-content:flex-start;"><div class="displaymath" id="p-1615">
\begin{equation*}
\begin{array}{c|cc}
& N & (12) N \\ \hline
N & N & (12) N \\
(12) N & (12) N & N
\end{array}
\end{equation*}
</div></div></div></div><p id="p-1616">This group is isomorphic to ${\mathbb Z}_2\text{.}$ At first, multiplying cosets seems both complicated and strange; however, notice that  $S_3 / N$ is a smaller group. The factor group displays a certain amount of information about $S_3\text{.}$  Actually, $N = A_3\text{,}$ the group of even permutations, and $(12) N = \{ (12), (13), (23) \}$ is the set of odd permutations. The information captured in $G/N$ is parity; that is, multiplying two even or two odd permutations results in an even permutation, whereas multiplying an odd permutation by an even permutation yields an odd permutation.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-normal-factor-z3"><h6 class="heading"><span class="type">Example</span><span class="codenumber">10.6</span></h6><p id="p-1617">Consider the normal subgroup $3 {\mathbb Z}$ of ${\mathbb Z}\text{.}$ The cosets of $3 {\mathbb Z}$ in ${\mathbb Z}$ are</p><div class="displaymath">
\begin{align*}
0 + 3 {\mathbb Z} & = \{ \ldots, -3, 0, 3, 6, \ldots \}\\
1 + 3 {\mathbb Z} & = \{ \ldots, -2, 1, 4, 7, \ldots \}\\
2 + 3 {\mathbb Z} & = \{ \ldots, -1, 2, 5, 8, \ldots \}.
\end{align*}
</div><p>The group ${\mathbb Z}/ 3 {\mathbb Z}$ is given by the multiplication table below.</p><div class="sidebyside" id="sidebyside-7"><div class="sbsrow" style="margin-left:0%;margin-right:0%;"><div class="sbspanel" style="width:100%;justify-content:flex-start;"><div class="displaymath" id="p-1618">
\begin{equation*}
\begin{array}{c|ccc}
+ & 0 + 3{\mathbb Z} & 1 + 3{\mathbb Z} & 2 + 3{\mathbb Z} \\\hline
0 + 3{\mathbb Z} & 0 + 3{\mathbb Z} & 1 + 3{\mathbb Z} & 2 + 3{\mathbb Z} \\
1 + 3{\mathbb Z} & 1 + 3{\mathbb Z} & 2 + 3{\mathbb Z} & 0 + 3{\mathbb Z} \\
2 + 3{\mathbb Z} & 2 + 3{\mathbb Z} & 0 + 3{\mathbb Z} & 1 + 3{\mathbb Z}
\end{array}
\end{equation*}
</div></div></div></div><p id="p-1619">In general, the subgroup $n {\mathbb Z}$ of ${\mathbb Z}$ is normal. The cosets of ${\mathbb Z } / n {\mathbb Z}$ are</p><div class="displaymath">
\begin{gather*}
n {\mathbb Z}\\
1 + n {\mathbb Z}\\
2 + n {\mathbb Z}\\
\vdots\\
(n-1) + n {\mathbb Z}.
\end{gather*}
</div><p>The sum of the cosets $k + {\mathbb Z}$ and $l + {\mathbb Z}$ is $k+l + {\mathbb Z}\text{.}$ Notice that  we have written our cosets additively, because the group operation is integer addition.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-normal-factor_dn"><h6 class="heading"><span class="type">Example</span><span class="codenumber">10.7</span></h6><p id="p-1620">Consider the dihedral group $D_n\text{,}$ generated by the two elements $r$ and $s\text{,}$ satisfying the relations</p><div class="displaymath">
\begin{align*}
r^n & = \identity\\
s^2 & = \identity\\
srs & = r^{-1}.
\end{align*}
</div><p>The element $r$ actually generates the cyclic subgroup of rotations, $R_n\text{,}$ of $D_n\text{.}$  Since $srs^{-1} = srs = r^{-1} \in R_n\text{,}$ the group of rotations is a normal subgroup of $D_n\text{;}$ therefore, $D_n / R_n$ is a group.  Since there are exactly two elements in this group, it must be isomorphic to ${\mathbb Z}_2\text{.}$</p></article></div>
