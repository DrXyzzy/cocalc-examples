<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 20.2 Subspaces"><span class="type">Section</span><span class="codenumber">20.2</span><span class="title">Subspaces</span></h2><a href="section-subspaces.ipynb" class="permalink">Â¶</a></div>

<div class="mathbook-content"><p id="p-3103">Just as groups have subgroups and rings have subrings, vector spaces also have substructures. Let $V$ be a vector space over a field $F\text{,}$ and $W$ a subset of $V\text{.}$ Then $W$ is a <dfn class="terminology">subspace</dfn> of $V$ if it is closed under vector addition and scalar multiplication; that is, if $u, v \in W$ and $\alpha \in F\text{,}$ it will always be the case that $u + v$ and $\alpha v$ are also in $W\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-vect-subspace-w"><h6 class="heading"><span class="type">Example</span><span class="codenumber">20.6</span></h6><p id="p-3104">Let $W$ be the subspace of ${\mathbb R}^3$ defined by $W = \{ (x_1, 2 x_1 + x_2, x_1 - x_2) : x_1, x_2 \in {\mathbb R} \}\text{.}$ We claim that $W$ is a  subspace of ${\mathbb R}^3\text{.}$  Since</p><div class="displaymath">
\begin{align*}
\alpha (x_1, 2 x_1 + x_2, x_1 - x_2) & =  (\alpha x_1, \alpha(2 x_1 + x_2), \alpha( x_1 - x_2))\\
& =  (\alpha x_1, 2(\alpha x_1) + \alpha x_2, \alpha x_1 -\alpha x_2),
\end{align*}
</div><p>$W$ is closed under scalar multiplication. To show that $W$ is closed under vector addition, let $u = (x_1, 2 x_1 + x_2, x_1 - x_2)$ and $v = (y_1, 2 y_1 + y_2, y_1 - y_2)$ be vectors in $W\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
u + v = (x_1 + y_1, 2( x_1 + y_1) +( x_2 + y_2), (x_1 + y_1) - (x_2+ y_2)).
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="example-like" id="example-vect-subspace-poly"><h6 class="heading"><span class="type">Example</span><span class="codenumber">20.7</span></h6><p id="p-3105">Let $W$ be the subset of polynomials of $F[x]$ with no odd-power terms. If $p(x)$ and $q(x)$ have no odd-power terms, then neither will  $p(x) + q(x)\text{.}$  Also, $\alpha p(x) \in W$ for $\alpha \in F$ and $p(x) \in W\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3106">Let $V$ be any vector space over a field $F$ and suppose that $v_1, v_2, \ldots, v_n$ are vectors in $V$ and $\alpha_1, \alpha_2, \ldots, \alpha_n$ are scalars in $F\text{.}$ Any vector $w$ in $V$ of the form</p><div class="displaymath">
\begin{equation*}
w = \sum_{i=1}^n \alpha_i v_i = \alpha_1 v_1 + \alpha_2 v_2 + \cdots + \alpha_n v_n
\end{equation*}
</div><p>is called a <dfn class="terminology">linear combination</dfn> of the vectors $v_1, v_2, \ldots, v_n\text{.}$ The <dfn class="terminology">spanning set</dfn> of vectors $v_1, v_2, \ldots, v_n$ is the set of vectors obtained from all possible linear combinations of $v_1, v_2, \ldots, v_n\text{.}$ If $W$ is the spanning set of $v_1, v_2, \ldots, v_n\text{,}$ then we say that $W$ is <dfn class="terminology">spanned</dfn> by $v_1, v_2, \ldots, v_n\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-40"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">20.8</span></h6><p id="p-3107">Let $S= \{v_1, v_2, \ldots, v_n \}$ be vectors in a vector space $V\text{.}$ Then the span of $S$ is a subspace of $V\text{.}$</p></article><article class="proof" id="proof-156"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3108">Let $u$ and $v$ be in $S\text{.}$ We can write both of these vectors as  linear combinations of the $v_i$'s:</p><div class="displaymath">
\begin{align*}
u & =  \alpha_1 v_1 + \alpha_2 v_2 + \cdots + \alpha_n v_n\\
v & =  \beta_1 v_1 + \beta_2 v_2 + \cdots + \beta_n v_n.
\end{align*}
</div><p>Then</p><div class="displaymath">
\begin{equation*}
u + v =( \alpha_1 + \beta_1) v_1 + (\alpha_2+ \beta_2) v_2 + \cdots + (\alpha_n + \beta_n) v_n 
\end{equation*}
</div><p>is a linear combination of the $v_i$'s. For $\alpha \in F\text{,}$</p><div class="displaymath">
\begin{equation*}
\alpha u = (\alpha \alpha_1) v_1 + ( \alpha \alpha_2) v_2 + \cdots + (\alpha \alpha_n ) v_n 
\end{equation*}
</div><p>is in the span of $S\text{.}$</p></article></div>
