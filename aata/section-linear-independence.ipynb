<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 20.3 Linear Independence"><span class="type">Section</span><span class="codenumber">20.3</span><span class="title">Linear Independence</span></h2><a href="section-linear-independence.ipynb" class="permalink">Â¶</a></div>

<div class="mathbook-content"><p id="p-3109">Let $S = \{v_1, v_2, \ldots, v_n\}$ be a set of vectors in a vector space $V\text{.}$ If there exist scalars $\alpha_1, \alpha_2 \ldots \alpha_n \in F$ such that not all of the $\alpha_i$'s are zero and</p><div class="displaymath">
\begin{equation*}
\alpha_1 v_1 + \alpha_2 v_2 + \cdots + \alpha_n v_n = {\mathbf 0 },
\end{equation*}
</div><p>then $S$ is said to be <dfn class="terminology">linearly dependent</dfn>. If the set $S$ is not linearly dependent, then it is said to be <dfn class="terminology">linearly independent</dfn>. More specifically, $S$ is a linearly independent set if</p><div class="displaymath">
\begin{equation*}
\alpha_1 v_1 + \alpha_2 v_2 + \cdots + \alpha_n v_n = {\mathbf 0 }
\end{equation*}
</div><p>implies that</p><div class="displaymath">
\begin{equation*}
\alpha_1 = \alpha_2 = \cdots = \alpha_n = 0
\end{equation*}
</div><p>for any set of scalars $\{ \alpha_1, \alpha_2 \ldots \alpha_n \}\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-41"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">20.9</span></h6><p id="p-3110">Let $\{ v_1, v_2, \ldots, v_n \}$ be a set of linearly independent vectors in a vector space. Suppose that</p><div class="displaymath">
\begin{equation*}
v = \alpha_1 v_1 + \alpha_2 v_2 + \cdots + \alpha_n v_n = \beta_1 v_1 + \beta_2 v_2 + \cdots + \beta_n v_n.
\end{equation*}
</div><p>Then $\alpha_1 = \beta_1, \alpha_2 = \beta_2, \ldots, \alpha_n = \beta_n\text{.}$</p></article><article class="proof" id="proof-157"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3111">If</p><div class="displaymath">
\begin{equation*}
v = \alpha_1 v_1 + \alpha_2 v_2 + \cdots + \alpha_n v_n = \beta_1 v_1 + \beta_2 v_2 + \cdots + \beta_n v_n,
\end{equation*}
</div><p>then</p><div class="displaymath">
\begin{equation*}
(\alpha_1 - \beta_1) v_1 + (\alpha_2 - \beta_2) v_2 + \cdots + (\alpha_n - \beta_n) v_n = {\mathbf 0}.
\end{equation*}
</div><p>Since $v_1, \ldots, v_n$ are linearly independent, $\alpha_i - \beta_i = 0$ for $i = 1, \ldots, n\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3112">The definition of linear dependence makes more sense if we consider the following proposition.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-42"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">20.10</span></h6><p id="p-3113">A set $\{ v_1, v_2, \dots, v_n \}$ of vectors in a vector space $V$ is linearly dependent if and only if one of the $v_i$'s is a linear combination of the rest.</p></article><article class="proof" id="proof-158"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3114">Suppose that $\{ v_1, v_2, \dots, v_n \}$ is a set of linearly dependent vectors.  Then there exist scalars $\alpha_1, \ldots, \alpha_n$ such that</p><div class="displaymath">
\begin{equation*}
\alpha_1 v_1 + \alpha_2 v_2 + \cdots + \alpha_n v_n = {\mathbf 0 },
\end{equation*}
</div><p>with at least one of the $\alpha_i$'s not equal to zero.  Suppose that $\alpha_k \neq 0\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
v_k = - \frac{\alpha_1}{\alpha_k} v_1 - \cdots - \frac{\alpha_{k - 1}}{\alpha_k} v_{k-1} - \frac{\alpha_{k + 1}}{\alpha_k} v_{k + 1} - \cdots  - \frac{\alpha_n}{\alpha_k} v_n.
\end{equation*}
</div><p id="p-3115">Conversely, suppose that</p><div class="displaymath">
\begin{equation*}
v_k = \beta_1 v_1 + \cdots + \beta_{k - 1} v_{k - 1} + \beta_{k + 1} v_{k + 1} + \cdots + \beta_n v_n.
\end{equation*}
</div><p>Then</p><div class="displaymath">
\begin{equation*}
\beta_1 v_1 + \cdots + \beta_{k - 1} v_{k - 1} - v_k + \beta_{k + 1} v_{k + 1} + \cdots + \beta_n v_n = {\mathbf 0}.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><p id="p-3116">The following proposition is a consequence of the fact that any system of homogeneous linear equations with more unknowns than equations will have a nontrivial solution.  We leave the details of the proof for the end-of-chapter exercises.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-linearly-independent"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">20.11</span></h6><p id="p-3117">Suppose that a vector space $V$ is spanned by $n$ vectors. If $m \gt n\text{,}$ then any set of $m$ vectors in $V$ must be linearly dependent.</p></article></div>

<div class="mathbook-content"><p id="p-3118">A set $\{ e_1, e_2, \ldots, e_n \}$ of vectors in a vector space $V$ is called a <dfn class="terminology">basis</dfn> for $V$ if $\{ e_1, e_2, \ldots, e_n \}$ is a linearly independent set that spans $V\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-vect-basis-r3"><h6 class="heading"><span class="type">Example</span><span class="codenumber">20.12</span></h6><p id="p-3119">The vectors $e_1 = (1, 0, 0)\text{,}$ $e_2 = (0, 1, 0)\text{,}$ and $e_3 =(0, 0, 1)$ form a basis for ${\mathbb R}^3\text{.}$  The set certainly spans ${\mathbb R}^3\text{,}$ since any arbitrary vector $(x_1, x_2, x_3)$ in ${\mathbb R}^3$ can be written as $x_1 e_1 + x_2 e_2 + x_3 e_3\text{.}$ Also, none of the vectors $e_1, e_2, e_3$ can be written as a linear combination of the other two; hence, they are linearly independent.  The vectors $e_1, e_2, e_3$ are not the only basis of ${\mathbb R}^3\text{:}$  the set $\{ (3, 2, 1), (3, 2, 0), (1, 1, 1) \}$ is also a basis for ${\mathbb R}^3\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-vect-basis-sqrt2"><h6 class="heading"><span class="type">Example</span><span class="codenumber">20.13</span></h6><p id="p-3120">Let ${\mathbb Q}( \sqrt{2}\, ) = \{ a + b \sqrt{2} : a, b \in {\mathbb Q} \}\text{.}$ The sets $\{1, \sqrt{2}\,  \}$ and $\{1 + \sqrt{2}, 1 - \sqrt{2}\,  \}$ are both bases of ${\mathbb Q}( \sqrt{2}\, )\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3121">From the last two examples it should be clear that a given vector space has several bases. In fact, there are an infinite number of bases for both of these examples. <em class="emphasis">In general, there is no unique basis for a vector space.</em>  However, every basis of ${\mathbb R}^3$ consists of exactly three vectors, and every  basis of ${\mathbb Q}(\sqrt{2}\, )$ consists of exactly two vectors. This is a consequence of the next proposition.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-44"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">20.14</span></h6><p id="p-3122">Let $\{ e_1, e_2, \ldots, e_m \}$ and $\{ f_1, f_2, \ldots, f_n \}$ be two bases for a vector space $V\text{.}$ Then $m = n\text{.}$</p></article><article class="proof" id="proof-159"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3123">Since $\{ e_1, e_2, \ldots, e_m \}$ is a basis, it is a linearly independent set. By PropositionÂ <a href="section-linear-independence.ipynb#proposition-linearly-independent" class="xref" alt="Proposition 20.11 " title="Proposition 20.11 ">20.11</a>, $n \leq m\text{.}$ Similarly, $\{ f_1, f_2, \ldots, f_n \}$ is a linearly independent set, and the last proposition implies that $m \leq n\text{.}$ Consequently, $m = n\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3124">If $\{ e_1, e_2, \ldots, e_n \}$ is a basis for a vector space $V\text{,}$ then we say that the <dfn class="terminology">dimension</dfn> of $V$ is $n$ and we write $\dim V =n\text{.}$  We will leave the proof of the following theorem as an exercise.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-109"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">20.15</span></h6><p id="p-3125">Let $V$ be a vector space of dimension $n\text{.}$ </p><ol class="decimal"><li id="li-699"><p id="p-3126">If $S = \{v_1, \ldots, v_n \}$ is a set of linearly independent vectors for $V\text{,}$ then $S$ is a basis for $V\text{.}$</p></li><li id="li-700"><p id="p-3127">If $S = \{v_1, \ldots, v_n \}$ spans $V\text{,}$ then $S$ is a basis for $V\text{.}$</p></li><li id="li-701"><p id="p-3128">If $S = \{v_1, \ldots, v_k \}$ is a set of linearly independent vectors for $V$ with $k \lt n\text{,}$ then there exist vectors $v_{k + 1}, \ldots, v_n$ such that</p><div class="displaymath">
\begin{equation*}
\{v_1, \ldots, v_k, v_{k + 1}, \ldots, v_n \}
\end{equation*}
</div><p>is a basis for $V\text{.}$</p></li></ol></article></div>
