<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 16.3 Ring Homomorphisms and Ideals"><span class="type">Section</span><span class="codenumber">16.3</span><span class="title">Ring Homomorphisms and Ideals</span></h2><a href="section-ring-homomorphisms.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-2352">In the study of groups, a homomorphism is a map that preserves the operation of the group.  Similarly, a homomorphism between rings preserves the operations of addition and multiplication in the ring. More specifically, if $R$ and $S$ are rings, then a <dfn class="terminology">ring homomorphism</dfn> is a map $\phi : R \rightarrow S$ satisfying</p><div class="displaymath">
\begin{align*}
\phi( a + b ) & = \phi( a ) + \phi(b)\\
\phi( a b ) & = \phi( a ) \phi(b)
\end{align*}
</div><p>for all $a, b \in R\text{.}$ If $\phi : R \rightarrow S$ is a one-to-one and onto homomorphism, then $\phi$ is called an <dfn class="terminology">isomorphism</dfn> of rings.</p></div>

<div class="mathbook-content"><p id="p-2353">The set of elements that a ring homomorphism maps to $0$ plays a fundamental role in the theory of rings. For any ring homomorphism $\phi : R \rightarrow S\text{,}$ we define the <dfn class="terminology">kernel</dfn> of a ring homomorphism to be the set</p><div class="displaymath">
\begin{equation*}
\ker \phi = \{ r \in R : \phi( r ) = 0 \}.
\end{equation*}
</div></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-homo"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.20</span></h6><p id="p-2354">For any integer $n$ we can define a ring homomorphism $\phi : {\mathbb Z} \rightarrow {\mathbb Z}_n$ by $a \mapsto a \pmod{n}\text{.}$ This is indeed a ring homomorphism, since</p><div class="displaymath">
\begin{align*}
\phi( a + b )  & = (a + b) \pmod{n}\\
& =  a \pmod{n} + b \pmod{n}\\
& = \phi( a ) + \phi(b)
\end{align*}
</div><p>and</p><div class="displaymath">
\begin{align*}
\phi( a b )  & =  ab \pmod{n}\\
& = a \pmod{n}\cdot  b \pmod{n}\\
& = \phi( a ) \phi(b).
\end{align*}
</div><p>The kernel of the homomorphism $\phi$ is $n {\mathbb Z}\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-continuous-function-homomorph"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.21</span></h6><p id="p-2355">Let $C[a, b]$ be the ring of continuous real-valued functions on an interval $[a,b]$ as in Example <a href="section-rings-definitions.ipynb#example-rings-functions" class="xref" alt="Example 16.5 " title="Example 16.5 ">16.5</a>. For a fixed  $\alpha \in [a, b]\text{,}$ we can define a ring homomorphism $\phi_{\alpha} : C[a, b] \rightarrow {\mathbb R}$ by $\phi_{\alpha} (f ) = f( \alpha)\text{.}$ This  is a ring homomorphism since</p><div class="displaymath">
\begin{gather*}
\phi_{\alpha}( f + g )  = (f + g)( \alpha) = f(\alpha) + g(\alpha) = \phi_{\alpha}( f ) + \phi_{\alpha}(g )\\
\phi_{\alpha}( f  g )  = (f  g)( \alpha) = f(\alpha)  g(\alpha) = \phi_{\alpha}( f )  \phi_{\alpha}(g ).
\end{gather*}
</div><p>Ring homomorphisms of the type $\phi_{\alpha}$ are called <dfn class="terminology">evaluation homomorphisms</dfn>.</p></article></div>

<div class="mathbook-content"><p id="p-2356">In the next proposition we will examine some fundamental properties of ring homomorphisms. The proof of the proposition is left as an exercise.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-34"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">16.22</span></h6><p id="p-2357">Let $\phi : R \rightarrow S$ be a ring homomorphism. </p><ol class="decimal"><li id="li-514"><p id="p-2358">If $R$ is a commutative ring, then $\phi(R)$ is a commutative ring.</p></li><li id="li-515"><p id="p-2359">$\phi( 0 ) = 0\text{.}$</p></li><li id="li-516"><p id="p-2360">Let $1_R$ and $1_S$ be the identities for $R$ and $S\text{,}$ respectively. If $\phi$ is onto, then $\phi(1_R) = 1_S\text{.}$</p></li><li id="li-517"><p id="p-2361">If $R$ is a field and $\phi(R) \neq \{ 0 \}\text{,}$ then $\phi(R)$ is a field.</p></li></ol></article></div>

<div class="mathbook-content"><p id="p-2362">In group theory we found that normal subgroups play a special role. These subgroups have nice characteristics that make them more interesting to study than arbitrary subgroups.  In ring theory the objects corresponding to normal subgroups are a special class of subrings called ideals. An <dfn class="terminology">ideal</dfn> in a ring $R$ is a subring $I$ of $R$ such that if $a$ is in $I$ and $r$ is in $R\text{,}$ then both $ar$ and $ra$ are in $I\text{;}$ that is, $rI \subset I$ and $Ir \subset I$ for all $r \in R\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-trivial-ideal"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.23</span></h6><p id="p-2363">Every ring $R$ has at least two ideals, $\{ 0 \}$ and $R\text{.}$  These ideals are called the <dfn class="terminology">trivial ideals</dfn>.</p></article></div>

<div class="mathbook-content"><p id="p-2364">Let $R$ be a ring with identity and suppose that $I$ is an ideal in $R$ such that $1$ is in $I\text{.}$ Since for any $r \in R\text{,}$ $r1 = r \in I$ by the definition of an ideal, $I = R\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-principal-ideal"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.24</span></h6><p id="p-2365">If $a$ is any element in a commutative ring $R$ with identity, then the set</p><div class="displaymath">
\begin{equation*}
\langle a \rangle = \{ ar : r \in R \}
\end{equation*}
</div><p>is an ideal in $R\text{.}$ Certainly, $\langle a \rangle$ is nonempty since both $0 = a0$ and $a = a1$ are in $\langle a \rangle\text{.}$ The sum of two elements in $\langle a \rangle$ is again in $\langle a \rangle$ since $ar + ar' =  a(r + r')\text{.}$ The inverse of $ar$ is $-ar = a (-r) \in \langle a \rangle\text{.}$  Finally, if we multiply an element $ar \in \langle a \rangle$ by an arbitrary element $s \in R\text{,}$ we have $s(ar) = a(sr)\text{.}$  Therefore, $\langle a \rangle$ satisfies the definition of an ideal.</p></article></div>

<div class="mathbook-content"><p id="p-2366">If $R$ is a commutative ring with identity, then an ideal of the form $\langle a \rangle  = \{ ar : r \in R \}$  is called a <dfn class="terminology">principal ideal</dfn>.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-z-ideals"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">16.25</span></h6><p id="p-2367">Every ideal in the ring of integers ${\mathbb Z}$ is a principal ideal.</p></article><article class="proof" id="proof-107"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2368">The zero ideal $\{ 0 \}$ is a principal ideal since $\langle 0 \rangle = \{ 0 \}\text{.}$ If  $I$ is any nonzero ideal in ${\mathbb Z}\text{,}$ then  $I$ must contain some positive integer $m\text{.}$  There exists a least positive integer $n$ in $I$ by the Principle of Well-Ordering. Now let $a$ be any element in $I\text{.}$ Using the division algorithm, we know that there exist integers $q$ and $r$ such that</p><div class="displaymath">
\begin{equation*}
a = nq + r
\end{equation*}
</div><p>where $0 \leq r \lt n\text{.}$ This equation tells us that $r = a - nq \in I\text{,}$ but $r$ must be $0$ since $n$ is the least positive element in $I\text{.}$ Therefore, $a = nq$ and $I = \langle n \rangle\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-integer-ideal"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.26</span></h6><p id="p-2369">The set $n {\mathbb Z}$ is ideal in the ring of integers. If $na$ is in $n{\mathbb Z}$ and $b$ is in ${\mathbb Z}\text{,}$ then $nab$ is in  $n {\mathbb Z}$ as required. In fact, by Theorem <a href="section-ring-homomorphisms.ipynb#theorem-z-ideals" class="xref" alt="Theorem 16.25 " title="Theorem 16.25 ">16.25</a>, these are the only ideals of ${\mathbb Z}\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-35"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">16.27</span></h6><p id="p-2370">The kernel of any ring homomorphism $\phi : R \rightarrow S$ is an ideal in $R\text{.}$</p></article><article class="proof" id="proof-108"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2371">We know from group theory that $\ker \phi$ is an additive subgroup of $R\text{.}$ Suppose that $r \in R$ and $a \in \ker \phi\text{.}$ Then we must show that $ar$ and $ra$ are in $\ker \phi\text{.}$ However,</p><div class="displaymath">
\begin{equation*}
\phi(ar) = \phi(a) \phi(r) = 0 \phi(r) = 0
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
\phi(ra) = \phi(r) \phi(a) =  \phi(r)0 = 0.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="remark-like" id="remark-7"><h6 class="heading"><span class="type">Remark</span><span class="codenumber">16.28</span></h6><p id="p-2372">In our definition of an ideal we have required that $rI \subset I$ and $Ir \subset I$ for all $r \in R\text{.}$  Such ideals are sometimes referred to as <dfn class="terminology">two-sided ideals</dfn>.  We can also consider <dfn class="terminology">one-sided ideals</dfn>; that is, we may require only that either $rI \subset I$ or $Ir \subset I$ for $r \in R$ hold but not both. Such ideals are called <dfn class="terminology">left ideals</dfn> and <dfn class="terminology">right ideals</dfn>, respectively. Of course, in a commutative ring any ideal must be two-sided. In this text we will concentrate on two-sided ideals.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-factor-ring"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">16.29</span></h6><p id="p-2373">Let $I$ be an ideal of $R\text{.}$ The factor group $R/I$ is a ring with multiplication defined by</p><div class="displaymath">
\begin{equation*}
(r + I)(s + I) = rs + I.
\end{equation*}
</div></article><article class="proof" id="proof-109"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2374">We already know that $R/I$ is an abelian group under addition. Let $r+I$ and $s +I$ be in $R/I\text{.}$ We must show that the product $(r + I)(s + I) = rs + I$ is independent of the choice of coset; that is, if $r' \in r+I$ and $s' \in s+I\text{,}$ then $r's'$ must be in $rs+I\text{.}$ Since $r' \in r+I\text{,}$ there exists an element $a$  in $I$ such that $r' = r + a\text{.}$ Similarly, there exists a $b \in I$ such that $s' = s + b\text{.}$ Notice that</p><div class="displaymath">
\begin{equation*}
r' s' = (r+a)(s+b) = rs + as + rb + ab
\end{equation*}
</div><p>and $as + rb + ab \in I$ since $I$ is an ideal; consequently, $r' s' \in rs + I\text{.}$ We will leave as an exercise the verification of the  associative law for multiplication and the distributive laws.</p></article></div>

<div class="mathbook-content"><p id="p-2375">The ring $R/I$ in Theorem <a href="section-ring-homomorphisms.ipynb#theorem-factor-ring" class="xref" alt="Theorem 16.29 " title="Theorem 16.29 ">16.29</a> is called the <dfn class="terminology">factor</dfn> or <dfn class="terminology">quotient ring</dfn>. Just as with group homomorphisms and normal subgroups, there is a relationship between ring homomorphisms and ideals.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-81"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">16.30</span></h6><p id="p-2376">Let $I$ be an ideal of $R\text{.}$ The map $\phi : R \rightarrow R/I$ defined by $\phi( r ) = r + I$ is a ring homomorphism of $R$ onto $R/I$ with kernel $I\text{.}$</p></article><article class="proof" id="proof-110"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2377">Certainly $\phi : R \rightarrow R/I$ is a surjective abelian group homomorphism. It remains to show that $\phi$ works correctly under ring multiplication.  Let $r$ and $s$ be in $R\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
\phi(r) \phi(s) = (r + I)(s+I) = rs + I = \phi(rs),
\end{equation*}
</div><p>which completes the proof of the theorem.</p></article></div>

<div class="mathbook-content"><p id="p-2378">The map $\phi : R \rightarrow R/I$ is often called the <dfn class="terminology">natural</dfn> or <dfn class="terminology">canonical homomorphism</dfn>. In ring theory we have isomorphism  theorems relating ideals and ring homomorphisms similar to the isomorphism theorems for groups that relate normal subgroups and homomorphisms in Chapter <a href="homomorph.ipynb" class="xref" alt="Chapter 11 Homomorphisms" title="Chapter 11 Homomorphisms">11</a>. We will prove only the First Isomorphism Theorem for rings in this chapter and  leave the proofs of the other two theorems as exercises. All of the proofs are similar to the proofs of the isomorphism theorems for groups.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-82"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">16.31</span><span class="title">First Isomorphism Theorem</span></h6><p id="p-2379">Let $\psi : R \rightarrow S$ be a ring homomorphism. Then $\ker \psi$ is an ideal of $R\text{.}$ If $\phi : R \rightarrow R/\ker \psi$ is the canonical homomorphism, then there exists a unique isomorphism $\eta: R/\ker \psi \rightarrow \psi(R)$ such that $\psi = \eta \phi\text{.}$</p></article><article class="proof" id="proof-111"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2380">Let $K = \ker \psi\text{.}$ By the First Isomorphism Theorem for groups, there exists a well-defined group homomorphism $\eta: R/K \rightarrow \psi(R)$ defined by $\eta(r + K) = \psi(r)$ for the additive abelian groups $R$ and $R/K\text{.}$  To show that this is a ring homomorphism, we need only show that $\eta( (r + K)(s + K) ) = \eta(r + K) \eta( s + K)\text{;}$ but</p><div class="displaymath">
\begin{align*}
\eta( (r + K)( s +K )) & = \eta(r s +K )\\
& = \psi(r s)\\
& = \psi(r) \psi(s)\\
& = \eta( r + K ) \eta( s + K ).
\end{align*}
</div></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-83"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">16.32</span><span class="title">Second Isomorphism Theorem</span></h6><p id="p-2381">Let $I$ be a  subring of a ring $R$  and  $J$ an ideal of $R\text{.}$  Then $I \cap J$ is an ideal of $I$ and</p><div class="displaymath">
\begin{equation*}
I / I \cap J \cong (I+ J) /J.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-84"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">16.33</span><span class="title">Third Isomorphism Theorem</span></h6><p id="p-2382">Let $R$ be a ring and $I$ and $J$ be ideals of $R$ where $J \subset I\text{.}$  Then</p><div class="displaymath">
\begin{equation*}
R/I \cong \frac{R/J}{I/J}.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-correspondence-rings"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">16.34</span><span class="title">Correspondence Theorem</span></h6><p id="p-2383">Let $I$ be an ideal of a ring $R\text{.}$ Then $S \mapsto S/I$ is a one-to-one correspondence between the set of subrings $S$ containing $I$ and the set of subrings of $R/I\text{.}$ Furthermore, the ideals of $R$ containing $I$ correspond to ideals of $R/I\text{.}$</p></article></div>
