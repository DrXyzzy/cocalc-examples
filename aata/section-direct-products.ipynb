<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 9.2 Direct Products"><span class="type">Section</span><span class="codenumber">9.2</span><span class="title">Direct Products</span></h2><a href="section-direct-products.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-1445">Given two groups $G$ and $H\text{,}$ it is possible to construct a new group from the Cartesian product of $G$ and $H\text{,}$ $G \times H\text{.}$ Conversely, given a large group, it is sometimes possible to decompose the group; that is, a group is sometimes isomorphic to the direct product of two smaller groups.  Rather than studying a large group $G\text{,}$ it is often easier to study the component groups of $G\text{.}$</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  External Direct Products"><span class="type">Subsection</span><span class="codenumber" /><span class="title">External Direct Products</span></h3><a href="section-direct-products.ipynb#isomorph-subsection-external-products" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-1446">If $(G,\cdot)$ and $(H, \circ)$ are groups, then we can make the Cartesian product of $G$ and $H$ into a new group. As a set, our group is just the ordered pairs $(g, h) \in G \times H$ where $g \in G$ and $h \in H\text{.}$ We can define a binary operation on $G \times H$ by</p><div class="displaymath">
\begin{equation*}
(g_1, h_1)(g_2, h_2) = (g_1 \cdot g_2, h_1 \circ h_2);
\end{equation*}
</div><p>that is, we just multiply elements in the first coordinate as we do in $G$ and elements in the second coordinate as we do in $H\text{.}$  We have specified the particular operations $\cdot$ and $\circ$ in each group here for the sake of clarity; we usually just write $(g_1, h_1)(g_2, h_2) = (g_1  g_2, h_1 h_2)\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-isomorph-7"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">9.13</span></h6><p id="p-1447">Let $G$ and $H$ be groups. The set $G \times H$ is a group under the operation $(g_1, h_1)(g_2, h_2) = (g_1  g_2, h_1 h_2)$ where $g_1, g_2 \in G$ and $h_1, h_2 \in H\text{.}$</p></article><article class="proof" id="proof-62"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1448">Clearly the binary operation defined above is closed. If $e_G$ and $e_H$ are the identities of the groups $G$ and $H$ respectively, then $(e_G, e_H)$ is the identity of $G \times H\text{.}$  The inverse of $(g, h) \in G \times H$ is $(g^{-1}, h^{-1})\text{.}$  The fact that the operation is associative follows directly from the associativity of $G$ and $H\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-r2-product"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.14</span></h6><p id="p-1449">Let ${\mathbb R}$ be the group of real numbers under addition.  The Cartesian product of ${\mathbb R}$ with itself, ${\mathbb R} \times {\mathbb R} = {\mathbb R}^2\text{,}$ is also a group, in which the group operation is just addition in each coordinate; that is, $(a, b) + (c, d) = (a + c, b + d)\text{.}$  The identity is $(0,0)$ and the inverse of $(a, b)$ is $(-a, -b)\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-z2xz2"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.15</span></h6><p id="p-1450">Consider</p><div class="displaymath">
\begin{equation*}
{\mathbb Z}_2 \times {\mathbb Z}_2 = \{ (0, 0), (0, 1), (1, 0),(1, 1) \}.
\end{equation*}
</div><p>Although ${\mathbb Z}_2 \times {\mathbb Z}_2$ and ${\mathbb Z}_4$ both contain four elements, they are not isomorphic. Every element $(a,b)$ in ${\mathbb Z}_2 \times {\mathbb Z}_2$ has order 2, since $(a,b) + (a,b) = (0,0)\text{;}$ however, ${\mathbb Z}_4$ is cyclic.</p></article></div>

<div class="mathbook-content"><p id="p-1451">The group $G \times H$ is called the <dfn class="terminology">external direct product</dfn> of  $G$ and $H\text{.}$ Notice that there is nothing special about the fact that we have used only two groups to build a new group. The direct product</p><div class="displaymath">
\begin{equation*}
\prod_{i = 1}^n G_i = G_1 \times G_2 \times \cdots \times G_n
\end{equation*}
</div><p>of the groups $G_1, G_2, \ldots, G_n$ is defined in exactly the same manner. If $G = G_1 = G_2 = \cdots = G_n\text{,}$ we often write $G^n$ instead of $G_1 \times G_2 \times \cdots \times G_n\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-z2n"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.16</span></h6><p id="p-1452">The group ${\mathbb Z}_2^n\text{,}$ considered as a set, is just the set of all binary $n$-tuples. The group operation is the “exclusive or” of two binary $n$-tuples. For example,</p><div class="displaymath">
\begin{equation*}
(01011101) + (01001011) = (00010110).
\end{equation*}
</div><p>This group is important in coding theory, in cryptography, and in many areas of computer science.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-isomorph-lcm"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">9.17</span></h6><p id="p-1453">Let $(g, h) \in G \times H\text{.}$ If $g$ and $h$ have finite orders $r$ and $s$ respectively, then the order of $(g, h)$ in $G \times H$ is the least common multiple of $r$ and $s\text{.}$</p></article><article class="proof" id="proof-63"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1454">Suppose that $m$ is the least common multiple of $r$ and $s$ and let $n = |(g,h)|\text{.}$ Then</p><div class="displaymath">
\begin{gather*}
(g,h)^m  = (g^m, h^m) = (e_G,e_H)\\
(g^n, h^n)  = (g, h)^n = (e_G,e_H).
\end{gather*}
</div><p>Hence, $n$ must divide $m\text{,}$ and $n \leq m\text{.}$  However, by the second equation, both $r$ and $s$ must divide $n\text{;}$ therefore, $n$ is a common multiple of $r$ and $s\text{.}$ Since $m$ is the <em class="emphasis">least common multiple</em> of $r$ and $s\text{,}$ $m \leq n\text{.}$  Consequently, $m$ must be equal to $n\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-9"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">9.18</span></h6><p id="p-1455">Let $(g_1, \ldots, g_n) \in \prod G_i\text{.}$ If $g_i$ has finite order $r_i$ in $G_i\text{,}$ then the order of $(g_1, \ldots, g_n)$ in $\prod G_i$ is the least common multiple of $r_1, \ldots, r_n\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-z12xz60"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.19</span></h6><p id="p-1456">Let $(8, 56) \in {\mathbb Z}_{12} \times  {\mathbb Z}_{60}\text{.}$ Since $\gcd(8,12) = 4\text{,}$ the order of 8 is $12/4 = 3$ in ${\mathbb Z}_{12}\text{.}$ Similarly, the order of $56$ in ${\mathbb Z}_{60}$ is $15\text{.}$ The least common multiple of 3 and 15 is 15; hence, $(8, 56)$ has order 15 in ${\mathbb Z}_{12} \times  {\mathbb Z}_{60}\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-z2xz3"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.20</span></h6><p id="p-1457">The group ${\mathbb Z}_2 \times {\mathbb Z}_3$ consists of the pairs</p><div class="displaymath">
\begin{align*}
& (0,0), & & (0, 1), & & (0, 2), & & (1,0), & & (1, 1), & & (1, 2).
\end{align*}
</div><p>In this case, unlike that of ${\mathbb Z}_2 \times {\mathbb Z}_2$ and ${\mathbb Z}_4\text{,}$ it is true that ${\mathbb Z}_2 \times {\mathbb Z}_3 \cong {\mathbb Z}_6\text{.}$ We need only show that ${\mathbb Z}_2 \times {\mathbb Z}_3$ is cyclic.  It is easy to see that $(1,1)$ is a generator for ${\mathbb Z}_2  \times {\mathbb Z}_3\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1458">The next theorem tells us exactly when the direct product of two cyclic groups is cyclic.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-z-pq"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">9.21</span></h6><p id="p-1459">The group ${\mathbb Z}_m \times {\mathbb Z}_n$ is isomorphic to ${\mathbb Z}_{mn}$ if and only if $\gcd(m,n)=1\text{.}$</p></article><article class="proof" id="proof-64"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1460">We will first show that if ${\mathbb Z}_m \times {\mathbb Z}_n \cong {\mathbb Z}_{mn}\text{,}$ then $\gcd(m, n) = 1\text{.}$ We will prove the contrapositive; that is, we will show that if $\gcd(m, n) = d \gt 1\text{,}$ then ${\mathbb Z}_m \times {\mathbb Z}_n$ cannot be cyclic. Notice that $mn/d$ is divisible by both $m$ and $n\text{;}$ hence, for any element $(a,b) \in {\mathbb Z}_m \times {\mathbb Z}_n\text{,}$</p><div class="displaymath">
\begin{equation*}
\underbrace{(a,b) + (a,b)+ \cdots + (a,b)}_{mn/d \; \text{times}} = (0, 0).
\end{equation*}
</div><p>Therefore, no $(a, b)$ can generate all of ${\mathbb Z}_m \times {\mathbb Z}_n\text{.}$</p><p id="p-1461">The converse follows directly from Theorem <a href="section-direct-products.ipynb#theorem-isomorph-lcm" class="xref" alt="Theorem 9.17 " title="Theorem 9.17 ">9.17</a> since $\lcm(m,n) = mn$ if and only if $\gcd(m,n)=1\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-relatively-prime"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">9.22</span></h6><p id="p-1462">Let $n_1, \ldots, n_k$ be positive integers. Then</p><div class="displaymath">
\begin{equation*}
\prod_{i=1}^k {\mathbb Z}_{n_i} \cong {\mathbb Z}_{n_1 \cdots n_k}
\end{equation*}
</div><p>if and only if $\gcd( n_i, n_j) =1$ for $i \neq j\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-11"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">9.23</span></h6><p id="p-1463">If</p><div class="displaymath">
\begin{equation*}
m = p_1^{e_1} \cdots  p_k^{e_k},
\end{equation*}
</div><p>where the $p_i$s are distinct primes, then</p><div class="displaymath">
\begin{equation*}
{\mathbb Z}_m \cong {\mathbb Z}_{p_1^{e_1}} \times \cdots \times {\mathbb Z}_{p_k^{e_k}}.
\end{equation*}
</div></article><article class="proof" id="proof-65"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1464">Since the greatest common divisor of $p_i^{e_i}$ and $p_j^{e_j}$ is 1 for $i \neq j\text{,}$ the proof follows from Corollary <a href="section-direct-products.ipynb#corollary-relatively-prime" class="xref" alt="Corollary 9.22 " title="Corollary 9.22 ">9.22</a>.</p></article></div>

<div class="mathbook-content"><p id="p-1465">In Chapter <a href="struct.ipynb" class="xref" alt="Chapter 13 The Structure of  Groups" title="Chapter 13 The Structure of  Groups">13</a>, we will prove that all finite abelian groups are  isomorphic to direct products of the form</p><div class="displaymath">
\begin{equation*}
{\mathbb Z}_{p_1^{e_1}} \times \cdots \times {\mathbb Z}_{p_k^{e_k}}
\end{equation*}
</div><p>where $p_1, \ldots, p_k$ are (not necessarily distinct) primes.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Internal Direct Products"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Internal Direct Products</span></h3><a href="section-direct-products.ipynb#isomorph-subsection-internal-products" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-1466">The external direct product of two groups builds a large group out of two smaller groups.   We would like to be able to reverse this process and conveniently break down a group into its direct product components; that is, we would like to be able to say when a group is isomorphic to the direct product of two of its subgroups.</p></div>

<div class="mathbook-content"><p id="p-1467">Let $G$ be a group with subgroups $H$ and $K$ satisfying the following conditions. </p><ul class="disc"><li id="li-360"><p id="p-1468">$G = HK = \{ hk : h \in H, k \in K  \}\text{;}$</p></li><li id="li-361"><p id="p-1469">$H \cap K = \{ e \}\text{;}$</p></li><li id="li-362"><p id="p-1470">$hk = kh$ for all $k \in K$ and $h \in H\text{.}$</p></li></ul><p> Then $G$ is the <dfn class="terminology">internal direct product</dfn> of $H$ and $K\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-u8"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.24</span></h6><p id="p-1471">The group $U(8)$ is the internal direct product of</p><div class="displaymath">
\begin{equation*}
H  = \{1, 3 \} \quad \text{and} \quad K  = \{1, 5 \}.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-d6-product"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.25</span></h6><p id="p-1472">The dihedral group $D_6$ is an internal direct product of its two subgroups</p><div class="displaymath">
\begin{equation*}
H  = \{\identity, r^3  \} \quad \text{and} \quad K  = \{\identity, r^2, r^4, s, r^2s, r^4 s   \}.
\end{equation*}
</div><p>It can easily be shown that $K \cong S_3\text{;}$ consequently, $D_6 \cong {\mathbb Z}_2 \times S_3\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-s3-not-a-product"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.26</span></h6><p id="p-1473">Not every group can be written as the internal direct product of two of its proper subgroups.  If the group $S_3$ were an internal direct product of its proper subgroups $H$ and $K\text{,}$ then one of the  subgroups, say $H\text{,}$ would have to have order 3. In this case $H$ is the subgroup $\{ (1), (123), (132) \}\text{.}$ The subgroup $K$ must have order 2, but no matter which subgroup we choose for $K\text{,}$ the condition that $hk = kh$ will never be satisfied for $h \in H$ and $k \in K\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-isomorph-directproducts"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">9.27</span></h6><p id="p-1474">Let $G$ be the internal direct product of  subgroups $H$ and $K\text{.}$ Then $G$ is isomorphic to $H \times K\text{.}$</p></article><article class="proof" id="proof-66"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1475">Since $G$ is an internal direct product, we can write any element $g \in G$ as $g =hk$ for some $h \in H$ and some $k \in K\text{.}$ Define a map $\phi : G \rightarrow H \times K$ by $\phi(g) = (h,k)\text{.}$</p><p id="p-1476">The first problem that we must face is to show that $\phi$ is a well-defined map; that is, we must show that $h$ and $k$ are uniquely determined by $g\text{.}$ Suppose that $g =  hk=h'k'\text{.}$ Then $h^{-1} h'= k (k')^{-1}$ is in both $H$ and $K\text{,}$ so it must be the identity. Therefore, $h = h'$ and $k = k'\text{,}$ which proves that $\phi$ is, indeed, well-defined.</p><p id="p-1477">To show that $\phi$ preserves the group operation, let $g_1 = h_1 k_1$ and $g_2 = h_2 k_2$ and observe that</p><div class="displaymath">
\begin{align*}
\phi( g_1 g_2 ) & = \phi( h_1 k_1 h_2 k_2 )\\
& = \phi(h_1  h_2 k_1 k_2)\\
& = (h_1  h_2, k_1 k_2)\\
& = (h_1, k_1)( h_2, k_2)\\
& = \phi( g_1 ) \phi(  g_2 ).
\end{align*}
</div><p>We will leave the proof that $\phi$ is one-to-one and onto as an exercise.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-isomorph-z6-product"><h6 class="heading"><span class="type">Example</span><span class="codenumber">9.28</span></h6><p id="p-1478">The group ${\mathbb Z}_6$ is an internal direct product isomorphic to $\{ 0, 2, 4\} \times \{ 0, 3 \}\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1479">We can extend the definition of an internal direct product of $G$ to a collection of subgroups $H_1, H_2, \ldots, H_n$ of $G\text{,}$ by requiring that </p><ul class="disc"><li id="li-363"><p id="p-1480">$G = H_1 H_2 \cdots H_n = \{ h_1 h_2 \cdots h_n : h_i \in H_i \}\text{;}$</p></li><li id="li-364"><p id="p-1481">$H_i \cap \langle \cup_{j \neq i} H_j \rangle = \{ e \}\text{;}$</p></li><li id="li-365"><p id="p-1482">$h_i h_j = h_j h_i$ for all $h_i \in H_i$ and $h_j \in H_j\text{.}$</p></li></ul><p> We will leave the proof of the following theorem as an exercise.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-49"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">9.29</span></h6><p id="p-1483">Let $G$ be the internal direct product of subgroups $H_i\text{,}$ where $i = 1, 2, \ldots, n\text{.}$ Then $G$ is isomorphic to $\prod_i H_i\text{.}$</p></article></div>
