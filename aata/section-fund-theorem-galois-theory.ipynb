<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 23.2 The Fundamental Theorem"><span class="type">Section</span><span class="codenumber">23.2</span><span class="title">The Fundamental Theorem</span></h2><a href="section-fund-theorem-galois-theory.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-3609">The goal of this section is to prove the Fundamental Theorem of Galois Theory. This theorem explains the connection between the subgroups of $G(E/F)$ and the intermediate fields between $E$ and $F\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-54"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">23.13</span></h6><p id="p-3610">Let $\{\sigma_i : i \in I  \}$ be a  collection of automorphisms of a field $F\text{.}$  Then </p><div class="displaymath">
\begin{equation*}
F_{\{\sigma_i \}} = \{ a \in F : \sigma_i(a) = a \text{ for all } \sigma_i \}
\end{equation*}
</div><p>is a subfield of $F\text{.}$</p></article><article class="proof" id="proof-197"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3611">Let $\sigma_i(a) = a$ and $\sigma_i(b)=b\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
\sigma_i(a \pm b) = \sigma_i(a) \pm \sigma_i(b) = a \pm b
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
\sigma_i(a b) = \sigma_i(a) \sigma_i(b) = a  b.
\end{equation*}
</div><p>If $a \neq 0\text{,}$ then  $\sigma_i(a^{-1}) = [\sigma_i(a)]^{-1} = a^{-1}\text{.}$ Finally, $\sigma_i(0) = 0$ and $\sigma_i(1)=1$ since $\sigma_i$ is an automorphism.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-38"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">23.14</span></h6><p id="p-3612">Let $F$ be a field and let $G$ be a subgroup of $\aut(F)\text{.}$ Then </p><div class="displaymath">
\begin{equation*}
F_G = \{ \alpha \in F : \sigma( \alpha ) = \alpha \text{ for all } \sigma \in G \}
\end{equation*}
</div><p>is a subfield of $F\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3613">The subfield $F_{ \{\sigma_i \} }$ of $F$ is called the <dfn class="terminology">fixed field</dfn> of $\{ \sigma_i \}\text{.}$ The field fixed by a subgroup $G$ of $\aut(F)$ will be denoted by $F_G\text{.}$</p></div>

<div class="mathbook-content"><article class="example-like" id="example-241"><h6 class="heading"><span class="type">Example</span><span class="codenumber">23.15</span></h6><p id="p-3614">Let $\sigma : {\mathbb Q}(\sqrt{3}, \sqrt{5}\, ) \rightarrow {\mathbb Q}(\sqrt{3}, \sqrt{5}\, )$ be the automorphism that maps $\sqrt{3}$ to $-\sqrt{3}\text{.}$ Then ${\mathbb Q}( \sqrt{5}\, )$ is the subfield of  ${\mathbb Q}(\sqrt{3}, \sqrt{5}\, )$ left fixed by $\sigma\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-fixed-field"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">23.16</span></h6><p id="p-3615">Let $E$ be a splitting field over $F$ of a separable polynomial. Then $E_{G(E/F)} = F\text{.}$</p></article><article class="proof" id="proof-198"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3616">Let $G = G(E/F)\text{.}$ Clearly, $F \subset E_G \subset E\text{.}$ Also, $E$ must be a splitting field of $E_G$ and $G(E/F) = G(E/E_G)\text{.}$ By Theorem <a href="section-field-automorphisms.ipynb#theorem-extension-order" class="xref" alt="Theorem 23.7 " title="Theorem 23.7 ">23.7</a>,</p><div class="displaymath">
\begin{equation*}
|G| = [E: E_G] =[ E:F].
\end{equation*}
</div><p>Therefore, $[E_G : F ] =1\text{.}$ Consequently,  $E_G = F\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3617">A large number of mathematicians first learned Galois theory from Emil Artin's monograph on the subject [1]. The very clever proof of the following lemma is due to Artin.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-41"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">23.17</span></h6><p id="p-3618">Let $G$ be a finite group of automorphisms of $E$ and let $F = E_G\text{.}$ Then $[E:F] \leq |G|\text{.}$</p></article><article class="proof" id="proof-199"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3619">Let $|G| = n\text{.}$ We must show that any set of $n + 1$ elements $\alpha_1, \ldots, \alpha_{n + 1}$ in $E$ is linearly dependent over $F\text{;}$ that is, we need to find elements $a_i \in F\text{,}$ not all zero, such that</p><div class="displaymath">
\begin{equation*}
a_1 \alpha_1 + a_2 \alpha_2 + \cdots + a_{n + 1} \alpha_{n + 1} = 0.
\end{equation*}
</div><p>Suppose that $\sigma_1 = \identity, \sigma_2, \ldots, \sigma_n$ are the automorphisms in $G\text{.}$ The homogeneous system of linear equations</p><div class="displaymath">
\begin{align*}
\sigma_1( \alpha_1 ) x_1 + \sigma_1(\alpha_2) x_2 + \cdots + \sigma_1(\alpha_{n + 1} ) x_{n + 1} & = 0\\
\sigma_2( \alpha_1 ) x_1 + \sigma_2(\alpha_2) x_2 + \cdots + \sigma_2(\alpha_{n + 1} ) x_{n + 1} & = 0\\
& \vdots &\\
\sigma_n( \alpha_1 ) x_1 + \sigma_n(\alpha_2) x_2 + \cdots + \sigma_n(\alpha_{n + 1} ) x_{n + 1} & = 0
\end{align*}
</div><p>has more unknowns than equations. From linear algebra we know that this system has a nontrivial solution, say $x_i = a_i$ for $i = 1, 2, \ldots, n + 1\text{.}$ Since $\sigma_1$ is the identity, the first equation translates to</p><div class="displaymath">
\begin{equation*}
a_1 \alpha_1 +  a_2 \alpha_2  + \cdots + a_{n + 1} \alpha_{n + 1} = 0.
\end{equation*}
</div><p>The problem is that some of the $a_i$'s may be in $E$ but not in $F\text{.}$ We must show that this is impossible.</p><p id="p-3620">Suppose that at least one of the $a_i$'s is in $E$ but not in $F\text{.}$ By rearranging the $\alpha_i$'s we may assume that $a_1$ is nonzero. Since any nonzero multiple of a solution is also a solution, we can also assume that $a_1 = 1\text{.}$ Of all possible solutions fitting this description, we choose the one with the smallest number of nonzero terms. Again, by rearranging $\alpha_2, \ldots, \alpha_{n + 1}$ if necessary, we can assume that $a_2$ is in $E$ but not in $F\text{.}$ Since $F$ is the subfield of $E$ that is fixed elementwise by $G\text{,}$ there exists a $\sigma_i$ in $G$ such that $\sigma_i( a_2 ) \neq a_2\text{.}$ Applying $\sigma_i$ to each equation in the system, we end up with the same homogeneous system, since $G$ is a group. Therefore, $x_1 = \sigma_i(a_1) = 1\text{,}$ $x_2 = \sigma_i(a_2)\text{,}$ $\ldots\text{,}$ $x_{n + 1} = \sigma_i(a_{n+1} )$ is also a solution of the original system. We know that a linear combination of two solutions of a homogeneous system is also a solution; consequently,</p><div class="displaymath">
\begin{align*}
x_1 & = 1 -1 = 0\\
x_2 & = a_2 - \sigma_i(a_2)\\
& \vdots &\\
x_{n + 1} & = a_{n + 1} - \sigma_i(a_{n + 1})
\end{align*}
</div><p>must be another solution of the system. This is a nontrivial solution because $\sigma_i( a_2 ) \neq a_2\text{,}$ and has fewer nonzero entries than our original solution. This is a contradiction, since the number of nonzero solutions to our original solution was assumed to be minimal. We can therefore conclude that $a_1, \ldots, a_{n + 1} \in F\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3621">Let $E$ be an algebraic extension of $F\text{.}$ If every irreducible polynomial in $F[x]$ with a root in $E$ has all of its roots in $E\text{,}$ then $E$ is called a <dfn class="terminology">normal extension</dfn> of $F\text{;}$ that is, every irreducible polynomial in $F[x]$ containing a root in $E$ is the product of linear factors in $E[x]\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-133"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">23.18</span></h6><p id="p-3622">Let $E$ be a field extension of $F\text{.}$ Then the following statements are equivalent. </p><ol class="decimal"><li id="li-784"><p id="p-3623">$E$ is a finite, normal, separable extension of $F\text{.}$</p></li><li id="li-785"><p id="p-3624">$E$ is a splitting field over $F$ of a separable polynomial.</p></li><li id="li-786"><p id="p-3625">$F = E_G$ for some finite group $G$ of automorphisms of $E\text{.}$</p></li></ol></article><article class="proof" id="proof-200"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3626">(1) $\Rightarrow$ (2). Let $E$ be a finite, normal, separable extension of $F\text{.}$ By the Primitive Element Theorem, we can find an  $\alpha$ in $E$ such that $E = F(\alpha)\text{.}$ Let $f(x)$ be the minimal polynomial of $\alpha$ over $F\text{.}$ The field $E$ must contain all of the roots of $f(x)$ since it is a normal extension $F\text{;}$ hence, $E$ is a splitting field for $f(x)\text{.}$</p><p id="p-3627">(2) $\Rightarrow$ (3). Let $E$ be the splitting field over $F$ of a separable polynomial. By Proposition <a href="section-fund-theorem-galois-theory.ipynb#proposition-fixed-field" class="xref" alt="Proposition 23.16 " title="Proposition 23.16 ">23.16</a>, $E_{G(E/F)} = F\text{.}$ Since $| G(E/F)| = [E:F]\text{,}$ this is a finite group.</p><p id="p-3628">(3) $\Rightarrow$ (1). Let $F = E_G$ for some finite group of automorphisms $G$ of $E\text{.}$ Since $[E:F] \leq |G|\text{,}$ $E$ is a finite extension of $F\text{.}$ To show that $E$ is a finite, normal extension of  $F\text{,}$ let $f(x) \in F[x]$ be an irreducible monic polynomial that has a root $\alpha$ in $E\text{.}$ We must show that $f(x)$ is the product of distinct linear factors in $E[x]\text{.}$ By Proposition <a href="section-field-automorphisms.ipynb#proposition-roots-permute" class="xref" alt="Proposition 23.5 " title="Proposition 23.5 ">23.5</a>, automorphisms in $G$ permute the roots of $f(x)$ lying in $E\text{.}$ Hence, if we let $G$ act on $\alpha\text{,}$ we can obtain distinct roots $\alpha_1 = \alpha, \alpha_2, \ldots, \alpha_n$ in $E\text{.}$ Let $g(x) = \prod_{i = 1}^{n} (x -\alpha_i)\text{.}$ Then $g(x)$ is separable over $F$ and $g( \alpha ) = 0\text{.}$ Any automorphism $\sigma$ in $G$ permutes the factors of $g(x)$ since it permutes these roots; hence, when $\sigma$ acts on $g(x)\text{,}$ it must fix the coefficients of $g(x)\text{.}$ Therefore, the coefficients of $g(x)$ must be in $F\text{.}$ Since $\deg g(x) \leq \deg f(x)$ and $f(x)$ is the minimal polynomial of $\alpha\text{,}$ $f(x) = g(x)\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-39"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">23.19</span></h6><p id="p-3629">Let $K$ be a field extension of $F$ such that $F = K_G$ for some finite group of automorphisms $G$ of $K\text{.}$ Then $G = G(K/F)\text{.}$</p></article><article class="proof" id="proof-201"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3630">Since $F = K_G\text{,}$ $G$ is a subgroup of $G(K/F)\text{.}$ Hence,</p><div class="displaymath">
\begin{equation*}
[K : F ]  \leq |G| \leq |G(K/F)| = [K:F].
\end{equation*}
</div><p>It follows that $G = G(K/F)\text{,}$ since they must have the same order.</p></article></div>

<div class="mathbook-content"><p id="p-3631">Before we determine the exact correspondence between field extensions and automorphisms of fields, let us return to a familiar example.</p></div>

<div class="mathbook-content"><p id="p-3633">We are now ready to state and prove the Fundamental Theorem of Galois Theory.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-fundamental-galois"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">23.22</span><span class="title">Fundamental Theorem of Galois Theory</span></h6><p id="p-3634">Let $F$ be a finite field or a field of characteristic zero. If $E$ is a finite normal extension of $F$ with Galois group $G(E/F)\text{,}$ then the following statements are true. </p><ol class="decimal"><li id="li-787"><p id="p-3635">The map $K \mapsto G(E/K)$ is a bijection of subfields $K$ of $E$ containing $F$ with the subgroups of $G(E/F)\text{.}$</p></li><li id="li-788"><p id="p-3636">If $F \subset K \subset E\text{,}$ then</p><div class="displaymath">
\begin{equation*}
[E:K] = |G(E/K)| \text{ and } [K:F] = [G(E/F):G(E/K)].
\end{equation*}
</div></li><li id="li-789"><p id="p-3637">$F \subset K \subset L \subset E$ if and only if $\{ \identity \} \subset G(E/L) \subset G(E/K) \subset G(E/F)\text{.}$</p></li><li id="li-790"><p id="p-3638">$K$ is a normal extension of $F$ if and only if $G(E/K)$ is a normal subgroup of $G( E/F)\text{.}$ In this case</p><div class="displaymath">
\begin{equation*}
G(K/F) \cong G(E/F) / G( E/K ).
\end{equation*}
</div></li></ol></article><article class="proof" id="proof-202"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3639">(1) Suppose that $G(E/K) = G(E/L) = G\text{.}$ Both $K$ and $L$ are fixed fields  of $G\text{;}$ hence, $K=L$ and the map defined by $K \mapsto G(E/K)$ is one-to-one. To show that the map is onto, let $G$ be a subgroup of $G(E/F)$ and $K$ be the field fixed by $G\text{.}$ Then $F \subset K \subset E\text{;}$ consequently, $E$ is a normal extension of $K\text{.}$ Thus, $G(E/K) = G$ and the map $K \mapsto G(E/K)$ is a bijection.</p><p id="p-3640">(2) By Theorem <a href="section-field-automorphisms.ipynb#theorem-extension-order" class="xref" alt="Theorem 23.7 " title="Theorem 23.7 ">23.7</a>, $|G(E/K)| = [E:K]\text{;}$ therefore,</p><div class="displaymath">
\begin{equation*}
|G(E/F)| = [G(E/F):G(E/K)] \cdot |G(E/K)| = [E:F] = [E:K][K:F].
\end{equation*}
</div><p>Thus, $[K:F] = [G(E/F):G(E/K)]\text{.}$</p><p id="p-3641">(3) Statement (3) is illustrated in Figure <a href="section-fund-theorem-galois-theory.ipynb#figure-galois-2" class="xref" alt="Figure 23.23 " title="Figure 23.23 ">23.23</a>. We leave the proof of this property as an exercise.</p><p id="p-3642">(4) This part takes a little more work. Let $K$ be a normal extension of $F\text{.}$ If $\sigma$ is in $G(E/F)$ and $\tau$ is in $G(E/K)\text{,}$ we need to show that $\sigma^{-1} \tau \sigma$ is in $G(E/K)\text{;}$ that is, we need to show that $\sigma^{-1} \tau \sigma( \alpha) = \alpha$ for all $\alpha \in K\text{.}$ Suppose that $f(x)$ is the minimal polynomial of $\alpha$ over $F\text{.}$ Then $\sigma( \alpha )$ is also a root of $f(x)$ lying in $K\text{,}$ since $K$ is a normal extension of $F\text{.}$ Hence, $\tau( \sigma( \alpha )) = \sigma( \alpha )$ or $\sigma^{-1} \tau \sigma( \alpha) = \alpha\text{.}$</p><p id="p-3643">Conversely, let $G(E/K)$ be a normal subgroup of $G(E/F)\text{.}$ We need to show that $F = K_{G(K/F)}\text{.}$ Let $\tau \in G(E/K)\text{.}$ For all $\sigma \in G(E/F)$ there exists a $\overline{\tau} \in G(E/K)$ such that $\tau \sigma = \sigma \overline{\tau}\text{.}$  Consequently, for all $\alpha \in K$</p><div class="displaymath">
\begin{equation*}
\tau( \sigma( \alpha ) ) = \sigma( \overline{\tau}( \alpha ) ) = \sigma( \alpha );
\end{equation*}
</div><p>hence, $\sigma( \alpha )$ must be in the fixed field of $G(E/K)\text{.}$ Let $\overline{\sigma}$ be the restriction of $\sigma$ to $K\text{.}$ Then $\overline{\sigma}$ is an automorphism of $K$ fixing $F\text{,}$ since $\sigma( \alpha ) \in K$ for all $\alpha \in K\text{;}$ hence, $\overline{\sigma} \in G(K/F)\text{.}$ Next, we will show that the fixed field of $G(K/F)$ is $F\text{.}$ Let $\beta$ be an element in $K$ that is fixed by all automorphisms in $G(K/F)\text{.}$  In particular, $\overline{\sigma}(\beta) = \beta$ for all $\sigma \in G(E/F)\text{.}$  Therefore, $\beta$ belongs to the fixed field $F$ of $G(E/F)\text{.}$</p><p id="p-3644">Finally, we must show that when $K$ is a normal extension of $F\text{,}$</p><div class="displaymath">
\begin{equation*}
G(K/F) \cong G(E/F) / G(E/K).
\end{equation*}
</div><p>For $\sigma \in G(E/F)\text{,}$ let $\sigma_K$ be the automorphism of $K$ obtained by restricting $\sigma$ to $K\text{.}$ Since $K$ is a normal extension, the argument in the preceding paragraph shows that $\sigma_K \in G( K/F)\text{.}$ Consequently, we have a map $\phi:G(E/F) \rightarrow G(K/F)$ defined by $\sigma \mapsto \sigma_K\text{.}$ This map is a group homomorphism since</p><div class="displaymath">
\begin{equation*}
\phi( \sigma \tau )  = (\sigma \tau)_K  = \sigma_K \tau_K  = \phi( \sigma) \phi( \tau ).
\end{equation*}
</div><p>The kernel of $\phi$ is $G(E/K)\text{.}$  By (2),</p><div class="displaymath">
\begin{equation*}
|G(E/F)| / |G(E/K)| = [K:F] = |G(K/F)|.
\end{equation*}
</div><p>Hence, the image of $\phi$ is $G(K/F)$ and $\phi$ is onto. Applying the First Isomorphism Theorem, we have</p><div class="displaymath">
\begin{equation*}
G(K/F) \cong G(E/F) / G( E/K ).
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="example-like" id="example-galois-x4-2"><h6 class="heading"><span class="type">Example</span><span class="codenumber">23.24</span></h6><p id="p-3645">In this example we will illustrate the Fundamental Theorem of Galois Theory by determining the lattice of subgroups of the Galois group of $f(x) = x^4 - 2\text{.}$ We will compare this lattice to the lattice of field extensions of ${\mathbb Q}$ that are contained in the splitting field of $x^4-2\text{.}$ The splitting field of $f(x)$ is ${\mathbb Q}( \sqrt[4]{2}, i )\text{.}$ To see this, notice that $f(x)$ factors as $(x^2 + \sqrt{2}\, )(x^2 - \sqrt{2}\, )\text{;}$ hence, the roots of $f(x)$ are $\pm \sqrt[4]{2}$ and $\pm \sqrt[4]{2}\, i\text{.}$ We first adjoin the root $\sqrt[4]{2}$ to ${\mathbb Q}$ and then adjoin the root $i$ of $x^2 + 1$ to ${\mathbb Q}(\sqrt[4]{2}\, )\text{.}$ The splitting field of $f(x)$ is then ${\mathbb Q}(\sqrt[4]{2}\, )(i) = {\mathbb Q}( \sqrt[4]{2}, i )\text{.}$</p><p id="p-3646">Since $[ {\mathbb Q}( \sqrt[4]{2}\, ) : {\mathbb Q}] = 4$ and $i$ is not in ${\mathbb Q}( \sqrt[4]{2}\, )\text{,}$ it must be the case that $[ {\mathbb Q}( \sqrt[4]{2}, i ): {\mathbb Q}(\sqrt[4]{2}\, )] = 2\text{.}$ Hence, $[ {\mathbb Q}( \sqrt[4]{2}, i ):{\mathbb Q}] = 8\text{.}$ The set</p><div class="displaymath">
\begin{equation*}
\{ 1, \sqrt[4]{2}, (\sqrt[4]{2}\, )^2, (\sqrt[4]{2}\, )^3, i, i \sqrt[4]{2}, i (\sqrt[4]{2}\, )^2, i(\sqrt[4]{2}\, )^3 \}
\end{equation*}
</div><p>is a basis of ${\mathbb Q}( \sqrt[4]{2}, i )$ over ${\mathbb Q}\text{.}$ The lattice of field extensions of ${\mathbb Q}$ contained in ${\mathbb Q}( \sqrt[4]{2}, i)$ is illustrated in Figure <a href="section-fund-theorem-galois-theory.ipynb#figure-galois-3" class="xref" alt="Figure 23.25 " title="Figure 23.25 ">23.25</a>(a).</p><p id="p-3647">The Galois group $G$ of $f(x)$ must be of order 8. Let $\sigma$ be the automorphism defined by $\sigma( \sqrt[4]{2}\, ) = i \sqrt[4]{2}$ and $\sigma( i ) = i\text{,}$ and $\tau$ be the automorphism defined by complex conjugation; that is, $\tau(i ) = -i\text{.}$ Then $G$ has an element of order 4 and an element of order 2. It is easy to verify by direct computation that the elements of $G$ are $\{ \identity, \sigma, \sigma^2,  \sigma^3, \tau, \sigma \tau, \sigma^2 \tau, \sigma^3 \tau \}$ and that the relations $\tau^2 = \identity\text{,}$ $\sigma^4 = \identity\text{,}$ and $\tau \sigma \tau = \sigma^{-1}$ are satisfied; hence, $G$ must be isomorphic to $D_4\text{.}$ The lattice of subgroups of $G$ is illustrated in Figure <a href="section-fund-theorem-galois-theory.ipynb#figure-galois-3" class="xref" alt="Figure 23.25 " title="Figure 23.25 ">23.25</a>(b).</p></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Historical Note"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Historical Note</span></h3><a href="section-fund-theorem-galois-theory.ipynb#galois-subsection-historical-note" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-3648">Solutions for the cubic and quartic equations were discovered in the 1500s. Attempts to find solutions for the quintic equations puzzled some of history's best mathematicians.  In 1798, P. Ruffini submitted a paper that claimed no such solution could be found; however, the paper was not well received. In 1826, Niels Henrik Abel (1802–1829) finally offered the first correct proof that quintics are not always solvable by radicals.</p></div>

<div class="mathbook-content"><p id="p-3649">Abel inspired the work of Évariste Galois. Born in 1811, Galois began to display extraordinary mathematical talent at the age of 14. He applied for entrance to the École Polytechnique several times; however, he had great difficulty meeting the formal entrance requirements, and the examiners failed to recognize his mathematical genius. He was finally accepted at the École Normale in 1829.</p></div>

<div class="mathbook-content"><p id="p-3650">Galois worked to develop a theory of solvability for polynomials.  In 1829, at the age of 17, Galois presented two papers on the solution of algebraic equations to the Académie des Sciences de Paris. These papers were sent to Cauchy, who subsequently lost them. A third paper was submitted to Fourier, who died before he could read the paper. Another paper was presented, but was not published until 1846.</p></div>

<div class="mathbook-content"><p id="p-3651">Galois' democratic sympathies led him into the Revolution of 1830. He was expelled from school and sent to prison for his part in the turmoil. After his release in 1832, he was drawn into a duel possibly over a love affair. Certain that he would be killed, he spent the evening before his death outlining his work and his basic ideas for research in a long letter to his friend Chevalier. He was indeed dead the next day, at the age of 20.</p></div>
