<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 5.2 Dihedral Groups"><span class="type">Section</span><span class="codenumber">5.2</span><span class="title">Dihedral Groups</span></h2><a href="section-dihedral-groups.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-823">Another special type of permutation group is the dihedral group. Recall the symmetry group of an equilateral triangle in Chapter <a href="groups.ipynb" class="xref" alt="Chapter 3 Groups" title="Chapter 3 Groups">3</a>. Such groups consist of the  rigid motions of a regular $n$-sided polygon or $n$-gon. For $n = 3, 4, \ldots\text{,}$ we define the <dfn class="terminology">nth dihedral group</dfn> to be the group of rigid motions of a regular $n$-gon.  We will denote this group by $D_n\text{.}$   We can number the vertices of a regular $n$-gon by $1, 2, \ldots, n$ (Figure <a href="section-dihedral-groups.ipynb#figure-regular" class="xref" alt="Figure 5.19 " title="Figure 5.19 ">5.19</a>).  Notice that there are exactly $n$ choices to replace the first vertex.  If we replace the first vertex by $k\text{,}$ then the second vertex must be replaced either by vertex $k+1$ or by vertex $k-1\text{;}$ hence, there are $2n$ possible rigid motions of the $n$-gon.  We summarize these results in the following theorem.</p></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-regular"><img src="images/permute-ngon.svg" width="60%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">5.19</span>A regular $n$-gon</figcaption></figure></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-21"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">5.20</span></h6><p id="p-824">The dihedral group, $D_n\text{,}$ is a subgroup of $S_n$ of order $2n\text{.}$</p></article></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-rotations"><img src="images/permute-motions-ngon.svg" width="100%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">5.21</span>Rotations and reflections of a regular $n$-gon</figcaption></figure></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-types"><img src="images/permute-reflections-ngon.svg" width="100%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">5.22</span>Types of  reflections of a regular $n$-gon</figcaption></figure></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-dn-generator"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">5.23</span></h6><p id="p-825">The group $D_n\text{,}$ $n \geq 3\text{,}$ consists of all products of the two elements $r$ and $s\text{,}$ satisfying the relations</p><div class="displaymath">
\begin{align*}
r^n & = 1\\
s^2 & = 1\\
srs & = r^{-1}.
\end{align*}
</div></article><article class="proof" id="proof-34"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-826">The possible motions of a regular $n$-gon are either reflections or rotations (Figure <a href="section-dihedral-groups.ipynb#figure-rotations" class="xref" alt="Figure 5.21 " title="Figure 5.21 ">5.21</a>). There are exactly $n$ possible rotations:</p><div class="displaymath">
\begin{equation*}
\identity, \frac{360^{\circ} }{n}, 2 \cdot \frac{360^{\circ} }{n}, \ldots, (n-1) \cdot \frac{360^{\circ} }{n}.
\end{equation*}
</div><p>We will denote the rotation $360^{\circ} /n$ by $r\text{.}$ The rotation $r$ generates all  of the other rotations. That is,</p><div class="displaymath">
\begin{equation*}
r^k = k \cdot \frac{360^{\circ} }{n}.
\end{equation*}
</div><p id="p-827">Label the $n$ reflections $s_1, s_2, \ldots, s_n\text{,}$ where $s_k$ is the reflection that leaves vertex $k$ fixed. There are two cases of reflections, depending on whether $n$ is even or odd. If there are an even number of vertices, then 2 vertices are left fixed by a reflection, and $s_1 = s_{n/2 + 1}, s_2 = s_{n/2 + 2}, \ldots, s_{n/2}  = s_n\text{.}$ If there are an odd number of vertices, then only a single vertex is left fixed by a reflection and $s_1, s_2, \ldots, s_n$ are distinct (Figure <a href="section-dihedral-groups.ipynb#figure-types" class="xref" alt="Figure 5.22 " title="Figure 5.22 ">5.22</a>).  In either case, the order of each $s_k$ is two. Let $s = s_1\text{.}$ Then $s^2 = 1$ and $r^n = 1\text{.}$ Since any rigid motion $t$ of the $n$-gon replaces the first vertex by the vertex $k\text{,}$ the second vertex must be replaced by either $k+1$ or by $k-1\text{.}$ If the second vertex is replaced by $k+1\text{,}$ then $t = r^k\text{.}$ If the second vertex is replaced by $k-1\text{,}$ then $t = s r^k\text{.}$ Hence, $r$ and $s$ generate $D_n\text{.}$ That is, $D_n$ consists of all finite products of $r$ and $s\text{,}$</p><div class="displaymath">
\begin{equation*}
D_n = \{1, r, r^2, \ldots, r^{n-1}, s, sr, sr^2, \ldots, sr^{n-1}\}.
\end{equation*}
</div><p>We will leave the proof that $srs = r^{-1}$ as an exercise.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-permute-d4-group"><h6 class="heading"><span class="type">Example</span><span class="codenumber">5.24</span></h6><p id="p-828">The group of rigid motions of a square, $D_4\text{,}$ consists of eight elements. With the vertices numbered 1, 2, 3, 4 (Figure <a href="section-dihedral-groups.ipynb#figure-d4" class="xref" alt="Figure 5.25 " title="Figure 5.25 ">5.25</a>), the rotations are</p><div class="displaymath">
\begin{align*}
r   & = (1234)\\
r^2 & = (13)(24)\\
r^3 & = (1432)\\
r^4 & = (1)
\end{align*}
</div><p>and the reflections are</p><div class="displaymath">
\begin{align*}
s_1 & = (24)\\
s_2 & = (13).
\end{align*}
</div><p>The order of $D_4$ is 8. The remaining two elements are</p><div class="displaymath">
\begin{align*}
r s_1   & = (12)(34)\\
r^3 s_1 & = (14)(23).
\end{align*}
</div></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  The Motion Group of a Cube"><span class="type">Subsection</span><span class="codenumber" /><span class="title">The Motion Group of a Cube</span></h3><a href="section-dihedral-groups.ipynb#permute-subsection-motion-group-cube" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-829">We can investigate the groups of rigid motions of  geometric objects other than a regular $n$-sided polygon to obtain interesting examples of permutation groups. Let us consider the group of rigid motions of a cube. One of the first questions that we can ask about this group is “what is its order?”  A cube has 6 sides. If a particular side is facing upward, then there are four possible rotations of the cube that will preserve the upward-facing side. Hence, the order of the group is $6 \cdot 4 = 24\text{.}$ We have just proved the following proposition.</p></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-motions"><img src="images/permute-motions-cube.svg" width="60%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">5.26</span>The motion group of a cube</figcaption></figure></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-motions-cube"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">5.27</span></h6><p id="p-830">The group of rigid motions of a cube contains $24$ elements.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-23"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">5.28</span></h6><p id="p-831">The group of rigid motions of a cube is $S_4\text{.}$</p></article><article class="proof" id="proof-35"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-832">From Proposition <a href="section-dihedral-groups.ipynb#proposition-motions-cube" class="xref" alt="Proposition 5.27 " title="Proposition 5.27 ">5.27</a>, we already know that the motion group of the cube has 24 elements, the same number of elements as there are in $S_4\text{.}$  There are exactly four diagonals in the cube.  If we label these diagonals 1, 2, 3, and 4, we must show that the motion group of the cube will give us any permutation of the diagonals (Figure <a href="section-dihedral-groups.ipynb#figure-motions" class="xref" alt="Figure 5.26 " title="Figure 5.26 ">5.26</a>). If we can obtain all of these permutations, then $S_4$ and the group of rigid motions of the cube must be the same. To obtain a transposition we can rotate the cube $180^{\circ}$ about the axis joining the midpoints of opposite edges (Figure <a href="section-dihedral-groups.ipynb#figure-transpose" class="xref" alt="Figure 5.29 " title="Figure 5.29 ">5.29</a>). There are six such axes, giving all transpositions in $S_4\text{.}$ Since every element in $S_4$ is the product of a finite number of transpositions, the motion group of a cube must be $S_4\text{.}$</p></article></div>
