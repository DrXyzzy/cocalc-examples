<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 11.2 The Isomorphism Theorems"><span class="type">Section</span><span class="codenumber">11.2</span><span class="title">The Isomorphism Theorems</span></h2><a href="section-group-isomorphism-theorems.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-1729">Although it is not evident at first, factor groups correspond exactly to homomorphic images, and we can use factor groups to study homomorphisms. We already know that with every group homomorphism $\phi: G \rightarrow H$ we can associate a normal subgroup of $G\text{,}$ $\ker \phi\text{.}$  The converse is also true; that is, every normal subgroup of a group $G$ gives rise to homomorphism of groups.</p></div>

<div class="mathbook-content"><p id="p-1730">Let $H$ be a normal subgroup of $G\text{.}$ Define the <dfn class="terminology">natural</dfn> or <dfn class="terminology">canonical homomorphism</dfn></p><div class="displaymath">
\begin{equation*}
\phi : G \rightarrow G/H
\end{equation*}
</div><p>by</p><div class="displaymath">
\begin{equation*}
\phi(g) = gH.
\end{equation*}
</div><p>This is indeed a homomorphism, since</p><div class="displaymath">
\begin{equation*}
\phi( g_1 g_2 ) = g_1 g_2 H =  g_1 H g_2 H = \phi( g_1) \phi( g_2 ).
\end{equation*}
</div><p>The kernel of this homomorphism is $H\text{.}$ The following theorems describe the relationships between group homomorphisms, normal subgroups, and factor groups.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-first-isomorphism"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">11.10</span><span class="title">First Isomorphism Theorem</span></h6><p id="p-1731">If $\psi : G \rightarrow H$ is a group homomorphism with $K =\ker \psi\text{,}$ then $K$ is normal in $G\text{.}$ Let $\phi: G \rightarrow G/K$ be the canonical homomorphism.  Then there exists a unique isomorphism $\eta: G/K \rightarrow \psi(G)$ such that $\psi =  \eta \phi\text{.}$</p></article><article class="proof" id="proof-74"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1732">We already know that $K$ is normal in $G\text{.}$ Define $\eta: G/K \rightarrow \psi(G)$ by $\eta(gK) = \psi(g)\text{.}$  We first show that $\eta$ is a well-defined map.  If $g_1 K =g_2 K\text{,}$ then for some $k \in K\text{,}$ $g_1 k=g_2\text{;}$ consequently,</p><div class="displaymath">
\begin{equation*}
\eta(g_1 K) = \psi(g_1) = \psi(g_1) \psi(k) = \psi(g_1k) = \psi(g_2) = \eta(g_2 K).
\end{equation*}
</div><p>Thus, $\eta$ does not depend on the choice of coset representatives and the map $\eta: G/K \rightarrow \psi(G)$ is uniquely defined since $\psi =  \eta \phi\text{.}$ We must also show that $\eta$ is a homomorphism, but</p><div class="displaymath">
\begin{align*}
\eta( g_1K g_2K ) & = \eta(g_1 g_2K)\\
& = \psi(g_1 g_2)\\
& = \psi(g_1) \psi(g_2)\\
& = \eta( g_1K) \eta( g_2K ).
\end{align*}
</div><p>Clearly, $\eta$ is onto $\psi( G)\text{.}$ To show that $\eta$ is one-to-one, suppose that $\eta(g_1 K) = \eta(g_2 K)\text{.}$ Then $\psi(g_1) = \psi(g_2)\text{.}$ This implies that $\psi( g_1^{-1} g_2 ) = e\text{,}$ or $g_1^{-1} g_2$ is in the kernel of $\psi\text{;}$  hence, $g_1^{-1} g_2K = K\text{;}$ that is, $g_1K =g_2K\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="homo_cyclic"><h6 class="heading"><span class="type">Example</span><span class="codenumber">11.11</span></h6><p id="p-1734">Let $G$ be a cyclic group with generator $g\text{.}$ Define a map $\phi : {\mathbb Z} \rightarrow G$ by $n \mapsto g^n\text{.}$  This map is a surjective homomorphism since</p><div class="displaymath">
\begin{equation*}
\phi( m + n) = g^{m+n} = g^m g^n = \phi(m) \phi(n).
\end{equation*}
</div><p>Clearly $\phi$ is onto. If $|g| = m\text{,}$ then  $g^m = e\text{.}$ Hence, $\ker \phi = m {\mathbb Z}$ and ${\mathbb Z} / \ker \phi =  {\mathbb Z} / m {\mathbb Z} \cong G\text{.}$ On the other hand, if the order of $g$ is infinite, then $\ker \phi = 0$ and $\phi$ is an isomorphism of $G$ and ${\mathbb Z}\text{.}$ Hence, two cyclic groups are isomorphic exactly when they have the same order. Up to isomorphism, the only cyclic groups are ${\mathbb Z}$ and ${\mathbb Z}_n\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-second-isomorphism"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">11.12</span><span class="title">Second Isomorphism Theorem</span></h6><p id="p-1735">Let  $H$ be a subgroup of a group $G$ (not necessarily normal in $G$) and $N$ a normal subgroup of $G\text{.}$  Then $HN$ is a subgroup of $G\text{,}$ $H \cap N$ is a normal subgroup of $H\text{,}$ and</p><div class="displaymath">
\begin{equation*}
H / H \cap N \cong HN /N.
\end{equation*}
</div></article><article class="proof" id="proof-75"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1736">We will first show that $HN = \{ hn : h \in H, n \in N \}$ is a subgroup of $G\text{.}$  Suppose that  $h_1 n_1, h_2 n_2 \in HN\text{.}$ Since $N$ is normal, $(h_2)^{-1} n_1 h_2 \in N\text{.}$ So</p><div class="displaymath">
\begin{equation*}
(h_1 n_1)(h_2 n_2) = h_1 h_2 ( (h_2)^{-1} n_1 h_2 )n_2
\end{equation*}
</div><p>is in $HN\text{.}$ The inverse of $hn \in HN$ is in $HN$ since</p><div class="displaymath">
\begin{equation*}
( hn )^{-1} = n^{-1 } h^{-1} = h^{-1} (h n^{-1} h^{-1} ).
\end{equation*}
</div><p id="p-1737">Next, we prove that $H \cap N$ is normal in $H\text{.}$ Let $h \in H$ and $n \in H \cap N\text{.}$ Then $h^{-1} n h \in H$ since each element is in $H\text{.}$ Also, $h^{-1} n h \in N$ since $N$ is normal in $G\text{;}$ therefore, $h^{-1} n h \in H \cap N\text{.}$</p><p id="p-1738">Now define a map $\phi$ from $H$ to $HN / N$ by $h \mapsto h N\text{.}$ The map $\phi$ is onto, since any coset $h n N = h N$ is the image of $h$ in $H\text{.}$ We also know that $\phi$ is a homomorphism because</p><div class="displaymath">
\begin{equation*}
\phi( h  h')  = h h' N =  h N h' N =  \phi( h ) \phi( h').
\end{equation*}
</div><p>By the First Isomorphism Theorem, the image of $\phi$ is isomorphic to $H / \ker \phi\text{;}$ that is,</p><div class="displaymath">
\begin{equation*}
HN/N = \phi(H) \cong H / \ker \phi.
\end{equation*}
</div><p>Since</p><div class="displaymath">
\begin{equation*}
\ker \phi = \{ h \in H : h \in N \} = H \cap N,
\end{equation*}
</div><p>$HN/N = \phi(H) \cong H / H \cap N\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-correspondence"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">11.13</span><span class="title">Correspondence Theorem</span></h6><p id="p-1739">Let $N$ be a normal subgroup of a group $G\text{.}$ Then $H \mapsto H/N$ is a one-to-one correspondence between the set of subgroups $H$ containing $N$  and the set of subgroups of $G/N\text{.}$ Furthermore, the normal subgroups of $G$ containing $N$ correspond to normal subgroups of $G/N\text{.}$</p></article><article class="proof" id="proof-76"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1740">Let $H$ be a subgroup of $G$ containing $N\text{.}$ Since $N$ is normal in $H\text{,}$ $H/N$ makes sense.  Let $aN$ and $bN$ be elements of $H/N\text{.}$ Then $(aN)( b^{-1} N )= ab^{-1}N \in H/N\text{;}$ hence, $H/N$ is a subgroup of$G/N\text{.}$</p><p id="p-1741">Let $S$ be a subgroup of $G/N\text{.}$ This subgroup is a set of cosets of $N\text{.}$  If  $H= \{ g \in G : gN \in S \}\text{,}$ then for $h_1, h_2 \in H\text{,}$ we have that $(h_1 N)( h_2 N )= h_1 h_2 N \in S$ and $h_1^{-1} N \in S\text{.}$ Therefore, $H$ must be a subgroup of $G\text{.}$ Clearly, $H$ contains $N\text{.}$ Therefore, $S = H / N\text{.}$ Consequently, the map $H \mapsto H/N$ is onto.</p><p id="p-1742">Suppose that $H_1$ and $H_2$ are subgroups of $G$ containing $N$ such that $H_1/N = H_2/N\text{.}$ If $h_1 \in H_1\text{,}$ then $h_1 N \in H_1/N\text{.}$ Hence, $h_1 N = h_2 N \subset H_2$ for some $h_2$ in $H_2\text{.}$ However, since $N$ is contained in $H_2\text{,}$ we know that $h_1 \in H_2$ or $H_1 \subset H_2\text{.}$ Similarly, $H_2 \subset H_1\text{.}$  Since $H_1 = H_2\text{,}$ the map  $H \mapsto H/N$ is one-to-one.</p><p id="p-1743">Suppose that $H$ is normal in $G$ and $N$ is a subgroup of $H\text{.}$  Then it is easy to verify that the map $G/N \rightarrow G/H$ defined by $gN \mapsto gH$ is  a homomorphism.  The kernel of this homomorphism is $H/N\text{,}$ which proves that $H/N$ is normal in $G/N\text{.}$</p><p id="p-1744">Conversely, suppose that $H/N$ is normal in $G/N\text{.}$ The homomorphism given by</p><div class="displaymath">
\begin{equation*}
G \rightarrow G/N \rightarrow \frac{G/N}{H/N}
\end{equation*}
</div><p>has kernel $H\text{.}$ Hence, $H$ must be normal in $G\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1745">Notice that in the course of the proof of Theorem <a href="section-group-isomorphism-theorems.ipynb#theorem-correspondence" class="xref" alt="Theorem 11.13 Correspondence Theorem" title="Theorem 11.13 Correspondence Theorem">11.13</a>, we have also proved the following theorem.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-third-isomorphism"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">11.14</span><span class="title">Third Isomorphism Theorem</span></h6><p id="p-1746">Let $G$ be a group and $N$ and $H$ be normal subgroups of $G$ with $N \subset H\text{.}$  Then</p><div class="displaymath">
\begin{equation*}
G/H \cong \frac{G/N}{H/N}.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="example-like" id="example-homomorph-third-isomorphism"><h6 class="heading"><span class="type">Example</span><span class="codenumber">11.15</span></h6><p id="p-1747">By the Third Isomorphism Theorem,</p><div class="displaymath">
\begin{equation*}
{\mathbb Z} / m {\mathbb Z} \cong ({\mathbb Z}/ mn {\mathbb Z})/ (m {\mathbb Z}/ mn {\mathbb Z}).
\end{equation*}
</div><p>Since $| {\mathbb Z} / mn {\mathbb Z} | = mn$ and  $|{\mathbb Z} / m{\mathbb Z}| = m\text{,}$ we have $| m {\mathbb Z} / mn {\mathbb Z}| = n\text{.}$</p></article></div>
