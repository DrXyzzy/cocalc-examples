<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 13.1 Finite Abelian Groups"><span class="type">Section</span><span class="codenumber">13.1</span><span class="title">Finite Abelian Groups</span></h2><a href="section-finite-abelian-groups.ipynb" class="permalink">Â¶</a></div>

<div class="mathbook-content"><p id="p-1919">In our investigation of cyclic groups we found that every group of prime order was isomorphic to ${\mathbb Z}_p\text{,}$ where $p$ was a prime number.  We also determined that ${\mathbb Z}_{mn} \cong {\mathbb Z}_m \times {\mathbb Z}_n$ when $\gcd(m, n) =1\text{.}$ In fact, much more is true. Every finite abelian group is isomorphic to a direct product of cyclic groups of prime power order; that is, every finite abelian group is isomorphic to a group of the type</p><div class="displaymath">
\begin{equation*}
{\mathbb Z}_{p_1^{\alpha_1}} \times \cdots \times {\mathbb Z}_{p_n^{\alpha_n}},
\end{equation*}
</div><p>where each $p_k$ is prime (not necessarily distinct).</p></div>

<div class="mathbook-content"><p id="p-1920">First, let us examine a slight generalization  of finite abelian groups. Suppose that $G$ is a group and let $\{ g_i\}$ be a set of elements in $G\text{,}$ where $i$ is in some index set $I$ (not necessarily finite).  The smallest subgroup of $G$ containing all of the $g_i$'s is the subgroup of $G$ <dfn class="terminology">generated</dfn> by the $g_i$'s. If this subgroup of $G$ is in fact all of $G\text{,}$ then $G$ is generated by the set $\{g_i : i \in I \}\text{.}$ In this case the $g_i$'s are said to be the <dfn class="terminology">generators</dfn> of $G\text{.}$ If there is a finite set  $\{ g_i : i \in I \}$ that generates $G\text{,}$ then $G$ is <dfn class="terminology">finitely  generated</dfn>.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-finite-groups"><h6 class="heading"><span class="type">Example</span><span class="codenumber">13.1</span></h6><p id="p-1921">Obviously, all finite groups are finitely generated. For example, the group $S_3$ is generated by the permutations $(12)$ and $(123)\text{.}$ The group ${\mathbb Z} \times {\mathbb Z}_n$ is an infinite group but is finitely generated by $\{ (1,0), (0,1) \}\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-infinite-groups"><h6 class="heading"><span class="type">Example</span><span class="codenumber">13.2</span></h6><p id="p-1922">Not all groups are finitely generated.  Consider the rational numbers ${\mathbb Q}$ under the operation of addition. Suppose that ${\mathbb Q}$ is finitely generated with generators $p_1/q_1, \ldots, p_n/q_n\text{,}$ where each $p_i/q_i$ is a fraction expressed in its lowest terms.  Let $p$ be some prime that does not divide any of the denominators $q_1, \ldots, q_n\text{.}$ We claim that $1/p$ cannot be in the subgroup of ${\mathbb Q}$ that is generated by  $p_1/q_1, \ldots, p_n/q_n\text{,}$ since $p$ does not divide the denominator of any element in this subgroup. This fact is easy to see since the sum of any two generators is</p><div class="displaymath">
\begin{equation*}
p_i / q_i + p_j / q_j = (p_i q_j + p_j q_i)/(q_i q_j).
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-27"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">13.3</span></h6><p id="p-1923">Let $H$ be the subgroup of a group $G$ that is generated by $\{ g_i \in G : i \in I \}\text{.}$ Then $h \in H$ exactly when it is a product of the form</p><div class="displaymath">
\begin{equation*}
h = g_{i_1}^{\alpha_1} \cdots g_{i_n}^{\alpha_n},
\end{equation*}
</div><p>where the $g_{i_k}$s are not necessarily distinct.</p></article><article class="proof" id="proof-80"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1924">Let $K$ be the set of all products of the form $g_{i_1}^{\alpha_1} \cdots g_{i_n}^{\alpha_n}\text{,}$ where the $g_{i_k}$s are not necessarily distinct. Certainly $K$ is a subset of $H\text{.}$  We need only show that $K$ is a subgroup of $G\text{.}$ If this is the case, then $K=H\text{,}$ since $H$ is the smallest subgroup containing all the $g_i$s.</p><p id="p-1925">Clearly, the set $K$ is closed under the group operation. Since $g_i^0 = 1\text{,}$ the identity is in $K\text{.}$ It remains to show that the inverse of an element  $g =g_{i_1}^{k_1} \cdots g_{i_n}^{k_n}$ in $K$ must also be in $K\text{.}$ However,</p><div class="displaymath">
\begin{equation*}
g^{-1} = (g_{i_1}^{k_{1}} \cdots g_{i_n}^{k_n})^{-1} = (g_{i_n}^{-k_n} \cdots g_{i_{1}}^{-k_{1}}).
\end{equation*}
</div></article></div>

<div class="mathbook-content"><p id="p-1926">The reason that powers of a fixed $g_i$ may occur several times in the product is that we may have a nonabelian group. However, if the group is abelian, then the $g_i$s need occur only once. For example, a product such as $a^{-3} b^5 a^7$ in an abelian group could always be simplified (in this case, to $a^4 b^5$).</p></div>

<div class="mathbook-content"><p id="p-1927">Now let us restrict our attention to finite abelian groups. We can express any finite abelian group as a finite direct product of cyclic groups. More specifically, letting $p$ be prime, we define a group $G$ to be a <dfn class="terminology">$p$-group</dfn> if every element in $G$ has as its order a power of $p\text{.}$ For example, both ${\mathbb Z}_2 \times {\mathbb Z}_2$ and ${\mathbb Z}_4$ are $2$-groups, whereas ${\mathbb Z}_{27}$ is a $3$-group. We shall prove the Fundamental Theorem of Finite Abelian Groups which tells us that every finite abelian group is isomorphic to a direct product of cyclic $p$-groups.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-finite-abelian-groups"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">13.4</span><span class="title">Fundamental Theorem of Finite Abelian Groups</span></h6><p id="p-1928">Every finite abelian group $G$ is isomorphic to a direct product of cyclic groups of the form</p><div class="displaymath">
\begin{equation*}
{\mathbb Z}_{p_1^{ \alpha_1 }} \times {\mathbb Z}_{p_2^{ \alpha_2 }} \times \cdots \times {\mathbb Z}_{p_n^{ \alpha_n }}
\end{equation*}
</div><p>here the $p_i$'s are primes (not necessarily distinct).</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-abelian-540"><h6 class="heading"><span class="type">Example</span><span class="codenumber">13.5</span></h6><p id="p-1929">Suppose that we wish to classify all abelian groups of order $540=2^2 \cdot 3^3 \cdot 5\text{.}$  The Fundamental Theorem of Finite Abelian Groups tells us that we have the following six possibilities.</p><ul class="disc"><li id="li-446"><p id="p-1930">${\mathbb Z}_2 \times {\mathbb Z}_2 \times {\mathbb Z}_3 \times {\mathbb Z}_3 \times {\mathbb Z}_3 \times {\mathbb Z}_5\text{;}$</p></li><li id="li-447"><p id="p-1931">${\mathbb Z}_2 \times {\mathbb Z}_2 \times {\mathbb Z}_3 \times {\mathbb Z}_9 \times {\mathbb Z}_5\text{;}$</p></li><li id="li-448"><p id="p-1932">${\mathbb Z}_2 \times {\mathbb Z}_2 \times {\mathbb Z}_{27} \times {\mathbb Z}_5\text{;}$</p></li><li id="li-449"><p id="p-1933">${\mathbb Z}_4 \times {\mathbb Z}_3 \times {\mathbb Z}_3 \times {\mathbb Z}_3 \times {\mathbb Z}_5\text{;}$</p></li><li id="li-450"><p id="p-1934">${\mathbb Z}_4 \times {\mathbb Z}_3 \times {\mathbb Z}_9 \times {\mathbb Z}_5\text{;}$</p></li><li id="li-451"><p id="p-1935">${\mathbb Z}_4 \times {\mathbb Z}_{27} \times {\mathbb Z}_5\text{.}$</p></li></ul></article></div>

<div class="mathbook-content"><p id="p-1936">The proof of the Fundamental Theorem of Finite Abelian Groups depends on several lemmas.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-cauchy-finite-abelian"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">13.6</span></h6><p id="p-1937">Let $G$ be a finite abelian group of order $n\text{.}$  If $p$ is a prime that divides $n\text{,}$ then $G$ contains an element of order $p\text{.}$</p></article><article class="proof" id="proof-81"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1938">We will prove this lemma by induction.  If $n = 1\text{,}$ then there is nothing to show.  Now suppose that the order of $G$ is $n$ the lemma is true for all groups of order $k\text{,}$ where $k \lt n\text{.}$  Furthermore, let $p$ be a prime that divides $n\text{.}$</p><p id="p-1939">If $G$ has no proper nontrivial subgroups, then $G = \langle a \rangle\text{,}$ where $a$ is any element other than the identity.  By ExerciseÂ <a href="exercises-cyclic.ipynb#cyclic-exercise-subgroups-exist" class="xref" alt="Exercise 4.4.39 " title="Exercise 4.4.39 ">4.4.39</a>, the order of $G$ must be prime.  Since $p$ divides $n\text{,}$ we know that $p = n\text{,}$ and $G$ contains $p - 1$ elements of order $p\text{.}$</p><p id="p-1940">Now suppose that $G$ contains a nontrivial proper subgroup $H\text{.}$  Then $1 \lt |H| \lt n\text{.}$  If $p \mid |H|\text{,}$ then $H$ contains an element of order $p$ by  induction and the lemma is true.  Suppose that $p$ does not divide the order of $H\text{.}$  Since $G$ is abelian, it must be the case that $H$ is a normal subgroup of $G\text{,}$ and $|G| = |H| \cdot |G/H|\text{.}$  Consequently, $p$ must divide $|G/H|\text{.}$  Since $|G/H| \lt |G| = n\text{,}$ we know that $G/H$ contains an element $aH$ of order $p$ by the induction hypothesis.  Thus,</p><div class="displaymath">
\begin{equation*}
H = (aH)^p = a^pH,
\end{equation*}
</div><p>and $a^p \in H$ but $a \notin H\text{.}$  If $|H| = r\text{,}$ then $p$ and $r$ are relatively prime, and there exist integers $s$ and $t$ such that $sp + tr = 1\text{.}$  Furthermore, the order of $a^p$ must divide $r\text{,}$ and $(a^p)^r = (a^r)^p = 1\text{.}$</p><p id="p-1941">We claim that $a^r$ has order $p\text{.}$  We must show that $a^r \neq 1\text{.}$  Suppose $a^r = 1\text{.}$  Then</p><div class="displaymath">
\begin{align*}
a & = a^{sp + tr}\\
& = a^{sp} a^{tr}\\
& = (a^p)^s (a^r)^t\\
& = (a^p)^s 1\\
& = (a^p)^s.
\end{align*}
</div><p>Since $a^p \in H\text{,}$ it must be the case that $a= (a^p)^s \in H\text{,}$ which is a contradiction.  Therefore, $a^r \neq 1$ is an element of order $p$ in $G\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1942">LemmaÂ <a href="section-finite-abelian-groups.ipynb#lemma-cauchy-finite-abelian" class="xref" alt="Lemma 13.6 " title="Lemma 13.6 ">13.6</a> is a special case of Cauchy's Theorem (TheoremÂ <a href="section-sylow-theorems.ipynb#theorem-cauchy" class="xref" alt="Theorem 15.1 Cauchy" title="Theorem 15.1 Cauchy">15.1</a>), which states that if $G$ be a finite group and $p$ a prime such that $p$ divides the order of $G\text{,}$ then $G$ contains a subgroup of order $p\text{.}$  We will prove Cauchy's Theorem in ChapterÂ <a href="sylow.ipynb" class="xref" alt="Chapter 15 The Sylow Theorems" title="Chapter 15 The Sylow Theorems">15</a>.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-p-group-order"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">13.7</span></h6><p id="p-1943">A finite abelian group is a $p$-group if and only if its order is a power of $p\text{.}$</p></article><article class="proof" id="proof-82"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1944">If $|G| = p^n$ then by Lagrangeâs theorem, then the order of any $g \in G$ must divide $p^n\text{,}$ and therefore must be a power of $p\text{.}$ Conversely, if $|G|$ is not a power of $p\text{,}$ then it has some other prime divisor $q\text{,}$ so by LemmaÂ <a href="section-finite-abelian-groups.ipynb#lemma-cauchy-finite-abelian" class="xref" alt="Lemma 13.6 " title="Lemma 13.6 ">13.6</a>, $G$ has an element of order $q$ and thus is not a $p$-group.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-direct-product-p-groups"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">13.8</span></h6><p id="p-1945">Let $G$ be a finite abelian group of order $n = p_1^{\alpha_1} \cdots p_k^{\alpha_k}\text{,}$ where where $p_1, \ldots, p_k$ are distinct primes and $\alpha_1, \alpha_2, \ldots, \alpha_k$ are positive integers.  Then $G$ is the internal direct product of subgroups $G_1, G_2, \ldots, G_k\text{,}$ where $G_i$ is the subgroup of $G$ consisting of all elements of order $p_i^k$ for some integer $k\text{.}$</p></article><article class="proof" id="proof-83"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1946">Since $G$ is an abelian group, we are guaranteed that $G_i$ is a subgroup of $G$ for $i = 1, \ldots, n\text{.}$  Since the identity has order $p_i^0 = 1\text{,}$ we know that $1 \in G_i\text{.}$  If $g \in G_i$ has order $p_i^r\text{,}$ then $g^{-1}$ must also have order $p_i^r\text{.}$  Finally, if $h \in G_i$ has order $p_i^s\text{,}$ then</p><div class="displaymath">
\begin{equation*}
(gh)^{p_i^t} = g^{p_i^t} h^{p_i^t} = 1 \cdot 1 = 1,
\end{equation*}
</div><p>where $t$ is the maximum of $r$ and $s\text{.}$</p><p id="p-1947">We must show that</p><div class="displaymath">
\begin{equation*}
G = G_1 G_2 \cdots  G_k
\end{equation*}
</div><p>and $G_i \cap G_j = \{1 \}$ for $i \neq j\text{.}$  Suppose that $g_1 \in G_1$ is in the subgroup generated by $G_2, G_3, \ldots, G_k\text{.}$  Then $g_1 = g_2 g_3 \cdots g_k$ for $g_i \in G_i\text{.}$  Since $g_i$ has order $p^{\alpha_i}\text{,}$ we know that $g_i^{p^{\alpha_i}} = 1$ for $i = 2, 3, \ldots, k\text{,}$ and $g_1^{p_2^{\alpha_2} \cdots p_k^{\alpha_k}} = 1\text{.}$  Since the order of $g_1$ is a power of $p_1$ and $\gcd(p_1, p_2^{\alpha_2} \cdots p_k^{\alpha_k}) = 1\text{,}$ it must be the case that $g_1 = 1$ and the intersection of $G_1$ with any of the subgroups $G_2, G_3, \ldots, G_k$ is the identity.  A similar argument shows that $G_i \cap G_j = \{1 \}$ for $i \neq j\text{.}$</p><p id="p-1948">Next, we must show that it possible to write every $g \in G$ as a product $g_1 \cdots g_k\text{,}$ where $g_i \in G_i\text{.}$ Since the order of $g$ divides the order of $G\text{,}$ we know that</p><div class="displaymath">
\begin{equation*}
|g| = p_1^{\beta_1}  p_2^{\beta_2} \cdots p_k^{\beta_k}
\end{equation*}
</div><p>for some integers $\beta_1, \ldots, \beta_k\text{.}$ Letting $a_i = |g| / p_i^{\beta_i}\text{,}$ the $a_i$'s are relatively prime; hence, there exist integers $b_1, \ldots, b_k$ such that $a_1 b_1 + \cdots + a_k b_k = 1\text{.}$ Consequently,</p><div class="displaymath">
\begin{equation*}
g = g^{a_1 b_1 + \cdots + a_k b_k} = g^{a_1 b_1} \cdots  g^{a_k b_k}.
\end{equation*}
</div><p>Since</p><div class="displaymath">
\begin{equation*}
g^{(a_i b_i ) p_i^{\beta_i}} = g^{b_i |g|} = e,
\end{equation*}
</div><p>it follows that $g^{a_i b_i}$ must be in $G_{i}\text{.}$ Let $g_i = g^{a_i b_i}\text{.}$ Then $g = g_1 \cdots g_k \in G_1 G_2 \cdots G_k\text{.}$ Therefore, $G = G_1 G_2 \cdots G_k$ is an internal direct product of subgroups.</p></article></div>

<div class="mathbook-content"><p id="p-1949">If remains for us to determine the possible structure of each $p_i$-group $G_i$ in LemmaÂ <a href="section-finite-abelian-groups.ipynb#lemma-direct-product-p-groups" class="xref" alt="Lemma 13.8 " title="Lemma 13.8 ">13.8</a>.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-struct-finite-abelian"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">13.9</span></h6><p id="p-1950">Let $G$ be a finite abelian $p$-group and suppose that $g \in G$ has maximal order. Then $G$ is isomorphic to $\langle g \rangle \times H$ for some subgroup $H$ of $G\text{.}$</p></article><article class="proof" id="proof-84"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-1951">By LemmaÂ <a href="section-finite-abelian-groups.ipynb#lemma-p-group-order" class="xref" alt="Lemma 13.7 " title="Lemma 13.7 ">13.7</a>, we may assume that the order of $G$ is $p^n\text{.}$  We shall induct on $n\text{.}$ If $n= 1\text{,}$ then $G$ is cyclic of order $p$ and must be generated by $g\text{.}$ Suppose now that the statement of the lemma holds for all integers $k$ with $1 \leq k \lt n$ and let $g$ be of maximal order in $G\text{,}$ say $|g| = p^{m}\text{.}$  Then $a^{p^m} = e$ for all $a \in G\text{.}$ Now choose $h$ in $G$ such that $h \notin \langle g \rangle\text{,}$ where $h$ has the smallest possible order.  Certainly such an $h$ exists; otherwise, $G = \langle g \rangle$ and we are done.  Let $H = \langle h \rangle\text{.}$</p><p id="p-1952">We claim that $\langle g \rangle \cap H = \{ e \}\text{.}$ It suffices to show that $|H|=p\text{.}$  Since $|h^p| = |h| / p\text{,}$ the order of $h^p$ is smaller than the order of $h$ and must be in $\langle g \rangle$ by the minimality of $h\text{;}$ that is, $h^p = g^r$ for some number $r\text{.}$ Hence,</p><div class="displaymath">
\begin{equation*}
(g^r)^{p^{m - 1}} = (h^p)^{p^{m - 1}} = h^{p^{m}} = e,
\end{equation*}
</div><p>and the order of $g^r$ must be less than or equal to $p^{m-1}\text{.}$ Therefore, $g^r$ cannot generate $\langle g \rangle\text{.}$  Notice that $p$ must occur as a factor of $r\text{,}$ say $r = ps\text{,}$ and $h^p = g^r = g^{ps}\text{.}$ Define $a$ to be $g^{-s}h\text{.}$ Then $a$ cannot be in $\langle g \rangle\text{;}$ otherwise, $h$ would also have to be in $\langle g \rangle\text{.}$ Also,</p><div class="displaymath">
\begin{equation*}
a^p = g^{-sp} h^p = g^{-r} h^p = h^{-p} h^p = e.
\end{equation*}
</div><p>We have now formed an element $a$ with order $p$ such that $a \notin \langle g \rangle\text{.}$ Since $h$ was chosen to have the smallest order of all of the elements that are not in $\langle g\rangle\text{,}$ $|H|  = p\text{.}$</p><p id="p-1953">Now we will show that the order of $gH$ in the factor group $G/H$ must be the same as the order of $g$ in $G\text{.}$  If $|gH| \lt |g| =  p^m\text{,}$ then</p><div class="displaymath">
\begin{equation*}
H = (gH)^{p^{m-1}} =  g^{p^{m-1}} H;
\end{equation*}
</div><p>hence, $g^{p^{m-1}}$ must be in $\langle g \rangle \cap H = \{ e \}\text{,}$ which contradicts the fact that the order of $g$ is $p^m\text{.}$  Therefore, $gH$ must have maximal order in $G/H\text{.}$  By the Correspondence Theorem and our induction hypothesis,</p><div class="displaymath">
\begin{equation*}
G/H \cong \langle gH \rangle \times K/H
\end{equation*}
</div><p>for some subgroup $K$ of $G$ containing $H\text{.}$  We claim that $\langle g \rangle \cap K = \{ e \}\text{.}$ If $b \in \langle g \rangle \cap K\text{,}$ then $bH \in \langle gH \rangle \cap K/H =  \{ H \}$ and $b \in \langle g \rangle \cap H = \{ e \}\text{.}$ It follows that $G = \langle g \rangle K$ implies that $G \cong \langle g \rangle \times K\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-1954">The proof of the Fundamental Theorem of Finite Abelian Groups follows very quickly from LemmaÂ <a href="section-finite-abelian-groups.ipynb#lemma-struct-finite-abelian" class="xref" alt="Lemma 13.9 " title="Lemma 13.9 ">13.9</a>.  Suppose that $G$ is a finite abelian group and let $g$ be an element of maximal order in $G\text{.}$ If $\langle g \rangle = G\text{,}$ then we are done; otherwise, $G \cong {\mathbb Z}_{|g|} \times H$ for some subgroup $H$ contained in $G$ by the lemma. Since $|H| \lt |G|\text{,}$ we can apply mathematical induction.</p></div>

<div class="mathbook-content"><p id="p-1955">We now state the more general theorem for all finitely generated abelian groups.  The proof of this theorem can be found in any of the references at the end of this chapter.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-65"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">13.10</span><span class="title">The Fundamental Theorem of Finitely Generated Abelian Groups</span></h6><p id="p-1956">Every finitely generated abelian group $G$ is isomorphic to a direct product of cyclic groups of the form</p><div class="displaymath">
\begin{equation*}
{\mathbb Z}_{p_1^{ \alpha_1 }} \times {\mathbb Z}_{p_2^{ \alpha_2 }} \times \cdots \times {\mathbb Z}_{p_n^{ \alpha_n }} \times {\mathbb Z} \times \cdots \times {\mathbb Z},
\end{equation*}
</div><p>where the $p_i$'s are primes (not necessarily distinct).</p></article></div>
