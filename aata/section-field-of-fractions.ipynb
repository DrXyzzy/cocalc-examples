<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 18.1 Fields of Fractions"><span class="type">Section</span><span class="codenumber">18.1</span><span class="title">Fields of Fractions</span></h2><a href="section-field-of-fractions.ipynb" class="permalink">Â¶</a></div>

<div class="mathbook-content"><p id="p-2746">Every field is also an integral domain; however, there are many integral domains that are not fields.  For example, the integers ${\mathbb Z}$ form an integral domain but not a field.  A question that naturally arises is how we might associate an integral domain with a field.  There is a natural way to construct the rationals ${\mathbb Q}$ from the integers: the rationals can be represented as formal quotients of two integers.  The rational numbers are certainly a field.  In fact, it can be shown that the rationals are the smallest field that contains the integers.  Given an integral domain $D\text{,}$ our question now becomes how to construct a smallest field $F$ containing $D\text{.}$  We will do this in the same way as we constructed the rationals from the integers.</p></div>

<div class="mathbook-content"><p id="p-2747">An element $p/q \in {\mathbb Q}$ is the quotient of two integers $p$ and $q\text{;}$ however, different pairs of integers can represent the same rational number.  For instance, $1/2 = 2/4 = 3/6\text{.}$ We know that</p><div class="displaymath">
\begin{equation*}
\frac{a}{b} = \frac{c}{d}
\end{equation*}
</div><p>if and only if $ad = bc\text{.}$ A more formal way of considering this problem is to examine fractions in terms of equivalence relations.  We can think of elements in ${\mathbb Q}$ as ordered pairs in ${\mathbb Z} \times {\mathbb Z}\text{.}$  A quotient $p/q$ can be written as $(p, q)\text{.}$  For instance, $(3, 7)$ would represent the fraction $3/7\text{.}$  However, there are problems if we consider all possible pairs in ${\mathbb Z} \times {\mathbb Z}\text{.}$  There is no fraction $5/0$ corresponding to the pair $(5,0)\text{.}$  Also, the pairs $(3,6)$ and $(2,4)$ both represent the fraction $1/2\text{.}$  The first problem is easily solved if we require the second coordinate to be nonzero.  The second problem is solved by considering two pairs $(a, b)$ and $(c, d)$ to be equivalent if $ad = bc\text{.}$</p></div>

<div class="mathbook-content"><p id="p-2748">If we use the approach of ordered pairs instead of fractions, then we can study integral domains in general.  Let $D$ be any integral domain and let</p><div class="displaymath">
\begin{equation*}
S = \{ (a, b) : a, b \in D \text{ and } b \neq 0 \}.
\end{equation*}
</div><p>Define a relation on $S$ by $(a, b) \sim (c, d)$ if $ad=bc\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-22"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">18.1</span></h6><p id="p-2749">The relation $\sim$ between elements of $S$ is an equivalence relation.</p></article><article class="proof" id="proof-129"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2750">Since $D$ is commutative, $ab = ba\text{;}$ hence, $\sim$ is reflexive on $D\text{.}$ Now suppose that $(a,b) \sim (c,d)\text{.}$ Then $ad=bc$ or $cb = da\text{.}$ Therefore, $(c,d) \sim (a, b)$ and the relation is symmetric. Finally, to show that the relation is transitive, let $(a, b) \sim (c, d)$ and $(c, d) \sim (e,f)\text{.}$ In this case $ad = bc$ and $cf = de\text{.}$ Multiplying both sides of $ad=bc$ by $f$ yields</p><div class="displaymath">
\begin{equation*}
a f d = a d f = b c f = b d e = bed.
\end{equation*}
</div><p>Since $D$ is an integral domain, we can deduce that $af = be$ or $(a,b ) \sim (e, f)\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-2751">We will denote the set of equivalence classes on $S$ by $F_D\text{.}$ We now need to define the operations of addition and multiplication on $F_D\text{.}$  Recall how fractions are added and multiplied in ${\mathbb Q}\text{:}$</p><div class="displaymath">
\begin{align*}
\frac{a}{b} + \frac{c}{d} & = \frac{ad + b c}{b d};\\
\frac{a}{b} \cdot \frac{c}{d} & = \frac{ac}{b d}.
\end{align*}
</div><p>It seems reasonable to define the operations of addition and multiplication on $F_D$ in a similar manner.  If we denote the equivalence class of $(a, b) \in S$ by $[a, b]\text{,}$ then we are led to define the operations of addition and multiplication on $F_D$ by</p><div class="displaymath">
\begin{equation*}
[a, b] + [c, d] = [ad + b c,b d]
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
[a, b] \cdot [c, d] = [ac, b d],
\end{equation*}
</div><p>respectively.  The next lemma demonstrates that these operations are independent of the choice of representatives from each equivalence class.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-23"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">18.2</span></h6><p id="p-2752">The operations of addition and multiplication on $F_D$ are well-defined.</p></article><article class="proof" id="proof-130"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2753">We will prove that the operation of addition is well-defined.  The proof that multiplication is well-defined is left as an exercise. Let $[a_1, b_1] = [a_2, b_2]$ and $[c_1, d_1] =[ c_2, d_2]\text{.}$  We must show that</p><div class="displaymath">
\begin{equation*}
[a_1 d_1 + b_1 c_1,b_1 d_1] = [a_2 d_2 + b_2 c_2,b_2 d_2]
\end{equation*}
</div><p>or, equivalently, that</p><div class="displaymath">
\begin{equation*}
(a_1 d_1 + b_1 c_1)( b_2 d_2) = (b_1 d_1) (a_2 d_2 + b_2 c_2).
\end{equation*}
</div><p>Since  $[a_1, b_1] = [a_2, b_2]$ and $[c_1, d_1] =[ c_2, d_2]\text{,}$ we know that $a_1 b_2 = b_1 a_2$ and $c_1 d_2 = d_1 c_2\text{.}$  Therefore,</p><div class="displaymath">
\begin{align*}
(a_1 d_1 + b_1 c_1)( b_2 d_2)  & =  a_1 d_1 b_2 d_2 + b_1 c_1 b_2 d_2\\
& = a_1 b_2 d_1 d_2 + b_1 b_2 c_1 d_2\\
& = b_1 a_2 d_1 d_2 + b_1 b_2 d_1 c_2\\
& =  (b_1 d_1) (a_2 d_2 + b_2 c_2).
\end{align*}
</div></article></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-field-of-fractions"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">18.3</span></h6><p id="p-2754">The set of equivalence classes of $S\text{,}$ $F_D\text{,}$ under the equivalence relation $\sim\text{,}$ together with the operations of addition and  multiplication defined by</p><div class="displaymath">
\begin{align*}
[a, b] + [c, d] & = [ad + b c, b d]\\
[ a, b] \cdot [c, d] & = [ac, b d],
\end{align*}
</div><p>is a field.</p></article><article class="proof" id="proof-131"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2755">The additive and multiplicative identities are $[0,1]$ and $[1,1]\text{,}$ respectively. To show that $[0,1]$ is the additive identity, observe that</p><div class="displaymath">
\begin{equation*}
[a, b] + [0, 1] =  [ a 1 + b 0, b 1] = [a,b].
\end{equation*}
</div><p>It is easy to show that $[1, 1]$ is the multiplicative identity. Let $[a, b] \in F_D$ such that $a \neq 0\text{.}$ Then $[b, a]$ is also in $F_D$ and $[a,b] \cdot [b, a] = [1,1]\text{;}$ hence, $[b, a]$ is the multiplicative inverse for $[a, b]\text{.}$  Similarly, $[-a,b]$ is the additive inverse of $[a, b]\text{.}$  We leave as exercises the verification of the associative and  commutative properties of multiplication in $F_D\text{.}$ We also leave it to the reader to show that $F_D$ is an abelian group under addition.</p><p id="p-2756">It remains to show that the distributive property holds in $F_D\text{;}$ however,</p><div class="displaymath">
\begin{align*}
[a, b] [e, f] + [c, d][ e, f ] & = [a e, b f ] + [c e, d f]\\
& = [a e d f + b f c e, b d f^2 ]\\
& = [a e d + b c e, b d f ]\\
& = [a d e + b c e, b d f ]\\
& = ( [a, b]  + [c, d] ) [ e, f ] 
\end{align*}
</div><p>and the lemma is proved.</p></article></div>

<div class="mathbook-content"><p id="p-2757">The field $F_D$ in LemmaÂ <a href="section-field-of-fractions.ipynb#lemma-field-of-fractions" class="xref" alt="Lemma 18.3 " title="Lemma 18.3 ">18.3</a> is called the <dfn class="terminology">field of fractions</dfn> or <dfn class="terminology">field of quotients</dfn> of the integral domain $D\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-field-of-quotients"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">18.4</span></h6><p id="p-2758">Let $D$ be an integral domain.  Then $D$ can be embedded in a field of fractions $F_D\text{,}$ where any element in $F_D$ can be expressed as the quotient of two elements in $D\text{.}$  Furthermore, the field of fractions $F_D$ is unique in the sense that if $E$ is any field containing $D\text{,}$ then there exists a map $\psi : F_D \rightarrow E$ giving an isomorphism with a subfield of $E$ such that $\psi(a) = a$ for all elements $a \in D\text{,}$ where we identify $a$ with its image in $F_D\text{.}$</p></article><article class="proof" id="proof-132"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2759">We will first demonstrate that $D$ can be embedded in the field  $F_D\text{.}$  Define a map $\phi : D \rightarrow F_D$ by $\phi(a)  = [a, 1]\text{.}$  Then for $a$ and $b$ in $D\text{,}$</p><div class="displaymath">
\begin{equation*}
\phi( a + b ) = [a+b, 1] = [a, 1] + [b, 1] = \phi(a ) + \phi(b)
\end{equation*}
</div><p>and</p><div class="displaymath">
\begin{equation*}
\phi( a b ) = [a b, 1] = [a, 1]  [b, 1] = \phi(a ) \phi(b);
\end{equation*}
</div><p>hence, $\phi$ is a homomorphism.  To show that $\phi$ is one-to-one, suppose that $\phi(a) = \phi( b)\text{.}$  Then $[a, 1] = [b, 1]\text{,}$ or $a = a1 = 1b = b\text{.}$ Finally, any element of $F_D$ can be expressed as the quotient of two elements in $D\text{,}$ since</p><div class="displaymath">
\begin{equation*}
\phi(a) [\phi(b)]^{-1} = [a, 1] [b, 1]^{-1} = [a, 1] \cdot [1, b] = [a, b].
\end{equation*}
</div><p id="p-2760">Now let $E$ be a field containing $D$ and define a map $\psi :F_D \rightarrow E$ by $\psi([a, b]) = a b^{-1}\text{.}$  To show that $\psi$ is well-defined, let $[a_1, b_1] = [a_2, b_2]\text{.}$ Then $a_1 b_2 = b_1 a_2\text{.}$ Therefore, $a_1 b_1^{-1} = a_2 b_2^{-1}$ and $\psi( [a_1, b_1]) = \psi( [a_2, b_2])\text{.}$</p><p id="p-2761">If $[a, b ]$ and $[c, d]$ are in $F_D\text{,}$ then</p><div class="displaymath">
\begin{align*}
\psi( [a, b] + [c, d] )  & = \psi( [ad + b c, b d ] )\\
& =  (ad +b c)(b d)^{-1}\\
& = a b^{-1} + c d^{-1}\\
& = \psi( [a, b] ) + \psi( [c, d] )
\end{align*}
</div><p>and</p><div class="displaymath">
\begin{align*}
\psi( [a, b] \cdot [c, d] ) & = \psi( [ac, b d ] )\\
& =  (ac)(b d)^{-1}\\
& = a b^{-1}  c d^{-1}\\
& = \psi( [a, b] )  \psi( [c, d] ).
\end{align*}
</div><p>Therefore, $\psi$ is a homomorphism.</p><p id="p-2762">To complete the proof of the theorem, we need to show that $\psi$ is one-to-one. Suppose that $\psi( [a, b] ) = ab^{-1} = 0\text{.}$ Then $a = 0b = 0$ and $[a, b] = [0, b]\text{.}$  Therefore, the kernel of $\psi$ is the zero element $[ 0, b]$ in $F_D\text{,}$ and $\psi$ is injective.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-domains-rational-polys"><h6 class="heading"><span class="type">Example</span><span class="codenumber">18.5</span></h6><p id="p-2763">Since ${\mathbb Q}$ is a field, ${\mathbb Q}[x]$ is an integral domain. The field of fractions of ${\mathbb Q}[x]$ is the set of all rational expressions $p(x)/q(x)\text{,}$ where $p(x)$ and $q(x)$ are polynomials over the rationals and $q(x)$ is not the zero polynomial. We will denote this field by ${\mathbb Q}(x)\text{.}$ </p></article></div>
