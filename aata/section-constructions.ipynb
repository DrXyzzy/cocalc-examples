<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 21.3 Geometric Constructions"><span class="type">Section</span><span class="codenumber">21.3</span><span class="title">Geometric Constructions</span></h2><a href="section-constructions.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-3298">In ancient Greece, three classic problems were posed. These problems are geometric in nature and involve straightedge-and-compass constructions from what is now high school geometry; that is, we are allowed to use only a straightedge and compass to solve them. The problems can be stated as follows. </p><ol class="decimal"><li id="li-729"><p id="p-3299">Given an arbitrary angle, can one trisect the angle into three equal subangles using only a straightedge and compass?</p></li><li id="li-730"><p id="p-3300">Given an arbitrary circle, can one construct a square with the same area using only a straightedge and compass?</p></li><li id="li-731"><p id="p-3301">Given a cube, can one construct the edge of another cube having twice the volume of the original? Again, we are only allowed to use a straightedge and compass to do the construction.</p></li></ol><p> After puzzling mathematicians for over two thousand years, each of these constructions was finally shown to be impossible.  We will use the theory of fields to provide a proof that the solutions do not exist.  It is quite remarkable that the long-sought solution to each of these three geometric problems came from abstract algebra.</p></div>

<div class="mathbook-content"><p id="p-3302">First, let us determine more specifically what we mean by a straightedge and compass, and also examine the nature of these problems in a bit more depth.  To begin with, <em class="emphasis">a straightedge is not a ruler</em>. We cannot measure arbitrary lengths with a straightedge. It is merely a tool for drawing a line through two points. The statement that the trisection of an arbitrary angle is impossible means that there is at least one angle that is impossible to trisect with a straightedge-and-compass construction. Certainly it is possible to trisect an angle in special cases.  We can construct a $30^\circ$ angle; hence, it is possible to trisect a $90^\circ$ angle.  However, we will show that it is impossible to construct a  $20^\circ$ angle.  Therefore, we cannot trisect a $60^\circ$ angle.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Constructible Numbers"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Constructible Numbers</span></h3><a href="section-constructions.ipynb#fields-subsection-constructible-numbers" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-3303">A real number $\alpha$ is <dfn class="terminology">constructible</dfn> if we can construct a line segment of length $| \alpha |$ in a finite number of steps from a segment of unit length by using a straightedge and compass.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-constructible-numbers"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.35</span></h6><p id="p-3304">The set of all constructible real numbers forms a subfield $F$ of the field of real numbers.</p></article><article class="proof" id="proof-175"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3305">Let $\alpha$ and $\beta$ be constructible numbers.  We must show that $\alpha + \beta\text{,}$ $\alpha - \beta\text{,}$ $\alpha \beta\text{,}$ and $\alpha / \beta$ ($\beta \neq 0$) are also constructible numbers. We can assume that both $\alpha$ and $\beta$ are positive with $\alpha \gt \beta\text{.}$ It is quite obvious how to construct $\alpha + \beta$ and $\alpha - \beta\text{.}$ To find a line segment with length $\alpha \beta\text{,}$ we assume that $\beta \gt 1$ and construct the triangle in Figure <a href="section-constructions.ipynb#figure-multiplication-construction" class="xref" alt="Figure 21.36 " title="Figure 21.36 ">21.36</a> such that triangles $\triangle ABC$ and $\triangle ADE$ are similar. Since $\alpha / 1 = x / \beta\text{,}$ the line segment $x$ has length $\alpha \beta\text{.}$  A similar construction can be made if $\beta \lt 1\text{.}$ We will leave it as an exercise to show that the same triangle can be used to construct $\alpha / \beta$ for $\beta \neq 0\text{.}$</p></article></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-multiplication-construction"><img src="images/fields-multiply.svg" width="80%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">21.36</span>Construction of products</figcaption></figure></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-35"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">21.37</span></h6><p id="p-3306">If $\alpha$ is a constructible number, then $\sqrt{\alpha}$ is a constructible number.</p></article><article class="proof" id="proof-176"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3307">In Figure <a href="section-constructions.ipynb#figure-root-construction" class="xref" alt="Figure 21.38 " title="Figure 21.38 ">21.38</a> the triangles $\triangle ABD\text{,}$ $\triangle BCD\text{,}$ and $\triangle ABC$ are similar; hence, $1 /x = x / \alpha\text{,}$ or $x^2 = \alpha\text{.}$</p></article></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-root-construction"><img src="images/fields-root.svg" width="80%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">21.38</span>Construction of roots</figcaption></figure></div>

<div class="mathbook-content"><p id="p-3308">By Theorem <a href="section-constructions.ipynb#theorem-constructible-numbers" class="xref" alt="Theorem 21.35 " title="Theorem 21.35 ">21.35</a>, we can locate in the plane any point $P =( p, q)$ that has rational coordinates $p$ and $q\text{.}$ We need to know what other points can be constructed with a compass and straightedge from points with rational coordinates.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-36"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">21.39</span></h6><p id="p-3309">Let $F$ be a subfield of ${\mathbb R}\text{.}$ </p><ol class="decimal"><li id="li-732"><p id="p-3310">If a line contains two points in $F\text{,}$ then it has the equation $a x + by + c = 0\text{,}$ where $a\text{,}$ $b\text{,}$ and $c$ are in $F\text{.}$</p></li><li id="li-733"><p id="p-3311">If a circle has a center at a point with coordinates in $F$ and a radius that is also in $F\text{,}$ then it has the equation $x^2 + y^2 + d x + e y + f = 0\text{,}$ where $d\text{,}$ $e\text{,}$ and $f$ are in $F\text{.}$</p></li></ol></article><article class="proof" id="proof-177"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3312">Let $(x_1, y_1)$ and $(x_2, y_2)$ be points on a line whose coordinates are in $F\text{.}$  If $x_1 = x_2\text{,}$ then the equation of the line through the two points is $x - x_1 = 0\text{,}$ which has the form $a x + by + c = 0\text{.}$  If $x_1 \neq x_2\text{,}$ then the equation of the line through the two points is given by</p><div class="displaymath">
\begin{equation*}
y - y_1 = \left( \frac{y_2 - y_1}{x_2 - x_1} \right) (x - x_1),
\end{equation*}
</div><p>which can also be put into the proper form.</p><p id="p-3313">To prove the second part of the lemma, suppose that $(x_1, y_1)$ is the center of a circle of radius $r\text{.}$  Then the circle has the equation</p><div class="displaymath">
\begin{equation*}
(x - x_1)^2 + (y - y_1)^2 - r^2 = 0.
\end{equation*}
</div><p>This equation can easily be put into the appropriate form.</p></article></div>

<div class="mathbook-content"><p id="p-3314">Starting with a field of constructible numbers $F\text{,}$ we have three possible ways of constructing additional points in ${\mathbb R}$ with a compass and straightedge. </p><ol class="decimal"><li id="li-734"><p id="p-3315">To find possible new points in ${\mathbb R}\text{,}$ we can take the intersection of two lines, each of which passes through two known points with coordinates in $F\text{.}$</p></li><li id="li-735"><p id="p-3316">The intersection of a line that passes through two points that have coordinates in $F$ and a circle whose center has coordinates in $F$ with radius of a length in $F$ will give new points in ${\mathbb R}\text{.}$</p></li><li id="li-736"><p id="p-3317">We can obtain new points in ${\mathbb R}$ by intersecting two circles whose centers have coordinates in $F$ and whose radii are of lengths in $F\text{.}$</p></li></ol><p> The first case gives no new points in ${\mathbb R}\text{,}$ since the solution of two equations of the form $a x + by + c = 0$ having coefficients in $F$ will always be in $F\text{.}$ The third case can be reduced to the second case.  Let</p><div class="displaymath">
\begin{gather*}
x^2 + y^2 + d_1 x +e_1 y + f_1 = 0\\
x^2 + y^2 + d_2 x +e_2 y + f_2 = 0
\end{gather*}
</div><p> be the equations of two circles, where $d_i\text{,}$ $e_i\text{,}$ and $f_i$ are in $F$ for $i = 1, 2\text{.}$ These circles have the same intersection as the circle</p><div class="displaymath">
\begin{equation*}
x^2 + y^2 + d_1 x +e_1 x + f_1 = 0 
\end{equation*}
</div><p>and the line</p><div class="displaymath">
\begin{equation*}
(d_1 - d_2) x + b(e_2 - e_1)y + (f_2 - f_1) = 0.
\end{equation*}
</div><p>The last equation is that of the chord passing through the intersection points of the two circles. Hence, the intersection of two circles can be reduced to the case of an intersection of a line with a circle.</p></div>

<div class="mathbook-content"><p id="p-3318">Considering the case of the intersection of a line and a circle, we must determine the nature of the solutions of the equations</p><div class="displaymath">
\begin{align*}
a x + by + c & = 0\\
x^2 + y^2 + d x + e y + f & = 0.
\end{align*}
</div><p>If we eliminate $y$ from these equations, we obtain an equation of the form $Ax^2 + B x + C = 0\text{,}$ where $A\text{,}$ $B\text{,}$ and $C$ are in $F\text{.}$ The $x$ coordinate of the intersection points is given by</p><div class="displaymath">
\begin{equation*}
x = \frac{- B \pm \sqrt{B^2 - 4 A C} }{2 A}
\end{equation*}
</div><p>and is in $F( \sqrt{\alpha}\, )\text{,}$ where $\alpha = B^2 - 4 A C \gt 0\text{.}$ We have proven the following lemma.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-lines-and-circles"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">21.40</span></h6><p id="p-3319">Let $F$ be a field of constructible numbers. Then the points determined by the intersections of lines and circles in $F$ lie in the field $F( \sqrt{\alpha}\, )$ for some $\alpha$ in $F\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-124"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">21.41</span></h6><p id="p-3320">A real number $\alpha$ is a constructible number if and only if there exists a  sequence of fields</p><div class="displaymath">
\begin{equation*}
{\mathbb Q} = F_0 \subset F_1 \subset \cdots \subset F_k
\end{equation*}
</div><p>such that $F_i = F_{i-1}( \sqrt{ \alpha_i}\, )$ with $\alpha_i \in F_i$ and $\alpha \in F_k\text{.}$ In particular, there exists an integer $k \gt 0$ such that $[{\mathbb Q}(\alpha) : {\mathbb Q} ] = 2^k\text{.}$</p></article><article class="proof" id="proof-178"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3321">The existence of the $F_i$'s and the $\alpha_i$'s is a direct consequence of Lemma <a href="section-constructions.ipynb#lemma-lines-and-circles" class="xref" alt="Lemma 21.40 " title="Lemma 21.40 ">21.40</a> and of the fact that</p><div class="displaymath">
\begin{equation*}
[F_k: {\mathbb Q}] = [F_k : F_{k - 1}][F_{k - 1} : F_{k - 2}] \cdots [F_1: {\mathbb Q} ] = 2^k.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="theorem-like" id="corollary-34"><h6 class="heading"><span class="type">Corollary</span><span class="codenumber">21.42</span></h6><p id="p-3322">The field of all constructible numbers is an algebraic extension of ${\mathbb Q}\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-3323">As we can see by the field of constructible numbers, not every algebraic extension of a field is a finite extension.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Doubling the Cube and Squaring the Circle"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Doubling the Cube and Squaring the Circle</span></h3><a href="section-constructions.ipynb#fields-subsection-doubling-the-cube" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-3324">We are now ready to investigate the classical problems of doubling the cube and squaring the circle. We can use the field of constructible numbers to show exactly when a particular geometric construction can be accomplished.</p></div>

<div class="mathbook-content"><h6 class="heading"><span class="title">Doubling the cube is impossible</span></h6></div>

<div class="mathbook-content"><p id="p-3325">Given the edge of the cube, it is impossible to construct with a straightedge and compass the edge of the cube that has twice the volume of the original cube. Let the original cube have an edge of length 1 and, therefore, a volume of  1. If we could construct a cube having a volume of 2, then this new cube would have an edge of length $\sqrt[3]{2}\text{.}$ However, $\sqrt[3]{2}$ is a zero of the irreducible polynomial $x^3 -2$ over ${\mathbb Q}\text{;}$ hence,</p><div class="displaymath">
\begin{equation*}
[{\mathbb Q}(\sqrt[3]{2}\, ) : {\mathbb Q}] = 3
\end{equation*}
</div><p>This is impossible, since 3 is not a power of 2.</p></div>

<div class="mathbook-content"><h6 class="heading"><span class="title">Squaring the circle</span></h6></div>

<div class="mathbook-content"><p id="p-3326">Suppose that we have a circle of radius 1.  The area of the circle is $\pi\text{;}$ therefore, we must be able to construct a square with side $\sqrt{\pi}\text{.}$ This is impossible since $\pi$ and consequently $\sqrt{\pi}$ are both transcendental. Therefore, using a straightedge and compass, it is not possible to construct a square with the same area as the circle.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Trisecting an Angle"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Trisecting an Angle</span></h3><a href="section-constructions.ipynb#fields-subsection-trisect-angle" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-3327"><em class="emphasis">Trisecting an arbitrary angle is impossible</em>.  We will show that it is impossible to construct a $20^\circ$ angle. Consequently, a $60^{\circ}$ angle cannot be trisected. We first need to calculate the triple-angle formula for the cosine:</p><div class="displaymath">
\begin{align*}
\cos 3 \theta & = \cos( 2 \theta + \theta )\\
& = \cos 2 \theta \cos \theta - \sin 2 \theta \sin \theta\\
& = ( 2 \cos^2 \theta - 1) \cos \theta - 2 \sin^2 \theta \cos \theta\\
& = ( 2 \cos^2 \theta - 1) \cos \theta - 2 (1- \cos^2 \theta) \cos \theta\\
& = 4 \cos^3 \theta - 3 \cos \theta. 
\end{align*}
</div><p>The angle $\theta$ can be constructed if and only if $\alpha = \cos \theta$ is constructible. Let $\theta = 20^{\circ}\text{.}$ Then $\cos 3 \theta =  \cos 60^\circ = 1/2\text{.}$ By the triple-angle formula for the cosine,</p><div class="displaymath">
\begin{equation*}
4 \alpha^3 - 3 \alpha = \frac{1}{2}.
\end{equation*}
</div><p>Therefore, $\alpha$ is a zero of $8 x^3 - 6 x -1\text{.}$ This polynomial has no factors in ${\mathbb Z}[x]\text{,}$ and hence is irreducible over ${\mathbb Q}[x]\text{.}$ Thus, $[{\mathbb Q}( \alpha ) : {\mathbb Q }] = 3\text{.}$ Consequently, $\alpha$ cannot be a constructible number.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Historical Note"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Historical Note</span></h3><a href="section-constructions.ipynb#fields-subsection-historical-note" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-3329">Algebraic number theory uses the tools of algebra to solve problems in number theory. Modern algebraic number theory began with Pierre de Fermat (1601–1665). Certainly we can find many positive integers that satisfy the equation $x^2 + y^2 = z^2\text{;}$ Fermat conjectured that the equation $x^n + y^n = z^n$ has no positive integer solutions for $n \geq 3\text{.}$  He stated in the margin of his copy of the Latin translation of Diophantus' <span class="booktitle">Arithmetica</span> that he had found a marvelous proof of this theorem, but that the margin of the book was too narrow to contain it.    Building on work of other mathematicians, it was Andrew Wiles who finally succeeded in proving Fermat's Last Theorem in the 1990s. Wiles's achievement was reported on the front page of the <span class="booktitle">New York Times</span>.</p></div>

<div class="mathbook-content"><p id="p-3330">Attempts to prove Fermat's Last Theorem have led to important contributions to algebraic number theory by such notable mathematicians as Leonhard Euler (1707–1783). Significant advances in the understanding of Fermat's Last Theorem were made by Ernst Kummer (1810–1893). Kummer's student, Leopold Kronecker (1823–1891), became one of the leading algebraists of the nineteenth century. Kronecker's theory of  ideals and his study of algebraic number theory added much to the understanding of fields.</p></div>

<div class="mathbook-content"><p id="p-3331">David Hilbert (1862–1943) and Hermann Minkowski (1864–1909) were among the mathematicians who led the way in this subject at the beginning of the twentieth century.  Hilbert and Minkowski were both mathematicians at Göttingen University in Germany. Göttingen was truly one the most important centers of mathematical research during the last two centuries. The large number of exceptional mathematicians who studied there included Gauss, Dirichlet, Riemann, Dedekind, Noether, and Weyl.</p></div>

<div class="mathbook-content"><p id="p-3332">André Weil answered questions in number theory using algebraic geometry, a field of mathematics that studies geometry by studying commutative rings. From about 1955 to 1970, Alexander Grothendieck dominated the field of algebraic geometry. Pierre Deligne, a student of Grothendieck, solved several of Weil's number-theoretic conjectures. One of the most recent contributions to algebra and number theory is Gerd Falting's proof of the Mordell-Weil conjecture. This conjecture of Mordell and Weil essentially says that certain polynomials $p(x, y)$ in ${\mathbb Z}[x,y]$ have only a finite number of integral solutions.</p></div>
