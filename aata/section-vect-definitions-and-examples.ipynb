<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 20.1 Definitions and Examples"><span class="type">Section</span><span class="codenumber">20.1</span><span class="title">Definitions and Examples</span></h2><a href="section-vect-definitions-and-examples.ipynb" class="permalink">Â¶</a></div>

<div class="mathbook-content"><p id="p-3083">A <dfn class="terminology">vector space</dfn> $V$ over a field $F$ is an abelian group with a <dfn class="terminology">scalar product</dfn> $\alpha \cdot v$ or $\alpha v$ defined for all $\alpha \in F$ and all $v \in V$ satisfying the following axioms. </p><ul class="disc"><li id="li-690"><p id="p-3084">$\alpha(\beta v) =(\alpha \beta)v\text{;}$</p></li><li id="li-691"><p id="p-3085">$(\alpha + \beta)v =\alpha v + \beta v\text{;}$</p></li><li id="li-692"><p id="p-3086">$\alpha(u + v) = \alpha u + \alpha v\text{;}$</p></li><li id="li-693"><p id="p-3087">$1v=v\text{;}$</p></li></ul><p> where $\alpha, \beta \in F$ and $u, v \in V\text{.}$</p></div>

<div class="mathbook-content"><p id="p-3088">The elements of $V$ are called <dfn class="terminology">vectors</dfn>; the elements of $F$ are called <dfn class="terminology">scalars</dfn>.  It is important to notice that in most cases two vectors cannot be multiplied.  In general, it is only possible to multiply a vector with a scalar. To differentiate between the scalar zero and the vector zero, we will write them as 0 and ${\mathbf 0}\text{,}$ respectively.</p></div>

<div class="mathbook-content"><p id="p-3089">Let us examine several examples of vector spaces. Some of them will be quite familiar; others will seem less so.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-vect-space0-rn"><h6 class="heading"><span class="type">Example</span><span class="codenumber">20.1</span></h6><p id="p-3090">The $n$-tuples of real numbers, denoted by ${\mathbb R}^n\text{,}$ form a vector space over ${\mathbb R}\text{.}$ Given vectors $u = (u_1, \ldots, u_n)$ and $v = (v_1, \ldots, v_n)$ in ${\mathbb R}^n$ and $\alpha$ in ${\mathbb R}\text{,}$ we can define vector addition by</p><div class="displaymath">
\begin{equation*}
u + v = (u_1, \ldots, u_n) + (v_1, \ldots, v_n) = (u_1 + v_1, \ldots, u_n + v_n)
\end{equation*}
</div><p>and scalar multiplication by</p><div class="displaymath">
\begin{equation*}
\alpha u = \alpha(u_1, \ldots, u_n)= (\alpha u_1, \ldots, \alpha u_n).
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="example-like" id="example-vect-space-fx"><h6 class="heading"><span class="type">Example</span><span class="codenumber">20.2</span></h6><p id="p-3091">If $F$ is a field, then $F[x]$ is a vector space over $F\text{.}$ The vectors in $F[x]$ are simply polynomials, and vector addition is just polynomial addition. If $\alpha \in F$ and $p(x) \in F[x]\text{,}$ then scalar multiplication is defined by $\alpha p(x)\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-vect-space-cont-func"><h6 class="heading"><span class="type">Example</span><span class="codenumber">20.3</span></h6><p id="p-3092">The set of all continuous real-valued functions on a closed interval $[a,b]$ is a vector space over ${\mathbb R}\text{.}$  If $f(x)$ and $g(x)$ are continuous on $[a, b]\text{,}$ then $(f+g)(x)$ is defined to be $f(x) + g(x)\text{.}$  Scalar multiplication is defined by $(\alpha f)(x) = \alpha f(x)$ for $\alpha \in  {\mathbb R}\text{.}$ For example, if $f(x) = \sin x$ and $g(x)= x^2\text{,}$ then $(2f + 5g)(x) =2 \sin x + 5 x^2\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-vect-space-sqrt2"><h6 class="heading"><span class="type">Example</span><span class="codenumber">20.4</span></h6><p id="p-3093">Let $V = {\mathbb Q}(\sqrt{2}\, ) = \{ a + b \sqrt{2} : a, b \in  {\mathbb Q } \}\text{.}$ Then $V$ is a vector space over ${\mathbb Q}\text{.}$ If $u = a + b \sqrt{2}$ and $v = c + d \sqrt{2}\text{,}$ then $u + v = (a + c) + (b + d ) \sqrt{2}$ is again in $V\text{.}$ Also, for $\alpha \in {\mathbb Q}\text{,}$ $\alpha v$ is in $V\text{.}$  We will leave it as an exercise to verify that all of the vector space axioms hold for $V\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-39"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">20.5</span></h6><p id="p-3094">Let $V$ be a vector space over $F\text{.}$ Then each of the following statements is true. </p><ol class="decimal"><li id="li-694"><p id="p-3095">$0v ={\mathbf 0}$ for all $v \in V\text{.}$</p></li><li id="li-695"><p id="p-3096">$\alpha {\mathbf 0} = {\mathbf 0}$ for all $\alpha \in F\text{.}$</p></li><li id="li-696"><p id="p-3097">If $\alpha v = {\mathbf 0}\text{,}$ then either $\alpha = 0$ or $v = {\mathbf 0}\text{.}$</p></li><li id="li-697"><p id="p-3098">$(-1) v = -v$ for all $v \in V\text{.}$</p></li><li id="li-698"><p id="p-3099">$-(\alpha v) = (-\alpha)v = \alpha(-v)$ for all $\alpha \in F$ and all $v \in V\text{.}$</p></li></ol></article><article class="proof" id="proof-155"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-3100">To prove (1), observe that</p><div class="displaymath">
\begin{equation*}
0 v = (0 + 0)v = 0v + 0v;
\end{equation*}
</div><p>consequently, ${\mathbf 0} + 0 v = 0v + 0v\text{.}$ Since $V$ is an abelian group, ${\mathbf 0} = 0v\text{.}$</p><p id="p-3101">The proof of (2) is almost identical to the proof of (1). For (3), we are done if $\alpha = 0\text{.}$  Suppose that $\alpha \neq 0\text{.}$ Multiplying both sides of $\alpha v = {\mathbf 0}$ by $1/ \alpha\text{,}$ we have $v = {\mathbf 0}\text{.}$</p><p id="p-3102">To show (4), observe that</p><div class="displaymath">
\begin{equation*}
v + (-1)v = 1v + (-1)v = (1-1)v = 0v = {\mathbf 0},
\end{equation*}
</div><p>and so $-v = (-1)v\text{.}$ We will leave the proof of (5) as an exercise.</p></article></div>
