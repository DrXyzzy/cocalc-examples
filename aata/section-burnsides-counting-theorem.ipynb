<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 14.3 Burnside's Counting Theorem"><span class="type">Section</span><span class="codenumber">14.3</span><span class="title">Burnside's Counting Theorem</span></h2><a href="section-burnsides-counting-theorem.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"></div>

<div class="mathbook-content"><p id="p-2061">Suppose that we wish to color the vertices of a square with two different colors, say black and white.  We might suspect that there would be $2^4=16$ different colorings. However, some of these colorings are equivalent.  If we color the first vertex black and the remaining vertices white, it is the same as coloring the second vertex black and the remaining ones white since we could obtain the second coloring simply by rotating the square $90^\circ$ (Figure <a href="section-burnsides-counting-theorem.ipynb#figure-actions-coloring" class="xref" alt="Figure 14.17 " title="Figure 14.17 ">14.17</a>).</p></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-actions-coloring"><img src="images/actions-colorings-square.svg" width="60%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">14.17</span>Equivalent colorings of square</figcaption></figure></div>

<div class="mathbook-content"><p id="p-2062">Burnside's Counting Theorem offers a method of computing the number of distinguishable ways in which something can be done. In addition to its geometric applications, the theorem has interesting applications to areas in switching theory and chemistry. The proof of Burnside's Counting Theorem depends on the following lemma.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-gset"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">14.18</span></h6><p id="p-2063">Let $X$ be a $G$-set and suppose that $x \sim y\text{.}$ Then $G_x$ is isomorphic to $G_y\text{.}$  In particular, $|G_x| = |G_y|\text{.}$</p></article><article class="proof" id="proof-91"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2064">Let $G$ act on $X$ by $(g,x) \mapsto g \cdot x\text{.}$ Since $x \sim y\text{,}$ there exists a $g \in G$ such that $g \cdot x=y\text{.}$ Let $a \in G_x\text{.}$ Since</p><div class="displaymath">
\begin{equation*}
gag^{-1} \cdot y = ga \cdot g^{-1}y = ga \cdot x = g \cdot x = y,
\end{equation*}
</div><p>we can define a map $\phi: G_x \rightarrow G_y$ by $\phi(a) = gag^{-1}\text{.}$ The map $\phi$ is a homomorphism since</p><div class="displaymath">
\begin{equation*}
\phi(ab) = gabg^{-1} = gag^{-1} gbg^{-1} = \phi(a) \phi(b).
\end{equation*}
</div><p>Suppose that $\phi(a) = \phi(b)\text{.}$ Then $gag^{-1}= gbg^{-1}$ or $a=b\text{;}$ hence, the map is injective.  To show that $\phi$ is onto, let $b$ be in $G_y\text{;}$ then $g^{-1}bg$ is in $G_x$ since</p><div class="displaymath">
\begin{equation*}
g^{-1}bg \cdot x = g^{-1}b \cdot gx = g^{-1}b \cdot y = g^{-1} \cdot y = x; 
\end{equation*}
</div><p>and $\phi(g^{-1}bg ) = b\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-69"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">14.19</span><span class="title">Burnside</span></h6><p id="p-2065">Let $G$ be a  finite group acting on a set $X$ and let $k$ denote the number of orbits of $X\text{.}$ Then</p><div class="displaymath">
\begin{equation*}
k = \frac{1}{|G|} \sum_{g \in G} |X_g|.
\end{equation*}
</div></article><article class="proof" id="proof-92"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2066">We look at all the fixed points $x$ of all the elements in $g \in G\text{;}$ that is, we look at all $g$'s and all $x$'s such that $gx =x\text{.}$ If viewed in terms of fixed point sets, the number of all $g$'s fixing $x$'s is</p><div class="displaymath">
\begin{equation*}
\sum_{g \in G} |X_g|.
\end{equation*}
</div><p>However, if viewed in terms of the stabilizer subgroups, this number is</p><div class="displaymath">
\begin{equation*}
\sum_{x \in X} |G_x|;
\end{equation*}
</div><p>hence, $\sum_{g \in G} |X_g| = \sum_{x \in X} |G_x|\text{.}$ By Lemma <a href="section-burnsides-counting-theorem.ipynb#lemma-gset" class="xref" alt="Lemma 14.18 " title="Lemma 14.18 ">14.18</a>,</p><div class="displaymath">
\begin{equation*}
\sum_{y \in {\mathcal O}_x} |G_y|  =  | {\mathcal O}_x| \cdot |G_x|.
\end{equation*}
</div><p>By Theorem <a href="section-groups-acting-on-sets.ipynb#theorem-orbit" class="xref" alt="Theorem 14.11 " title="Theorem 14.11 ">14.11</a> and Lagrange's Theorem, this expression is equal  to $|G|\text{.}$ Summing over all of the $k$ distinct orbits, we conclude that</p><div class="displaymath">
\begin{equation*}
\sum_{g \in G} |X_g| = \sum_{x \in X} |G_x| = k \cdot |G|.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><article class="example-like" id="example-actions-burnside"><h6 class="heading"><span class="type">Example</span><span class="codenumber">14.20</span></h6><p id="p-2067">Let $X = \{1, 2, 3, 4, 5 \}$ and suppose that $G$ is the permutation group $G= \{(1), (1 3), (1 3)(2 5), (2 5) \}\text{.}$ The orbits of $X$ are $\{1, 3\}\text{,}$ $\{2, 5\}\text{,}$ and $\{4\}\text{.}$ The fixed point sets are</p><div class="displaymath">
\begin{align*}
X_{(1)} & =  X\\
X_{(1 3)} & =  \{2, 4, 5 \}\\
X_{(1 3)(2 5)} & =  \{4\}\\
X_{(2 5)} & =  \{1, 3, 4 \}.
\end{align*}
</div><p>Burnside's Theorem says that</p><div class="displaymath">
\begin{equation*}
k = \frac{1}{|G|} \sum_{g \in G} |X_g| = \frac{1}{4}(5 + 3 + 1 + 3) = 3.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  A Geometric Example"><span class="type">Subsection</span><span class="codenumber" /><span class="title">A Geometric Example</span></h3><a href="section-burnsides-counting-theorem.ipynb#actions-subsection-geometric-example" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-2068">Before we apply Burnside's Theorem to switching-theory problems, let us examine the number of ways in which the vertices of a square can be colored black or white. Notice that we can sometimes obtain equivalent colorings by simply applying a rigid motion to the square. For instance, as we have pointed out, if we color one of the vertices black and the remaining three white, it does not matter which vertex was colored black since a rotation will give an equivalent coloring.</p></div>

<div class="mathbook-content"><p id="p-2069">The  symmetry group of a square, $D_4\text{,}$ is given by the following permutations:</p><div class="displaymath">
\begin{align*}
&(1)    & &(13)     & & (24)     & & (1432)\\
&(1234) & &(12)(34) & & (14)(23) & & (13)(24)
\end{align*}
</div><p>The group $G$ acts on the set of vertices $\{ 1, 2, 3, 4\}$ in the usual manner. We can describe the different colorings by mappings from $X$ into $Y = \{ B, W \}$ where $B$ and $W$ represent the colors black and white, respectively. Each map $f : X \rightarrow Y$ describes a way to color the corners of the square. Every $\sigma \in D_4$ induces a permutation $\widetilde{ \sigma }$ of the possible colorings given by $\widetilde{\sigma}(f) = f \circ \sigma$ for $f : X \rightarrow Y\text{.}$ For example, suppose that $f$ is defined by</p><div class="displaymath">
\begin{align*}
f(1) & =  B\\
f(2) & =  W\\
f(3) & =  W\\
f(4) & =  W
\end{align*}
</div><p>and $\sigma = (1 2)(3 4)\text{.}$ Then $\widetilde{\sigma}(f) = f \circ \sigma$ sends vertex 2 to $B$ and the remaining vertices to $W\text{.}$ The set of all such $\widetilde{\sigma}$ is a permutation group $\widetilde{G}$ on the set of possible colorings. Let $\widetilde{X}$   denote the set of all possible colorings; that is, $\widetilde{X}$ is the set of all possible maps from $X$ to $Y\text{.}$  Now we must compute the number of $\widetilde{G}$-equivalence classes. </p><ol class="decimal"><li id="li-467"><p id="p-2070">$\widetilde{X}_{(1)} = \widetilde{X}$ since the identity fixes every possible coloring. $|\widetilde{X}| = 2^4 =~16\text{.}$</p></li><li id="li-468"><p id="p-2071">$\widetilde{X}_{(1 2 3 4)}$ consists of all $f \in \widetilde{X}$ such that $f$ is unchanged by the permutation $(1 23 4)\text{.}$ In this case $f(1) = f(2) = f(3) = f(4)\text{,}$ so that all values of $f$ must be the  same; that is, either $f(x)= B$ or $f(x)= W$ for every vertex $x$ of  the square. So $|\widetilde{X}_{(1 2 3 4)}| = 2\text{.}$</p></li><li id="li-469"><p id="p-2072">$|\widetilde{X}_{(1 4 3 2)}| = 2\text{.}$</p></li><li id="li-470"><p id="p-2073">For $\widetilde{X}_{(1 3)(2 4)}\text{,}$ $f(1) = f(3)$ and $f(2) = f(4)\text{.}$ Thus, $|\widetilde{X}_{(13)(24)}| = 2^2 = 4\text{.}$</p></li><li id="li-471"><p id="p-2074">$|\widetilde{X}_{(1 2)(3 4)}| = 4\text{.}$</p></li><li id="li-472"><p id="p-2075">$|\widetilde{X}_{(1 4)(2 3)}| = 4\text{.}$</p></li><li id="li-473"><p id="p-2076">For $\widetilde{X}_{(1  3 )}\text{,}$ $f(1) = f(3)$ and the other corners can be of any color; hence, $|\widetilde{X}_{(1 3)}| = 2^3 = 8\text{.}$</p></li><li id="li-474"><p id="p-2077">$|\widetilde{X}_{(2 4)}| = 8\text{.}$</p></li></ol><p> By Burnside's Theorem, we can conclude that there are exactly</p><div class="displaymath">
\begin{equation*}
\frac{1}{8} ( 2^4 + 2^1 + 2^2 + 2^1  + 2^2 + 2^2 +2^3 + 2^3) = 6
\end{equation*}
</div><p>ways to color the vertices of the square.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-action-function"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">14.21</span></h6><p id="p-2078">Let $G$ be a permutation group of $X$ and $\widetilde{X}$ the set of functions from $X$ to $Y\text{.}$ Then there exists a permutation group  $\widetilde{G}$ acting on $\widetilde{X}\text{,}$ where $\widetilde{\sigma}  \in \widetilde{G}$ is defined by $\widetilde{\sigma}(f) = f \circ \sigma$ for $\sigma \in G$ and $f \in \widetilde{X}\text{.}$ Furthermore, if $n$ is the number of cycles in the cycle decomposition of $\sigma\text{,}$ then $|\widetilde{X}_{\sigma}| = |Y|^n\text{.}$</p></article><article class="proof" id="proof-93"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2079">Let $\sigma \in G$ and $f \in  \widetilde{X}\text{.}$ Clearly, $f \circ \sigma$ is also in $\widetilde{X}\text{.}$ Suppose that $g$ is another function from $X$ to $Y$ such that $\widetilde{\sigma}(f) = \widetilde{\sigma}(g)\text{.}$ Then for each $x \in X\text{,}$</p><div class="displaymath">
\begin{equation*}
f( \sigma(x )) = \widetilde{\sigma}(f)(x) = \widetilde{\sigma}(g)(x) = g( \sigma(x )).
\end{equation*}
</div><p>Since $\sigma$ is a permutation of $X\text{,}$ every element $x'$ in $X$ is the image of some $x$ in $X$ under $\sigma\text{;}$ hence, $f$ and $g$ agree on all elements of $X\text{.}$ Therefore, $f=g$ and $\widetilde{\sigma}$ is injective.  The map $\sigma \mapsto \widetilde{\sigma}$ is onto, since the two sets are the same size.</p><p id="p-2080">Suppose that $\sigma$ is a permutation of $X$ with cycle decomposition $\sigma = \sigma_1 \sigma_2 \cdots \sigma_n\text{.}$ Any $f$ in ${\widetilde{X}}_{\sigma}$ must have the same value on each cycle of $\sigma\text{.}$ Since there are $n$ cycles and $|Y|$ possible values for each cycle, $|{\widetilde{X}}_{\sigma}| = |Y|^n\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-actions-burnside-x7"><h6 class="heading"><span class="type">Example</span><span class="codenumber">14.22</span></h6><p id="p-2081">Let $X = \{1, 2, \ldots, 7\}$ and suppose that $Y = \{ A, B, C \}\text{.}$ If $g$ is the permutation of $X$ given by $(1 3)(2 4 5) = (1 3)(2 4 5)(6)(7)\text{,}$ then $n = 4\text{.}$ Any $f \in \widetilde{X}_g$ must have the same value on each cycle in $g\text{.}$ There are $|Y|=3$ such choices for any value, so $|\widetilde{X}_g|  = 3^4 = 81\text{.}$</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-actions-color-square"><h6 class="heading"><span class="type">Example</span><span class="codenumber">14.23</span></h6><p id="p-2082">Suppose that we wish to color the vertices of a square using four different colors. By Proposition <a href="section-burnsides-counting-theorem.ipynb#proposition-action-function" class="xref" alt="Proposition 14.21 " title="Proposition 14.21 ">14.21</a>, we can immediately decide that there are</p><div class="displaymath">
\begin{equation*}
\frac{1}{8} (4^4 + 4^1 + 4^2 + 4^1 + 4^2 + 4^ 2 + 4^3 + 4^3) = 55
\end{equation*}
</div><p>possible ways.</p></article></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Switching Functions"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Switching Functions</span></h3><a href="section-burnsides-counting-theorem.ipynb#actions-subsection-switching-functions" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-2083">In switching theory we are concerned with the design of electronic circuits with binary inputs and outputs. The simplest of these circuits is a switching function that has $n$ inputs and a single output (Figure <a href="section-burnsides-counting-theorem.ipynb#figure-switching-n-variables" class="xref" alt="Figure 14.24 " title="Figure 14.24 ">14.24</a>). Large electronic circuits can often be  constructed by combining smaller modules of this kind. The inherent problem here is that even for a simple circuit a large number of different switching functions can be constructed.  With only four inputs and a single output, we can construct 65,536 different switching functions. However, we can often replace one switching function with another merely by permuting the input leads to the circuit (Figure <a href="section-burnsides-counting-theorem.ipynb#figure-switching-two-variables" class="xref" alt="Figure 14.25 " title="Figure 14.25 ">14.25</a>).</p></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-switching-n-variables"><img src="images/actions-switching-function.svg" width="80%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">14.24</span>A switching function of $n$ variables</figcaption></figure></div>

<div class="mathbook-content"><p id="p-2084">We define a <dfn class="terminology">switching</dfn> or <dfn class="terminology">Boolean function</dfn> of $n$ variables to be a function from ${\mathbb Z}_2^n$ to ${\mathbb Z}_2\text{.}$ Since  any switching function can have two possible values for each binary $n$-tuple and there are $2^n$ binary $n$-tuples, $2^{2^n}$ switching functions are possible for $n$ variables. In general, allowing permutations of the inputs greatly reduces the number of different kinds of modules that are needed to build a large circuit.</p></div>

<div class="mathbook-content"><figure class="figure-like" id="figure-switching-two-variables"><img src="images/actions-switching-two-variables.svg" width="100%" alt="" /><figcaption><span class="type">Figure</span><span class="codenumber">14.25</span>A switching function of two variables</figcaption></figure></div>

<div class="mathbook-content"><p id="p-2085">The possible switching functions  with two input variables $a$ and $b$ are listed in Table <a href="section-burnsides-counting-theorem.ipynb#table-switching-2-variables" class="xref" alt="Table 14.26 " title="Table 14.26 ">14.26</a>. Two switching functions $f$ and $g$ are equivalent if $g$ can be obtained from $f$ by a permutation of the input variables. For example, $g(a, b, c) = f(b, c, a)\text{.}$ In this  case $g \sim f$ via the permutation $(acb)\text{.}$ In the case of switching functions of two variables, the permutation $(ab)$ reduces 16 possible switching functions to 12 equivalent functions since</p><div class="displaymath">
\begin{align*}
f_2 & \sim  f_4\\
f_3 & \sim  f_5\\
f_{10} & \sim f_{12}\\
f_{11} & \sim f_{13}.
\end{align*}
</div></div>

<div class="mathbook-content"><p id="p-2086">For three input variables there are $2^{2^3} = 256$ possible switching functions; in the case of four variables there are $2^{2^4} =\text{65,536}\text{.}$ The number of equivalence classes is too large to reasonably calculate directly. It is necessary to employ  Burnside's Theorem.</p></div>

<div class="mathbook-content"><p id="p-2087">Consider a switching function with three possible inputs, $a\text{,}$ $b\text{,}$ and $c\text{.}$ As we have mentioned, two switching functions $f$ and $g$ are equivalent if a permutation of the input variables of $f$ gives $g\text{.}$ It is important to notice that a permutation of the switching functions is not simply a permutation of the input values $\{a, b, c\}\text{.}$ A switching function is a set of output values for the inputs $a\text{,}$ $b\text{,}$ and $c\text{,}$ so when we consider equivalent switching functions, we are permuting $2^3$ possible outputs, not just three input values. For example, each binary triple $(a, b, c)$ has a specific output associated with it. The  permutation $(acb)$ changes outputs as follows:</p><div class="displaymath">
\begin{align*}
(0, 0, 0) & \mapsto  (0, 0, 0)\\
(0, 0, 1) & \mapsto  (0, 1, 0)\\
(0, 1, 0) & \mapsto  (1, 0, 0)\\
& \vdots \\
(1, 1, 0) & \mapsto  (1, 0, 1)\\
(1, 1, 1) & \mapsto  (1, 1, 1).
\end{align*}
</div></div>

<div class="mathbook-content"><p id="p-2088">Let $X$ be the set of output values for a switching function in $n$ variables. Then $|X|=2^n\text{.}$ We can enumerate these values as follows:</p><div class="displaymath">
\begin{align*}
(0, \ldots, 0, 1) & \mapsto  0\\
(0, \ldots, 1, 0) & \mapsto  1\\
(0, \ldots, 1, 1) & \mapsto  2\\
& \vdots \\
(1, \ldots, 1, 1) & \mapsto  2^n-1.
\end{align*}
</div><p>Now let us consider a circuit with four input variables and a single output. Suppose that we can permute the leads  of any circuit according to the following permutation group:</p><div class="displaymath">
\begin{gather*}
(a)    \quad (ac)     \quad (bd)     \quad (adcb)\\
(abcd) \quad (ab)(cd) \quad (ad)(bc) \quad (ac)(bd).
\end{gather*}
</div><p>The permutations of the four possible input variables induce the permutations of the output values in Table <a href="section-burnsides-counting-theorem.ipynb#table-switching-permutation" class="xref" alt="Table 14.27 " title="Table 14.27 ">14.27</a>.</p></div>

<div class="mathbook-content"><p id="p-2089">Hence, there are</p><div class="displaymath">
\begin{equation*}
\frac{1}{8} (2^{16} + 2 \cdot 2^{12} + 2 \cdot 2^6 + 3 \cdot 2^{10}) = 9616
\end{equation*}
</div><p>possible switching functions of four variables under this group of permutations. This number will be even smaller if we consider the full symmetric group on four letters.</p></div>

<div class="mathbook-content"><h3 class="heading hide-type" alt="Subsection  Historical Note"><span class="type">Subsection</span><span class="codenumber" /><span class="title">Historical Note</span></h3><a href="section-burnsides-counting-theorem.ipynb#actions-subsection-historical-note" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-2091">William Burnside was born in London in 1852. He attended Cambridge University from 1871 to 1875 and won the Smith's Prize in his last year. After his graduation he lectured at Cambridge.  He was made a member of the Royal Society in 1893. Burnside wrote approximately 150 papers on topics in applied mathematics, differential geometry, and probability, but his most famous contributions were in group theory.  Several of Burnside's conjectures have stimulated research to this day. One such conjecture was that every group of odd order is solvable; that is, for a group $G$ of odd order, there exists a sequence of subgroups</p><div class="displaymath">
\begin{equation*}
G = H_n \supset H_{n-1} \supset \cdots \supset H_1 \supset H_0 = \{ e \}
\end{equation*}
</div><p>such that $H_i$ is normal in $H_{i+1}$ and $H_{i+1} / H_i$ is abelian. This conjecture was finally proven by W. Feit and J. Thompson in 1963. Burnside's <span class="booktitle">The Theory of Groups of Finite Order</span>, published in 1897, was one of the first books to treat groups in a modern context as opposed to permutation groups. The second edition, published in 1911, is still a classic.</p></div>
